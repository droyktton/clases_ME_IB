# Densidad de Estados de **1 partícula**

La densidad de estados $g(\epsilon)$ se define así

$$g(\epsilon)dϵ= \text{# estados de 1 partícula con energía en }[\epsilon,ϵ+dϵ]$$

Es el análogo al $\Omega(E)$ del microcanónico, con la diferencia que en el anterior caso nos referiamos al los microestados de **todo el sistema**, no solo los de **una partícula**.

Tener $g(\epsilon)$ es muy práctico porque muchas propiedades de los sistemas de partículas idénticas se reducen integrales sobre funciones que solo dependen de $\epsilon$.

La densidad de estados $g(\epsilon)$ puede ser a veces muy complicada de calcular, puede tener gaps (rangos de energía donde no hay estados de una partícula) y aún divergencias.
Para los casos que veremos en esta materia sin embargo, serán funciones simples de calcular o será un dato, que podremos usar sin problemas para hacer mecánica estadística.



## Densidad de estados de una partícula en una caja con espín

Este caso, los estados estan etiquetatos por ${\bf k}$ y $\sigma$. A la función que da la energía del estado en función de los números cuánticos  

$$\epsilon_{\bf k,\sigma}$$

se la llama **relación de dispersión**, y puede ser *en general anisotrópica y no monótona*.

Podemos escribir, formalmente, en el caso más general

$$g(ϵ)=\sum_{\sigma} L^d \int_{{\bf k}} \frac{dk_1 \dots dk_d}{(2\pi)^d}
\delta(ϵ-\epsilon_{\bf k,\sigma})
$$

> Estamos usando condiciones periódicas de contorno, por eso usamos $(2\pi/L)^d$. Si queremos condiciones cerradas, como en una caja, hay que multiplicar todos los resultados por $2^d$, pues el elemento en el espacio $k$ es $(\pi/L)^d$ en ese caso. Más adelante, cuando incluyan la influencia de la red de iones verán que reaparece $(2\pi/L)^d$ en la $g(\epsilon)$, mas otras sorpresas...

Si hay degeneración de espín, podemos escribir

$$g(ϵ)=g_s L^d \int_{{\bf k}} \frac{dk_1 \dots dk_d}{(2\pi)^d}
\delta(ϵ-\epsilon_{\bf k})
$$

Y si además, hay simetría esférica en la relación de dispesión, $ϵ_{\bf k}=ϵ_k$,
con $k=\sqrt{k_1^2+\dots+k_d^2}$, entonces

$$g(ϵ)= g_s L^d\int_{{\bf k}} \frac{\Omega_d k^2 dk}{(2\pi)^d}
\delta(ϵ-\epsilon_k).
$$

Si además $ϵ_k$ es biyectiva, la evaluación es más sencilla, pues

$$g(ϵ)dϵ=g(k)dk = \frac{g_s \Omega_d k^2 dk}{(2\pi/L)^d}$$

de modo que

$$g(ϵ)=\frac{g_s \Omega_d k^2}{(2\pi/L)^d} \frac{dk}{d\epsilon}|_{k=\epsilon_k^{-1}(\epsilon)}$$

La cuenta es sencilla y en $d=3$ da

\begin{align}
g(\epsilon)d\epsilon &=
g_s \frac{4\pi k^2 dk}{(2\pi)^3 V^{-1}}
=
g_s V \frac{2 k^2 dk}{(2\pi)^2} \\
&=
\frac{g_s V}{2 \pi^2} k^2 dk
= \frac{g_s V}{2 \pi^2} (2m\epsilon/\hbar^2)d((2m\epsilon/\hbar^2)^{1/2}) \\
&=  \frac{g_s V}{4 \pi^2} (2m/\hbar^2)^{3/2} \epsilon^{1/2} d\epsilon
\end{align}

Y en cualquier dimensión (ejercicio)

\begin{align}
g(\epsilon)d\epsilon &=
\frac{g_s L^d \Omega_d}{2(2\pi)^d} (2m/\hbar^2)^{d/2} \epsilon^{d/2-1} d\epsilon
\end{align}

Podemos ver fácilmente que electrones con relaciones de dispersión monótonas en dimensión arbitraria con simetría esférica del tipo $\epsilon_k \propto k^\alpha$ degenerada en espín tienen en general la forma

$$g(\epsilon)=A V \epsilon^{\gamma}$$

donde $A$ es una constante, $V=L^d$ y $\gamma$ a determinar, función de $\alpha$ y $d$.


**ATENCION**
* No confundir $g(\epsilon)$ con $\langle n(\epsilon) \rangle$. La densidad de estados es una propiedad de la estructura de estados de una partícula y no depende del estado del sistema para nada. En particular $g(\epsilon)$ no depende de la temperatura, ni de $\mu$, pero si depende de $V$ porque afecta a la estructura de niveles de energía.

* Notar que la función de Fermi $\langle n(\epsilon) \rangle$ es la ocupación media de un estado de una partícula con energía $\epsilon$, que no es lo mismo que la ocupación media del rango de energía $[\epsilon, \epsilon+d\epsilon]$, que sería $\langle n(\epsilon) \rangle g(\epsilon)d\epsilon$.

* Notar que $A=A(g_s,m)$. Depende de la degeneración, de la masa, de la dimensión del espacio, y de la relación de dispersión. En este caso el de una partícula masiva no relativista en tres dimensiones, con $g=2s+1$ y $s$ el espín.


* Notar que $g(\epsilon) \propto V$: cuanto más volumen más densidad de estados pues los niveles se acercan más.

* La densidad de estados no tiene información sobre la estadística de las partículas, BE o FD.

* Si los niveles son discretos uno pensaría que no puede ser una función continua, pero en muchos casos prácticamente lo es pues la separación entre niveles es ínfima, $\propto L^{-2}$ para el caso anterior.

* Si fuera una trampa armónica en cambio, la separación está dada por $\hbar\omega_0$, que para trampas macroscópicas es también muy chiquita.

La idea es que puedan hacer este cambio de variables automaticamente

\begin{align}
\sum_k \equiv \int \frac{d^3 k}{(2\pi)^3 V^{-1}} \dots
\equiv \int_0^\infty d\epsilon\; g(\epsilon) \dots
\end{align}

y pasar a integrar sobre energías.



## Ejemplo de uso

Por ejemplo, si queremos calcular la energía interna en 3d

$$E = \sum_{\sigma} \int_{\bf k} \frac{d^3 k}{(2\pi/L)^3} \langle n_{\bf k,\sigma}\rangle ϵ_{\bf k,\sigma}
= \int d\epsilon g(ϵ) \epsilon n(ϵ)
= \int d\epsilon \frac{g(ϵ) \epsilon}{z^{-1}e^{\beta \epsilon}-\eta}
$$

Y si usamos $g(\epsilon)=A(T,m) g_s V ϵ^{1/2}$ para partículas no relativistas en 3d con degeneración de espín $g_s$, es fácil llegar al resultados que obtuvimos anteriormente

$$E
= A g_s V \int d\epsilon \frac{\epsilon^{3/2}}{z^{-1}e^{\beta \epsilon}-\eta}
= A g_s V \beta^{-5/2}\int dx \frac{x^{3/2}}{z^{-1}e^{x}-\eta}
= A g_s V \beta^{-5/2} f_{5/2}^\eta(z)
$$

donde el prefactor depende de $T$ y $m$. Para verlo, solo hacemos un cambio de variables

# Fermiones

[¿Quiénes son fermiones?](https://en.wikipedia.org/wiki/List_of_particles)



Una regla bastante general es que *una partícula que contiene un número impar de fermiones es un fermión*. Sean fermiones identicos o no. Incluso si hay una mezcla de bosones y fermiones idénticos o no entre sí, funciona la regla.  

Algunos Ejemplos de Fermiones compuestos:

* Protón (3 quarks).
* Neutrón (3 quarks).
* Carbono 13 (6 protones, 7 neutrones).
* Helio 3 (2 protones, 1 neutrón, 2 electrones).
* Deuterio (1 protón, 1 neutrón, 1 electrón).

Para *átomos neutros* es sencillo, porque tienen igual número de protones y de electrones, de modo que quien decide es el *número de neutrones*: si es par el isotopo es un bosón, y si es impar un fermión.

Cabe aclarar que matemáticamente hay varios tipos de fermiones, de Dirac, de Weyl, de Majorana, pero los del modelo standard son de Dirac. Los otros son más complicados y siguen siendo tema activo de investigación.

#  Gas de Fermiones Idénticos no interactuantes

Vamos a considerar fermiones (de cualquier tipo, elementales o compuestos) sin interacciones. Se le llama "degenerado" al gas, no por ser un gas inmoral, sino porque hay "degeneración de intercambio" debido a la indistinguibilidad de partículas idénticas.

Se usa degenerado para enfatizar el hecho de que la simetría de intercambio (en este caso anti-simetría) no se puede despreciar, ya sea porque la temperatura es baja o la densidad es alta, o más bien porque $N \lambda^3 /V \gg 1$ (muchas partículas entran en una cajita de De Broglie de volumen $\lambda^3$).

Para fijar ideas vamos a considerar a los fermiones en un potencial confinante o caja de volumen $V$, de modo que la única energía del gas es cinética, y caracterizada por el vector de onda ${\bf k}$. Inicialmente consideraremos que el espín solo produce degeneración $g_s$ de cada estado de una partícula, pero al final agregaremos un campo magnético para romper esta degeneración de espín.

## Números de ocupación medios
Los fermiones corresponden al caso $\eta=-1$.

\begin{align}
\langle n_k \rangle_{FD} = \frac{1}{z^{-1}e^{\beta {\cal E}(k)}+1}
=
\frac{1}{e^{\beta ({\cal E}(k)-\mu)}+1}
\end{align}

* A veces la escriben en términos de $f(x)$, que no es más que la misma distribución pero centrada en $0$ en vez de $\mu$.
\begin{align}
f(\epsilon-\mu)=\langle n(\epsilon) \rangle_{FD} =
\frac{1}{e^{\beta (\epsilon-\mu)}+1}
\end{align}
pero *no hay que perder de vista que representa el número medio en un estado de una partícula, y no en una autoenergía de una partícula, ya que esta puede estar degenerada.*

* Tiene forma de escaloncito.

* A $T$ baja satura a $f(\epsilon)=1$, lo cual está bien porque más no se pueden poner por Pauli.

* Vale $f(0)=1/2$ y para $\epsilon \gg \mu$ decae a cero exponencialmente.

* El ancho del escalón es de orden $k_B T$.

* Tiene una simetría respecto de $\mu$,  
$f(-x)=1-f(x)$.

* Es el mismo número de ocupación medio de partículas adsorbidas en una red donde se puede poner como máximo una partícula, cada sitio con energía $\epsilon$.

```python
#@title Visualización interactiva de la función de Fermi
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def fermi_function(epsilon, mu, T):
  """Calculates the Fermi-Dirac distribution."""
  k_B = 1  # Boltzmann constant (set to 1 for simplicity)
  return 1 / (np.exp((epsilon - mu) / (k_B * T)) + 1)

def plot_fermi_function(T, mu):
  """Plots the Fermi function with given T and mu."""
  epsilon = np.linspace(-5, 5, 100)
  f_epsilon = fermi_function(epsilon, mu, T)
  plt.plot(epsilon, f_epsilon)
  plt.xlabel('ε')
  plt.ylabel('f(ε-μ)')
  plt.title('Fermi-Dirac Distribution')
  plt.grid(True)
  plt.show()

# Create sliders for T and mu
interact(plot_fermi_function,
          T=FloatSlider(value=0.2, min=0.1, max=5, step=0.1, description='T:'),
          mu=FloatSlider(value=0.5, min=0, max=5, step=0.1, description='μ:'))
```

## T=0, Estado Fundamental

El estado fundamental de los fermiones ya es sorprendente, pues tiene una energía por partícula muy por encima de la del estado fundamental de una partícula, debido al principio de exclusión de Pauli.

En el caso de partículas en una caja, la energía es puramente cinética, con un rango muy amplio de energías de una partícula, desde practicamente cero, hasta valores grandes que dependen de la densidad de fermiones.



### Energía y vector de onda de Fermi

A $T=0$, $\langle n_k \rangle$ se vuelve un escalón abrupto, que salta de $\langle n_k \rangle=1$ a $\langle n_k \rangle=0$ en $\epsilon_F = \mu(T=0)$, que se denomina **energía de Fermi**. Todos los estados con $\epsilon_k<\epsilon_F$ están en el **mar de Fermi**.

Como $\epsilon_k \equiv \frac{\hbar^2 k^2}{2m}$ podemos definir el **vector de onda de Fermi**

$$k_F = \sqrt{\frac{2 m \epsilon_F}{\hbar^2}}$$

que es el máximo valor de $k$ posible a $T=0$, porque la energíua es monótonicamente creciente con $k$.

Para calcular $k_F$ en términos de variables termodinámicas, podemos usar la simetría esférica de la relación de dispersión
y que $g=2s+1$ y escribir

\begin{align}
N &= g\int_0^{k_F} \frac{4\pi k^2 dk}{(2\pi)^3 V^{-1}}
= \frac{2 g V}{(2\pi)^2}
\int_0^{k_F} k^2 dk = \frac{2 g V}{(2\pi)^2} \frac{k_F^3}{3} \\
n &= N/V = \frac{g}{2 \pi^2} \frac{k_F^3}{3} \\
\end{align}

Entonces

\begin{align}
k_F = \left( \frac{6\pi^2 n}{g} \right)^{1/3}
\end{align}

y como $ϵ_F=\hbar^2 k_F^2/2m$,

\begin{align}
\epsilon_F = \frac{\hbar^2}{2m}\left( \frac{6\pi^2 n}{g} \right)^{2/3}
\end{align}

La dependencia con la densidad es esperada, ya que

* Por una lado si hay más fermiones, más estados de una partícula y más altos en energía puedo llenar. Y el 2/3 tiene que ver con la esfera de Fermi.
* Por otro lado, si mantengo el número de fermiones fijo pero bajo el volumen, los niveles de energía se separan más (recordar que un nivel de una partícula tiene una energía $\epsilon_k = \hbar^2 k^2/2m \sim \hbar^2 (\pi /L)^2 (n_x^2+n_y^2+n_z^2)/2m \sim 1/L^2$), y el nivel de Fermi queda más alto.

Para que se den una idea, en metales, la **temperatura de Fermi** $T_F = \epsilon_F/k_B \approx 80000 K$, asi que la función de Fermi parece un escalón aún a $T\sim 293K$ ambiente, porque $T/T_F \ll 1$.

### Estado Fundamental (¡de todo el gas!)

Notar entonces que el estado fundamental del gas de Fermi, alcanzado en $T=0$ tiene a todos los estados ocupados hasta el nivel de Fermi. En esa última capa, si no está completa, hay cierta libertad de cambiar de microestado sin cambiar de energía total. Hay del orden de $\Gamma = g_s 4 \pi k_F^2/(\pi/L)^3$ estados para elegir, para las $M \ll N$ partículas que quedan en la última capa, o sea $\Gamma$ tomados de a $M$. Esto dará una contribución al número de microestados en el estado fundamental $\Omega(E_0)$, y una entropía a $T=0$,
$S = k_B \ln(\Omega(E_0))$, pero esta será tal que $S/N \to 0$ cuando $N\to \infty$. Entonces vale la tercer ley de la termodinámica, o Ley de Nernst, $\lim_{T\to 0} S/N = 0$.

Este resultado contrasta fuertemente con el caso clásico, donde vimos que $\Omega \propto V^N E^{3N/2}/N!$, y teniamos problemas con la ley de Nernst,
$\lim_{T\to 0} S/N \to -\infty$ (!). Así que ya vemos como la estadística cuántica de Fermiones en este caso nos salva de semejante catástrofe.

El fundamental tiene entonces una energía *total*
\begin{align}
E = g \int_0^{k_F} 4\pi k^2 dk \frac{\hbar^2 k^2}{2m}
= 2\pi g \frac{k_F^5}{5} \frac{\hbar^2}{m} \equiv \frac{3}{5}
N \epsilon_F
\end{align}
y crece con la densidad.
Notar que  $E < N\epsilon_F$, como debería ser ya que $N\epsilon_F$ sería poner todas las partículas en $\epsilon_F$, y no sería un mínimo. Por otro lado es notable qeu $3/5 = 0.6$ no es tan pequeño, cosa que es entendible por como crece el número de estados con $k$: hay $4\pi k^2 dk$ estados para la energía $\hbar^2 k^2/2m$, y pesan más las energías más altas.  

Recordemos también que, independientemente de que sean Fermiones o Bosones, las partículas no relativistas en 3d cumplen, a toda $T$, incluída $T=0$, que

$$P V = (2/3)E$$

Pues bien, a $T=0$ los fermiones tienen una $E$ apreciable,

$$P V = (2/3)(3/5)N ϵ_F = (2/5)N ϵ_F $$

entonces obtenemos la presión de Fermi, $P_F = P(T=0,n)$

$$P_F = (2/5)n ϵ_F \propto n^{5/3} $$

Este resultado es bastante sorprendente, una presión a $T=0$, y es una consecuencia directa del principio de exclusión de Pauli. Al no poder poner a todos los fermiones en el estado fundamental de una partícula, donde tienen un momento practicamente nulo, nos vemos obligados a tener estados ocupados con momentum finito $\hbar k$, que pueden contribuir a la presión aún a $T=0$.

> O sea, la hinchada fermiónica aunque este super tranquila, ejerce presión.

**No confundir degeneración del estado fundamental de una partícula con degeneración del estado fundamental de todo el sistema**

En tres dimensiones para partículas no relativistas, tenemos

$$g(\epsilon)=V A \epsilon^{1/2}$$  

con $A=\left[g (2m/\hbar^2)^{3/2}  
\;/(2\pi)^2 \right]=(g m^{3/2})/(\hbar^3 2^{1/2} \pi^2)$.

Entonces

\begin{align}
E = \int_0^{\epsilon_F} g(\epsilon)d\epsilon\; \epsilon
= A \int_0^{\epsilon_F} d\epsilon\; \epsilon^{3/2}
= \frac{2}{3} A \epsilon_F^{5/2}
\end{align}

y metiendo $A$ reobtenemos lo anterior, y

\begin{align}
N = \int_0^{\epsilon_F} g(\epsilon)d\epsilon\;  
= A \int_0^{\epsilon_F} d\epsilon\; \epsilon^{1/2}
= 2 A \epsilon_F^{3/2}
\end{align}



### Superficie de Fermi

Es simplemente la capa de estados a la energía de Fermi. En el caso de un gas de fermiones no interactuantes y en ausencia de un potencial externo es una *esfera*, la esfera de Fermi, y los vectores de onda pueden tener cualquier magnitud.

Pero, cuando uno incorpora por ejemplo el efecto de la red de núcleos en un sólido sobre los electrones, o considera el caso discreto en una red (es decir en la aproximación en que los electrones solo pueden saltar entre sitios, pero no estar en medio, conocida como "tight binding"), la superficie de Fermi puede tomar otras formas y pueden confinarse los valores permitidos de ${\bf k}$ a una región finita... La forma de la superficie es afectada por la red considerada también.

```python
#@title Superficie de Fermi red cubica simple
import numpy as np
import plotly.graph_objects as go

# Define the dispersion relation for a BCC lattice
def E_dispersion_bcc(kx, ky, kz, E0=0, t=1, a=1):
    return E0 - 2 * t * (np.cos(kx * a) * np.cos(ky * a) +
                         np.cos(ky * a) * np.cos(kz * a) +
                         np.cos(kz * a) * np.cos(kx * a))


# Define the dispersion relation for a BCC lattice
def E_dispersion_sc(kx, ky, kz, E0=0, t=1, a=1):
    return E0 - 2 * t * (np.cos(kx * a) +
                         np.cos(ky * a) +
                         np.cos(kz * a) )

# Parameters
t = 1.0     # Hopping energy (can be adjusted)
a = 1.0     # Lattice constant (can be adjusted)
k_range = np.linspace(-np.pi/a, np.pi/a, 50)  # Define the range of k-values

# Create a 3D grid of k-points
kx, ky, kz = np.meshgrid(k_range, k_range, k_range)
#E_values = E_dispersion_bcc(kx, ky, kz, E0=0, t=t, a=a)
E_values = E_dispersion_sc(kx, ky, kz, E0=0, t=t, a=a)

# Create the initial figure
E_F_initial = -0.1  # Initial value of Fermi energy
fig = go.Figure()

# Add the initial isosurface for the initial Fermi energy
isosurface = go.Isosurface(
    x=kx.flatten(),
    y=ky.flatten(),
    z=kz.flatten(),
    value=E_values.flatten(),
    isomin=E_F_initial,
    isomax=E_F_initial,
    surface_count=1,   # Only show the Fermi surface
    colorscale="Viridis",
    caps=dict(x_show=False, y_show=False, z_show=False)
)
fig.add_trace(isosurface)

# Set layout options
fig.update_layout(
    title=f'Fermi Surface for SIMPLE CUBIC Lattice',
    scene=dict(
        xaxis_title="k_x",
        yaxis_title="k_y",
        zaxis_title="k_z"
    ),
    sliders=[{
        "active": 10,
        "currentvalue": {"prefix": "E_F: "},
        "pad": {"t": 50},
        "steps": [
            {
                "method": "update",
                "label": f"{E_F:.2f}",
                "args": [{"isomin": E_F, "isomax": E_F}]
            } for E_F in np.linspace(-6, 6, 20)  # Slider values from 0 to 1
        ]
    }]
)

# Show the plot
fig.show()
```

## Fermiones a bajas temperaturas $T \ll T_F$

Cuando uno dice baja $T$, tiene que decir respecto de que y para eso hay que encontrar en el problema temperaturas características finitas. En el problema de Fermiones hay una sola, $T_F=\epsilon_F/k_B$.

> Uno podría decir que algo como $\hbar^2 (\pi/L)^2/2m$ es otra escala de energía y por tanto de temperatura, pero no es relevante en el límite termodinámico.

Por lo tanto, cuando decimos bajas temperaturas, nos referimos a $T \ll T_F$ y veremos que es consistente: todas las expansiones resultan ser en $T/T_F$.

## Desarrollo de Sommerfeld

El comportamiento de una propiedad de la forma

$${\cal I}=\int_{-\infty}^\infty d\epsilon f(\epsilon-\mu) \phi(\epsilon)$$

con la función de Fermi

$$f(x)=(1+e^{\beta x})^{-1}$$

se puede estudiar en general con un desarrollo en potencias de $k_B T$, cuando $T \ll T_F$, cuando los efectos de degeneración de intercambio y el principio de exclusión de Pauli son muy relevantes. O sea, cuando $f(x)$ tiene aproximadamente la forma de escalón, apenas redondeado.

El desarrollo en potencias se llama desarrollo de Sommerfeld, y a continuación va la receta.

### Receta

Esta es la receta de Sommerfeld para aproximar integrales de este tipo

\begin{align}
\boxed{
{\cal I}=\int_{-\infty}^\infty d\epsilon f(\epsilon-\mu) \phi(\epsilon)
\approx \int_{-\infty}^\mu d\epsilon  \phi(\epsilon) + \frac{\pi^2}{6}(k_B T)^2 \phi'(\mu) + \frac{7\pi^4}{360}(k_B T)^4 \phi'''(\mu)+\dots
}
\end{align}

Este tipo general de integrales aparece muy a menudo cuando calculamos propiedades del gas de Fermi.
Una obervación importante que ya se puede hacer es que toda la expansión está controlada por $\phi$ y sus derivadas en $\mu$, que a bajas $T$, es $\mu \approx \epsilon_F$. Vemos nuevamente que lo que ocurre en la superficie de Fermi es importante.

> Para usar el desarrollo de Sommerfeld necesitamos una $\phi(\epsilon)$ *analítica*. Sino hay que usar otro método.


### Idea de la demostración

Este desarrollo explota el hecho de que $f(x)$ es *casi* un escalón, y su derivada casi una delta de Dirac en $x=0$.

La parte clave de la demostración es hacer aparecer entonces $f'(x)$, que se parece a una gausianita angosta, y para eso podemos proceder así.

* Definimos $\Phi(\epsilon)\equiv \int_{-\infty}^\epsilon d\epsilon' \phi(\epsilon')$, tal que $\phi(\epsilon)=\Phi'(\epsilon)$.

* Tenemos entonces
\begin{align}
{\cal I}&=\int_{-\infty}^{\infty}d\epsilon f(\epsilon-\mu)d\Phi/d\epsilon \\
&=-\int_{-\infty}^{\infty}d\epsilon f'(\epsilon-\mu)\Phi(\epsilon) \\
&=-\int_{-\infty}^{\infty}d\epsilon f'(\epsilon)\Phi(\epsilon+\mu)
\end{align}
donde en la segunda igualdad usamos que $|f(\epsilon-\mu)\Phi(\epsilon)|_{-\infty}^\infty=0$ (cuando no se anula una se anula la otra).

* $f'(\epsilon)$ es una función tipo pico de ancho $\sim k_B T$ que tiene colas exponenciales, ya que es la derivada de la función de Fermi a $T$ finita. Esto indica que las contribuciones más importantes a la integral estarán concentradas cerca de $\epsilon\sim 0$. Entonces puedo desarrollar en serie de potencias, entonces

\begin{align}
{\cal I}&=
-\int_{-\infty}^{\infty}d\epsilon f'(\epsilon)
[\Phi(\mu)+\epsilon \Phi'(\mu) + (\epsilon^2/2)\Phi''(\mu)+\dots] \\
&=
-\Phi(\mu)\int_{-\infty}^{\infty}d\epsilon f'(\epsilon)
- \Phi'(\mu) \int_{-\infty}^{\infty}d\epsilon\; \epsilon f'(\epsilon)
- (\Phi''(\mu)/2) \int_{-\infty}^{\infty}d\epsilon\; \epsilon^2 f'(\epsilon)
+ \dots
\end{align}

* Recordemos que $f(\epsilon)=(1+e^{\beta \epsilon})^{-1}$. Si hacemos un cambio de variables $x=\beta \epsilon$, aparecen automáticamente las potencias de $k_B T$ acompañadas de prefactores adimensionales que se pueden determinar.

* Como $f'(\epsilon)$ es una función par, todas las integrales $\int_{-\infty}^{\infty} d\epsilon\; \epsilon^p f'(\epsilon) = 0$ si $p$ es impar. Entonces solo sobreviven los términos con derivadas pares de $\Phi$.

\begin{align}
{\cal I}&=
-\Phi(\mu)\int_{-\infty}^{\infty}d\epsilon f'(\epsilon)
- (\Phi''(\mu)/2) \int_{-\infty}^{\infty}d\epsilon\; \epsilon^2 f'(\epsilon)
+ \dots
\end{align}

* Hacemos ahora un cambio de variables, $x=\beta \epsilon$, y hacemos $g(x)=f(\epsilon)=1/(e^x+1)$

\begin{align}
{\cal I}&=
-\Phi(\mu) \int_{-\infty}^{\infty}dx g'(\epsilon)
- (\Phi''(\mu)/2) \beta^{-2} \int_{-\infty}^{\infty}dx\; x^2 g'(x)
+ \dots
\end{align}

* Todas las potencias de $T$ que aparecen son pares, asi como el orden de las derivadas de $\Phi$.

* Las integrales en $x$ dan números que se pueden calcular, dando lugar así a la expansión de Sommerfeld.

> El desarrollo de Sommerfeld nos da así una aproximación a muchas propiedades que se escriben convolucionando con la función de Fermi, es decir pesando por el número de ocupación de cada estado de una partícula, en función de $\mu$, $V$ y $T$, como es de esperar en el ensemble gran-canónico.

### ¿Cuándo usar el desarrollo de Sommerfeld?

Lo pueden usar siempre y cuando $\phi(\mu)$ sea una función analítica. En el problema del semiconductor de la guía con la energía de Fermi en el gap, $\phi(\mu)=0$ a $T=0$, así como todas sus derivadas, de modo que se me anulan todos los términos del desarrollo. No aporta nada más que la solución a $T=0$, sin correcciones.

En cambio, en el caso que acabamos de analizar, donde $\epsilon_F$ queda en medio de la banda, donde $\phi(\mu)$ es continua, se dan las condiciones perfectas para usar desarrollo de Sommerfeld.



## Número de partículas para $T\ll T_F$ a $\mu$, $V$ fijos

Así como está escrita la receta del desarrollo más arriba puede estar difícil despejar $\mu$ en función de por ejemplo $N$ y $T$, pero podemos usar el truco perturbativo de usar $\mu \sim \epsilon_F$ en las potencias como primer aproximación en $N$ y así ir "invirtiendo la serie".

Como ejemplo, veamos el número de partículas.

\begin{align}
N = \int_{-\infty}^\infty d\epsilon\; g(\epsilon) \Theta(\epsilon) f(\epsilon-\mu)
\end{align}

Identificando $\phi(\epsilon)=g(\epsilon)\Theta(\epsilon)$, y usando que $g(\epsilon)=A V \sqrt{\epsilon}$, nos queda

\begin{align}
N &= \int_{0}^\mu d\epsilon\;
g(\epsilon)  
+ \frac{\pi^2}{6}(k_B T)^2 g'(\mu) + \dots \\
&= \int_{0}^\mu d\epsilon\;
A V \epsilon^{1/2}
+ \frac{\pi^2}{6}(k_B T)^2 \frac{AV}{2\mu^{1/2}} + \dots \\
&= (2 A V/3) \mu^{3/2}
+ \frac{\pi^2}{6}(k_B T)^2 \frac{AV}{2\mu^{1/2}} + \dots
\end{align}

* A orden cero es lo esperado $N=(2 A V/3) \epsilon_F^{3/2}$ por definición de $\epsilon_F$ a $T=0$.

Más adelante usaremos esto para ver como cambia $\mu$ con $T$.


> Es importante enfatizar que $f(\epsilon-\mu) \equiv \langle n_{{\bf k},\alpha}\rangle$ es la ocupación de *un* estado. **¡No de un nivel de energía!** De hecho, la ocupación media de un nivel de energía $\in [\epsilon,\epsilon+d\epsilon]$ sería $g(\epsilon) f(\epsilon-\mu) d\epsilon$, siendo $f(\epsilon-\mu)$ la ocupación media de al menos **uno** de los estados con energía $\epsilon$.




## Corrección al Potencial químico para $T\ll T_F$, a $N$, $V$ fijos

Calculemos ahora las correcciones a $\mu$ a bajas $T$.

Hay dos formas de encararlo. Una es calcular $\mu = k_B T \ln z$ reemplazando el desarrollo (7.45) de arriba, de $\ln z$ a $T$ finita pero chica....
La otra es usar el desarrollo más general de Sommerfeld, como hacemos a continuación.

Recordemos el desarrollo de Sommerfeld

\begin{align}
\boxed{
\int_{-\infty}^\infty d\epsilon f(\epsilon-\mu) \phi(\epsilon)
\approx \int_{-\infty}^\mu d\epsilon  \phi(\epsilon) + \frac{\pi^2}{6}(k_B T)^2 \phi'(\mu) + \frac{7\pi^4}{360}(k_B T)^4 \phi'''(\mu)+\dots
}
\end{align}

Si ponemos $\mu = ϵ_F + \delta \mu$ en Sommerfeld, identificando $\phi(\epsilon)=g(\epsilon)$ nos queda
\begin{align}
N = \int_{0}^\infty g(\epsilon)d\epsilon + g(ϵ_F) \delta \mu  
+\frac{\pi^2}{6}(k_B T)^2 g'(\epsilon_F) + \dots
\end{align}

de donde sale que

\begin{align}
\delta \mu = - \frac{\pi^2}{6}(k_B T)^2 \frac{g'(\epsilon_F)}{g(\epsilon_F)} + {\cal O}(T^4)
\end{align}

Usando que $g(\epsilon)\sim \epsilon^{1/2}$ (el prefactor proporcional al volumen es irrelevante porque aparece multiplicando y dividiendo), llegamos a

\begin{align}
\mu \approx \epsilon_F - \frac{\pi^2}{12} \frac{(k_B T)^2}{\epsilon_F}
= \epsilon_F - \frac{\pi^2}{12} (T/T_F)^2
\end{align}

donde recordemos $\epsilon_F = \frac{\hbar^2}{2m} (3\pi^2 n)^{2/3}$ para la relación de dispersión y $d=3$.

$N = \int_{0}^\infty d\epsilon g(\epsilon)f(\epsilon-\mu)$ aumentaría con $T$, porque $f$ se ensancha y ocupa más estados por encima que los que desocupa por debajo de $\mu$. Por lo tanto, si queremos que el número medio total de partículas no cambie, tenemos que asegurarnos que el área "unoccupied" y el área "occupied" en el dibujo de abajo, sean iguales. Por el comportamiento de $g(\epsilon)$ es evidente que si dejo $\mu=\epsilon_F$ aumentaría $N$. Para evitar esto, tenemos que mover $\mu$ hacia la izquierda de $\epsilon_F$, como resulta del desarrollo de Sommerfeld.

## Energía interna y calor específico para $T \ll T_F$, a $N$, $V$ fijos

Tenemos que calcular

\begin{align}
E = \int_{-\infty}^\infty d\epsilon f(\epsilon-\mu) g(\epsilon)\epsilon
\end{align}

Recordemos nuevamente el desarrollo de Sommerfeld

\begin{align}
\int_{-\infty}^\infty d\epsilon f(\epsilon-\mu) \phi(\epsilon)
\approx \int_{-\infty}^\mu d\epsilon  \phi(\epsilon) + \frac{\pi^2}{6}(k_B T)^2 \phi'(\mu) + \frac{7\pi^4}{360}(k_B T)^4 \phi'''(\mu)+\dots
\end{align}

de donde podemos identificar $\phi(\epsilon)=g(\epsilon)\epsilon$ (ligeramente diferente del cálculo de $N$ por el $\epsilon$ multiplicando).
Entonces

\begin{align}
E &\approx \int_{-\infty}^\mu d\epsilon  g(\epsilon)\epsilon + \frac{\pi^2}{6}(k_B T)^2 \frac{d}{d\epsilon} [g(\epsilon)\epsilon]_\mu + \frac{7\pi^4}{360}(k_B T)^4 \frac{d^3}{d\epsilon^3} [g(\epsilon)\epsilon]_\mu+\dots \\
&= E_0(\mu) + \frac{\pi^2}{6}(k_B T)^2 [g'(\mu)\mu + g(\mu)] + \dots
\end{align}

Pero como nos gustaría fijar $N/V$ en vez de $\mu$ podemos escribir $\mu=\epsilon_F+\delta \mu$ y nos queda, al mas bajo orden (recordar que $\delta \mu \sim T^2$ asi que podemos truncar en todos los lugares donde aparece)

\begin{align}
E &\approx  E_0 + g(\epsilon_F)\delta \mu + \frac{\pi^2}{6}(k_B T)^2 [g'(\epsilon_F)\epsilon_F + g(\epsilon_F)] + \dots
\end{align}

donde $E_0=(3/5) N ϵ_F=(3/5) n V ϵ_F$ es la energía del estado fundamental.
Cuando $g(\epsilon)=A V \epsilon^{1/2}$, $g'(\epsilon)\epsilon=A V \epsilon^{1/2}/2=g(\epsilon)/2$, y usando que
$\delta \mu \approx - \frac{\pi^2}{12} \frac{(k_B T)^2}{\epsilon_F}$
queda

\begin{align}
E &\approx  E_0 + \frac{\pi^2}{3}(k_B T)^2 g(\epsilon_F) + \dots \\
C_V &= \frac{2 \pi^2}{3}k_B^2 T g(\epsilon_F) + \dots
\end{align}

Particularizando $g(\epsilon)$ para el caso no relativista en $d=3$ nos queda finalmente (ejercicio para el lector!)

\begin{align}
\frac{E}{V} &\approx \frac{3}{5} \frac{N}{V} k_B T_F \left[ 1 +\frac{5}{12} \left(\frac{T}{T_F} \right)^2  + \dots \right] \\
&= \frac{\pi^2}{2} \left( \frac{k_B T}{\epsilon_F} \right)n k_B + \dots
= \frac{\pi^2}{2} \left( \frac{T}{T_F} \right)n k_B + \dots
\end{align}

## Calor específico para $T\ll T_F$ a $N$ y $V$ fijos

Entonces, usando

\begin{align}
\frac{E}{V} &\approx \frac{3}{5} \frac{N}{V} k_B T_F \left[ 1 +  \frac{5}{12}\pi^2 \left(\frac{T}{T_F} \right)^2  + \dots \right]
= const + \frac{3*5 }{5*12} \frac{N}{V} k_B T_F \left(\frac{T}{T_F} \right)^2
= const + \frac{1}{4} \frac{N}{V} k_B T_F \left(\frac{T}{T_F} \right)^2
\end{align}

tenemos la capacidad calorífica al mas bajo orden

\begin{align}
C_{V,N} \approx
\left( \partial_T E \right)_{V,N} = \frac{\pi^2}{2} N k_B \left( \frac{T}{T_F}\right) + {\cal O}[(T/T_F)^3]
\end{align}




* Vemos ya que el calor específico de un gas de Fermiones tiene que is lineal, $C_V \sim T$.

Bueno, el estado fundamental es el mar de Fermi lleno, correspondiente a una función escalón para $f(\epsilon)$ que salta de 1 a 0 exactamente en $\epsilon_F=\mu$ y a un número de fermiones medio que $N\sim \epsilon_F^{3/2}$. A esa distancia hay
$g(\epsilon_F)k_B T \sim V \epsilon_F^{3/2} (k_B T/\epsilon_F)\sim N (T/T_F)$. Como cada uno gana $k_B T$, la energía interna aumenta en $\Delta E \sim N(T/T_F)k_B T$, explicando el comportamiento lineal de $C_V$.



## Presión para $T \ll T_F$





La presión es proporcional a $E/V$, así que usamos los resultados anteriores para el desarrollo de Sommerfeld de P. En particular $PV = (2/3)E$ para el gas no relativista y podemos usar el desarrollo anterior de $E$.

> Para otros casos también vale $PV \propto E$. Es una relación que se usa mucho, y es más general que el teorema de equipartición, vale para gases degenerados de fermiones o bosones.

**Observaciones**:

* La presión de Fermi es interesante, porque muestra que a diferencia del caso clásico la presión no cae a cero en $T\to 0$ ¡sino que mantiene un valor finito! (esto estabiliza algún tipo de estrellas, llamadas enanas blancas).

* Vamos a ver que la presión del estado fundamental de bosones en cambio tiende a cero cuando $T\to 0$.

* Como la densidad de energía es proporcional a la presión, esto también significa que la energía es finita y positiva en el fundamental.

* Si uno pudiera dibujar la presión vs $T$, vería algo como lo de arriba. Empezamos con una presión finita, y luego se va a una función lineal en $T$, ya que debe tender a $P = n k_B T$.

* La presión de Fermi ¡salva a ciertas estrellas con alta densidad de fermiones (electrones o neutrones) de convertirse en agujeros negros! También afecta y estabiliza a los átomos de muchos electrones.

#### Paramagnetismo de Pauli

Como los electrones tienen espín $s=1/2$, también tienen un momento magnético, $\vec \mu \approx g \mu_0 \vec S/\hbar$. Su proyección puede escribirse $\mu_z=-\mu_B \sigma$, con $\sigma=\pm 1$, las dos proyecciones de espín, y donde el signo menos aparece de considerar que el electrón tiene carga negativa.
El momento magnético puede acoplarse con un campo magnético externo $H$, que podemos asumir está en la dirección $\hat z$ también, en lo que se denomina *acoplamiento Zeeman*.
Este acoplamiento origina un comportamiento paramagnético, y la estadística de Fermi-Dirac modifica el comportamiento magnético esperado clasicamente.

Considerar electrones no interactuantes en presencia de un campo magnético $H$ en la dirección $\hat z$. En ausencia de acoplamiento espín-órbita (que acopla el momento angular con el espín), la relación de dispersión es
\begin{align}
    {\cal E}({\vec k},\sigma) = \frac{\hbar^2 k^2}{2m}
    + \mu_B H \sigma
\end{align}
con $\sigma=\pm 1$, correspondiente a las dos proyecciones posibles para $S_z=(\hbar/2)\sigma$ para partículas de espín $1/2$, y donde $\mu_B$ es el magnetón de Bohr que relaciona $S_z$ con la correspondiente proyección del momento magnético del electrón.

Usando el desarrollo de Sommerfeld se puede obtener la susceptibilidad del gas de electrones. La idea central es darse cuenta de que podemos separar los electrones de distinto espín en dos especies, pero con una energía de Fermi que depende de $H$.

Para darse cuenta de donde viene el paramagnetismo, ver la figura.
Lo que sucede al prender $B$ es que desaparece la degeneración de espín y se corren los niveles de energía según espín. Si conservamos el número de espines con cada proyección a $T=0$, llegamos a la contradicción de no estar en el estado de más baja energía por un lado, y que la población electrónica no esta descripta por la función escalón de Fermi con una $\epsilon_F$ bien definida. Necesitamos entonces invertir la proyección de algunos electrones desalineados con el campo, de modo de rellenar y llegar a una energía de Fermi común. Esto da lugar a una magnetización, pues hay más espines alineados que no alineados, y es una respuesta paramagnética, llamada paramagnetismo de Pauli.


En un ejercicio de la guía van a ver que la suceptibilidad a baja temperatura es
\begin{align}
\boxed{
\chi = \mu_B^2 g(\epsilon_F)
}
\end{align}
de modo que nuevamente la densidad de estados a la energía de Fermi controla la coas.


Este resultado que contrasta mucho con la de alta temperatura, la *susceptibilidad de Curie*
\begin{align}
\chi = N \mu_B^2 / k_B T
\end{align}
Interesantemente sin embargo, si usamos la expresión clásica reemplazando $N$ por el número efectivo de electrones $N(T/T_F)$, reobtenemos el resultado cuántico. Esto se debe a que solo un pequeño número de electrones participa de la excitación con el campo externo. Ver la discusión sobre la capacidad calorífica a bajas  temperaturas.


image.png

El desarrollo de Sommerfeld se puede aplicar sin problemas notando que la función de Fermi es común para los dos espines, pero no así la densidad de estados.
Por ejemplo, para $N$ identificando:
\begin{align}
\phi(\epsilon) &\equiv g_{⇓}(\epsilon,B)-g_{⇑}(\epsilon,B) \\
&= g(\epsilon-\mu_B B)\Theta(\epsilon-\mu_B B)/2-g(\epsilon+\mu_B B)\Theta(\epsilon+\mu_B B)/2
\end{align}
ya podemos aplicarla, y luego hacer la aproximación $\mu_B B \ll \epsilon_F$. Podemos ver así las correcciones de $T$ y $B$ simultáneamente. Si $T=0$ obtenemos el cálculo anterior, puesla magnetización media es $M = \mu_B (N_⇓-N_⇑)$ y
\begin{align}
N_⇓-N_⇑
&=
\int_{-\infty}^{\mu} \phi(\epsilon)d\epsilon + \frac{\pi^2}{6}(k_B T)^2 \phi'(\mu) + \dots \\
&=
\int_{-\mu_B B}^{\mu} g(\epsilon+\mu_B B)d\epsilon/2 -
\int_{\mu_B B}^{\mu} g(\epsilon-\mu_B B)d\epsilon/2
+ \frac{\pi^2}{6}(k_B T)^2 (g'(\mu+\mu_B B)-g'(\mu-\mu_B B)) + \dots \\
&= \int_{\mu-\mu_B B}^{\mu+\mu_B B} g(x)dx/2
+ \frac{\pi^2}{12}(k_B T)^2 (g'(\mu+\mu_B B)-g'(\mu+\mu_B B)) + \dots \\
N_⇓-N_⇑ &\approx \left[g(\mu)
+ \frac{\pi^2}{6}(k_B T)^2 g''(\mu)\right] \mu_B B + \dots
\end{align}
y asi podemos sacar las correcciones por temperatura de la susceptibilidad, que provienen no solo del ensanchamiento de la función de Fermi sino también a través de $\mu$. Ya vemos de arriba que para $T=0$, se reobtiene la susceptibilidad de Pauli, ya que $\mu\to ϵ_F$.

\begin{align}
M = \mu_B (N_⇓-N_⇑)=g(ϵ_F) \mu_B^2 B
\end{align}

Si queremos las correcciones a $T$ finita, ya tenemos la magnetización pero como función de $\mu$, $B$ y $T$. Para tener las correciones a $N$ fijo, tenemos que calcular la dependencia de $\mu$ con $B$ y $T$. Eso lo podemos obtener del cálculo de $N=N_⇓+N_⇑$. Lo que se encuentra son dos correciones negativas que se suman, una que va como $(T/T_F)^2$ como en el caso sin campo, y otra que va como $(\mu_B H/k_B T_F)^2$. Este resultado tiene sentido, ya que debería haber simetría $B \to -B$ en $\mu$. Es interesante notar que aún a $T=0$ hay corrección negativa debida al campo $B$, y se la puede entender notando que si no hubiera corrección a $\mu$, el número medio de electrones aumentaría (ver [notas de Arovas](https://courses.physics.ucsd.edu/2010/Spring/physics210a/LECTURES/210_COURSE.pdf) para mas detalles).

#### Diamagnetismo de Landau

Cabe mencionar que el comportamiento magnético del gas de electrones no interactuantes no solo se debe al espín del electrón. También existe una contribución, pero en este caso diamagnética, debido a su movimiento orbital en presencia de un campo magnético.

Clásicamente uno pensaría que se trata del movimiento circular o más bien hélico que hace una carga con cierta velocidad en un campo magnético. Como tiene carga negativa, esta órbita genera un campo magnético opuesto al aplicado, resultando en una respuesta diamagnética. *Sin embargo*, si uno resuelve la mecánica estadística clásica, con un hamiltoniano que incorpore el campo magnético a partir del potencial vector (recordar mecánica!), no se obtiene ninguna respuesta diamagnética! Resultado que se conoce como *teorema de Bohr-van Leeuwen*.


El problema debe resolverse con mecánica cuántica, y como es de esperarse, la energía asociada con el movimiento circular queda cuantizada en unidades de $e\hbar B/mc$, y también aparece una energía de punto cero. El movimiento en la vertical de la hélice, en $\hat z$ por ejemplo, también se cuantiza,  pero los gaps son tan pequeños que puede considerarse clásicamente, es decir que es lícito sumar estados así $\int dz p_z/h$.
La relación de dispersión es

\begin{align}
\epsilon_{j,p_z}=\frac{e\hbar B}{mc}(j+1/2)+\frac{p_z^2}{2m}
\end{align}

Lo que se hace es plantear la gran función de partición, permitiendo solo $n=0,1$ para cada estado $(j,p_z)$ y luego en la energía libre sumando sobre todos los posibles valores de $j$ e integrando en $p_z$.

Considerando mecánica cuántica, si aparece una respuesta diamagnética

\begin{align}
M &= -N \mu_{ef} \left( \coth(\beta \mu_{ef} B)
-\frac{1}{\beta \mu_{ef} B}\right) \\
\mu_{ef}&=\frac{eh}{4\pi m c}
\end{align}


donde la función entre paréntesis se parece bastante a una $\tanh(x)$, pero es distinta, claro...

¡Notar que todo desaparece cuando $h \to 0$!, consistente con el *teorema de Bohr-van Leeuwen*. Este basicamente dice que

\begin{align}
H &= \sum_{i=1}^N \frac{({\vec p}_i-e{\vec A})^2}{2m} + U({\vec q}_1,\dots, {\vec q}_N) \\
{\vec B} &= \nabla \times {\vec A}
\end{align}

da $Z({\vec B})=Z({\vec 0})$, ya que la transformación canónica ${\vec p}'_i=({\vec p}_i-e{\vec A})$, ${\vec q}'_i={\vec q}_i$ elimina ${\vec A}$.

y por lo tanto el problema es independiente de ${\vec B}$.
¡La cuantización de los niveles de energía cambia la cosa!

(ver [notas de Arovas](https://courses.physics.ucsd.edu/2010/Spring/physics210a/LECTURES/210_COURSE.pdf) para mas detalles).

#### Conductividad eléctrica

La conductividad eléctrica es la respuesta del gas de electrones a un campo eléctrico pequeño. Es una propiedad dinámica, para la cual necesitamos saber algo sobre la estadística cerca del equilibrio. Es crucial, para derivar la ley de Ohm, meter algún mecanismo de disipación que impida que los electrones se aceleren continuamente. La red cristalina no ayuda, pues se puede mostrar que no contribuye al calor de Joule. Es necesario introducir interacciones de los electrones o más bien dispersión o scattering (la red periódica no lo consigue pese a la alta densidad de núcleos). Por ejemplo interacciones electrón-fonón (mas importante), o electrón-electrón. Pero tampoco es suficiente electrón-fonón porque depende de $T$ y no explica lo que sucede a $T$ baja. Para explicarlo en general tenemos que tener en cuenta la interacción con impurezas o defectos. Ya lo verán en materias de teoría de la materia condensada.


Solo para dar una idea, las siguientes son expresiones aproximadas para la conductividad eléctrica suponiendo un material cristalino isótropo(estrictamente la conductividad es un tensor). No la vamos a derivar aquí, pero lo harán en introducción a la teoría de los sólidos o de la materia condensada.


En las expresiones $e$ es la carga del electrón, $\tau({\bf k})$ es un tiempo característico de relajación, $v_x({\bf k})=\partial_{k_x}\epsilon_{{\bf k},\alpha}$ la velocidad de grupo en la dirección $x$ y $f$ la función de Dirac, que ya conocemos bien. La primera expresión es una suma sobre estados, la cual a menudo puede aproximarse por una integral sobre energías de una partícula, en la segunda expresión.

### Enanas blancas y agujeros negros

La primer aplicación de la estadística de Fermi no fue para los electrones en un metal, ¡sino para explicar el efecto de los electrones en una estrella! Fowler, en 1926, el mismo año que Fermi y Dirac presentaron la estadística...

* Las enanas blancas son estrellas viejas y blancas, pero de intensidad muy baja.

* Ya consumieron casi todo el hidrógeno necesario para hacer reacciones nucleares en su interior, y se quedaron con un montón de helio.

* Un modelo idealizado de enana blanca es una masa de M(∼ $10^{33}$ g) de helio empaquedato en una bola con densidad de masa ρ(∼ $10^7$ g cm^−3 ), a una temperatura central  T (∼ $10^7$ K).
A esa $T$, todo el helio está casi completamente ionizado.

* La estrella entonces puede pensarse como un conjunto de $N$ electrones de masa $m$ y $N/2$ núcleos de helio (dos protones y dos neutrones, asi que su masa es de $\sim 4m_p$, ya que el neutrón solo es un $\%1$ mas pesado), así que la masa total es
$$M=N(m+2 m_p)\approx 2 N m_p$$

* La densidad de electrones es
$$n = N/V \approx \frac{(M/2m_p)}{M/\rho} \sim {\cal O}(10^{30}) cm^{-3}$$. Comparar con la densidad de electrones en un sólido, donde hay unas decenas de electrones por cada $(10^{-8} cm)^3 \sim 10^{-24} cm^{3}$...

* El momento de Fermi $p_F = \hbar k_F$ se puede estimar a partir de $n$, y usando $g=(2s+1)=2$, como
$$p_F = \left( \frac{3n}{8\pi} \right)^{1/3} h = {\cal O}(10^{-17}) \text{g cm s}^{-1}$$
y es del orden de $mc$, o sea que la energía de Fermi es del orden de la energía en reposo $mc^2$, indicando que el problema es relativista. Y si uno calcula $ϵ_F \sim 10^6 \text{eV}$ y $T_F\sim 10^{10} K$ y vemos que $T \ll T_F$, asi que esta totalmente degenerado...

* Visto y considerando, es lícito considerar a los electrones como relativistas, y usar el límite de la estadística de FD para $T\to 0$. Es decir, nos podemos quedar con el estado fundamental.

* ¿y los núcleos de helio? Se puede ver que no contribuyen apreciablemente al menos a la presión, que es lo que más nos interesa ahora...
