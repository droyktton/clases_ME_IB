- Con **partículas idénticas**, las distintas formas incluyen sus **permutaciones** → el estado cuántico debe ser simétrico (bosones) o antisimétrico (fermiones).  

- Ejemplo: dispersión de dos electrones idénticos:  
  $$
  P_1 = |f(\theta) + e^{i\delta} f(\pi-\theta)|^2
  $$  
  mientras que para partículas distinguibles:  
  $$
  P_1 = |f(\theta)|^2 + |f(\pi-\theta)|^2.
  $$  

- La fase de intercambio solo puede ser $\delta=0$ (bosones) o $\delta=\pi$ (fermiones).


Es decir, *lo que se mide en física cuántica es distinto si las partículas son o no idénticas*, y si son idénticas, el resultado a su vez ¡depende de si éstas son fermionicas o bosónicas!

Esto no se limita a colisiones, sino a muchísimos otros fenómenos a distintas escalas, incluso en el límite termodinámico. La simetría o degeneración de intercambio es una propiedad fundamental de la física, con profundas implicancias en particular en mecánica estadística.


## Partículas idénticas

- **Dos partículas son idénticas si sus propiedades intrínsecas son las mismas**  
  (masa, carga, espín, etc.).  
  Ejemplo: todos los electrones son idénticos entre sí, lo mismo que los protones o los muones.

- La identidad no requiere que sean elementales:  
  **partículas compuestas también pueden ser idénticas** (ej. protones hechos de quarks, o átomos de hidrógeno o moléculas).

- Partículas idénticas **pueden ocupar estados distintos** → no confundir identidad con “estar en el mismo estado”.

- **Principio clave:** intercambiar dos partículas idénticas **no puede cambiar la física**.  

---

### Clásico vs Cuántico

- **Mecánica clásica**:  
  Las partículas idénticas son **distinguibles en principio**.  
  Un microestado clásico especifica posición y momento de cada una, y se podría “seguir” a una partícula en particular.  
  El factor $1/N!$ se introduce “a mano” porque las propiedades macroscópicas no dependen de esa identidad → así se resuelve la paradoja de Gibbs.

- **Mecánica cuántica**:  
  Las partículas idénticas son **indistinguibles por naturaleza**.  
  En un microestado cuántico no podemos asignar de manera consistente “qué partícula está en qué estado”.  
  Los procesos de intercambio hacen que la identidad individual se pierda intrínsecamente.  





### Partículas idénticas: clásico vs cuántico

| Aspecto | Mecánica clásica | Mecánica cuántica |
|---------|------------------|-------------------|
| Identidad | Partículas idénticas **son distinguibles en principio** | Partículas idénticas **son indistinguibles por naturaleza** |
| Microestado | Se especifica posición y momento de **cada partícula** | El estado cuántico no permite asignar “qué partícula está en qué estado” |
| Intercambio | Intercambiar partículas no cambia la física, pero **se podrían seguir individualmente** | Intercambiar partículas **cambia la simetría del estado** (bosones: simétrico, fermiones: antisimétrico) |
| Factor $1/N!$ | Se introduce “a mano” para evitar la paradoja de Gibbs | Surge automáticamente del principio de indistinguibilidad |
| Ejemplo | Gas clásico de moléculas idénticas | Gas de electrones, protones, fotones, etc. |


## Postulado de simetrización

En principio, la función de onda de dos particulas $\Psi(x_1,x_2)$ me da la probabilidad $|\Psi(x_1,x_2)|^2$ de encontrar la primera en $x_1$ y la segunda en $x_2$. Es algo que podemos medir.

Pero porque no podemos distinguirlas, necesariamente

$$
|\Psi(x_1,x_2)|^2 = |\Psi(x_2,x_1)|^2
$$

y esto implica que las función de onda permutada

$$\hat P \Psi(x_1,x_2)=\Psi(x_2,x_1)$$

difiere de la original en un factor de fase global $\phi$,

$$
\Psi(x_1,x_2)= e^{i\phi} \Psi(x_2,x_1)
$$

Pero, si permuto de nuevo (notar que es la misma operación aplicada de nuevo),

$$\hat P \hat P \Psi(x_1,x_2)=\Psi(x_1,x_2)$$

no nos queda otra que $e^{2i\phi}=1$, entonces $\phi=0,\pi$, asi que

$$
\Psi(x_1,x_2)= \pm \Psi(x_2,x_1)
$$

o lo que es lo mismo permutar dos veces las mismas partículas es como no hacer nada, $\hat P^2 = 1$.

Esta simetría/antisimetría de intercambio es la condición que tienen que cumplir *todos los estados físicos permitidos de un sistema de dos partículas idénticas*. Es un espacio de Hilbert *más restringido* que el simple producto tensorial de espacios de Hilbert de una partícula.  
Este es el **postulado de simetrización**.



## Como construir una base simétrica/antisimétrica

* Una forma simple de construir estados asimétricos es superponer los anteriores adecuadamente
$$|k_1 \dots k_N \rangle_{-} = \frac{1}{\sqrt{N_{-}}} \sum_P (-1)^P P|k_1 \dots k_N \rangle
$$
donde $N_{-}$ cuenta el número de permutaciones posibles.
parece complicado, pero es facil ver que ocurre con dos partículas,
$$
| k_1, k_2 \rangle_{-}=
\frac{| k_1, k_2 \rangle - | k_2, k_1 \rangle}{\sqrt{2}}
$$
de donde es evidente que si permuto un par cambia de signo. *Notar que si dos partículas están en el mismo estado, no hay estado posible. Los fermiones no pueden estar en el mismo estado, resultados que se conoce como principio de exclusión de Fermi*. Esto implica que $N_{-}=N!$ exactamente, ya que muchas permutaciones se anulan. O sea que $|k_1 \dots k_N \rangle$ no puede tener estados repetidos, y por lo tanto solo $N!$ permutaciones de estados distintos me dan estados asimetrizados válidos.

* Análogamente, si quiero construirme estados simétricos de $N$ partículas,
$$|k_1 \dots k_N \rangle_{+} = \frac{1}{\sqrt{N_+}} \sum_P P|k_1 \dots k_N \rangle
$$
  donde ${N_+}=N! \prod_k n_k!$ con $n_k$ el número de veces que se repite el estado $k$. El número de permutaciones es más grande porque el estado $|k_1 \dots k_N \rangle$ puede contener repeticiones de estados. Notar que mientras los fermiones no pueden compartir estado, los bosones si pueden.  


Veamos con dos partículas
$$
| k_1, k_2 \rangle_{+}=
\frac{| k_1, k_2 \rangle + | k_2, k_1 \rangle}{\sqrt{2}}
$$
El cual es evidentemente simétrico.
La normalización es $\sqrt{N_+}=
\sqrt{2}$, que es $N!$ si $k_1 \neq k_2$. Si se repitieran sería $\sqrt{2!1!0!}=\sqrt{2}$, lo que es correcto.

## Bases de autoestados de 2 partículas en una caja unidimensional

Vamos a ignorar por el momento los números cuánticos de espín que tienen todos los fermiones, y casi todos los bosones salvo aquellos con espín zero.

$$\hat H = \frac{\hat p_1^2}{2m} + \frac{\hat p_2^2}{2m}$$

Las autofunciones

$$\hat H \psi(x_1,x_2) = E \psi(x_1,x_2)$$

las podemos construir por analogía al caso bidimensional, $x_1 \sim x$, $x_2 \sim y$,  en una caja con $L_1=L_2=L$ pues el problema matemático es el mismo.

$$\frac{\hbar^2 (\partial^2_{x_1}+\partial^2_{x_2})}{2m}
\psi(x_1,x_2) = E \psi(x_1,x_2)
$$

por lo tanto las soluciones son

$$\psi_{n_1,n_2}(x_1,x_2)=\sqrt{\frac{4}{L^2}}\sin(n_1 \pi x_1/L)\sin(n_2 \pi x_2/L)$$

con

$$E_{n_1,n_2} = \frac{\hbar^2}{2m}\frac{\pi^2}{L^2}[n_1^2+n_2^2]$$

Sin embargo, esta $\psi_{n_1,n_2}(x_1,x_2)$ es aceptable físicamente si las partículas no son idénticas. Si son idénticas debemos construir una función simetrizada o antisimetrizada (vamos a ignorar el espín de las partículas por ahora). Es decir,

* Bosones idénticos:

$$
\psi^B_{n_1,n_2}(x_1,x_2)=\frac{1}{\sqrt{2(1+\delta_{n_1,n_2})}}[\psi_{n_1,n_2}(x_1,x_2)+\psi_{n_1,n_2}(x_2,x_1)]
$$

* Fermiones idénticos:

$$
\psi^F_{n_1,n_2}(x_1,x_2)=\frac{1}{\sqrt{2}}[\psi_{n_1,n_2}(x_1,x_2)-\psi_{n_1,n_2}(x_2,x_1)]
$$

* Partículas distintas (un bosón y un fermión, dos bosones distintos, dos fermiones distintos):

$$
\psi_{n_1,n_2}(x_1,x_2)
$$

Y todas ellas son autofunciones del hamiltoniano de las dos partículas, pues son combinaciones de la última.

El espacio de Hilbert de bosones se puede describir por la base de todos los $\psi^B_{n_1,n_2}$, la de fermiones por la base de todos los $\psi^F_{n_1,n_2}$, y la de partículas no idénticas (pero con igual masa) por $\psi_{n_1,n_2}(x_1,x_2)$.

Es decir, cualquier estado bosónico arbitrario está descripto por

$$\Psi^B(x_1,x_2) = \sum_{n_1,n_2} c_{n_1,n_2} \psi^B_{n_1,n_2}(x_1,x_2)
$$

cualquier estado fermiónico arbitrario por

$$\Psi^F(x_1,x_2) = \sum_{n_1,n_2} c_{n_1,n_2} \psi^F_{n_1,n_2}(x_1,x_2)
$$

y cualquier estado arbitrario de partículas distinguibles,

$$
\Psi(x_1,x_2) = \sum_{n_1,n_2} c_{n_1,n_2} \psi_{n_1,n_2}(x_1,x_2)
$$


Hay técnicas experimentales para medir $|\psi(x,y)|^2$ simultáneamente, de dos tipos:

* Con microscopía o dispersión de luz u otras partículas.
* Abriendo la caja, detectando las partículas afuera e infiriendo donde estaban en la caja.

Para armarse el mapa $P(x_1,x_2)=|\Psi(x,y)|^2$ necesitamos repetir muchas veces el experimento, desde las mismas condiciones iniciales.

¡Hagamos la predicción de lo que medirían!


```python
#@title Funcion de onda estacionaria de 2 particulas en una caja unidimensional
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Parámetros de la caja
L = 1.0
N = 200

# Funciones de onda de una partícula
def phi(k, x, L=1.0):
    return np.sqrt(2/L) * np.sin(k * np.pi * x / L)

# Estado de dos partículas
def psi_2p(x1, x2, k1, k2, particle_type="boson"):
    psi = phi(k1, x1) * phi(k2, x2)
    psi_perm = phi(k2, x1) * phi(k1, x2)

    if particle_type == "boson":
        return (psi + psi_perm) / np.sqrt(2)
    elif particle_type == "fermion":
        return (psi - psi_perm) / np.sqrt(2)
    elif particle_type == "distinguishable":
        return psi  # sin simetrización
    else:
        raise ValueError("particle_type debe ser 'boson', 'fermion' o 'distinguishable'")

# Mallado
x = np.linspace(0, L, N)
X1, X2 = np.meshgrid(x, x)

# Función interactiva
@interact(
    k1=widgets.IntSlider(min=1, max=5, step=1, value=1, description="k1"),
    k2=widgets.IntSlider(min=1, max=5, step=1, value=2, description="k2"),
    particle_type=widgets.Dropdown(
        options=["boson", "fermion", "distinguishable"],
        value="boson",
        description="Partículas"
    )
)
def plot_wavefunction(k1, k2, particle_type):
    Psi = psi_2p(X1, X2, k1, k2, particle_type)
    Prob = np.abs(Psi)**2

    plt.figure(figsize=(6,5))
    plt.pcolormesh(X1, X2, Prob, shading="auto", cmap="viridis")
    plt.colorbar(label=r"$|\Psi(x_1,x_2)|^2$")
    plt.xlabel(r"$x_1$")
    plt.ylabel(r"$x_2$")
    plt.title(f"Autoestado de 2 {particle_type}es, k1={k1}, k2={k2}")
    plt.show()
```

**¡Atencion!** *El plot anterior no es una función de onda
estacionaria de una partícula en dos dimensiones, es la función de onda estacionaria de dos partículas en una dimensión*

## Una mezcla arbitraria

Para el caso más general, donde el sistema esta en cualquier estado puro, debemos combinar distintos autoestados.

Examinemos que pasa si hacemos esta familia de combinaciones parametrizadas por una variable $\theta$,

\begin{align}
\Psi^B(x_1,x_1) =
\left[
\cos(\theta) \psi_{k_1}(x_1)\psi_{k_2}(x_2) +
\sin(\theta) \psi_{k_3}(x_1)\psi_{k_4}(x_2)  
\right]/{\cal N} \\
+
\left[
\cos(\theta) \psi_{k_1}(x_2)\psi_{k_2}(x_1) +
\sin(\theta) \psi_{k_3}(x_2)\psi_{k_4}(x_1)  
\right]/{\cal N}
\end{align}


para bosones,

\begin{align}
\Psi^F(x_1,x_1) =
\left[
\cos(\theta) \psi_{k_1}(x_1)\psi_{k_2}(x_2) +
\sin(\theta) \psi_{k_3}(x_1)\psi_{k_4}(x_2)  
\right]/{\cal N} \\
-
\left[
\cos(\theta) \psi_{k_1}(x_2)\psi_{k_2}(x_1) +
\sin(\theta) \psi_{k_3}(x_2)\psi_{k_4}(x_1)  
\right]/{\cal N}
\end{align}


para fermiones, y

$$\Psi(x_1,x_1) =
\left[
\cos(\theta) \psi_{k_1}(x_1)\psi_{k_2}(x_2) +
\sin(\theta) \psi_{k_3}(x_1)\psi_{k_4}(x_2)  
\right]
$$

para distinguibles, con $\theta$ un ángulo arbitrario.
La normalización ${\cal N}=\sqrt{2}$ solo si $(k_1,k_2) \neq (k_3,k_4)$, sino es ${\cal N}=\sqrt{|\cos(\theta)+\sin(\theta)|}$.



```python
#@title Ejemplo de Funcion de onda arbitraria de 2 particulas en una caja unidimensional
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Parámetros de la caja
L = 1.0
N = 200

# Funciones de onda de una partícula
def phi(k, x, L=1.0):
    return np.sqrt(2/L) * np.sin(k * np.pi * x / L)

# Estado de dos partículas
def psi_2p(x1, x2, k1, k2, k3, k4, ang, particle_type="boson"):

    #random_angle_radians = np.random.uniform(0, 2 * np.pi)
    #c = np.cos(random_angle_radians)
    #s = np.sin(random_angle_radians)
    c = np.cos(ang*np.pi/180.)
    s = np.sin(ang*np.pi/180.)

    psi = (c*phi(k1, x1) * phi(k2, x2) + s*phi(k3, x1) * phi(k4, x2))/np.sqrt(2)
    psi_perm = (c*phi(k1, x2) * phi(k2, x1) + s*phi(k3, x2) * phi(k4, x1))/np.sqrt(2)

    if particle_type == "boson":
        return (psi + psi_perm) / np.sqrt(2)
    elif particle_type == "fermion":
        return (psi - psi_perm) / np.sqrt(2)
    elif particle_type == "distinguishable":
        return psi  # sin simetrización
    else:
        raise ValueError("particle_type debe ser 'boson', 'fermion' o 'distinguishable'")

# Mallado
x = np.linspace(0, L, N)
X1, X2 = np.meshgrid(x, x)

# Función interactiva
@interact(
    k1=widgets.IntSlider(min=1, max=5, step=1, value=1, description="k1"),
    k2=widgets.IntSlider(min=1, max=5, step=1, value=2, description="k2"),
    k3=widgets.IntSlider(min=1, max=5, step=1, value=1, description="k3"),
    k4=widgets.IntSlider(min=1, max=5, step=1, value=2, description="k4"),
    ang=widgets.IntSlider(min=0, max=360, step=1, value=0, description="ang"),
    particle_type=widgets.Dropdown(
        options=["boson", "fermion", "distinguishable"],
        value="boson",
        description="Partículas"
    )
)
def plot_wavefunction(k1, k2, k3, k4, ang, particle_type):
    Psi = psi_2p(X1, X2, k1, k2, k3, k4, ang, particle_type)
    Prob = np.abs(Psi)**2

    plt.figure(figsize=(6,5))
    plt.pcolormesh(X1, X2, Prob, shading="auto", cmap="viridis")
    plt.colorbar(label=r"$|\Psi(x_1,x_2)|^2$")
    plt.xlabel(r"$x_1$")
    plt.ylabel(r"$x_2$")
    plt.title(f"Dos {particle_type}, k1={k1}, k2={k2}, k3={k3}, k4={k4}")
    plt.show()
```

**¡Atencion!** *El plot anterior no es una función de onda
arbitraria de una partícula en dos dimensiones, es una función de onda arbitraria de dos partículas en **una** dimensión*

**Muy Importante**

* El **postulado de simetrización** para partículas idénticas debe cumplirse siempre. Por ejemplo, al hacer evolucionar una condición inicial simétrica o antisimétrica según corresponda. Como el Hamiltoniano tiene simetría bajo permutaciones de partículas, la evolución preservará automáticamente esta simetría.

* Este postulado no se limita a microestados cuánticos de partículas idénticas en mecánica estadística, sino que es fundamental para cualquier sistema cuántico con múltiples partículas idénticas. Afecta el cálculo de cualquier observable dinámico, como por ejemplo la **sección eficaz de colisión** de dos partículas idénticas.


## Teorema espín-estadística

Hasta ahora no hemos considerado el **espín** de las partículas. Si no hay campos magnéticos ni interacciones asociadas al espín, este solo introduce un **factor de degeneración** $g = 2s + 1$, donde $s$ es el espín de la partícula. Por ejemplo:  
- $s=1/2$ para electrones, protones, neutrones, quarks y neutrinos.  
- $s=1$ para fotones, bosones $W/Z$ y gluones, mediadores de interacciones.  

El **estado cuántico de una partícula** se describe por:  
1. El número cuántico de traslación $k$ (energía/momento).  
2. La proyección del espín $m = -s, \dots, s$.  

Si se trata de moléculas, puede haber otros números cuánticos (vibración, rotación, excitaciones electrónicas o nucleares), lo que complica la clasificación porque pueden coexistir fermiones y bosones.

El **espín determina la naturaleza de la partícula** según el **teorema espín-estadística**:  
- **Espín semi-entero ($s = 1/2, 3/2, …$) → fermiones** ($\eta=-1$).  
  - Obedecen el **principio de exclusión de Pauli**: no puede haber dos fermiones idénticos en el mismo estado cuántico.  
- **Espín entero ($s = 0, 1, 2, …$) → bosones** ($\eta=1$).  
  - No hay restricción de ocupación: cualquier número de bosones puede compartir el mismo estado, lo que permite fenómenos como la **condensación de Bose-Einstein**.

Ejemplos:  
- **Fermiones compuestos**: protones y neutrones (3 quarks de espín 1/2) → espín 1/2, fermión.  
- **Bosones compuestos**: mesones (2 quarks) → espín entero, bosón.  
- **Cuasipartículas** como fonones o magnones: sin espín intrínseco, pero su polarización puede generar degeneración.  

En resumen, en 3D todas las partículas se clasifican como **fermiones o bosones**, sin excepciones.


## Simetría de intercambio de funciones de onda de muchas partículas
La condición anterior se generaliza para un número $N$ de partículas, ante cualquiera de las $N!$ permutaciones posibles.

En tres dimensiones (en dos dimensiones fueron predichas otras) hay solo dos tipos de partículas en la naturaleza. Si $P$ es el operador de permutación arbitraria de muchas partículas:

* **Bosones:** $P |\psi_S(1,\dots,N) \rangle = +|\psi_S(1,\dots,N) \rangle$. El estado es *simétrico*.

* **Fermiones:** $P |\psi_A(1,\dots,N) \rangle = (-1)^p|\psi_A(1,\dots,N) \rangle$. El estado es *antisimétrico*.

donde $|\psi_{S,A}(1,\dots,N) \rangle$ es un estado de ese espacio de Hilbert simétrizado (simétrico o antisimétrico).  
$p$ es la paridad de la permutación: es el número de pares de partículas que permuté, y siempre cualquier $P$ se puede descomponer en permutaciones de pares.

## resumen

| Espín (s)         | Tipo de partícula | Signo de simetría η | Principio de ocupación / comportamiento |
|------------------|-----------------|------------------|---------------------------------------|
| Semi-entero (1/2, 3/2, …) | Fermión         | -1               | No más de una partícula idéntica por estado (Pauli) |
| Entero (0, 1, 2, …)       | Bosón           | +1               | Cualquier número puede ocupar el mismo estado (condensación posible) |

**Notas importantes:**
- El espín determina la naturaleza cuántica de la partícula.  
- El factor de degeneración de espín es $g = 2s + 1$.  
- Partículas compuestas también siguen el teorema:  
  - Proton/neutrón: fermiones, espín 1/2  
  - Mesones: bosones, espín entero  
- Cuasipartículas (fonones, magnones) no tienen espín intrínseco, pero pueden tener degeneración por polarización.  
- No hay otras posibilidades en 3D: todas las partículas son **fermiones o bosones**.


## ¿Cómo se demuestra el teorema espín-estadística?

Richard Feynman, en *The Feynman Lectures on Physics Vol. III*, lo explica así:

> “Why is it that particles with half-integral spin are Fermi particles whose amplitudes add with the minus sign, whereas particles with integral spin are Bose particles whose amplitudes add with the positive sign? We apologize for the fact that we cannot give you an elementary explanation. An explanation has been worked out by Pauli from complicated arguments of quantum field theory and relativity. He has shown that the two must necessarily go together, but we have not been able to find a way of reproducing his arguments on an elementary level. It appears to be one of the few places in physics where there is a rule which can be stated very simply, but for which no one has found a simple and easy explanation. The explanation is deep down in relativistic quantum mechanics. This probably means that we do not have a complete understanding of the fundamental principle involved. For the moment, you will just have to take it as one of the rules of the world.”

**En palabras simples:**
- Las partículas de espín semi-entero son fermiones, las de espín entero son bosones.  
- No hay una explicación “simple” accesible a nivel introductorio.  
- La demostración completa requiere mecánica cuántica relativista y teoría cuántica de campos (Pauli).  
- Por ahora, debemos aceptarlo como una regla fundamental de la naturaleza.


## Ejemplos sencillos

### Dos espines

#### Dos espines 1/2

Consideremos dos partículas con espín 1/2, cada una con dos posibles estados:

$$
|\uparrow\rangle , \quad |\downarrow\rangle
$$

La base producto tensorial de los cuatro estados posibles es:

$$
|\uparrow \uparrow \rangle, \quad
|\uparrow \downarrow \rangle, \quad
|\downarrow \uparrow \rangle, \quad
|\downarrow \downarrow \rangle
$$

donde la primer flecha indica el espín de una partícula y la siguiente de la otra partícula.

Si las partículas son idénticas (fermiones), el único estado permitido **antisimétrico** es el **singlete**:

$$
|\Psi\rangle_- = \frac{1}{\sqrt{2}} \left( |\uparrow \downarrow \rangle - |\downarrow \uparrow \rangle \right)
$$

> En este estado se pierde la información de qué partícula está en qué estado de espín.  
> Si el Hamiltoniano no diferencia partículas, cualquier evolución conservará la simetría de intercambio, y no podremos saber que partícula tiene cada spín.

>Notar que de los 4 estados posibles para dos partículas distinguibles, si son fermiones solo queda 1.

Si tenemos además dos sitios distintos A y B (sin interacción entre espín y posición), podemos formar un **estado global antisimétrico**:

$$
|\Psi\rangle_- = \frac{1}{\sqrt{2}} \left( |\uparrow \downarrow \rangle + |\downarrow \uparrow \rangle \right)
\otimes \frac{1}{\sqrt{2}} \left( |AB\rangle - |BA\rangle \right)
$$

Al intercambiar partículas, cada estado producto $|a b \rangle$ pasa a $|b a \rangle$. La parte de espín permanece igual, pero la parte espacial (o referida a otros números cuánticos) cambia de signo. Por lo tanto la función de onda completa es antisimétrica.

No es

$$
|\Psi\rangle_- = \frac{1}{\sqrt{2}} \left( |\uparrow \downarrow \rangle + |\downarrow \uparrow \rangle \right)
\otimes \frac{1}{\sqrt{2}} \left( |AB\rangle - |BA\rangle \right)
$$

#### Dos espines 1

Consideremos dos partículas con espín 1, cada una con tres posibles estados:

$$
|\uparrow\rangle, \quad |0\rangle, \quad |\downarrow\rangle
$$

La base producto tensorial de los 9 estados posibles es:

$$
|\uparrow \uparrow \rangle, \quad |\uparrow 0 \rangle, \quad |\uparrow \downarrow \rangle, \quad
|0 \uparrow \rangle, \quad |0 0 \rangle, \quad |0 \downarrow \rangle, \quad
|\downarrow \uparrow \rangle, \quad |\downarrow 0 \rangle, \quad |\downarrow \downarrow \rangle
$$

Para construir la base del **subespacio simétrico** (permitido para bosones), combinamos estados de manera simétrica:

$$
|\uparrow \uparrow \rangle, \quad
\frac{1}{\sqrt{2}}(|\uparrow 0 \rangle + |0 \uparrow \rangle), \quad
\frac{1}{\sqrt{2}}(|\uparrow \downarrow \rangle + |\downarrow \uparrow \rangle), \quad
|0 0 \rangle, \quad
\frac{1}{\sqrt{2}}(|\downarrow 0 \rangle + |0 \downarrow \rangle), \quad
|\downarrow \downarrow \rangle
$$

Si añadimos **dos sitios distintos** A y B, podemos formar un **estado global simétrico** para bosones:

$$
|\Psi\rangle_+ = \frac{1}{\sqrt{2}} (|\uparrow \downarrow \rangle - |\downarrow \uparrow \rangle)
\otimes \frac{1}{\sqrt{2}} (|AB\rangle - |BA\rangle)
$$

Este estado es simétrico respecto al intercambio de partículas y, por lo tanto, permitido para **dos bosones idénticos**.


## Formulación canónica

Implementar el ensemble microcanónico o canónico considerando los estados bosónicos o fermiónicos es laborioso, porque necesitamos generar permutaciones sobre estados con $N$ partículas exactas. Esto hace que el problema combinatorio se complique rápidamente.

Echemos un vistazo al ensemble **canónico** con $N$ fijo. Cada partícula tiene un número cuántico $k_i$ (suponemos que el espín está incluido en $k_i$ o se mantiene fijo) y energía asociada ${\cal E}(k_i)$.  

Un autoestado de $N$ partículas simetrizado,

$$
|k_1, \dots, k_N \rangle_\eta
$$

tiene energía

$$
H | k_1, \dots, k_N \rangle_\eta
= \left( \sum_{n=1}^{N} {\cal E}(k_n) \right) | k_1, \dots, k_N \rangle_\eta
$$

> Notar que la autoenergía es **la misma que la de los estados producto no simetrizados**, porque la superposición de permutaciones no cambia la suma de energías. Sin embargo, los estados producto **no son físicamente aceptables** para partículas idénticas, aunque matemáticamente sean autoestados.

La función de partición canónica se obtiene sumando sobre todos los posibles conjuntos de números cuánticos de las $N$ partículas. En el caso de fermiones, la suma se hace **respetando la restricción de no tener dos partículas en el mismo estado**:

$$
Z = \text{Tr}[e^{-\beta H}]
= \sum'_{k_1, k_2, \dots, k_N} {}_\eta\langle k_1, \dots, k_N | e^{-\beta H} | k_1, \dots, k_N \rangle_\eta
$$

## Efectos de la degeneración de intercambio: dos partículas no interactuantes

El **postulado de simetrización** tiene efectos importantes en la estadística de las partículas. Por ejemplo, para **dos partículas idénticas** de masa $m$ en el ensemble canónico, las funciones de partición son diferentes según sean bosones o fermiones (ejercicio de la guía):

$$
Z_2^B = \frac{1}{2} \left[ Z_1^2(m, T, V) + Z_1(m/2, T, V)\right], \quad
Z_2^F = \frac{1}{2} \left[ Z_1^2(m, T, V) - Z_1(m/2, T, V)\right]
$$

> Aquí $Z_1(m,T,V)$ es la función de partición de una sola partícula.  
> En contraste, el límite clásico es
> $$
> Z_2^{\rm cl} = \frac{Z_1^2(m,T,V)}{2}.
> $$

El factor $2 = 2! = N!$ aparece **naturalmente** a partir del postulado de simetrización, y generaliza a cualquier número $N$ de partículas.

Estos efectos se llaman **degeneración de intercambio**, y son más importantes cuando aumenta la probabilidad de que dos o más partículas ocupen el mismo estado cuántico, por ejemplo a **bajas temperaturas** o **altas densidades**.

> Notar que incluso si las partículas **no interactúan**, las funciones de partición difieren entre bosones y fermiones, y ambas difieren de la clásica.  
> Si la corrección de intercambio es pequeña, se puede interpretar como un desarrollo de virial clásico, donde los bosones parecen "atraerse" y los fermiones "repelerse".  
> Pero esto **no es interacción real**, sino un efecto de la restricción del espacio de Hilbert según el teorema spin-estadística.


## Representación de Fock o de número de ocupación

En mecánica estadística cuántica, imponer el postulado de simetrización sobre muchas partículas idénticas es complicado. Una forma más conveniente de tratarlo es usar **números de ocupación** de los estados de una partícula, en lugar de enumerar microestados individuales.

### Idea clave

* Cada estado de una partícula puede estar ocupado por **0, 1, 2…** partículas (bosones) o **0 o 1** (fermiones).
* El microestado de $N$ partículas queda definido por la lista de ocupaciones de todos los estados posibles:

$$
|n_1, n_2, \dots, n_\infty \rangle
$$

con la restricción $\sum_\alpha n_\alpha = N$.

* Por ejemplo, para dos partículas en estados $q \neq q'$:

$$
|\phi \rangle = \frac{1}{\sqrt{2}} \left[|q, q'\rangle \pm |q', q \rangle\right]
\quad \longleftrightarrow \quad |0,\dots,1,\dots,1,\dots,0\rangle
$$

donde los `1` indican que los estados $q$ y $q'$ están ocupados.

### Ventajas

* Evita tener que sumar sobre todas las permutaciones explícitas de partículas.
* Implementa de manera natural las restricciones de **Fermi-Dirac** ($n_k=0,1$) y **Bose-Einstein** ($n_k=0,\dots,\infty$).
* Permite expresar la función de partición como suma sobre ocupaciones:

$$
Z = \sum_{ \{n_k\} }^{\mathrm{N,FD/BE}} e^{-\beta \sum_k n_k \mathcal{E}(k)}
$$

donde la suma está filtrada para cumplir las reglas de ocupación y, si corresponde, el número fijo de partículas $N$.

* De manera similar, en el ensemble microcanónico:

$$
\Omega(E,N) = \sum_{ \{n_k\} }^{\mathrm{N, E, FD/BE}} 1
$$

donde las restricciones son sobre $N$ y $E$ fijos, y también sobre los microestados cuánticos con la simetría de intercambio que corresponda.

### Operadores de número

* Definimos operadores de número $\hat n_k$ para cada estado de una partícula $k$:

$$
\hat n_k | n_1, n_2, \dots \rangle = n_k | n_1, n_2, \dots \rangle
$$

* El hamiltoniano de partículas no interactuantes se escribe como suma sobre estados de partículas:

$$
\hat H = \sum_k \hat n_k \mathcal{E}(k)
$$

y los estados de Fock son autoestados de $\hat H$:

$$
\hat H | n_1, n_2, \dots \rangle = \left(\sum_k n_k \mathcal{E}(k)\right) | n_1, n_2, \dots \rangle
$$

### Analogía con osciladores armónicos

* Cada estado de una partícula es equivalente a un **oscilador independiente unidimensional** con frecuencia $\omega_k = \mathcal{E}(k)/\hbar$.
* Una partícula en un estado $k$ representa un cuanto de energía del oscilador. Y tiene sentido, porque esta crece linealmente con el número de ocupación del estado / número cuántico del oscilador asociado.  
* Un sistema de partículas idénticas no interactuantes se puede ver como un conjunto de osciladores armónicos independientes (cada uno representando un estado de partícula, pero **no** una partícula).

### Creación y destrucción

* Los estados de Fock se pueden generar a partir del vacío mediante **operadores de creación y aniquilación**.
* La **álgebra de estos operadores** depende de la estadística:

  * Bosones: conmutadores $[a_i, a_j^\dagger] = \delta_{ij}$
  * Fermiones: anticonmutadores $\{c_i, c_j^\dagger\} = \delta_{ij}$

> Reflexión: el concepto central es que en mecánica cuántica, hablar de “una partícula en particular” pierde sentido. Solo tiene sentido el **número de partículas en cada estado posible**.



## Formulación Gran canónica


Es una gran simplificación el haber pasado de estados simetrizados de $N$ partículas (que requieren superponer permutaciones de estados producto atendiendo a la paridad de cada una) a estados en el espacio de Fock, definidos solamente con los números de ocupación de los estados de una partícula, porque este último nos permite imponer fácilmente el caracter fermiónico o bosónico.

Aún así, la representación de los estados con número de ocupación resulta incómoda si fijamos el número de partículas, pues aparece un vínculo $\sum_k n_k = N$ que impide que factoricemos la función de partición canónica.

Por suerte, en el ensemble gran canónico todo se vuelve muy simple porque removemos la restricción en los $n_k$ (que denotamos con el tilde en la suma $\sum'$).
La gran función de partición queda

\begin{align}
{\cal Z}(T,\mu,V)&=\sum_{N=0}^\infty e^{\beta \mu N}\sum^{\mathrm{FD/BE}}_{n_{k_1},\dots}
\exp \left[ -\beta \sum_k {\cal E}(k)n_k \right] \\
&= \sum^\eta_{n_{k_1},\dots} \prod_k \exp \left[ -\beta({\cal E}(k)-\mu)n_k\right]
\end{align}

donde $\sum^\eta$ significa que restringimos el número según sean bosones $\sum^\eta \equiv \sum_0^\infty$ o fermiones
$\sum^\eta \equiv \sum_0^1$, o sea según la simetría.

La función de partición queda factorizada, una para cada $k$, y podemos hacer la suma sobre el número de ocupación $n_k$,

\begin{align}
{\cal Z}(T,\mu,V)&=
\left[\sum^\eta_{n} \exp \left[ -\beta({\cal E}(k_1)-\mu)n\right]\right].\\
&\left[\sum^\eta_{n} \exp \left[ -\beta({\cal E}(k_2)-\mu)n\right]\right] \dots \\
&= \prod_k Z_k
\end{align}

* Fermiones, $n_k=0,1$
\begin{align}
{\cal Z}_- =  \prod_k [1+\exp \left(-\beta({\cal E}(k)-\mu)\right)]
\end{align}

* Bosones, $n_k=0,\dots,\infty$ y hacemos la suma geométrica
\begin{align}
{\cal Z}_+ =  \prod_k [1-\exp \left(-\beta({\cal E}(k)-\mu)\right)]^{-1}
\end{align}
pero **ojo**, para que  esta suma converja, hemos supuesto que $\mu<\epsilon_i$ $\forall i$, o $\mu< \epsilon_0$ donde $\epsilon_0$ es la menor energía del estado de una partícula. Vamos a ver luego que los estados de equilibrio de bosones siempre satisfacen esta condición.

Ambos resultados pueden ser condensados en uno solo

\begin{align}
\ln {\cal Z}_\eta = -\eta \sum_k \ln \left[ 1- \eta \exp(\beta \mu - \beta {\cal E}(k))\right]
\end{align}

con $\eta=-1$ para fermiones y $\eta=1$ para bosones.

Notar que el problema es muy parecido a uno de adsorción. La analogía es la siguiente
* Los sitios de adsorción ahora son estados de una partícula, y estos pueden aceptar hasta 1 partícula (caso fermiones) o infinitas (caso bosones)
* Los distintos estados de una partícula tienen número de ocupación que son independientes.

La probabilidad conjunta de los números de ocupación es

\begin{align}
p_\eta(\{ n_k \}) = \frac{1}{{\cal Z}_\eta}\prod_k \exp \left[ -\beta({\cal E}(k)-\mu)n_k \right]
\end{align}

Y el número medio de ocupación de cada estado de una partícula es

\begin{align}
\langle n_k \rangle_\eta = -\frac{\partial{\cal Z}_\eta}{\partial(\beta {\cal E}(k))}
= \frac{1}{z^{-1}e^{\beta {\cal E}(k)}-\eta}
\end{align}

donde $z=e^{\beta \mu}$ es la fugacidad.

* Notar que si $\eta=1$, como ${\cal E}(k)>0$, y $\langle n_k \rangle \geq 0$, necesariamente $e^{\beta(\epsilon-\mu)}>1$
o  $\epsilon > \mu$ para todo $\epsilon$. Como hay $\epsilon$ muy cerca de cero, el potencial químico debe ser negativo $\mu < 0$.

* Recordar que clásicamente $\mu=k_B T \ln (n\lambda^3)$, y que la validez del mismo es justamente $n\lambda^3\ll 1$, de moso que $\mu$ viene desde valores negativos grandes. Los efectos de simetría se empezaran cuando $\mu$ se acerca a los niveles más bajos de energía por debajo...

* Notar que para fermiones $\eta=1$ y no hay restricción para $\mu$, ya que $\langle n_k \rangle \geq 0$ siempre, para cualquier $z$ y $T$.

* Notar que si $\mu=0$, las relaciones obtenidas son identicas que para el caso de fotones y fonones, si pensamos ${\cal E}(k)\equiv \hbar \omega_k$ y tenemos en cuenta que la energía de punto cero no aparece en los promedios.


Dos cantidades importantes son el número medio total de partículas y la energía media

\begin{align}
N_\eta &=
\sum_k \langle n_k \rangle_\eta =
\sum_k \frac{1}{z^{-1}e^{\beta {\cal E}}-\eta} \\
E_\eta &=  
\sum_k \langle n_k \rangle_\eta {\cal E}(k)=
\sum_k \frac{{\cal E}(k)}{z^{-1}e^{\beta {\cal E}}-\eta} \\
\end{align}

para calcularlas necesitamos definir ${\cal E}(k)$, el análogo de la relación de dispersión.



### Estadísticas de Fermi-Dirac, de Bose-Einstein y de Maxwell-Boltzmann

Del cálculo anterior vemos que hay una estadística de Fermi-Dirac con $\eta=-1$, una de Bose-Einstein cuando $\eta=1$.

La de Maxwell-Boltzmann es en realidad un límite, a la que tienden las dos anteriores cuando el número de ocupación de cada estado de una partícula es pequeño, o lo que es lo mismo cuando $z^{-1}e^{\beta {\cal E}_k}-\eta \gg 1$. En ese límite, las estadísticas de FD y BE tienden a la de MB

\begin{align}
\langle n_k \rangle_\eta \approx
z e^{-\beta {\cal E}(k)} = e^{-\beta [{\cal E}(k)-\mu]}
\end{align}

En síntesis  


| **Estadística**   | Número de ocupación del estado $i$ de 1 partícula                                              |
|------------------|-----------------------------------------------------------------------------|
| Fermi-Dirac      | $n_i = \frac{1}{e^{\frac{\epsilon_i - \mu}{k_B T}} + 1}$                |
| Bose-Einstein    | $n_i = \frac{1}{e^{\frac{\epsilon_i - \mu}{k_B T}} - 1}$                |
| Maxwell-Boltzmann| $n_i = e^{-\frac{\epsilon_i - \mu}{k_B T}}$                             |

Tener en cuenta,
* Sólo existen 2 estadísticas de partículas idénticas, FD y BE. MB es un límite común de las anteriores.
* MB no es estadística clásica, porque todavía puede tener en cuenta la discretización de la energía. De hecho, es la que usamos para derivar el calor específico de los gases poliatómicos.
* La estadística clásica es MB pero en el límite en el que puede considerarse que los niveles de energía están prácticamente en un continuo, y reemplazar sumas por integrales.

* **Como problema de adsorción**

  Finalmente, lo que hemos hecho es convertir al problema del gas no interactuante de fermiones o bosones, en un problema efectivo de *adsorción*

# Gas no relativista en tres dimensiones

Vamos a considerar que las autoenergías (cinéticas) de las partículas en una caja,  

\begin{align}
{\cal E}(k)&=\hbar^2 k^2/2m
\end{align}

La energía de todos los estados es independiente del espín, que entrará sin embargo como factor de degeneración $g=2s+1$.

Vamos a tener que hacer sumas sobre estados $k$.
Suponiendo **condiciones de contorno periódicas** tenemos

\begin{align}
k&=2\pi(n_x/L_x,n_y/L_y, n_z/L_z) \\
&n_x,n_y,n_z \in [-\infty, \infty]
\end{align}

Por lo tanto, como $L_x,L_y,L_z$ son macroscópicos, los niveles están practicamente en el continuo y podemos aproximar sumas por integrales (luego veremos que hay que tener cuidado con esto si la $T$ es relativamente baja y la densidad alta).

> Si queremos condiciones cerradas, solo cambia un prefactor en todas las integrales que hagamos sobre ${\bf k}$. Las de periodicas se deben multiplicar por $2^d$.

La idea general es pasar de sumas sobre estados de una partícula, a integrales. Por ejemplo:

\begin{align}
N_\eta &=
g \sum_k \langle n_k \rangle_\eta  
\approx g\int \frac{d^3 k}{(2\pi/L)^3} \langle n_k \rangle_\eta
\\
E_\eta &=  
g \sum_k \langle n_k \rangle_\eta {\cal E}(k)
\approx g\int \frac{d^3 k}{(2\pi/L)^3} \langle n_k \rangle_\eta {\cal E}(k)
\end{align}

donde ya hemos usado que la energía no depende del espín.

También con el logaritmo de la gran función de partición o el potencial gran canónico podemos hacer lo mismo

\begin{align}
\ln {\cal Z}_\eta &= -\eta g\sum_k \ln \left[ 1- \eta \exp(\beta \mu - \beta {\cal E}(k))\right] \\
&\approx
-\eta g \int \frac{d^3 k}{(2\pi/L)^3} \ln \left[ 1- \eta \exp(\beta \mu - \beta {\cal E}(k))\right]
\end{align}

y luego usar simetrías en el espacio $k$ (por ejemplo la esférica) para reducir aún más estas integrales.




Las integrales pueden hacerse usando $g(\epsilon)$ o bien hacer explicitamente el cambio de variables explotando la simetría esferica, como se hace en el libro de Kardar. La única diferencia es que se usa la variable $x=\beta \epsilon$.

### Límite "no degenerado": Maxwell Boltzmann

Se llama límite no degenerado:

* Al límite donde la ocupación media de cada estado de una partícula es $\langle n_k \rangle \to 0$. Es decir, que las partículas están diluídas sobre los niveles y entonces los efectos de simetrización son despreciables y reaparece la estadística de Maxwell-Boltzmann.

* Al de alta $T$ y baja densidad, de modo que los efectos de simetrización de la función de onda sean despreciables.

* Se llama *degeneración de intercambio*, pero tiene que quedar claro que es justamente la que ya eliminamos al considerar un espacio de Hilbert simetrizado.

Recordemos que

\begin{align}
\langle n_k \rangle_\eta = -\frac{\partial{\cal Z}_\eta}{\partial(\beta {\cal E}_k)}
= \frac{1}{z^{-1}e^{\beta {\cal E}_k}-\eta}
\end{align}


donde $z=e^{\beta \mu}$ es la fugacidad. Lo que queremos es que el número medio sea chico y desaparezca el efecto de $\eta$, ya que en MB todas las partículas son tratadas de la misma forma. **Eso ocurre cuando $z \to 0$** y

\begin{align}
\langle n_k \rangle_\eta \approx z e^{-\beta {\cal E}}
= e^{-\beta ({\cal E}-\mu)} = \langle n_k \rangle
\end{align}

que es lo que sale en un problema de adsorción cuando la probabilidad de ocupacíon múltiple de un sitio es muy baja, o la dilución del gas adsorbido es alta...

**Notar que para que $z\to 0$ puedo aumentar $T$ o hacer más negativo $\mu$**

De esa forma se ensancha $\langle n_k^2 \rangle_c$ y se despueblan los niveles....
