# Métodos aproximados para tratar interacciones

Vamos a discutir ahora algunos métodos de aproximación para sistemas *interactuantes clásicos*, siempre en el límite $N \gg 1$. Algunas de estas aproximaciones permiten corregir las propiedades del sistema no interactuante dentro de una fase dada, mientras que otras permiten predecir transiciones de fase y aproximar sus propiedades.


## Método Variacional

Las expansiones perturbativas son útiles para incorporar el efecto de las interacciones. Sin embargo, cuando las interacciones son fuertes dejan de ser prácticas, ya que necesitaríamos calcular una gran cantidad de términos —por ejemplo, para describir correctamente la transición de condensación de partículas con interacciones atractivas.

El método variacional es siempre aproximado, pero ofrece una estrategia para tratar interacciones fuertes. El método puede aplicarse a cualquier función de partición, aunque para fijar ideas trabajaremos con el ensamble canónico.

Supongamos que tenemos un hamiltoniano $H$ cuya función de partición $Z$ es difícil de calcular, pero conocemos un hamiltoniano $H_0$ similar, para el cual sí podemos calcular su función de partición $Z_0$. Podemos entonces definir una interpolación lineal entre ambos:

$$H(\lambda)=H_0 + \lambda (H-H_0)$$

donde $\lambda$ es un parámetro artificial. Si $\lambda=0$ nos da el $H_0$, si $\lambda=1$ nos da el $H$. La funcion de partición de este nuevo hamiltoniano interpolado es

$$Z(\lambda)=\sum_{\omega} e^{-\beta [H_0 + \lambda (H-H_0)]}$$

es fácil ver que

$$\frac{d^2 \ln Z(\lambda)}{d\lambda^2}=\beta^2 \langle (H-H_0)^2 \rangle_c \geq 0$$

<blockquote>
porque

$$Z^{-1}(\lambda)\frac{d^n Z(\lambda)}{d\lambda^n} =
Z^{-1}(\lambda) \sum_{\omega} e^{-\beta [H_0 + \lambda (H-H_0)]} (-1)^n\beta^n (H-H_0)^n = \langle (-1)^n\beta^n (H-H_0)^n \rangle
$$

y

$$\frac{d^2 \ln Z(\lambda)}{d\lambda^2} =Z(\lambda)^{-2} \left[\frac{d Z(\lambda)}{d\lambda}\right]^2 + Z(\lambda)^{-1} \frac{d^2 Z(\lambda)}{d\lambda^2}
=
\left[\frac{d \ln Z(\lambda)}{d\lambda}\right]^2 + Z(\lambda)^{-1} \frac{d^2 Z(\lambda)}{d\lambda^2}
= -\langle \beta (H-H_0) \rangle^2 + \langle \beta^2 (H-H_0)^2 \rangle
= \beta^2 \langle (H-H_0)^2 \rangle_c
$$

Matemáticamente, $F(\gamma)=\ln \langle e^{\gamma x} \rangle$ es la generadora de cumulantes de $x$. Es decir, la que me da $\langle x^m \rangle_c$ derivando $m$ veces $F$.

En realidad, este cálculo lo hicieron muchas veces. Para refrescar la memoria
* En un sistema magnético con acoplamiento $-MB$ a un campo $B$ en el canónico de Gibbs, teníamos,
$$\frac{d^2 \ln Z}{dB^2}= \beta^2 [\langle M^2 \rangle - \langle M \rangle^2]$$
* En un problema de adsorción en el gran canónico con el término $-\mu N$ a un campo $B$ teníamos
$$\frac{d^2 \ln Z}{d\mu^2}=  \beta^2 [\langle N^2 \rangle - \langle N \rangle^2]$$
* El el canónico de Gibbs de un sistema $PVT$ teníamos
$$\frac{d^2 \ln Z}{dP^2}=  \beta^2 [\langle V^2 \rangle - \langle V \rangle^2]$$
* Etc.
Y está bueno recordar, que a su vez, estas fluctuaciones son proporcionales a funciones de respuesta, $dM/dB$ (suceptibilidad), $dN/d\mu$ (proporcional a la compresibilidad), $dV/dp$ (proporcional a la compresibilidad), etc. Estos resultados son casos especiales del *Teorema de fluctuación disipación*, válido cuando la respuesta del sistema el campo variado sea lineal.

</blockquote>

Es decir que $Z(\lambda)$ es *convexa* en función de $\lambda$ y coincide con $Z_0$ en $\lambda=0$ y con $Z$ en en $\lambda=1$. Podemos entonces escribir

$$\ln Z(\lambda) \geq \ln Z(0) + \lambda \frac{d\ln Z(\lambda)}{d\lambda}|_{\lambda=0}$$

Para convencerse, dibujar $\ln Z(\lambda)$ como una función convexa, y luego la recta que representa el término del lado derecho de la desigualdad, la cual es tangente a $\ln Z(\lambda)$ en $\lambda=0$.



Screenshot from 2025-11-04 11-53-40.png

ya habíamos visto mas arriba que $d\ln(Z)/d\lambda = -\beta \langle ( H - H_0 ) \rangle$. Si fijamos $\lambda = 0$ luego de derivar, $d\ln(Z)/d\lambda|_{\lambda=0}=-\beta \langle (H-H_0) \rangle_0$, porque la función de partición es $Z_0$. Asi que si ahora fijamos $\lambda=1$ en la desigualdad llegamos a la famosa *desigualdad de Gibbs* (o de Bogoliubov??)

$$\ln Z \geq \ln Z(0) + \beta \langle H_0 \rangle_0 - \beta \langle H \rangle_0$$

Mientras que el lado izquierdo es difícil, el derecho es fácil, porque tenemos $Z(0)\equiv Z_0$, para un $H_0$ dado. Notar que la igualdad vale evidentemente cuando $H_0 \to H$. Como $H_0$ puede depender de muchos parámetros, podemos usar estos para encontrar el $H_0$ que *maximiza* el lado derecho, y asi obtener una aproximación a la verdadera $\ln Z$.   

Formas sugerentes de escribir lo mismo es

$$F \leq
F_0 + \langle (H - H_0) \rangle_0
= -T S_0 + \langle H \rangle_0
$$

y en este caso lo que podemos hacer es *minimizar* el lado derecho en función de los parámetros del hamiltoniano fácil, para aproximar mejor la verdadera $F$, la verdadera termodinámica (del modelo claro).

*No lo demostraremos pero el mismo método funciona para las otras funciones de partición, de otros ensembles*

Lo que es importante notar aquí es que el lado izquierdo es difícil de calcular,

$$\ln Z = \ln[\sum_\omega e^{-\beta H_\omega}]$$

mientras que el derecho es fácil, porque conocemos $Z_0$.
Hay que calcular por un lado,

$$\langle H_0 \rangle_0 = -\partial_{\beta} \ln Z_0$$

lo que es trivial conociendo $Z_0$.

Por el otro, debemos calcular

$$\langle H \rangle_0 = \frac{1}{Z_0}\sum_\omega e^{-\beta H_0} H_\omega$$

que parece más difícil, pero en la práctica no lo es, porque hay mucha libertad para elegir $H_0$, lo que termina determinando la forma de la distribución de probabilidad $e^{-\beta H_0}/Z_0$. Vamos con ejemplos entonces.

### Teorema variacional en el gran canónico
El teorema variacional, general, dice que

$$\ln Z \geq \ln Z(0) + \beta \langle H_0 \rangle^0 - \beta \langle H \rangle^0$$


donde $Z$ es la función de partición "posta", $Z(0)$ es la función de partición de un modelo soluble que se usa como primera aproximación al problema, con hamiltoniano $H_0$ pero parámetros óptimos a ser obtenidos por maximización del lado derecho de la desigualdad. Los promedios $\langle \dots \rangle^0$ son realizados sobre el ensemble canónico con $Z(0)$.

El teorema vale también oir ejemplo para el ensemble gran canónico, ya que podemos pasar en la demostración de $H \to H - \mu N$. Como queremos mantener $N$ fijo, tenemos lo mismo para la gran función de partición ${\cal Z}$,

$$\ln {\cal Z} \geq \ln {\cal Z}(0) + \beta \langle H_0 \rangle^0 - \beta \langle H \rangle^0$$

Entonces para el gran potencial $\Omega = -k_B T \ln({\cal Z}) \equiv -PV$, tenemos

$$\Omega \leq \Omega_0 +  \langle (H-H_0) \rangle^0$$

o también

$$\Omega \leq (F_0 - \mu N) + \langle (H-H_0) \rangle^0$$

en términos de la energia libre de Helmoltz.

---


### Teorema variacional en el microcanónico

Si tenemos un ansatz para la probabilidad de un microestado $\omega$,
$\rho(\omega)$, que satisface las condiciones del microcanónico (energía y número de partículas constante), entonces se cumple que la verdadera $\rho_{micro}(\omega)=\delta(E-H(\omega))/\Omega(E)$ tiene siempre mayor entropía de Shannon, pues fue así como la derivamos!
Es decir,

$$S[\rho] \leq S(E) = k_B \ln(\Omega(E))$$

Entonces, hay que proponer $\rho(\omega)$ y maximizarla respecto a sus parámetros libres.

Otra forma de verlo es usando que $x \ln x \geq x-1$ para todo $x>0$, entonces dadas $\rho, \rho'$, uno puede demostrar que

$$\int d\omega \rho(\omega) \ln(\rho(\omega)/\rho'(\omega))\geq 0$$

y sive como medida de distancia $D(\rho||\rho')$ (de Kullback-Leibler) entre dos distribuciones ya que la igualdad se cumple solo sin son iguales.Si $\rho$ es la exacta, esto implica que

$$-k_B \int d\omega \rho'(\omega) \ln(\rho'(\omega))\leq
-k_B \int d\omega \rho'(\omega) \ln(\rho(\omega))
$$

pero eso es

$$S[\rho']\leq
k_B \ln(\Omega)=S_{micro}
$$



* En general, **todos los teoremas variacionales de la mecánica estadística, para los distintos ensembles, derivan del mismo principio de maximización de la entropía de Shannon**.

* El principio de máxima entropía va más alla de la física porque es un principio de inferencia universal, de la teoría de la información. Uno pone restricciones, en base a la información parcial disponible, e infiere distribuciones de probabilidad.

* Como en general, uno no puede maximizar la entropía de una $\rho$ con infinitos parámetros, modela $\rho$ de la forma más sencilla posible.


## Comentarios sobre el método variacional


* **El método variacional es válido también en estadística cuántica**. Por ejemplo, en el canónico usando que $Z=Tr[e^{-\beta \hat H}]$ (o lo que corresponda al ensemble elegido) llegamos a la misma desigualdad.

* ¿Por qué si el valor númerico de la energía libre derivada del ansatz se aproxima más a la exacta significa que todas las propiedades termodinámicas también?
Una posible respuesta es notar que $F[\rho]-F_{exacta}=D(\rho || \rho_{exacta})$, donde la $D(\rho||\rho')$ es una distancia entre distribuciones (no simétrica pero positiva). Pero $\rho$ determina todas las propiedades que podamos calcular.

* Puede pasar que una peor $\rho'$ de mejor un promedio que la $\rho$, pero globalmente va a estar mejor $\rho$. Es como cuando ajustamos una curva con un modelo, algunos puntos quedaran lejos, pero el error global sera el minimo (si usamos por ejemplo cuadrados minimos).



## Como se usa el método variacional en física


Lo que hacemos usualmente con los teoremas variacionales es una parametrización informada por intuición física del espacio de distribuciones plausibles, codificando los parámetros de $\rho$ en un hamiltoniano efectivo, con interpretación física.

La idea es pensar a $H_0$ como un hamiltoniano fácil pero efectivo, con parámetros en principio desconocidos. Es un ansatz, y buscamos el mejor hamiltoniano de esa familia descripta por variaciones de esos parámetros.

Al optimizar los parámetros tenemos al mejor de la familia del ansatz. Pero si el ansatz es muy malo, hasta el mejor será malo, así que aquí es necesario desarrollar una intuición.

Es ajustar un modelo de hamiltoniano en vez de ajustar una función.

**¿Qué hamiltoniano fácil para hacer estadística pero suficientemente expresivo podemos proponer?**


> *Comentarios al margen*: Hacer aproximaciones con el teorema variacional es parecido a lo que hacemos cuando entrenamos una red neuronal en IA. Allí también se minimiza una "función de costo" variando parámetros que son los pesos de las conexiones sinápticas entre neuronas.
>* Minimizar en deep learning es entrenar una función óptima, para que de con gran probabilidad los output correctos a determinado conjunto de inputs.
>* En mecánica estadística de equilibrio, buscamos que las energías libres o las distribuciones de probabilidad den outputs lo más correctos posibles a los inputs $(N,T,V)$, pero ajustando en realidad un hamiltoniano microscópico.
>* En aprendizaje supervisado en IA necesitamos ejemplos para entrenar, pero también existe el no supervisado donde no necesitamos ejemplos.
>* En mecánica estadística no necesitamos ejemplos, la única guía es minimizar o maximizar.
>* En deep learning se ajustan un montón de parámetros que describen la función.
>* En mecánica estadística se ajustan unos pocos parámetros que describen al hamiltoniano efectivo fácil, y con este hacemos predicciones.
>* Hace poco aparecieron las redes neuronales informadas por física, que consisten en aprender los parámetros de una ecuación diferencial dadas observaciones parciales. Esto si se parece más a aprender un hamiltoniano efectivo en un ensemble.









## Campo Medio

Una aplicación directa del método variacional anterior es la de campo medio. Vamos a resolver dos problemas, el modelo de Ising y el de partículas interactuantes con interacciones atractivas a grandes distancias y fuertemente repulsivas a corto.










### Modelo de Ising Ferromagnético

En cualquier red tenemos

$$H= -J \sum_{\langle i,j \rangle} \sigma_{i}\sigma_{j} -  B \sum_i \sigma_i$$

como ya vimos, es muy dificil de resolver en $d>1$. Peeeero, sabemos resolver exactamente este

$$H_0= -(b+B) \sum_i \sigma_i$$

donde podemos tratar a **$b$ como un parámetro** y $B$ esta puesto a propósito.

El plan entonces es maximizar respecto de $b$, el lado derecho de la siguiente desigualdad:
$$
\boxed{
\ln Z \geq \ln Z(0) + \beta \langle H_0 \rangle_0 - \beta \langle H \rangle_0
}
$$
y para eso necesitamos $Z(0)$, $\langle H_0 \rangle_0$ y $\langle H \rangle_0$.

Sabemos que

$$Z_0 = \sum_{\sigma_1,\dots, \sigma_N} e^{-\beta H_0} =
Z_1^N = \left[\sum_{\sigma=\pm 1} e^{\beta (b+B) \sigma}\right]^N
= \left[2\cosh[\beta (b+B)]\right]^N,\\
\ln Z_0 = N \ln Z_1 = N \ln \left[2\cosh[\beta (b+B)]\right]
$$

y que

$$\langle H_0 \rangle_0 = -\partial_\beta \ln Z
= -N \partial_\beta \ln Z_1 = -N  (b+B) \tanh[\beta (b+B)]] \\
\langle \sigma \rangle_0 = -\frac{\langle H_0 \rangle_0}{N(b+B)}
= \tanh[\beta (b+B)]
$$

Por otro lado

$$\langle H \rangle_0 = \langle H_0 \rangle_0 +
\langle -J \sum_{\langle i,j \rangle} \sigma_{i}\sigma_{j}
+ b \sum_i \sigma_i
\rangle_0,\\
\langle (H_0-H) \rangle_0 =
J \sum_{\langle i,j \rangle}
\langle \sigma_{i}\sigma_{j} \rangle_0
-b\sum_i \langle \sigma_i \rangle_0
= J \sum_{\langle i,j \rangle} \langle \sigma_i \rangle^2
-b\sum_i \langle \sigma_i \rangle_0
=  J (N z/2) \langle \sigma \rangle_0^2
- b N  \langle \sigma \rangle_0
$$

reemplazamos entonces en la desigualdad

$$\ln Z \geq \ln Z(0) + \beta \langle H_0 \rangle_0 - \beta \langle H \rangle_0$$

y obtenemos

$$\ln Z \geq  
N \ln \left[2\cosh[\beta (b+B)]\right]
+ \beta [ J (N z/2) \langle \sigma \rangle_0^2
- b N  \langle \sigma \rangle_0] \\
= N \ln \left[2\cosh[\beta (b+B)]\right]
+ \beta [ J (N z/2)  \tanh^2[\beta (b+B)]
- b N  \tanh[\beta (b+B)]
$$

podemos entonces maximizar el lado derecho con respecto a $b$.
En realidad es mas facil maximizar respecto de $x=\beta(b+B)$, ya que queda

$$\frac{\ln Z}{N
} \geq \ln [2\cosh(x)]+\beta (J z/2) \tanh[x]^2 - (x-B) \tanh[x]$$

La funcion de la derecha tiene un máximo en

$$
x^*  = \beta J z \tanh(x^*)+ \beta B \\
\beta(b^*+B) = \beta J z \tanh(\beta(b^*+B)) + \beta B \\
b^* = \beta J z \tanh(\beta(b^*+B))\\
$$
es decir que
$$
b^* = \beta J z \langle \sigma \rangle_0
$$
en otras palabras, el mejor $H_0$, el que da la mejor $Z_0$ que aproxima a $Z$, es aquel donde agregamos un *campo efectivo* autoconsistente,

$$H_0 = -(B+\beta J z \langle \sigma \rangle_0) \sum_i \sigma_i = -B_{ef} \sum_i \sigma_i$$

Si comparamos con el hamiltoniano original, es evidente porque le llamamos **campo medio**. El campo efectivo es una suma de un campo aplicado y un campo "medio" que proviene de las interacciones.

La buena noticia es que si podemos resolver este problema. Eso si, para resolverlo completamente deberemos pedir *autoconsistencia*. Es decir, en vez de obtenerlo directamente, obtendremos una ecuación para $\langle \sigma \rangle_0$.

Por supuesto, la solución que encontramos depende del *ansatz*, $H_0$, que propusimos, y su calidad como aproximación también.


* En general, cualquier $H_0$ que *desacople los grados de libertad* (porque eso sabemos resolver!), da lugar a una aproximación de *campo medio*. Pero no es obligación usar tal ansatz, y pueden haber mejores que no desacoplen.

* En otras palabras *el campo medio es la mejor aproximación de producto (factorizada) que maximiza la entropía*.



### Derivar rápidamente el hamiltoniano de campo medio

**Receta muy útil**

Hay una receta que permite rápidamente derivar la teoría de campo medio de casi cualquier modelo.

Para ver como funciona, veramos que en nuestro caso del Ising le podemos dar una interpretación física al campo efectivo.
Para eso escribamos

$$
H
=-\sum_{\langle i,j \rangle} J \sigma_i \sigma_j - B \sum_i \sigma_i
=-\sum_{\langle i,j \rangle} J (\langle \sigma \rangle + \delta \sigma_i) (\langle \sigma \rangle + \delta \sigma_j)- B \sum_i \sigma_i
$$

y supongamos que las fluctuaciones de los espines respecto del valor medio son relativamente pequeñas, de modo que términos del orden $\delta \sigma_i \delta \sigma_j$ puedan ser despreciados,  

$$
H
\approx
-J\sum_{\langle i,j \rangle} (\langle \sigma \rangle \delta \sigma_j+\langle \sigma \rangle \delta \sigma_i + \langle \sigma \rangle^2)- B \sum_i \sigma_i
=
-J\langle \sigma \rangle^2 \sum_{\langle i,j \rangle}
-2J\sum_{\langle i,j \rangle} \langle \sigma \rangle \delta \sigma_i - B \sum_i \sigma_i\\
=
-J\langle \sigma \rangle^2 Nz/2 +
(-2J)(z/2)\langle \sigma \rangle  \sum_j  \delta  \sigma_j- B \sum_i \sigma_i
=
-J\langle \sigma \rangle^2 Nz/2
-Jz\langle \sigma \rangle  \sum_j  \delta  \sigma_j - B \sum_i \sigma_i\\
= J\langle \sigma \rangle^2 Nz/2
-Jz\langle \sigma \rangle  \sum_j  \sigma_j- B \sum_i \sigma_i
$$

donde en el último paso usamos $\delta\sigma_i=\sigma_i-\langle \sigma \rangle$.
La cantidad $Nz/2$ es el número de bonds total.

Notar que salvo la constante extra $H \approx \sum_i H_i$
con

$$H_i=-(Jz\langle \sigma \rangle \sigma_i+ B) \sigma_i$$

el espin $\sigma_i$ está rodeado e interactua con $z$ espines, $\sigma_{j_i}$, con $j_i=1,\dots, z$. Si aproximamos la suma sobre espines vecinos con su valor medio $\sum_{j_i} \sigma_{j_i} \approx \langle \sum_{j_i} \sigma_{j_i} \rangle = z\langle \sigma \rangle$ nos queda el campo efectivo mencionado.

En síntesis la aproximación basada en el ansatz optimizado es equivalente a tener el mismo campo efectivo autoconsistente $Jz\langle \sigma \rangle$. Tambien notar que hay una constante adicional a la energia $-J\langle \sigma \rangle^2 Nz/2$, que no va a ser relevante para los valores medios pero si para otros desarrollos...

Definimos entonces el *hamiltoniano de campo medio* de $H$ como

$$H \approx H_{CM} = J\langle \sigma \rangle^2 \frac{Nz}{2}
-Jz\langle \sigma \rangle  \sum_j  \sigma_j- B \sum_i \sigma_i$$

La aproximación de campo medio siempre se aplica igual en una red.

* Se escriben las variables ${S}_i$ (puede ser discretas y tener cualquier número de posibles valores, o continuas, pueden ser escalares, vectoriales o tensoriales) como ${S}_i = \langle {S}_i \rangle + \delta {S}_i$.

* Se escribe
$H(\{{S}_i\})=H(\{ \langle {S}_i \rangle  + \delta {S}_i \}) \approx H(\langle {S}_i \rangle) +
\sum_i \partial_{S_i} H(\langle {S}_i \rangle)\delta {S}_i
+ {\cal O}(\delta {S}^2) \equiv H_{CM} + {\cal O}(\delta {S}^2)
$

* Se resuelve el problema no interactuante autoconsistente.

### Interpretación física del campo medio

La aproximación de campo medio es equivalente a reemplazar el hamiltoniano interactuante $H$, por uno de espines independientes, $H_0=\sum H_0^i$, con $H_0^i=-B_{ef} \sigma_i$, donde $B_{ef}=Jz \langle \sigma \rangle$. Es decir, aproximamos el campo exacto generado por los vecinos del espin $i$ sobre el mismo, $h_i = J\sum_{j=vec(i)} \sigma_j \approx Jz \langle \sigma \rangle$.


### Solución autoconsistente

Sea

$$H_{CM} = J\langle \sigma \rangle^2 \frac{Nz}{2}
-Jz\langle \sigma \rangle  \sum_j  \sigma_j - B\sum_j \sigma_j$$

Ya sabemos muy bien resolver este modelo, ya que son espines independientes en un campo efectivo. La función de partición es $Z=Z_1^N$ con

$$Z_1 =
e^{-\beta J\langle \sigma \rangle^2 \frac{z}{2}} \sum_{\sigma=\pm 1}
e^{\beta (Jz\langle \sigma \rangle+B) \sigma}
= e^{-\beta J\langle \sigma \rangle^2 \frac{Nz}{2}} 2\cosh[\beta (Jz\langle \sigma \rangle+B)] \\
= e^{-\beta [J\langle \sigma \rangle^2 \frac{z}{2} - \beta^{-1}\ln[2\cosh[\beta (Jz\langle \sigma \rangle+B)]]}
\equiv e^{-\beta \tilde{F_1}}
$$

donde hemos definido, la pseudo-energía libre por partícula

$$\boxed{\tilde{F_1}(\langle \sigma \rangle,T) = J\langle \sigma \rangle^2 \frac{z}{2} - \beta^{-1}\ln[2\cosh[\beta (Jz\langle \sigma \rangle+B)]}$$

La ecuación de autoconsistencia puede encontrarse minimizando $\tilde{F_1}(m,T)$ respecto a $m$, o bien usando la magnetización por partícula conocida para espines independientes, $\langle \sigma \rangle = \beta^{-1}\partial_B \ln Z_1$
asi que

$$
\boxed{
\langle \sigma \rangle = \tanh[\beta (Jz\langle \sigma \rangle+B)]
}$$

Notar que:

* Si ponemos $z=1$ tenemos la misma solución autoconsistente que sale de la solución exacta del modelo de Ising totalmente conectado (si ponemos $z=1$) ! Es decir, *la solución de campo medio es la solución exacta en el caso totalmente conectado*.

* En el caso general en cambio, donde cada espín tiene unos pocos vecinos, es solo una aproximación. Veremos sin embargo, que la aproximación se vuelve buena a medida que aumentamos la dimensión del espacio. Es más, algunos resultados, como los exponentes críticos se vuelven exactos por encima de una *dimensión crítica superior*.

* La pseudo energía libre es idéntica a la calculada en la guía usando la aproximación de Hubbard-Stratonovich (el truco de las integrales gausianas), pero diferente de la de Bragg-Williams (mirar por ejemplo donde y como aparece $B$ en cada caso, o el signo del término cuártico). Sin embargo, para cualquier pseudo-energía libre los mínimos son idénticos si tomamos $z=1$.

* Que $z=1$ corresponda al caso totalmente conectado puede parecer raro, pero es razonable, ya que en realidad en el caso totalmente conectado la constante de acoplamiento es $J'=J/N$ en vez de $J$, y el número de vecinos es $z'=(N-1)\approx N$, de modo que aplicando el resultado de la aproximación (que ahora es exacta) de campo medio,  $J'z' \equiv (J/N)N = J$.

* La solución de campo medio implica entonces siempre una transición de fase en $k_B T_c = J z$. Para una red cuadrada $k_B T_c = J 2d$, es decir que depende de la dimensión. Por lo demás, tiene un comportamiento análogo al que ya discutimos.

* Todos los extremos de la pseudo energía satisfacen la ecuación de autoconsistencia, pero algunos son inestables o metaestables. Para derivar la ${\tilde F_1}$ a partir $H_{CM}$ fue necesario retener la constante $Jz\langle \sigma \rangle^2/2$.

Como veremos más adelante, el comportamiento crítico de la solución de campo medio es idéntico al del modelo totalmente conectado. Es más, los exponentes críticos de la solución de campo medio se vuelven exactos ya por encima de la dimensión crítica superior, $d \geq 4$.  


```python
#@title Visualización de la pseudo-energía libre de campo medio

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


# Define the function f(x, y)
def f(M, T, h):
    # Bragg Williams era:
    #return -M**2/2 - h*M + T*((1 + M) * np.log((1 + M)/2)/2 + (1 - M) * np.log((1 - M)/2)/2)

    # Campo medio es:
    return M**2/2 - T*np.log(2*np.cosh((M+h)/T))

# Define the function f'(x, y)
def f1(M, T, h):
    #return -h - M +T*(-np.log((1-M)/2)+np.log((1+M)/2))/2
    return M - np.tanh((M+h)/T)

# Define the function f''(x, y)
def f2(M, T, h):
    #return - 1 + T*(1/(1-M)+1/(1+M))/2
    return 1 - (1/np.cosh((M+h)/T)**2)/T

def plot_free_energy(t,h):
  m = np.linspace(-0.999 , 0.999, 400)
  y = f(m,t,h)
  plt.figure(figsize=(6, 4))
  plt.plot(m, y, label=r"$F(m,t)$")
  plt.show()

def plot_contour(h):
  # Create a grid of (x, y) points
  t = np.linspace(0, 2, 400)
  m = np.linspace(-0.999 , 0.999, 400)
  M, T = np.meshgrid(m, t)

  # Evaluate the function on the grid
  Z = f1(M, T,h)

  # second derivative
  Z2 = f2(M, T,h)

  stable = Z2 > 0
  unstable = Z2 < 0

  plt.figure(figsize=(6, 4))

  filtered_Z = np.where(stable,Z,np.nan)
  plt.contour(M, T, filtered_Z, levels=[0], colors='blue')

  filtered_Z = np.where(unstable,Z,np.nan)
  plt.contour(M, T, filtered_Z, levels=[0], colors='red',linestyles='--')

  #print(stable.shape)

  # Customize plot
  plt.title(r"Contorno de $df/dm =  0$")
  plt.xlabel("M")
  plt.ylabel("T")
  plt.grid(True)
  plt.show()

interact(plot_contour, h=(-2.0, 2.0, 0.001))

interact(plot_free_energy,h=(-0.5, 0.5, 0.001),t=(0, 3.0, 0.1))
```

### Pseudo energía libre de Landau

Como ya discutimos en relación del modelo totalmente conectado, la pseudo-energía libre de campo medio no es muy práctica para hacer cálculos analíticos para el comportamiento crítico y las transiciones.

Landau propuso desarrollar como un polinomio a la misma, suponiendo que el parámetro de orden $m$ se mantiene pequeño. La serie se trunca en la potencia par más baja con coeficiente positivo que permita obtener todas las raíces, y el desarrollo de la pseudo-energía libre este acotado por debajo. En el caso del modelo de Ising basta con llegar la la potencia 4.

La pseudo-energía de Landau debe tener la misma simetría que el hamiltoniano microscópico. Por ejemplo, si $B=0$, solo pueden haber potencias pares pues $H$ es simétrico respecto a $\sigma_i \to -\sigma_i$, y por tanto ${\tilde F}_1(m,T) ={\tilde F}_1(-m,T)$. Esta es una simetría de inversión temporal (ojo, no es inversión espacial, ya que la el momento magnético es un pseudo-vector, como el momento angular, de modo que invertir espacialmente lo deja igual).

Landau fue más lejos, y propuso modelar las transiciones de fase usando solamente simetrías y parámetros de orden adecuados.


```python
#@title Desarrollo de la pseudo energia Libre, a lo Landau

from sympy import *

m = Symbol('m')
J = Symbol('J')
z = Symbol('z')
b = Symbol('\\beta')
B = Symbol('B')

def f(m,J,b,B):
    return (J*z/2)*(m**2)  -(1/b)*log(2*cosh(b*(J*z*m+B)))

print("Si B=0")
display(series(f(m,J,b,0),m))
print("\n\n")

expr=f(m,J,b,B)

print("Si B<>0")
expr.series(m, 0, 5).removeO().series(B, 0, 2).removeO()

```

$${\tilde F}_1 \approx r m^2 (T-T_c) + u m^4$$

donde $r>0$, $u>0$, y que para $B>0$,

$${\tilde F}_1 \approx r m^2 (T-T_c) + u m^4 - B m$$

donde podemos tirar el término con $m^3$.

```python
#@title Visualización de la pseudo-energía libre de Landau ($T_c=1$)

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


# Define the function f(x, y)
def f(M, T, h):
    # Bragg Williams era:
    #return -M**2/2 - h*M + T*((1 + M) * np.log((1 + M)/2)/2 + (1 - M) * np.log((1 - M)/2)/2)
    # Campo medio es:
    #return M**2/2 - T*np.log(2*np.cosh((M+h)/T))
    return M**2*(T-1)/2 + M**4/4 - h*M

# Define the function f'(x, y)
def f1(M, T, h):
    #return -h - M +T*(-np.log((1-M)/2)+np.log((1+M)/2))/2
    #return M - np.tanh((M+h)/T)
    return M*(T-1) + M**3 - h

# Define the function f''(x, y)
def f2(M, T, h):
    #return - 1 + T*(1/(1-M)+1/(1+M))/2
    #return 1 - (1/np.cosh((M+h)/T)**2)/T
    return (T-1) + 3*M**2

def plot_free_energy(t,h):
  m = np.linspace(-0.999 , 0.999, 400)
  y = f(m,t,h)
  plt.figure(figsize=(6, 4))
  plt.plot(m, y, label=r"$F(m,t)$")
  plt.show()

def plot_contour(h):
  # Create a grid of (x, y) points
  t = np.linspace(0, 2, 400)
  m = np.linspace(-0.999 , 0.999, 400)
  M, T = np.meshgrid(m, t)

  # Evaluate the function on the grid
  Z = f1(M, T,h)

  # second derivative
  Z2 = f2(M, T,h)

  stable = Z2 > 0
  unstable = Z2 < 0

  plt.figure(figsize=(6, 4))

  filtered_Z = np.where(stable,Z,np.nan)
  plt.contour(M, T, filtered_Z, levels=[0], colors='blue')

  filtered_Z = np.where(unstable,Z,np.nan)
  plt.contour(M, T, filtered_Z, levels=[0], colors='red',linestyles='--')

  #print(stable.shape)

  # Customize plot
  plt.title(r"Contorno de $df/dm =  0$")
  plt.xlabel("M")
  plt.ylabel("T")
  plt.grid(True)
  plt.show()


interact(plot_contour, h=(-2.0, 2.0, 0.001))
interact(plot_free_energy,h=(-0.5, 0.5, 0.001),t=(0, 3.0, 0.1))
```

### Modelo de Ising AntiFerromagnético

Si $J < 0$, es de esperar que *en una red cuadrada* el orden sea antiferromagnético a bajas temperaturas, ya que espines vecinos prefieren estar antiparalelos y se debería formar un tablero de ajedrez (blanco/negativo, negro/positivo). Es decir, un orden del tipo

+.-.+.-.+.-.+.-.+.-.+.-  
-.+.-.+.-.+.-.+.-.+.-.+  
+.-.+.-.+.-.+.-.+.-.+.-  
-.+.-.+.-.+.-.+.-.+.-.+  

pues ese minimiza la energía de cada link entre primeros vecinos. Notar que esta configuración tiene magnetización neta cero, pero es claramente distinta de la paramagnética.

Si usamos sin cuidado el resultado que derivamos antes con $J < 0$ tendríamos

$$\langle \sigma \rangle = \tanh[\beta (-|J|z\langle \sigma \rangle+B)]$$

Pensandolo gráficamente es fácil ver que hay una sola raíz, asi que parecería ser que no hay transición de fase, y que para $B=0$ la única solución es $\langle \sigma \rangle=0$. ¿Está bien esto? ¿Qué pasó con el orden antiferro?

Por un lado es verdad que $\langle \sigma \rangle=0$, aún por debajo de la transición antiferro (si la hubiera), porque la magnetización de un orden tipo tablero de ajedrez es cero. Así que ambos estados fundamentales tienen magnetización cero. En otras palabras, **la magnetización por sitio no es buen parámetro de orden para la fase antiferromagnética.**

Aplicar lo del caso ferromagnético directamente no es conveniente. La razón es que el campo medio de los vecinos es distinto para las dos subredes, y cada una se magnetiza en dirección opuesta para minimizar la energía libre a bajas temperaturas. En otras palabras $\langle \sigma_i \rangle$ ya no es independiente de $i$, y en este caso depende la subred a la que $i$ pertenece.

La solución más general al problema de campo medio cuando el valor medio de las magnetización de cada sitio es distinta es la siguiente

$$m_i = \tanh[\beta(J\sum_{j_i} m_j+h)]$$

Notar que si los sitios son equivalentes $\sum_{j_i} m_j=zm$ y reobtenemos el caso ferromagnético. El anterior resultado que se puede obtener del hamiltoniano de campo medio (obtenido con la receta de siempre)

$$H_{CM} = C + \sum_{\langle i,j \rangle} 2J \sigma_i \langle \sigma_j \rangle
=
C + \sum_{i} \sigma_i \left[\sum_{j_i} J \langle \sigma_j \rangle \right]
$$

El asunto es que para aplicar esta fórmula necesitamos saber a priori los patrones de magnetización por debajo de $T< T_c$. La pseudo energía libre se vuelve complicada en el caso general, pues ${\tilde f}(m_1,\dots,m_N,T,h)$ (en en modelo de Hopfield, este es el punto de partida de hecho, pero luego se reduce a un conjunto de parámetros de orden que miden el overlap con los patrones memorizados).

Sin embargo, en caso antiferro, sabemos que la mitad de los espines tiene una magnetización media positiva y la otra mitad negativa, de la misma magnitud, así que vamos a poder reducir el problema a un solo parámetro de orden, gracias a conocer la simetría. En otras palabras la pseudo-energía tiene solo dos mínimos a bajas temperaturas, uno en el que todos los sitios blancos tienen $m_i=m>0$ y los negros $m_i=-m< 0$, y viceversa.

Ahora si, podemos aplicar la solución de campo medio a cada subred $A$ y $B$

$$
m_A = \tanh[\beta(-Jz m_B+h)]\\
m_B = \tanh[\beta(-Jz m_A+h)]\\
$$

y de aqui si sale una transición de fase ferromagnetica para $T_c=Jz$, con parámetro de orden *staggered* $m_S=(m_A-m_B)/2$. Entonces podemos directamente
escribir $m=m_A=-m_B \equiv m_S$ arriba, ya que las subredes son equivalentes salvo el signo de diferencia, y las dos ecuaciones se vuelven una sola

$$
m = \tanh[\beta(Jz m+h)]
$$

diciendo que la transición ocurre a la misma $T_c$ y que $m_S$ varía de la misma forma que la del ferro, pero el orden es diferente: consiste en dos subredes ferromagnéticamente ordenadas, pero con magnetizaciones opuestas. Es decir que $m \sim (T_c-T)^{1/2}$ y $m=0$ para $T>T_c$.

La única cosa que si cambia notoriamente es la susceptibilidad, porque cada subred responde distinto a $h$, así que el conjunto no responde como el caso ferro. *No confundir $m$ con la verdadera magnetización, que sin campo es $m=0$, para cualquier $T$, incluso $T<T_c$*.

Derivando las ecs de arriba, y tomando el límite $h\to 0$

$$
\chi_A = \beta(-J z \chi_B +1 )\text{sech}^2(\beta J z m_0) \\
\chi_B = \beta(-J z \chi_A +1 )\text{sech}^2(\beta J z m_0)
$$

donde usamos la notación $m_0=m|_{h=0}\equiv m_0(T)$.
La susceptibilidad es $\chi=(\chi_A+\chi_B)/2$ porque es la respuesta de la magnetización total, porque así se mide,

$$
\chi = \beta(-J z \chi +1 )(1-m_0^2)
$$

usando que $\text{sech}(x)^2=1-\tanh(x)^2$.

de donde se puede ver que

$$
\chi = \frac{1}{T+Jz}, \;\; T > T_c \\
\chi = \frac{1-m_0^2}{T+Jz(1-m_0^2)}, \;\; T < T_c
$$

y la novedad aqui es que no hay divergencia, aunque si una singularidad a la $T_c$, que la llamaremos $T_N$ o **temperatura de Neel**, para diferenciarla de la temperatura de Curie de los ferromagnetos.

Notas al margen:

* En $d=2$ en *una red cuadrada* hay transición pero el campo medio no da $T_c$ ni los exponentes críticos con mucha precisión. El parámetro de orden es el mismo. Es más, la $T_c$ exacta y los exponentes críticos son los mismos que para el caso ferromagnético. En  [ésta simulación](https://editor.p5js.org/droyktton/full/NIiD5ztFG) podemos ver como se mapea un problema a otro. Eso si, son equivalentes si usamos para cada uno su parámetro de orden. Si usamos la verdadera magnetización por sitio, el comportamiento de la susceptibilidad cambia.


* Si se midiera la respuesta a un campo staggered que cambia de signo como un tablero de ajedrez, entonces la correspondiente susceptibilidad divergiría como el ferromagneto. Pero esta no es la definición de respuesta $\chi$.

* La interacción antiferromagnética no siempre da lugar a un antoferromagneto tipo tablero de ajedrez. En redes triangulares por ejemplo hay *frustración* del orden antiferro, y queda una fase de bajas temperaturas desordenada o *vidrio*.

* En el modelo de Hopfield, la pseudo-energía libre, a diferencia del caso ferro o antiferro, tiene múltiples mínimos correspondientes a distintos patrones memorizados en las conexiones sinápticas. El parámetro de orden se vuelven "overlaps" de la magnetización con los patrones.

## Teoría variacional de la condensación

Vamos a considerar que el *hamiltoniano simple* $H_0$ es el que no tiene las interacciones atractivas, pero si los efectos de volumen excluido y la energía cinética.

Asociada a $H_0$ esta la gran función de partición ${\cal Z}(0)$, y a ella la energía libre de Landau o gran potencial $\Omega(0)=F_0-\mu N$, donde $F_0$ es la energía de Helmoltz asociada al hamiltoniano simple.

La $F_0$ va a ser casi la de un gas ideal pero debemos incorporar los efectos de exclusión de volumen. Estos efectos son difíciles de calcular precisamente, pero nos bastara con una aproximacion,

$$Z(0)\approx \frac{1}{N! \lambda^{3N}} V(V-\omega)\dots(V-(N-1)\omega) \approx  \frac{1}{N! \lambda^{3N}} (V-N\omega/2)^N$$

donde cada termino indica cuanto volumen va quedando a medida que agregamos particulas.

Bajo esta aproximación, la probabilidad de encontrar una particula en cualquier posicion ${\bf x}$ en $V-N\omega/2$ es constante, y por tanto la densidad $\rho({\bf x})=1/(V-N\omega/2)=cte$.

Que la densidad sea constante facilita muchos las cosas pues,

$$\langle H-H_0 \rangle_0=
\frac{1}{2}\sum_{i\neq j} \langle u_{at}({\bf x}_i-{\bf x}_j) \rangle_0
= \frac{1}{2} \int d{\bf r}_1 d{\bf r}_2
\langle  \rho({\bf r}_1)\rho({\bf r}_2) \rangle_0
u_{at}({\bf r}_1-{\bf r}_2)
\approx \frac{n^2}{2}V \int d{\bf r} u_{at}({\bf r})
\equiv -\frac{N^2}{2V} u
$$

donde $u>0$ es el modulo de la energía promedio de interacción por unidad de volumen.

Juntando todo en el teorema variacional,

$$\Omega \leq (F_0 - \mu N) + \langle (H-H_0) \rangle^0$$

llegamos a

$$\Omega \leq -k_B T N \ln(V-N\omega/2) +k_B T [N\ln(N)-N] + 3 k_B T N \ln(\lambda_T) - \mu N - \frac{N^2}{2V} u$$

Conviene ahora hacer aparecer la densidad $n=N/V$,

$$\frac{\Omega}{k_B T V} \leq -n \ln(n^{-1}-\omega/2) -n (1+ 3 \ln(\lambda_T) + \beta \mu)  - \beta \frac{n^2}{2} u$$


Al lado derecho lo podemos llamar

$$\frac{\tilde \Omega(n,\mu, T)}{k_B T V} =  -n \ln(n^{-1}-\omega/2) -n (1+ 3 \ln(\lambda_T) + \beta \mu)  - \beta \frac{n^2}{2} u$$

a la cual tenemos que *minimizar* respecto a $n$ para que se acerque lo mas posible a la verdadera $\Omega$.

Lo que acabamos de hacer es encontrar una pseudoenergía libre $\tilde \Omega(n,\mu, T)$ con paramétro de orden $n$, cuyos mínimos corresponden a los estados de equilibrio permitidos para cada $\mu$, y $T$, es decir $n_{eq}(\mu, T)$.

Podemos encontrar la ecuacion que satisfaces los extremos, y ver cuales corresponden a minimos estables, y finalmente ver cual es el minimo global, pero como el objetivo es entender cualitativamente haremos una visualizacion.



















```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Constants
kB = 1.0
Omega = 1.0   # parameter (not the grand potential itself)
u = 1.0

mu = -0.6 # liquido-gas 1er orden aumentando T
#mu = 0.2 # crossover loquido gas

# Function definition
def plot_tilde_Omega(T=0.15):
    beta = 1 / (kB * T)
    lambda_T = 1/np.sqrt(2 * np.pi / beta)  # just for illustration; can be any value

    def tilde_Omega(n):
        term1 = -n * np.log(1/n - Omega/2)
        term2 = -n * (1 + 3*np.log(lambda_T) + beta*mu)
        term3 = -beta * (n**2) * u / 2
        return term1 + term2 + term3

    n_vals = np.linspace(0.01, 1.9, 1000)
    Omega_vals = tilde_Omega(n_vals)

    plt.figure(figsize=(7, 5))
    plt.plot(n_vals, Omega_vals, color='navy')
    plt.xlabel("n")
    plt.ylabel(r"$\tilde{\Omega}(n, \mu, T)/k_B T V$")
    plt.title(rf"$\tilde{{\Omega}}(n,\mu,T)/k_B T V$ for $\Omega=1$, $\mu={mu}$, $u={u}$,  $T={T:.2f}$")
    plt.grid(True)
    plt.show()

# Interactive slider for T between 0.05 and 0.5
interact(plot_tilde_Omega, T=(0.05, 0.5, 0.01))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
kB = 1.0
Omega = 1.0   # parameter (not the grand potential itself)
u = 1.0

mu = -0.6  # liquid-gas 1st order transition increasing T
#mu = 0.2  # crossover liquid-gas

# Function definition
def tilde_Omega(n, T):
    beta = 1 / (kB * T)
    lambda_T = 1 / np.sqrt(2 * np.pi / beta)  # illustrative
    term1 = -n * np.log(1/n - Omega/2)
    term2 = -n * (1 + 3 * np.log(lambda_T) + beta * mu)
    term3 = -beta * (n**2) * u / 2
    return term1 + term2 + term3

# Domain
n_vals = np.linspace(0.01, 1.9, 1000)

# Temperatures to plot
T_values = [0.14, 0.145, 0.150, 0.155, 0.160, 0.165, 0.170, 0.175, 0.180, 0.185, 0.190, 0.195, 0.2, 0.205]

# Colormap for visual distinction
colors = plt.cm.viridis(np.linspace(0, 1, len(T_values)))

plt.figure(figsize=(7, 5))
for T, c in zip(T_values, colors):
    Omega_vals = tilde_Omega(n_vals, T)
    plt.plot(n_vals, Omega_vals, color=c, label=f"T = {T:.3f}")

plt.xlabel("n")
plt.ylabel(r"$\tilde{\Omega}(n, \mu, T)$")
plt.title(rf"$\Omega=1$, $\mu={mu}$, $u={u}$")
plt.grid(True)
plt.legend(title="Temperature", fontsize=9)
plt.show()
```

### Aproximación de campo medio en el ensemble canónico y Van der Walls

Veamos que lo que hicimos es equivalente a despreciar fluctuaciones de la densidad. La parte de interacciones la podemos escribir, usando que $\rho({\bf r})=\sum_i \delta({\bf r}-{\bf x}_i)$,

$$U=\frac{1}{2}\sum_{i\neq j} u_{at}({\bf x}_i-{\bf x}_j)
= \frac{1}{2} \int d{\bf r}_1 d{\bf r}_2
 \rho({\bf r}_1)\rho({\bf r}_2)
u_{at}({\bf r}_1-{\bf r}_2)
$$

ahora proponemos $\rho({\bf r})\approx n + \delta n({\bf r})$, y hacemos $\rho({\bf r})\rho({\bf r'}) \approx n^2 + n\delta n({\bf r})+ n\delta n({\bf r'})+{\cal O}(\delta n^2)$

Si reemplazamos y tiramos los órdenes superiores tenemos

$$
U \approx  
\frac{1}{2} \left[
  n^2 V u
  +
  2\int d{\bf r}_1 d{\bf r}_2 \delta \rho({\bf r})
  u_{at}({\bf r}_1-{\bf r}_2)
\right]= -\frac{n^2 V u}{2} +
\int d{\bf r}_1 \delta \rho({\bf r}_1)
  \int d{\bf r}_2  u_{at}({\bf r}_1-{\bf r}_2)
= -\frac{n^2 V u}{2}
$$

donde usamos que $\int \delta \rho({\bf r})=0$ ya que $\langle \rho({\bf r}) \rangle=n$, y que $u$ es el módulo de la interacción promedio atractiva por unidad de volumen.

Asi que el principio variacional es equivalente al campo medio, donde cada partícula interactúa con el campo medio producida por otras, y donde despreciamos las fluctuaciones de densidad.

Es interesante notar que si usamos el modelo de campo medio en el canónico

$$Z(T,N,V) \approx \int d{\bf r}_1\dots d{\bf r}_N e^{-β[\sum_i p_i^2/2m+ \sum_{i<j} u_{rep}({\bf r}_i-{\bf r}_j)]}
e^{-β[\sum_{i<j} u_{at}({\bf r}_i-{\bf r}_j)]}$$

$$Z(T,N,V)\approx \frac{(V-N\omega/2)^N}{N! \lambda_T^{3N}} e^{\beta u N^2/2V}$$

entonces

$$F(T,N,V)=-k_B T \ln(Z)=-k_B T
\left[
N \ln(V-N\omega/2) - N\ln N + N - N\ln(\lambda_T^3)  
+ \beta \frac{u N^2}{2V}
\right]
$$



y por lo tanto la presión

$$P= -\partial_V F =
k_B T
\left[
\frac{N}{V-N\omega/2} - \beta \frac{u N^2}{2V^2}
\right]
= \frac{N k_B T}{V-N\omega/2} - \frac{u N^2}{2V^2}
$$

Entonces obtenemos

$$(P+\frac{u N^2}{2V^2})(V-N\omega/2)= N k_B T
$$

que es la ecuación de Van der Walls!. Recordemos que el la obtuvo de forma totalmente fenomenológica.

Esta ecuación tiene isotermas con inestabilidades. **El problema viene de que supone que la densidad es homogénea**, cosa que no ocurre en el ensemble canónico, pues en algún momento hay separación de fases. En cada fase en cambio, en el líquido como en el gas, despreciar las fluctuaciones de densidad es una muy buena aproximación.

## Campo medio es la aproximación de punto-silla

Cuando resolvimos el modelo de Ising completamente conectado, en algun punto usamos la aproximación de punto silla, y mostramos que esta es esencialmente *exacta* en el límite termodinámico.

En el modelo de Ising a primeros vecinos podemos usar la misma aproximación, aunque ya no sea exacta en el límite termodinámico. Lo interesante es que conduce directamente al método de campo medio.

Lo que se usa es la transformación de Hubbard-Stratonovich para convertir $Z$ en una integral funcional del tipo $\int {\cal D}\phi e^{-S[\phi]}$, y luego aproximar que la misma está completamente dominada por el mínimo de la "acción" $S[\phi]$, que además es una solución uniforme. Esta solución, corresponde a la solución de campo medio, porque desprecia fluctuaciones.

Veamos como:

**Función de partición**

$$
H[s] = -J\sum_{\langle i j\rangle} s_i s_j - h\sum_i s_i .
$$

$$
Z = \sum_{\{s_i=\pm1\}} e^{-\beta H[s]}
= \sum_{\{s\}} \exp\!\Big(\beta J\sum_{\langle i j\rangle} s_i s_j + \beta h\sum_i s_i\Big).
$$

$$
\sum_{\langle i j\rangle} s_i s_j = \tfrac{1}{2}\sum_{i,j} J_{ij}s_i s_j,
$$

$$
Z = \sum_{\{s\}} \exp\!\Big(\tfrac{\beta}{2}\sum_{ij} J_{ij} s_i s_j + \beta h\sum_i s_i\Big).
$$

**Hubbard–Stratonovich (HS)**

Para una matriz definida positiva $M$ vale:
$$
\exp\!\Big(\tfrac{1}{2} x^\top M x\Big)
= \frac{1}{\sqrt{\det(2\pi M^{-1})}}\int D\phi \;
\exp\!\Big(-\tfrac{1}{2}\phi^\top M^{-1}\phi + x^\top \phi\Big).
$$

Identificamos $x_i=s_i$ and $M_{ij}=\beta J_{ij}$.
$$
\exp\!\Big(\tfrac{\beta}{2}\sum_{ij}J_{ij}s_i s_j\Big)
\propto\int D\phi\,\exp\!\Big(-\tfrac{1}{2}\sum_{ij}\phi_i(\beta J)^{-1}_{ij}\phi_j+\sum_i \phi_i s_i\Big).
$$

$$
Z \propto \int D\phi\; \exp\!\Big(-\tfrac{1}{2}\phi^\top (\beta J)^{-1}\phi\Big)
\sum_{\{s\}}\exp\!\Big(\sum_i s_i(\phi_i+\beta h)\Big).
$$


**Factorización de Suma sobre espines**

Se factoriza porque es como la funcion de partición de espines independientes
$$
\sum_{s_i=\pm1} e^{s_i(\phi_i+\beta h)} = 2\cosh(\phi_i+\beta h).
$$
Entonces
$$
Z \propto \int D\phi\; \exp\!\Big(-\tfrac{1}{2}\phi^\top (\beta J)^{-1}\phi + \sum_i \ln\big[2\cosh(\phi_i+\beta h)\big]\Big).
$$

Finalmente definimos una *accion*

$$
S[\phi]=\tfrac{1}{2}\phi^\top (\beta J)^{-1}\phi - \sum_i \ln 2\cosh(\phi_i+\beta h),
$$

$$Z\propto\int D\phi\, e^{-S[\phi]}$$

Hasta aquí todo exacto, solo hicomos un cambio de variables, de las $s_i = \pm 1$ a las variables continuas $\phi_i$.

**Aproximacion de punto-silla**

Asumimos que la integral esta dominada por el mínimo de la acción,

$$\delta S/\delta \phi_i=0$$

entonces

$$
\sum_j (\beta J)^{-1}_{ij}\phi_j - \tanh(\phi_i+\beta h)=0.
$$

multiplicamos por  $(\beta J_{k i})$ y sumamos sobre $i$:

$$
\phi_k = \beta\sum_j J_{k j}\,\tanh(\phi_j+\beta h).
$$

Asumimos $\phi_i\equiv\phi$, entonces $\sum_j J_{k j}=J z$ y llegamos a

$$
\phi = \beta J z\,\tanh(\phi+\beta h).
$$

La magnetización por sitio es, derivando respecto a $h$ a $\ln(Z)$,
$$
m\equiv\langle s_i\rangle \approx \tanh(\phi+\beta h).
$$

entonces

$$
\boxed{\,m=\tanh\!\big(\beta (J z m + h)\big)\, .}
$$

**Hamiltoniano y energía libre de campo medio**

$\phi=\beta J z m$ de $Z$ se ve que
$F=-\beta^{-1}\ln Z$ es

$$
\frac{F_{\rm MF}}{N} = -\tfrac{1}{2}J z m^2 - \frac{1}{\beta}\ln\big[2\cosh(\beta(J z m + h))\big].
$$

$$
\boxed{\,H_{\rm MF} = -\sum_i (J z m + h)\, s_i + \tfrac{1}{2}N J z m^2\, .}
$$

* HS es exacto.
* La aproximación es la de punto-silla.
* $J_{ij}$ debe ser invertible.
* Fluctuaciones gausianas alrededor del punto-silla dan correcciones al campo medio que se llaman "random-phase approximation".
