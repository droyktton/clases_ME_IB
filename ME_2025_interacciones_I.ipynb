## Ensembles

*¿Qué quiere decir resolver un problema en mecánica estadística de equilibrio?*

Básicamente, derivar la termodinámica de un modelo microscópico. Operativamente se reduce en general a hacer una suma sobre microestados, para calcular la densidad de estados del sistema en el microcanónico o bien la función de partición en los otros ensembles, ya que eso nos permite en principio obtener cualquier valor medio o distribución de probabilidad, en equilibrio, y encontrar la entropía o una energía libre en función de sus variables naturales.

\begin{align}
\Omega(E,V,N)&=\sum_\omega \delta_{E_\omega,E} ⇒ S(E,V,N)=k_B \ln(\Omega) \\
  Z_F(T,V,N)&=\sum_\omega e^{-\beta E_\omega} = \sum_E \Omega(E,V,N)e^{-\beta E} ⇒ F(T,V,N)=-k_B T \ln(Z_F)\\
  Z_{\Theta}(T,\mu,V)&=\sum_\omega e^{-\beta (E_\omega-\mu N_\omega)}
  =\sum_N  Z(T,V,N) e^{\beta \mu N} ⇒
  \Theta(T,\mu, V)=-k_B T \ln(Z_{\Theta})
  \\
  Z_G(\beta, P, N)&=\sum_\omega e^{-\beta (E_\omega + PV_\omega)}
  =\sum_V Z(T,V,N) e^{\beta PV} ⇒
  G(T,P,N)=-k_B T \ln(Z_G)
\end{align}

En cada caso $\sum_\omega$ contempla microestados compatibles con las condiciones impuestas. En el caso mas general $PV$ debe ser reemplazado por $-{\bf J}.{\bf X}$, las variables conjugadas que aparezcan en el trabajo hecho sobre el sistema, siendo ${\bf X}$ la extensiva (Ejemplo: en el polímero es $\tau L$, y en un sistema magético $-{\bf B}.{\bf M}$).


Cada ensemble tiene su conveniencia. En particular, muchas veces resulta práctico eliminar la restricción de una suma, sumando sobre más variables. Por ejemplo, en el microcanónico a veces es difícil filtrar los microestados con $E$ fija. El canónico permite evitar ese filtro al sumar sobre energías distintas, usando el factor de Boltzmann. El canónico por otro lado, necesita filtrar los estados con $N$ fijo, lo cual a veces se torna complicado (en particular en estadística cuántica), y el gran canónico elimina la necesidad de ese filtro. Etc. Esto se refiere a la termodinámica, pues en ese caso todos los ensembles son equivalentes y la elección es por conveniencia práctica. Si no es termodinámica, la elección está motivada por razones físicas propias del sistema.


Lo importante de las sumas de arriba es que nos dan la entropía, energía libre de Helmholtz, energía libre de Landau, energía libre de Gibbs, siempre en *función de sus variables naturales*, es decir, la información termodinámica completa del modelo (no hace falta complementar con ninguna propiedad sacada de otro lado..). Y por supuesto, funciona para cualquier conjunto de variables termodinámicas, sea el sistema un gas, un sistema magnético, un sólido, un líquido, o cualquier otro que pueda describirse termodinámicamente.

*En principio*, tenemos todo lo que necesitamos para extraer la termodinámica de cualquier sistema microscópico. El problema es que para la mayoría de los modelos microscópicos no triviales, hacer las cuentas es un gran desafío técnico y requiere de un gran esquema de aproximaciones, incluídas las resoluciones numéricas.



# Interacciones en ME clásica

El formalismo de los ensembles estadísticos de equilibrio es  universal: podemos *en principio* aplicarlo a cualquier modelo microscópico en principio, definido por sus variables microscópicas y su hamiltoniano o energía por microestado.

La resolución del problema termodinámico, es decir la obtención de las energías libres en términos de sus variables naturales a partir de un modelo microscópico, es sin embargo difícil en general, salvo por unos pocos pero instructivos casos, como los gases ideales y los paramagnetos. La principal dificultad que se presenta en el cálculo son las interacciones entre las componentes del sistema, porque atentan contra una factorización sencilla del sistema, necesaria dado el enorme número de microestados.

Pero son precisamente las interacciones las que dan lugar a las correlaciones espaciales y al comportamiento colectivo de la materia, y en particular a sus distintas fases y transiciones de fase, de modo que no podemos despreciarlas si queremos modelar estos fenómenos.   

Aunque algunos pocos modelos con interacciones no triviales son exactamente solubles e instructivos, en general para resolver el problema con muchos grados de libertad hay que recurrir a distintos métodos de  aproximación o métodos numéricos para obtener las propiedades termodinámicas.



## Dos espines interactuantes

  Para fijar ideas, consideremos un ejemplo extremadamente simple de dos espines distinguibles $\sigma_1=\pm 1$ y $\sigma_2=\pm 1$ interactuantes

  \begin{align}
  E = -J \sigma_1 \sigma_2
  \end{align}

con $J>0$. Físicamente, es una interacción que tiende a alinear los espines, y es el tipo de interacción responsable por el ferromagnetismo en muchos materiales. Es evidente que el estado fundamental del sistema es degenerado, $\sigma_1=\sigma_2=\pm 1$, con $E_0=-J$ en ambos casos. Está claro que los espines no son independientes, al menos a baja temperatura, ya que prefieren ponerse de acuerdo para lograr configuraciones mucho más probables.

En el ensemble de Gibbs, tenemos que la probabilidad de un microestado es

\begin{align}
P(\sigma_1,\sigma_2) =\frac{\exp[\beta(J \sigma_1 \sigma_2 + B_1 \sigma_1 + B_2 \sigma_2)]}{Z(T,B_1,B_2,J)}
\end{align}

donde $B_1=B_2=B$ pero lo mantendremos en general por conveniencia computacional:

\begin{align}
\langle \sigma_i \rangle &= \frac{\beta^{-1}\partial_{B_i}Z}{Z}= \beta^{-1}\partial_{B_i}\ln(Z) &=
-\partial_{B_i}\ln(G) \\
\langle \sigma_1 \sigma_2 \rangle &= \frac{\beta^{-1}\partial_{J}Z}{Z}= \beta^{-1}\partial_{J}\ln(Z) &=
-\partial_{J}\ln(G)
\end{align}


El sistema tiene 4 microestados $(\sigma_1,\sigma_2)\in \{(-1,-1), (-1,1), (1, -1), (1,1)\}$ con magnetizaciones $\mu_0(\sigma_1+\sigma_2)\in \mu_0\{-2, 0, 0, 2\}$ respectivamente y energías $-J\sigma_1\sigma_2=J\{-1, 1, 1, -1\}$.


| Microestado | S1 | S2 | E  | M  | E-S1 B1-S2 B2     | Peso             |
|-------------|----|----|----|----|-------------------|------------------|
| 1           | 1  | 1  | -J | 2  | -(J+B1+B2)        | exp(b(J+B1+B2))  |
| 2           | -1 | -1 | -J | -2 | -(J-B1-B2)        | exp(b(J-B1-B2))  |
| 3           | 1  | -1 | J  | 0  | J - S1 B1 + S2 B2 | exp(-b(J-B1+B2)) |
| 4           | -1 | 1  | J  | 0  | J + S1 B1 - S2 B2 | exp(-b(J+B1-B2)) |


La función de partición de Gibbs con un campo $B$ aplicado es una suma sobre los cuatro microestados posibles

\begin{align}
Z &= 2 e^{\beta  (-J)} \cosh [\beta  (B_1-B_2)]+2 e^{\beta  J} \cosh [\beta  (B_1+B_2)]
\end{align}


>  Notar que en ausencia de interacción
  \begin{align}
  Z|_{J=0} &= 2 \cosh [\beta  (B_1-B_2)]+2 \cosh [\beta  (B_1+B_2)]
  \end{align}
  por otro lado,
  \begin{align}
  Z_1 = \sum_{\sigma_1} e^{-\beta(B_1 \sigma_1)}
  = 2\cosh(\beta B_1) \\
  Z=Z_1 Z_2= 4\cosh(\beta B_1)\cosh(\beta B_2)
  \end{align}
  y para que coincidan
  $4\cosh(\beta B_1)\cosh(\beta B_2)=2\cosh [\beta  (B_1-B_2)]+2 \cosh [\beta  (B_1+B_2)]$


Se pueden dibujar los "niveles de energía", el estado fundamental y los excitados.

Si $B/J<1$


>| E-MB   | Niveles | Estados |
|-------|----------------|----------------|
| J     | ============== | (1,-1), (-1,1) |
| -J+2B | --------------------------- | (-1,-1)        |
| -J-2B | --------------------------- | (1,1)          |

Si $B/J>1$,

>| E-MB   | Niveles | Estados |
|-------|----------------|----------------|
| -J+2B | --------------------------- | (-1,-1)        |
| J     | ============== | (-1,1), (1,-1) |
| -J-2B | --------------------------- | (1,1)          |

Si $B=0$ con $J>0$,

>| E-MB   | Niveles | Estados |
|-------|----------------|----------------|
|  J     | ============== | (-1,1), (1,-1) |
| -J | ============== | (1,1),(-1,-1)          |


Si $J/B=0$, con $B>0$, tenemos

>| E-MB   | Niveles | Estados |
|-------|----------------|----------------|
| 2B     | --------------------------- | (-1,-1) |
| 0     | ============== | (-1,1), (1,-1) |
| -2B    | --------------------------- | (1,1)          |


* **Dificultad** La solución con interacción es más dificil que la de sin interacción, pero esto es mucho más serio cuando el número de espines crece, porque en el caso interactuante para obtener $Z$ necesitamos sumar $2^N$ microestados y conocer la magnetización y energía de cada uno, mientras, mientras que en el caso sin interacción la suma $Z_1$ sigue siendo la de los dos microestados de un solo espín y $Z=Z_1^N$. Por ejemplo, si fuera una red de 3x3 espines tendríamos que sumar $2^{9}=512$ microestados. Para una red tan pequeña como de $10x10$ ya tenemos $2^{100}=1267650600228229401496703205376 \sim 10^{30}$ configuraciones o microestados.

* **Correlaciones** La solución con interacción tiene fluctuaciones correlaciones (ver [notebook de mathematica](https://www.wolframcloud.com/obj/cd6b4654-8c45-48bb-8542-d5e5e2b3fb30))
\begin{align}
g_{12}&=\langle \sigma_1 \sigma_2 \rangle - \langle \sigma_1 \rangle \langle \sigma_2 \rangle \\
&=
\frac{e^{4 \beta  J}-1}{\left(\cosh (\beta  ({B_1}-{B_2}))+e^{2 \beta  J} \cosh (\beta  ({B_1}+{B_2}))\right)^2}
\end{align}
Notar que $g_{12}=0$ si $J=0$ y que $g_{12}=g_{21}>0$. Este tipo de correlaciones, cuando se extienden a grandes distancias, dan lugar a las transiciones de fase.

  Una consecuencia es que la susceptibilidad del sistema, cambia debido a las correlaciones o al comportamiento colectivo. Si $\langle M \rangle =\langle \sigma_1 + \sigma_2 \rangle$, y tomamos $B_1=B_2=B$,

  \begin{align}
  \chi = \partial_B \langle M \rangle|_{B=0} = \frac{4 \beta  e^{2 \beta  J}}{e^{2 \beta  J}+1}
  \end{align}

  diferente de la ley de Curie que derivamos para espines no interactuantes. Por otro lado sabemos que

  \begin{align}
  \chi = \beta^{-1} [\langle M^2 \rangle - \langle M \rangle^2] =
  \beta^{-1} [g_{11}+g_{22}+g_{12}+g_{21}]
  \end{align}

  de modo que la susceptibilidad tiene dos contribuciones extra debidas a las correlaciones.
  Fisicamente, significa que el sistema como un todo responde diferente debido a las interacciones. En este caso, la respuesta es mas alta pues $g_{12}=g_{21}>0$.

  Notar que a pesar de estar los dos estados de más baja energía completamente magnetizados $\langle M \rangle=0$, como en el caso con  $B=J=0$. Lo que diferencia al caso interactuante es que $P(M=0)\to 0$ para $T\to 0$, debido justamente a las correlaciones entre los espines. Para verlo más claramente, calculemos $P(M)$. Para eso hay que tener claro que en este caso de un par de espines los posibles valores son $M=-2,0,2$ (para un sistema de muchos espines interactuantes ¡hay muchos valores mas!) .

  \begin{align}
  P(M=0)&=\frac{e^{-\beta J}}{Z} \\
  P(M=\pm 2)&=\frac{e^{\beta J}}{Z} \\
  \end{align}

  de modo que $P(M=0)\to 0$ y $P(M=\pm 1)\to 1/2$ cuando $T\to 0$. Notar que si no hubieran interacciones $P(M=0)= 1/2$ y $P(M=0)=1/4$ cuando $T\to 0$. Esta es otra forma de ver el efecto de las correlaciones entre espines.


## Dos broches interactuantes

En principio, es posible resolver el problema de 2 partículas interactuantes de masa $m$ en $d=1$ en una caja resolviendo integrales en un espacio de fases de cuatro dimensiones (en vez de $2dN$ dimensiones).

\begin{align}
Z &= \int \frac{dx_1 dx_2 dp_1 dp_2}{2! h^2} e^{-\beta[p_1^2/2m + p_2^2/2m + U(|x_2-x_1|)]} \\
&= \frac{2\pi m/\beta}{2h^2}
\int {dx_1 dx_2} e^{-\beta[p_1^2/2m + p_2^2/2m + U(|x_2-x_1|)]} \\
&= \frac{\lambda_T^{-2}}{2}
\int {dx_1 dx_2} e^{-\beta U(|x_2-x_1|)}
\end{align}

Uno podría en principio dar un potencial de inteacción $U$ y resolver las integrales y obtener distintas propiedades, por ejemplo, la probabilidad de que las partículas estén separadas una cierta distancia $r$.

Por simplicidad, en vez del caso más general podemos considerar que las partículas son 2 broches de ancho $2R$, con centros en $x_1$ y $x_2$, en una soga de longitud $L$. Eso quiere decir que $U(r)=\infty$ cuando $|x_2-x_1|< 2R$, pues no puede solaparse.
La integral es sencilla en este caso, pues basta cambiar los límites de integración para sacar las configuraciones improbables (de energía infinita) y suponer un ordenamiento de las partículas (la 1 atras, la 2 adelante, pero multiplicando todo por un factor 2 para tener en cuenta la otra situación)
\begin{align}
Z
&= {\lambda_T^{-2}}
\int_{R}^{L-3R} dx_1 \int_{x_1+2R}^{L-R} dx_2 \\
&=  {\lambda_T^{-2}}
\int_{R}^{L-3R} dx_1 (L-3R-x_1) \\
&= {\lambda_T^{-2}}
\frac{(L-3R-x_1)^2}{2}|_{L-3R}^R =
{\lambda_T^{-2}}
\frac{(L-4R)^2}{2}
\end{align}

así

$$F(T,L,N=2)=-k_B T [2 \ln (L-4R) -2\ln(\lambda_T)-\ln 2]$$

recordando que $p=-(\partial_L F)_{T,N}$, tenemos finalmente que

$$p=\frac{2 k_B T}{L-4 R}$$

que es practicamente la ecuación de estado del gas ideal unidimensional de 2 partículas, salvo por el "volumen excluido" $4R$ que viene de la interacción fuerte entre los broches.


La densidad de probabilidad de encontrar el par de partículas a distancia $r$, ¿que pinta tendrá?

Hagamos el cálculo y veamos si la intuición queda satisfecha (¡o no!)

\begin{align}
g(r)&=\frac{\lambda_T^{-2}}{Z}
\int_{R}^{L-3R} dx_1 \int_{x_1+2R}^{L-R} dx_2\;
\delta(x_2-x_1-r) \\
&=
\frac{\lambda_T^{-2}}{Z}\int_{R}^{L-3R} dx_1 \Theta(r-2R)\Theta(L-R-r-x_1)
\\
&=
\frac{\lambda_T^{-2}}{Z}\Theta(r-2R) \int_{R}^{L-R-r} dx_1 \\
&=
\frac{\lambda_T^{-2}}{Z} \Theta(r-2R)\Theta(L-2R-r)(L-2R-r) \\
&=
\frac{2\Theta(r-2R)\Theta(L-2R-r)(L-2R-r)}{(L-4R)^2}
\end{align}

## Sistemas interactuantes con muchos grados de libertad:

* Sistema de $N$ partículas interactuantes de a pares en tres dimensiones

  $$H=\sum_i p_i^2/2m + \sum_{i=1}^N \sum_{j=i+1}^N U(r_{ij})$$

  con $r_{ij}=\sqrt{|{\bf x}_i-{\bf x}_j|^2}$.
  Función de partición canónica:

  \begin{align}
  Z = \frac{1}{N! h^{3N}} \int \prod_i d^3x_i d^3p_i e^{-\beta[\sum_i p_i^2/2m + \sum_{i=1}^N \sum_{j=i+1}^N U(r_{ij})]} \\
  = Z_{ideal} V^{-N} \int \prod_i d^3x_i e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N U(r_{ij})]} = Z_{ideal} V^{-N} Q
  \end{align}

  donde $Q(T,V,N)$ **no es trivialmente separable** (¡convencerse!), salvo que $U$ sea "diagonalizable", por ejemplo, una forma cuadrática como la que aparece en pequeñas oscilaciones.

* Sistema de $N$ espines interactuantes:

  $$H= \sum_{i=1}^N \sum_{j=i+1}^N J_{ij} \sigma_i \sigma_j$$
  
  Función de partición canónica:

  \begin{align}
  Z = \sum_{\sigma_1,\dots,\sigma_N} e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N J_{ij} \sigma_i \sigma_j]} \\
  = \sum_{\sigma_1,\dots,\sigma_N}
  e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N J_{ij} \sigma_i \sigma_j]}
  \end{align}

  la cual **no es trivialmente separable** para una $J_{ij}$ general, ni siquiera para $J_{ij}=J$.  

* Sistemas de partículas adsorbidas interactuantes en $M$ sitios

  $$H= \sum_{j=i+1}^M J_{ij} n_i n_j$$
  
  Función de partición gran-canónica:

  \begin{align}
  Z = \sum_{n1_1,\dots,n_N} e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N J_{ij} n_i n_j - \mu \sum_{i=1}^M n_i]}
  \end{align}

  la cual **no es trivialmente separable** a menos que por ejemplo $J_{ij}$ sea cero a menudo tal de poder separar el sistema en clusters independientes.

* **La excepción** Pequeñas oscilaciones de un sistema con cualquier número de grados de libertad, alrededor de un mínimo de su energía. Por ejemplo, considerar la lagrangiana de este sistema mecánico:

  $${\cal L}= \frac{1}{2}\sum_{i,j}[\dot{x_i} M_{ij} \dot{x_j} - x_i K_{ij} x_j] $$

  donde las matrices $K$ y $M$ son matrices semidefinidas positivas, asi como las formas cuadráticas que se forman con ellas. El caso usual de pequeñas oscilaciones en torno a un equilibrio estable o semiestable. Un solo cambio de variables puede diagonalizar simultáneamente la matriz de masas y de resortes y llevarnos a  

  $${\cal L}= \sum_{\alpha}\frac{m_\alpha}{2}[\dot Q^2_\alpha - \omega_\alpha^2 Q_\alpha^2]$$

  y en particular al hamiltoniano,

  $${\cal H}= \sum_{\alpha}[P_\alpha^2/2m_\alpha + m_\alpha \omega_\alpha^2 Q_\alpha^2]$$

  donde las coordenadas normales están completamente desacopladas, de modo que

  $$Z=\prod_\alpha Z_\alpha$$

  donde $Z_\alpha$ es la función de partición del modo $\alpha$, y es fácilmente calculable ya que es la de *un solo oscilador armónico unidimensional* con frecuencia $\omega_\alpha$. Gracias a esto vamos a poder resolver fácilmente el problema de las vibraciones de los sólidos (fonones en el tratamiento cuántico) y otras excitaciones de baja energía en una red llamadas cuasipartículas (magnones, excitones, polaritones, plasmones, polarones, anyones, etc). Vamos a ver sin embargo que la misma magia aplica a los modos electromagnéticos en una cavidad (fotones) y en última instancia a las teoría de campos que dan origen a las partículas, en cierta aproximación (acción cuadrática, o "teorías libres").

  En conclusión, la mayoría pero no todos los sistemas con interacciones son complicados en el límite $N\gg 1$.





## Generalidades

En el ejemplo anterior vimos que además de la aparición de correlaciones, el problema es que las sumas o el conteo se puede volver muy difícil en los casos en que hay interacciones y muchos microestados o grados de libertad.

* **Dificultad**: El número de microestados sobre el que hay que sumar para obtener las energías libres crece exponencialmemte rápido con $N$, ya que a energía fija $\Omega \sim e^{-S/k_B}$ con $S\propto N$.
  * $N$ espines 1/2 por ejemplo, tienen $2^N$ estados, o $e^{N \log(2)}$, crece rapidisimo con $N$, mucho mas rapido que cualquier ley de potencias $N^a$, con $a$ cualquier exponente.
  * $N$ partículas en un gas ideal, ya vimos que crece como $V^N E^{3N/2}$.

  *La conclusión es que la enumeración completa no es viable.*

* **Factorización** ¿Qué son los problemas factorizables? (contexto canónico para fijar ideas)

  Por ejemplo, en el canónico, un sistema factorizable es uno con $N$ grados de libertad, donde $E(\omega)=E_1(ω_1)+E_2(ω_2)+\dots+E_M(ω_M)$, tal que un microestado $\omega$ de todo el sistema está dado por el producto externo de los microestados de los $M \propto N$ subsistemas, $\omega \equiv \{\omega_1,\omega_2,\dots,\omega_M\}$. Un ejemplo son un sistema de $N$ átomos que forman $M$ moléculas no interactuantes, cada una de ellas con posiblemente varios átomos interactuantes.
  Entonces
  
  $$Z=\sum_\omega e^{-\beta E_\omega}
  =
  \sum_{\omega_1} e^{-\beta E_{1}(\omega_1)}  \sum_{\omega_2} e^{-\beta E_{2}(\omega_2)}\dots \sum_{\omega_M} e^{-\beta E_{M}(\omega_M)}
  \propto Z_1 Z_2 Z_3 \dots Z_M
  $$

  Notar que puede o no haber un $M!$ dividiendo dependiendo la indistinguibilidad de las partes, pero lo importante aquí es notar que podemos evaluar $Z$ facilmente para $N$ grande si cada $Z_i$ es computable. Notar que la factorización implica independencia de los subsistemas, $p(\omega_1, \dots, \omega_N)=p_1(\omega_1)\dots p_M(\omega_M)$.

  Ejemplos: Gas ideal, gases de moléculas no interactuantes, espines o dímeros no interactuantes, pequeñas vibraciones de un sólido, el campo electromagnético en una cavidad. En todos $Z \propto Z_1^N$ o $Z \propto Z_1 Z_2\dots Z_N$, con cada $Z_i$ calculable analíticamente (a veces es más sencillo aún porque todos los $Z_i$ iguales).
  
  ¿Que tipo de factores $Z_i$ hay? ¿representa cada uno necesariamente una particula o grado de libertad? En cualquier caso, lo importante es poder calcular los $Z_i$ analíticamente. Los $Z_i$ puede ser funciones de partición de sistemas con un número finito de grados de libertad, y por lo tanto computables.

* **Interacciones** ¿Que tipos de interacciones hay y cual es la dificultad para obtener $Z$?
  * Hay interacciones de a pares de partículas, pero también de a más partículas.
  * Hay interacciones de corto y largo alcance, isotrópicas e anisotrópicas.
  * Las interacciones vinculan grados de libertad y nos obliga a sumar sobre todas las combinaciones de ellos.

* **Interacciones simples** ¿Todas las interacciones son complicadas? ¿Y los sistemas lineales? (vibraciones/fonones, campos electromagnéticos/fotones y otras cuasipartículas libres)
  * Hay interacciones que se diagonalizan, es decir que un cambio de variables deja al sistema desacoplado. Ejemplo: los osciladores acoplados se resuelven como una superposición de modos normales independientes, y no importa cuantas partículas agregue.
  * Hay interacciones de tipo campo medio, que se reducen el problema a uno de pocos grados de libertad o separable, a ser resuelto autoconsistentemente.

* **Fenomenología** ¿Que fenomenología nueva aparece con las interacciones?
  * Conceptualmente, las interacciones siempre están, sino sería imposible llegar al equilibrio termodinámico en un sistema aislado. Aún si no estuviera aislado, si el contacto con el baño térmico ocurre solo en una superficie, necesitamos las interacciones para que fluya la energía a todo el sistema completo y equilibre térmicamente.
  * Correlaciones y comportamiento colectivo (más que la "suma de las partes").
  * No analiticidades, transiciones de fase, rupturas de simetría, saltos en propiedades termodinámicas.
  * Sin las interacciones no habría la complejidad que hay en la naturaleza.
  * ¿Son las transiciones de fase exclusivas de los sistemas interactuantes? En general sí, salvo
  la condensación de Bose, aunque clásicamente puede verse como una interacción efectiva atractiva.

## Transiciones de Fase

Las transiciones de fase, continuas o discontinuas, se manifiestan como cambios no analíticos propiedades termodinámicas. A veces las derivadas primeras o segundas de una energía libre son discontinuas, pero hay otros casos igualmente no analíticos.

En general las transiciones tienen que ver con un cambio de estabilidad de una fase de partida variando algún parámetro de control, como por ejemplo la temperatura.   

Las de primer orden en general son bruscas, las energías libres tienen derivadas primeras discontinuas, calor latente, presentan coexistencia, nucleación, etc.

Las transiciones continuas en particular, están caracterizadas por un comportamiento tipo ley de potencia de las propiedades termodinámicas cerca de un punto crítico el en espacio de los parámetros de control, asi como la existencia de longitudes de correlación divergentes, y habitualmente una ruptura de simetría del estado termodinámico respecto de la simetría microscópica del hamiltoniano.  

Dos ejemplos clásicos son las transiciones en sistemas de partículas, líquido, vapor, sólido, y la transición ferromagnética.

# Modelo de Ising

Los sistemas interactuantes descriptos por variables discretas definidas sobre una red son a menudo muy convenientes tanto para los cálculos analíticos como para las simulaciones numéricas, y arrojan luz sobre muchas cosas generales, que van mucho más alla de los detalles de los modelos.

Entre los modelo más estudiados esta el [**modelo de Ising** ](https://en.wikipedia.org/wiki/Ising_model). Este modela la energía de un sistema de espines 1/2, o lo que es equivalente, de sistemas de dos niveles localizados en sitios de una red dada. En realidad es el modelo que inventó Lenz y se lo dió a Ising para que lo resuelva, quien concluyó erróneamente que no había transición de fase en ninguna circunstancia. El modelo de Ising ferromagnético es de hecho el primer modelo de mecánica estadística que predijo una transición de fase.  

Supongamos que tenemos variables $\sigma_i=\pm 1$ sobre una red, en presencia de un campo magnético $B$ y agregamos una interacción o *acoplamiento* de magnitud $J_{ij}$ entre los sitios $i$ y $j$. La probabilidad de un microestado es $p_\omega = e^{-\beta H}/Z$ con $H$ dado por

$$H = -\sum_{i<j} J_{ij} \sigma_i \sigma_j - \sum_i B \sigma_i$$

donde $J_{ij}$ es la constante de interacción entre el sitio $i$ y el $j$, y la tomamos positiva/negativa para favorecer/desfavorecer alineamiento del espín $i$ con el $j$. El campo $B$ es una notación abreviada para $\mu B'$, donde $\mu$ es el momento magnético del espín y $B'$ el campo magnético. Este tipo de acoplamiento se suele llamar acoplamiento $Zeeman$. Físicamente los $\sigma_i$ corresponden a los posibles valores que puede tomar la proyección sobre un eje el espín $1/2$ de una partícula. Cuando la energía depende solo de esta proyección es porque existe una fuerte anisotropía.

Formulado así, el problema es muy general, y las soluciones dependen crucialmente de la estructura de los acoplamientos $J_{ij}$. Es tan general que todavía no dijimos en que red están localizados estos espines, ni en que dimensión. De hecho, así escrito es un modelo de mecánica estadística en un grafo arbitrario cuyas conexiones son de magnitud $J_{ij}$. Por lo tanto, hay una cuestión de **topología** que hay que definir bien antes de resolver, así como la asignación de los $J_{ij}$.
Así como esta escrito, puede servir tanto para describir un ferromagneto o antiferromagneto, o un fluído sobre una red, o una aleación binaria, o una memoria asociativa en una red neuronal, o la opinión binaria en una red social de agentes que se influencian, o el modelo de un vidrio de espines, etc, etc.

Si vamos a modelar por ejemplo un ferromagneto, es natural suponer que $J_{ij}=J>0$ para espines vecinos en la red asi favorecemos la alineación de espines cercanos ($J<0$ favorece orden local antiferromagnético), y tomar $J_{ij}=0$ para todos los otros pares de espines. Así llegamos al famoso [modelo de Ising ferromagnético](https://es.wikipedia.org/wiki/Modelo_de_Ising).

$$H = - J\sum_{\langle i,j \rangle} \sigma_i \sigma_j - \sum_i B \sigma_i$$

donde la notación $\langle i,j \rangle$ significa que solo vamos a sumar sobre pares de espines vecinos en una red regular de algún tipo (hay que tener mucho cuidado de no sumar el mismo par de espines dos veces). Es decir, que vamos a considerar interacciones de corto alcance para empezar.

*¿Que podemos decir de este sistema sin resolver nada?*

Por un lado es claro que a $T\to \infty$, o más precisamente $k_B T \gg J$, los espines se vuelven independientes y el sistema se comporta como un *paramagneto*. Es decir, tiene magnetización neta solamente como respuesta a un campo $B$ aplicado.

Por otro lado es de esperar que a $T\to 0$ el sistema busque minimizar $H$. Para eso, si $B=0$, es muy conveniente que $\sigma_i=\sigma_j$, ya que eso minimiza la energía de interacción, que vale $-J$ por cada par. Por lo tanto, el estado fundamental a $B=0$ es degenerado:

$$E_{min}=-J N z/2,\;\; \sigma_i=1 \forall i,\; o \;\sigma_i=-1 \forall i,$$

donde $z$ es el número de vecinos de un espín (la red es regular, por lo tanto cada espín tiene el mismo número de vecinos), y el $1/2$ se usa para corregir el doble conteo de pares de espines interactuantes. Los estados que tienen esta energía mínima son dos, según $\sigma_i=\pm 1$. Es decir el *estado fundamental* del sistema tiene degeneración $g=2$. En otro lenguaje, podríamos decir que el modelo tiene dos memorias que se recuperan cuando $T\to 0$.

El estado fundamental degenerado esta totalmente magnetizado, y sus magnetizaciones son $m=\pm N$ respectivamente.
Si prendemos el campo magnético $B>0$ la energía de uno de estos dos estados, el de $\sigma_i=1$ se hace menor a la del otro, y se rompe la degeneración, quedando un estado fundamental único,  

$$E_{min}=-J N z/2 -  B N,\;\; \sigma_i=1 \forall i$$

Este estado está totalmente magnetizado, y su magnetización es máxima $m=N$.




## Magnetización espontánea

Concentremosnos de nuevo en $B=0$. Supongamos que estamos en el estado totalmentre magnetizado con $\sigma_i=1 \forall i$. ¿Que ocurre si aumentamos la temperatura? Una cosa que sabemos que ocurre seguro es que a $T\to \infty$, $m=0$, porque todos los estados se vuelven equiprobables. En particular, los estados de un sitio son equiprobables, y entonces $\langle \sigma_i \rangle=0, \forall i$. Entonces tenemos que $M=\mu N$ para $T\to 0$, y $M=0$ para $T\to \infty$.

¿Que ocurre a temperaturas intermedias?

* ¿como decae $\langle M \rangle=N\langle \sigma_i \rangle$ aumentando $T$ desde cero?
* ¿Cambia suavemente o bruscamente?
* ¿Hay alguna temperatura especial?

Para saber la respuesta necesitaríamos calcular,

$$\langle M \rangle = N \langle \sigma_i \rangle \equiv \partial_B \ln(Z)$$

vs $T$, por ejemplo, usando el ensemble canónico. El primer paso es calcular la función de partición canónica

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1} e^{-\beta H(\sigma_1,\dots,\sigma_N)}=
\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1} e^{\beta (J\sum_{\langle i,j \rangle} \sigma_i \sigma_j + \sum_i B \sigma_i)}
$$

Esta es una suma de $2^N$ configuraciones que no podemos factorizar de manera sencilla culpa de las interacciones. Si no fuera por ellas, ya vimos que $Z=Z_1^N$. Debemos adoptar entonces otra estrategia para poder hacer esta suma. Vamos a ver de hecho muchas estrategias, algunas exactas, aplicables para ciertas circunstancias muy especiales, y otras aproximadas, pero aplicables a muchas circunstancias.

Hay sin embargo, un problema muy importante que salta a la vista a $B=0$, que es que una configuración dada tiene la misma probabilidad que la obtenida invirtiendo todos los espines $\sigma_i \to -\sigma_i$, y que tiene la misma magnetización pero opuesta. Esto significa que $m=0$. Entonces no habría nunca una magnetización espontanea, ni una transición de fase según la mecánica estadística. Y no es una propiedad del modelo de Ising solamente. Matemáticamente, aplicar la teoría de ensembles da siempre $m=0$ si hay una simetría de inversión *temporal* (ojo, es una simetría temporal y no espacial; Se puede pensar como analogía para el espín, que invertir el tiempo es como invertir la dirección de la corriente en una bobina, su momento magnético cambia de dirección.). Pero la ausencia de magnetización espontánea, es decir sin campo $B$ aplicado, implica que ¡no puede haber ferromagnetismo!

**Error: ¡acabamos de demostrar que no existe el ferromagnetismo o magnetización espontánea para cualquier modelo con simetría de inversión temporal!**

Tenemos que volver a la física y tratar de interpretar correctamente las cosas.

Consideremos la probabilidad de observar una magnetización

\begin{align}
p(m_0)=\frac{\sum_{\sigma_1,\dots, \sigma_N=\pm 1} e^{-\beta H(\sigma_1,\dots, \sigma_N)} \delta
\left(\sum_{i=1}^N \sigma_i-m_0 \right)}{Z}
\end{align}

Como $m$ es una variable aleatoria, podemos pensar a $Q(m,B,T)=p(m)Z$ como una función de partición a magnetización fija,

$$Z(B,T)=\sum_{M=-N}^N Q(M,B,T)$$.

y a $A(m_0,T)=-kT \ln (Q(m_0))$ como una energía libre *con el vínculo de magnetizacion fija*, que representa el trabajo reversible requerido para cambiar la magnetización a temperatura fija, entre dos estados de equilibrio a distintas magnetizaciones.

Si dibujamos $A(m,T)$ para $B=0$ tiene típicamente forma de doble pozo, con una barrera en $M=0$, reflejando el hecho de que hay dos estados igualmente estables termodinámicamente. Lo interesante es que en general va a tener una barrera separando los dos mínimos y **la magnitud de esta barrera es crucial**.

En otras palabras, $p(m_0)$ tiene a bajas temperaturas dos picos, uno cerca de $N\mu$ y el otro cerca de $-N\mu$. Si el sistema tiene magnetización positiva, para pasar a magnetización negativa debe necesariamente pasar espontáneamente por una configuración de magnetización cero, pero esta configuración es poco probable cuando $T\to 0$ y $N\to \infty$. Esa es la barrera. En algunos casos la barrera es practicamente infranqueable a $T$ suficientemente baja, y tenemos magnetización espontánea y ferromagnetismo, y en otros no, y no tenemos ferromagnetismo a ninguna $T$.


Pensemos no en el ensemble, sino en un sistema en particular. Si en un momento dado el sistema esta cerca de un mínimo de la energía libre con una magnetización espontánea, para ir al otro mínimo es necesario que absorba energía suficiente como para pasar la barrera. Esto tiene que ocurrir como resultado de una fluctuación térmica. El punto crucial es el siguiente: *Los sistemas con transición de fase ferromagnética tienen una barrera que escalea como el número de espines*, por ejemplo como $N^{1/2}$ para $d=2$, o $N^{2/3}$ para $d=3$. Por lo tanto, en el límite termodinámico, tal fluctuación es improbable, porque la varianza de $m$ es pequeña comparada con $m$. En particular, $p(m=0)\to 0$, y el sistema se pasa muchísimo tiempo cerca del mínimo inicial. Por lo tanto, a los fines prácticos, presenta magnetización espontánea. Sistemas con un número pequeño de espines tienen una barrera relativamente baja, y la magnetización fluctúa mucho más fuertemente que en un sistema de $N \gg 1$. Esto implica, que existe una probabilidad no nula de invertirse en el tiempo. En $d=1$, sin embargo, *para cualquier número de espines $N$* veremos que la barrera tiende a cero por una cuestión entrópica, de modo que la magnetización se pasea ampliamente y se promedia a cero. La barrera tiene que ver con el fénomeno de núcleación y crecimiento que necesitamos para que un sistema invierta su magnetización. Este es un fenómeno dinámico, pero podemos servirnos de la mecánica estadística para entender las escalas de energía que entran en juego.

Pero como, ¿no era que los sistemas que considerabamos para hacer mecánica estadística eran ergódicos, que exploraban todo su espacio de fases o de configuraciones? Bueno, no, **en las transiciones de fase continuas hay usualmente una ruputura efectiva de la ergodicidad** y esta viene acompañada de una **ruptura espontánea de la simetría** .

Cuando hay campo $B$, en un sistema ferromagnético, uno de los dos mínimos de la energía libre a $m$ fijo esta más bajo que el otro, y correponde a una mayoría de los espines apuntando en la dirección del campo. Este es el mínimo estable, el estado termodinámico de equilibrio. Sin embargo, si preparamos el sistema en el otro mínimo, y $B$ no es muy grande, la barrera todavía estará. Esta situación es metastable y es la responsable de la histéresis de os ferromagnetos cuando uno hace un ciclado en campo, desde valores negativos grandes a positivos grandes, y de vuelta. El campo donde la magnetización salta, es el campo coercitivo.

Veremos en lo que sigue que

* El modelo de Ising ferromagnético en $d=1$ tiene solución exacta en el límite termodinámico, pero **no** tiene transición de fase.

* El modelo de Ising ferromagnético en $d=2$ tiene solución exacta en el límite termodinámico, y tiene transición de fase.

* El modelo de Ising ferromagnético en $d>2$ no tiene solución exacta en el límite termodinámico pero tiene transición de fase.

* El modelo de Ising ferromagnético en $d=\infty$ tiene solución exacta en el límite termodinámico y tiene transición de fase.

## Solución exacta del modelo de Ising en d=1 para $N \gg 1$

Las soluciones exactas de sistemas interactuantes no factorizables de muchos grados de libertad son escasas y usualmente representan casos demasiado idealizados. Sin embargo son muy valiosas ya que permiten poner a prueba esquemas de aproximación analítica y numérica.

Empecemos con el caso de una red regular de sitios equiespaciados en una red.
En este caso

$$H= -J \sum_i \sigma_i \sigma_{i+1} - \mu B \sum_i\sigma_i$$

y notar que los primeros vecinos son los de la derecha y la izquierda, y que la suma sobre pares se puede hacer moviéndonos solamente en la dirección creciente del índice del sitio.

La función de partición queda

$$Z=
\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1} e^{\beta (J \sum_i \sigma_i \sigma_{i+1} + \mu B \sum_i\sigma_i)}
$$

y todavía parece difícil de sumar. Sin embargo, existe en este caso un truquito muy ingenioso. Rescribimos primero el hamiltoniano

$$H= -J \sum_{i=1}^N\sigma_i \sigma_{i+1} - \mu B \sum_{i=1}^N (\sigma_i+\sigma_{i+1})/2$$

y pensemos para empezar en condiciones de contorno periódicas, de tal forma que $\sigma_{N+1}\equiv \sigma_1$. Es decir, que la red tiene forma de anillo. Se pueden elegir otras condiciones de contorno sin problemas, pero esta nos va a resultar más simple por ahora y por otro lado, esperamos que las condiciones de contorno no sean tan importantes para el comportamiento colectivo cuando $N\gg 1$. La simetrización del término de acoplamiento con $B$, o acoplamiento Zeeman, nos va a servir para escribir $Z$ en una forma muy compacta.

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1}
T_{\sigma_1,\sigma_2}T_{\sigma_2,\sigma_3}\dots T_{\sigma_N,\sigma_1}
$$

donde

$$T_{\sigma_i, \sigma_{i+1}}
= e^{\beta [J \sigma_i \sigma_{i+1}+B\mu(\sigma_i + \sigma_{i+1})/2]}
$$

Si pensamos a estas cantidades como elementos de una matriz, variando el valor de $\sigma_i=\pm 1$ y $\sigma_j=\pm 1$, tenemos una matriz de 2x2

$${\cal T}=
\begin{pmatrix}
T_{++} & T_{+-}\\
T_{-+} & T_{--}
\end{pmatrix}
=
\begin{pmatrix}
e^{\beta (J + B\mu)} & e^{-\beta J}\\
e^{-\beta J} & e^{\beta (J - B\mu)}
\end{pmatrix}
$$

que se denomina *matriz de transferencia*. Para ver la ventaja de representar en términos de esta matriz, notemos que los dos últimos factores en $Z$, los únicos que involucran la última sumatoria en $\sigma_N$, se pueden escribir como

$$
\sum_{\sigma_N=\pm 1} T_{\sigma_{N-1},\sigma_{N}}T_{\sigma_{N},\sigma_{1}}
\equiv [{\cal T}^2]_{\sigma_{N-1},\sigma_1}
$$

ahora, si hacemos la suma en $\sigma_{N-1}$ tenemos

$$
\sum_{\sigma_{N-1}=\pm 1} T_{\sigma_{N-2},\sigma_{N-1}}{\cal T}^2_{\sigma_{N-1},\sigma_{1}} \equiv {\cal T}^3_{\sigma_{N-2},\sigma_{1}}
$$

y asi uno podría seguir hasta la última suma y encontrar que


$$Z = \sum_{\sigma_1=\pm1} T^{N}_{\sigma_1,\sigma_1} \equiv \text{Tr}[{\cal T}^N]$$

<blockquote>
Una forma piola de ver lo mismo es pensar a la matriz de transferencia como un operador $\hat T$ sobre el espacio de estados de $\sigma$ y usar la notación de bra-kets de mecánica cuántica

$$T_{\sigma_i,\sigma_j}=\langle \sigma_i | {\hat T} | \sigma_j  \rangle$$

entonces la función de partición adquiere una forma muy sugestiva

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1}
\langle \sigma_1| {\hat T} | \sigma_2 \rangle
\langle \sigma_2| {\hat T} | \sigma_3 \rangle
…
\langle \sigma_N| {\hat T} | \sigma_1 \rangle
$$

que se puede escribir así

$$Z=
\sum_{\sigma_1=\pm 1} \langle \sigma_1| {\hat T}
\left[\sum_{\sigma_2=\pm 1} | \sigma_2 \rangle
\langle \sigma_2| \right] {\hat T} \left[\sum_{\sigma_3=\pm 1}
| \sigma_3 \rangle \langle \sigma_3|\right]
…
\left[\sum_{\sigma_N=\pm 1}| \sigma_N \rangle \langle \sigma_N|\right] {\hat T} | \sigma_1 \rangle \\
=
\sum_{\sigma_1=\pm 1} \langle \sigma_1| {\hat T}
\left[\mathbb{1}\right] {\hat T} \left[\mathbb{1}\right]
…
\left[\mathbb{1}\right] {\hat T} | \sigma_1 \rangle

$$
y obtenemos el mismo resultado. Sin embargo, la última expresión es muy sugerente porque es muy parecida a la que aparece en la formulación de integrales de camino de la mecánica cuántica, donde se divide el operador de evolución $e^{-iHt}$ en $N$ pedacitos...
</blockquote>

$\lambda_2$ son los autovalores de $\cal{T}$, llegamos a

$$Z= \lambda_1^N + \lambda_2^N$$

\sqrt{\exp(-2\beta J)+\exp(2\beta J)\sinh^2(\beta \mu B)}\right]^N
$$

Notar que al final pudimos factorizar la función de partición, pero los factores son no triviales, a diferencia del caso interactuante.

Podemos ahora obtener la magnetización media $\langle M \rangle=\partial_{\beta B} \log(Z)=\beta^{-1}\partial_{B} \log(Z) = \beta^{-1}\partial_{B} (-F/\beta)$,

$$\langle M \rangle =\frac{N\mu \sinh(\beta \mu B)}{\sqrt{\exp(-4\beta J)+\sinh^2(\beta \mu B)}}$$

Notemos que si $B=0$, entonces $\langle M \rangle = 0$, a cualquier temperatura. Como ya discutimos antes, esto no significa necesariamente que no haya magnetización espontánea.
Notemos también que si $J=0$ se reduce a lo conocido para espines no interactuantes: $\langle M \rangle = N\mu \tanh[\beta \mu B]$.

Por otro lado la susceptilidad magnética es

$$\chi= \lim_{B\to 0} \partial_B \langle M \rangle = \beta N \mu^2 \exp(2\beta J)$$

donde $M=\sum_{i=1}^N \mu \sigma_i$ es la magnetización instantánea o fluctuante. Notar que $\chi=\beta N \mu^2 \exp(2\beta J)$ diverge exponencialmente rápido.

> **Control** Notar que si hacemos $J\to 0$ deberíamos reobtener resultados muy conocidos, $M=N\mu \tanh(\beta \mu B)$ y $\chi=\beta N \mu^2$ (ley de Curie). Vale la pena notar que la ley de Curie se cumple entonces para temperaturas suficientemente grandes, aún en presencia de interacciones.

### Longitud de Correlación

Una de las propiedades emergentes más interesantes de las interacciones es la emergencia de correlaciones espaciales.
Calculemos la correlación spin-spin en función de la distancia,

\begin{aligned}
\left\langle\sigma_i \sigma_{i+r}\right\rangle & =\frac{1}{Z} \sum_{\sigma_1= \pm 1} \sum_{\sigma_2= \pm 1} \cdots \sum_{\sigma_N= \pm 1} \sigma_i \sigma_{i+r} \prod_{j=1}^N \exp \left(K \sigma_j \sigma_{j+1}\right) \\
& =\frac{1}{Z} \operatorname{tr}\left[T^{i-1} \hat{\sigma}_z T^r \hat{\sigma}_z T^{N-i-r+1}\right]=\frac{1}{Z} \operatorname{tr}\left[\hat{\sigma}_z T^r \hat{\sigma}_z T^{N-r}\right]
\end{aligned}

En la segunda igualdad usamos el mismo truco de la matriz de Pauli que usamos cuando calculamos $\langle \sigma_i \rangle$.


\begin{aligned}
\left\langle\sigma_i \sigma_{i+r}\right\rangle & =\frac{1}{Z} \operatorname{tr}\left[\left(\begin{array}{cc}
0 & 1 \\
1 & 0
\end{array}\right)\left(\begin{array}{cc}
\lambda_{+}^r & 0 \\
0 & \lambda_{-}^r
\end{array}\right)\left(\begin{array}{cc}
0 & 1 \\
1 & 0
\end{array}\right)\left(\begin{array}{cc}
\lambda_{+}^{N-r} & 0 \\
0 & \lambda_{-}^{N-r}
\end{array}\right)\right] \\
& =\frac{1}{Z} \operatorname{tr}\left(\begin{array}{cc}
\lambda_{+}^{N-r} \lambda_{-}^r & 0 \\
0 & \lambda_{-}^{N-r} \lambda_{+}^r
\end{array}\right)=\frac{\lambda_{+}^{N-r} \lambda_{-}^r+\lambda_{-}^{N-r} \lambda_{+}^r}{\lambda_{+}^N+\lambda_{-}^N} .
\end{aligned}

\begin{equation}
\left\langle\sigma_i \sigma_{i+r}\right\rangle \approx\left(\frac{\lambda_{-}}{\lambda_{+}}\right)^r \equiv \mathrm{e}^{-r / \xi}
\end{equation}

Es decir que

\begin{equation}
\xi = [\ln(\lambda_+/\lambda_-)]^{-1} = -\frac{1}{\ln \tanh (\beta J)} = \frac{1}{\ln \coth (\beta J)}
\end{equation}

Este resultado es superinteresante. Por un lado, vemos que el mas grande de los autovalores de la matriz de transferencia nos da $Z$ y por el otro el cociente entre el más chico y el más grande nos da la longitud de correlación, la cual decrece aumentando $T$ como es de esperarse.

Es notorio que $\xi$ diverge solo cuando $T\to 0$. Veremos luego que cuando $\xi$ diverge a una temperatura finita $T_c$, es la señal de una transición de fase a esa temperatura.


Hay una conexión precisa entre la susceptibilidad magnética y la función de correlación espacial. Recordemos que la suceptibilidad esta conectada a la varianza de la magnetización y que ese resultado es muy general e independiente de la forma de $H$ siempre y cuando aparezca un acoplamiento del tipo $M B$.
Es decir que
$$\chi = \frac{1}{k_B T}[\langle M^2 \rangle-\langle M \rangle^2]$$
En el caso de espines no interactuantes usamos el teorema del límite central para escribir $\chi=N\chi_1$, con $\chi_1$ la susceptibilidad de un spin aislado. Con correlaciones, ya no podemos usar el mismo argumento y a cambio obtenemos,

\begin{align}
\chi = \frac{1}{k_B T}[\langle M^2 \rangle-\langle M \rangle^2]
=  
\frac{1}{k_B T}[
  \langle (\sum \sigma_i)^2 \rangle
  -
  \langle (\sum \sigma_i) \rangle^2
  ]
=
\frac{1}{k_B T}[
  \sum_{i,j} \langle \sigma_i \sigma_j \rangle
  -
  \sum_{i,j} \langle \sigma_i \rangle \langle \sigma_j \rangle
  ]
= \frac{1}{k_B T}[
  \sum_{i,j} [\langle \sigma_i \sigma_j \rangle - \langle \sigma_i \rangle \langle \sigma_j \rangle
  ]
  =
  \frac{1}{k_B T}[
  \sum_{i,j} g(|j-i|)
  ]
  =  
  \frac{N}{k_B T} \sum_r g(r) \sim \frac{N}{k_B T} \xi(T,J)
\end{align}


donde
$$g(r)=\langle \sigma_i \sigma_{i+r} \rangle- \langle \sigma_i \rangle \langle \sigma_j \rangle = \langle \sigma_i \sigma_{i+r} \rangle$$ usando ya que $\langle \sigma_i \rangle=0$ y $\xi=[\ln \coth(\beta J)]^{-1}$.

\begin{align}
\chi &= \frac{1}{k_B T}\left[\langle M^2 \rangle-\langle M \rangle^2\right] \\
&= \frac{1}{k_B T}
  \sum_{i,j} [\langle \sigma_i \sigma_j \rangle - \langle \sigma_i \rangle \langle \sigma_j \rangle
  ] \\
&=
  \frac{1}{k_B T}
  \sum_{i,j} g(|j-i|) \\
&=  
  \frac{N}{k_B T} \sum_r g(r) \sim \frac{N}{k_B T} \xi(T,J)
\end{align}



Notar que en un sistema infinito la función de correlación depende sólo de la distancia entre los dos espines en consideración, es decir depende de $r$ y no de $i$, como resultado de la invariancia discreta de traslación. Es decir, si nos trasladamos un parámetro de red, al siguiente sitio, el resultado no puede cambiar. Además, podemos decir que $g(r) \equiv g(|r|)$ por simetría especular, es decir la correlación no distingue el sentido del vector de desplazamiento.

### Orden espacial y magnetización espontánea

Vimos que a $B=0$, tenemos $\langle M \rangle =0$, para toda $T$. Esto **no** puede ser usado como argumento en contra de un orden ferromagético, ya que ocurre en cualquier sistema con simetría de inversión temporal, en cualquier dimensión. Necesitamos entonces otros argumentos.

* Un argumento en contra de la transición de fase, es que la solución exacta no tiene signularidades a $T$ finita. En particular, la susceptibilidad, que debería divergir a $T_c$, diverge pero a $T=0$, como tambien lo hace el sistema no interactuante.

* Otro argumento que muestra que no hay orden ferromagnético es notar que la probabilidad de que un par de espines separados $r<\xi$ estén orientados en la misma dirección es alta por $g(r)$ es positiva y alta. Es decir, hay *dominios* con espines con el mismo signo. Sin embargo, para $r>\xi$, la correlación se va a cero, significando hay una probabilidad apreciable de que a esa distancia el segundo espín este invertido la mitad de las veces. Pero si es así, es probable que exista un dominio de magnetización opuesta empezando más allá. Si tenemos un montón de dominios de tamaño finito, con magnetización positiva o negativa, no queda otra que $\langle |M| \rangle =0$ si $B=0$.
Una forma de ver esto, es *renormalizar* el sistema, es decir, mirarlo desde mas lejos cada vez, es decir ir generando una serie de funciones de partición con menos grados de libertad (ver notas de E. Jagla).

* Supongamos por el contrario que hay una magnetización neta $\langle M \rangle>0$ a $T>0$. Entonces $\langle \sigma_i \sigma_{i+r} \rangle \sim 1$ para cualquier $r$, contradiciendo que la longitud de correlación es finita.

* Otro argumento para la *ausencia de transición de fase en $d=1$* es que
si partimos de todo magnetizado, el costo energético de dos paredes que generen un dominio de espines up en un mar de espines down cuesta solamente $4J$. Por otro lado, como puedo poner esas paredes en cualquier posición de las $N$ posibles, puedo ganar una entropía que crece con $N$ a cualquier $T=\epsilon>0$ para $N \gg 1$, con lo cual se hace altamente probable la configuración excitada. Por lo tanto, tener paredes minimiza $F$. Pero tener muchas paredes no es consistente con un estado magnetizado. Es decir, si parto a $T=0$ totalmente magnetizado en un sentido, a $T>0$ ese orden se destruye.




**Conclusión: No hay transición de fase ferromagnética en $d=1$**

### Método de la Matriz de Transferencia

El método que usamos para resolver el modelo de Ising 1d es muy poderoso y muy general, y adecuado para resolver una familia grande de problemas en $d=1$ con interacciones a primeros vecinos.

Para cualquier función de partición que podamos escribir como

$$Z=\sum_{s_1,\dots,s_N} e^{\sum_{i=1}^N B(s_i, s_{i+1})}
= \sum_{s_1,\dots,s_N} \prod_{i=1}^N e^{B(s_i, s_{i+1})}$$

podemos definir su matriz de transferencia

\begin{equation}
\left\langle s_i\right| T\left|s_j\right\rangle=\mathrm{e}^{B\left(s_i, s_j\right)}
\end{equation}

y calcular, en el caso de condiciones de contorno periódicas (si son abiertas hay una pequeña corrección que no es importante si $N\gg 1$),

$$Z=tr[T^N] \approx \lambda_{max}^N$$

y la función de correlación haciendo el cociente entre autovalores.

>El [teorema de Perron-Frobenius](https://en.wikipedia.org/wiki/Perron%E2%80%93Frobenius_theorem) nos garantiza que $\lambda_{max}$ no es degenerado porque $T$ es finita y con elementos positivos. Si por ejemplo tenemos esta matriz simétrica,
$$\left(
\begin{array}{cc}
 a & b \\
 b & c \\
\end{array}
\right)$$
sus autovalores son
$$\frac{1}{2} \left(-\sqrt{a^2-2 a c+4 b^2+c^2}+a+c\right),\frac{1}{2} \left(\sqrt{a^2-2 a c+4 b^2+c^2}+a+c\right)$$
El discriminante es $a^2-2 a c+4 b^2+c^2 = (a-c)^2 +4 b^2>0$, por lo tanto los autovalores son reales y diferentes. Como $a$ y $c$ son positivos, y la raíz cuadrada es siempre menor a $a+c$, porque el determinante de la matriz es positivo, ambos autovalores son reales, positivos y diferentes.  


Por lo tanto, $Z$ será una función analítica de los parámetros del hamiltoniano y de $T$, y solo podrá exhibir una singularidad a $T=0$. Por lo tanto, no hay transición de fase en $d=1$ en este tipo de modelos.

El formalismo no se limita a variables discretas. Puede ser aplicado sin problemas al hamiltoniano de Heisenberg por ejemplo

$$Z=\sum_{{\vec s}_1,\dots,{\vec s}_N} e^{-\beta[-J\sum_{i=1}^N {\vec s}_i.{\vec s}_{i+1}]} = tr[T^N]$$

donde ${\vec s}_i=(s_i^x,s_i^y,s_i^z)$ son vectores con componentes continuas. El gran desafío es diagonalizar $T$ que tiene dimensión $\infty \times \infty$, pero esto puede hacerse sin problemas encontrando los autovectores que en realidad son autofunciones en este caso...




### $Z$ como una suma sobre caminos...

¿Y que tiene que ver con la formulación de la cuántica de Feynman?

El formalismo de la matriz de transferencia en mecánica estadística tiene además una conexión con la mecánica cuántica. Se puede mostrar por ejemplo, que el modelo de Ising 1d es isomorfo al problema de **un solo** espín cuántico en un campo transversal. La conexión se ve cuando uno usa la formulación de la mecánica cuántica en términos de integrales de camino de Feynman (ver por ejemplo cap 5.8 del libro de Chandler).  

Para ver que hay "caminos" escondidos recordemos esta forma

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1}
\langle \sigma_1| {\hat T} | \sigma_2 \rangle
\langle \sigma_2| {\hat T} | \sigma_3 \rangle
…
\langle \sigma_N| {\hat T} | \sigma_1 \rangle
$$

Y piensen que $\sigma_i$ es el estado de un espín solito al "tiempo imaginario" $t=i$, es decir $\sigma(t=i)=\sigma_i$.
Viendolo así, $\hat T$ es un "operador de evolución", que lleva al espín de un tiempo $i$ al siguente, y asi sucesivamente.

La suma entonces es una suma sobre todos los caminos posibles que salen y llegan al mismo estado $\sigma_1$ y al final sumo sobre estos estados. Es decir, la expresión de $Z$ en esta analogía es una suma sobre caminos seguidos por $\sigma(t)$m entre $t=1$ a $t=N$.
