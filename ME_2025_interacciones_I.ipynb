## Ensembles

*Qu quiere decir resolver un problema en mecnica estadstica de equilibrio?*

Bsicamente, derivar la termodinmica de un modelo microscpico. Operativamente se reduce en general a hacer una suma sobre microestados, para calcular la densidad de estados del sistema en el microcannico o bien la funcin de particin en los otros ensembles, ya que eso nos permite en principio obtener cualquier valor medio o distribucin de probabilidad, en equilibrio, y encontrar la entropa o una energa libre en funcin de sus variables naturales.

\begin{align}
\Omega(E,V,N)&=\sum_\omega \delta_{E_\omega,E}  S(E,V,N)=k_B \ln(\Omega) \\
  Z_F(T,V,N)&=\sum_\omega e^{-\beta E_\omega} = \sum_E \Omega(E,V,N)e^{-\beta E}  F(T,V,N)=-k_B T \ln(Z_F)\\
  Z_{\Theta}(T,\mu,V)&=\sum_\omega e^{-\beta (E_\omega-\mu N_\omega)}
  =\sum_N  Z(T,V,N) e^{\beta \mu N} 
  \Theta(T,\mu, V)=-k_B T \ln(Z_{\Theta})
  \\
  Z_G(\beta, P, N)&=\sum_\omega e^{-\beta (E_\omega + PV_\omega)}
  =\sum_V Z(T,V,N) e^{\beta PV} 
  G(T,P,N)=-k_B T \ln(Z_G)
\end{align}

En cada caso $\sum_\omega$ contempla microestados compatibles con las condiciones impuestas. En el caso mas general $PV$ debe ser reemplazado por $-{\bf J}.{\bf X}$, las variables conjugadas que aparezcan en el trabajo hecho sobre el sistema, siendo ${\bf X}$ la extensiva (Ejemplo: en el polmero es $\tau L$, y en un sistema magtico $-{\bf B}.{\bf M}$).


Cada ensemble tiene su conveniencia. En particular, muchas veces resulta prctico eliminar la restriccin de una suma, sumando sobre ms variables. Por ejemplo, en el microcannico a veces es difcil filtrar los microestados con $E$ fija. El cannico permite evitar ese filtro al sumar sobre energas distintas, usando el factor de Boltzmann. El cannico por otro lado, necesita filtrar los estados con $N$ fijo, lo cual a veces se torna complicado (en particular en estadstica cuntica), y el gran cannico elimina la necesidad de ese filtro. Etc. Esto se refiere a la termodinmica, pues en ese caso todos los ensembles son equivalentes y la eleccin es por conveniencia prctica. Si no es termodinmica, la eleccin est motivada por razones fsicas propias del sistema.


Lo importante de las sumas de arriba es que nos dan la entropa, energa libre de Helmholtz, energa libre de Landau, energa libre de Gibbs, siempre en *funcin de sus variables naturales*, es decir, la informacin termodinmica completa del modelo (no hace falta complementar con ninguna propiedad sacada de otro lado..). Y por supuesto, funciona para cualquier conjunto de variables termodinmicas, sea el sistema un gas, un sistema magntico, un slido, un lquido, o cualquier otro que pueda describirse termodinmicamente.

*En principio*, tenemos todo lo que necesitamos para extraer la termodinmica de cualquier sistema microscpico. El problema es que para la mayora de los modelos microscpicos no triviales, hacer las cuentas es un gran desafo tcnico y requiere de un gran esquema de aproximaciones, includas las resoluciones numricas.



# Interacciones en ME clsica

El formalismo de los ensembles estadsticos de equilibrio es  universal: podemos *en principio* aplicarlo a cualquier modelo microscpico en principio, definido por sus variables microscpicas y su hamiltoniano o energa por microestado.

La resolucin del problema termodinmico, es decir la obtencin de las energas libres en trminos de sus variables naturales a partir de un modelo microscpico, es sin embargo difcil en general, salvo por unos pocos pero instructivos casos, como los gases ideales y los paramagnetos. La principal dificultad que se presenta en el clculo son las interacciones entre las componentes del sistema, porque atentan contra una factorizacin sencilla del sistema, necesaria dado el enorme nmero de microestados.

Pero son precisamente las interacciones las que dan lugar a las correlaciones espaciales y al comportamiento colectivo de la materia, y en particular a sus distintas fases y transiciones de fase, de modo que no podemos despreciarlas si queremos modelar estos fenmenos.   

Aunque algunos pocos modelos con interacciones no triviales son exactamente solubles e instructivos, en general para resolver el problema con muchos grados de libertad hay que recurrir a distintos mtodos de  aproximacin o mtodos numricos para obtener las propiedades termodinmicas.



## Dos espines interactuantes

  Para fijar ideas, consideremos un ejemplo extremadamente simple de dos espines distinguibles $\sigma_1=\pm 1$ y $\sigma_2=\pm 1$ interactuantes

  \begin{align}
  E = -J \sigma_1 \sigma_2
  \end{align}

con $J>0$. Fsicamente, es una interaccin que tiende a alinear los espines, y es el tipo de interaccin responsable por el ferromagnetismo en muchos materiales. Es evidente que el estado fundamental del sistema es degenerado, $\sigma_1=\sigma_2=\pm 1$, con $E_0=-J$ en ambos casos. Est claro que los espines no son independientes, al menos a baja temperatura, ya que prefieren ponerse de acuerdo para lograr configuraciones mucho ms probables.

En el ensemble de Gibbs, tenemos que la probabilidad de un microestado es

\begin{align}
P(\sigma_1,\sigma_2) =\frac{\exp[\beta(J \sigma_1 \sigma_2 + B_1 \sigma_1 + B_2 \sigma_2)]}{Z(T,B_1,B_2,J)}
\end{align}

donde $B_1=B_2=B$ pero lo mantendremos en general por conveniencia computacional:

\begin{align}
\langle \sigma_i \rangle &= \frac{\beta^{-1}\partial_{B_i}Z}{Z}= \beta^{-1}\partial_{B_i}\ln(Z) &=
-\partial_{B_i}\ln(G) \\
\langle \sigma_1 \sigma_2 \rangle &= \frac{\beta^{-1}\partial_{J}Z}{Z}= \beta^{-1}\partial_{J}\ln(Z) &=
-\partial_{J}\ln(G)
\end{align}


El sistema tiene 4 microestados $(\sigma_1,\sigma_2)\in \{(-1,-1), (-1,1), (1, -1), (1,1)\}$ con magnetizaciones $\mu_0(\sigma_1+\sigma_2)\in \mu_0\{-2, 0, 0, 2\}$ respectivamente y energas $-J\sigma_1\sigma_2=J\{-1, 1, 1, -1\}$.


| Microestado | S1 | S2 | E  | M  | E-S1 B1-S2 B2     | Peso             |
|-------------|----|----|----|----|-------------------|------------------|
| 1           | 1  | 1  | -J | 2  | -(J+B1+B2)        | exp(b(J+B1+B2))  |
| 2           | -1 | -1 | -J | -2 | -(J-B1-B2)        | exp(b(J-B1-B2))  |
| 3           | 1  | -1 | J  | 0  | J - S1 B1 + S2 B2 | exp(-b(J-B1+B2)) |
| 4           | -1 | 1  | J  | 0  | J + S1 B1 - S2 B2 | exp(-b(J+B1-B2)) |


La funcin de particin de Gibbs con un campo $B$ aplicado es una suma sobre los cuatro microestados posibles

\begin{align}
Z &= 2 e^{\beta  (-J)} \cosh [\beta  (B_1-B_2)]+2 e^{\beta  J} \cosh [\beta  (B_1+B_2)]
\end{align}


>  Notar que en ausencia de interaccin
  \begin{align}
  Z|_{J=0} &= 2 \cosh [\beta  (B_1-B_2)]+2 \cosh [\beta  (B_1+B_2)]
  \end{align}
  por otro lado,
  \begin{align}

  \end{align}
  y para que coincidan
  $4\cosh(\beta B_1)\cosh(\beta B_2)=2\cosh [\beta  (B_1-B_2)]+2 \cosh [\beta  (B_1+B_2)]$

* **Dificultad** La solucin con interaccin es ms dificil que la de sin interaccin, pero esto es mucho ms serio cuando el nmero de espines crece, porque en el caso interactuante para obtener $Z$ necesitamos sumar $2^N$ microestados y conocer la magnetizacin y energa de cada uno, mientras, mientras que en el caso sin interaccin la suma $Z_1$ sigue siendo la de los dos microestados de un solo espn y $Z=Z_1^N$. * **Correlaciones** La solucin con interaccin tiene fluctuaciones correlaciones (ver [notebook de mathematica](https://www.wolframcloud.com/obj/cd6b4654-8c45-48bb-8542-d5e5e2b3fb30))
\begin{align}
g_{12}&=\langle \sigma_1 \sigma_2 \rangle - \langle \sigma_1 \rangle \langle \sigma_2 \rangle \\
&=
\frac{e^{4 \beta  J}-1}{\left(\cosh (\beta  ({B_1}-{B_2}))+e^{2 \beta  J} \cosh (\beta  ({B_1}+{B_2}))\right)^2}
\end{align}
Notar que $g_{12}=0$ si $J=0$ y que $g_{12}=g_{21}>0$. Este tipo de correlaciones, cuando se extienden a grandes distancias, dan lugar a las transiciones de fase.

  Una consecuencia es que la susceptibilidad del sistema, cambia debido a las correlaciones o al comportamiento colectivo. Si $\langle M \rangle =\langle \sigma_1 + \sigma_2 \rangle$, y tomamos $B_1=B_2=B$,

  \begin{align}
  \chi = \partial_B \langle M \rangle|_{B=0} = \frac{4 \beta  e^{2 \beta  J}}{e^{2 \beta  J}+1}
  \end{align}

  diferente de la ley de Curie que derivamos para espines no interactuantes. Por otro lado sabemos que

  \begin{align}
  \chi = \beta^{-1} [\langle M^2 \rangle - \langle M \rangle^2] =
  \beta^{-1} [g_{11}+g_{22}+g_{12}+g_{21}]
  \end{align}

  de modo que la susceptibilidad tiene dos contribuciones extra debidas a las correlaciones.
  Fisicamente, significa que el sistema como un todo responde diferente debido a las interacciones. En este caso, la respuesta es mas alta pues $g_{12}=g_{21}>0$.

  Notar que a pesar de estar los dos estados de ms baja energa completamente magnetizados $\langle M \rangle=0$, como en el caso con  $B=J=0$. Lo que diferencia al caso interactuante es que $P(M=0)\to 0$ para $T\to 0$, debido justamente a las correlaciones entre los espines. Para verlo ms claramente, calculemos $P(M)$. Para eso hay que tener claro que en este caso de un par de espines los posibles valores son $M=-2,0,2$ (para un sistema de muchos espines interactuantes hay muchos valores mas!) .

  \begin{align}
  P(M=0)&=\frac{e^{-\beta J}}{Z} \\
  P(M=\pm 2)&=\frac{e^{\beta J}}{Z} \\
  \end{align}

  de modo que $P(M=0)\to 0$ y $P(M=\pm 1)\to 1/2$ cuando $T\to 0$. Notar que si no hubieran interacciones $P(M=0)= 1/2$ y $P(M=0)=1/4$ cuando $T\to 0$. Esta es otra forma de ver el efecto de las correlaciones entre espines.

## Sistemas interactuantes con muchos grados de libertad:

* Sistema de $N$ partculas interactuantes de a pares en tres dimensiones

  $$H=\sum_i p_i^2/2m + \sum_{i=1}^N \sum_{j=i+1}^N U(r_{ij})$$

  con $r_{ij}=\sqrt{|{\bf x}_i-{\bf x}_j|^2}$.
  Funcin de particin cannica:

  \begin{align}
  Z = \frac{1}{N! h^{3N}} \int \prod_i d^3x_i d^3p_i e^{-\beta[\sum_i p_i^2/2m + \sum_{i=1}^N \sum_{j=i+1}^N U(r_{ij})]} \\
  = Z_{ideal} V^{-N} \int \prod_i d^3x_i e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N U(r_{ij})]} = Z_{ideal} V^{-N} Q
  \end{align}

  donde $Q(T,V,N)$ **no es trivialmente separable** (convencerse!), salvo que $U$ sea "diagonalizable", por ejemplo, una forma cuadrtica como la que aparece en pequeas oscilaciones.

* Sistema de $N$ espines interactuantes:

  $$H= \sum_{i=1}^N \sum_{j=i+1}^N J_{ij} \sigma_i \sigma_j$$
  
  Funcin de particin cannica:

  \begin{align}
  Z = \sum_{\sigma_1,\dots,\sigma_N} e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N J_{ij} \sigma_i \sigma_j]} \\
  = \sum_{\sigma_1,\dots,\sigma_N}
  e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N J_{ij} \sigma_i \sigma_j]}
  \end{align}

  la cual **no es trivialmente separable** para una $J_{ij}$ general, ni siquiera para $J_{ij}=J$.  

* Sistemas de partculas adsorbidas interactuantes en $M$ sitios

  $$H= \sum_{j=i+1}^M J_{ij} n_i n_j$$
  
  Funcin de particin gran-cannica:

  \begin{align}
  Z = \sum_{n1_1,\dots,n_N} e^{-\beta[\sum_{i=1}^N \sum_{j=i+1}^N J_{ij} n_i n_j - \mu \sum_{i=1}^M n_i]}
  \end{align}

  la cual **no es trivialmente separable** a menos que por ejemplo $J_{ij}$ sea cero a menudo tal de poder separar el sistema en clusters independientes.

* **La excepcin** Pequeas oscilaciones de un sistema con cualquier nmero de grados de libertad, alrededor de un mnimo de su energa. Por ejemplo, considerar la lagrangiana de este sistema mecnico:

  $${\cal L}= \frac{1}{2}\sum_{i,j}[\dot{x_i} M_{ij} \dot{x_j} - x_i K_{ij} x_j] $$

  donde las matrices $K$ y $M$ son matrices semidefinidas positivas, asi como las formas cuadrticas que se forman con ellas. El caso usual de pequeas oscilaciones en torno a un equilibrio estable o semiestable. Un solo cambio de variables puede diagonalizar simultneamente la matriz de masas y de resortes y llevarnos a  

  $${\cal L}= \sum_{\alpha}\frac{m_\alpha}{2}[\dot Q^2_\alpha - \omega_\alpha^2 Q_\alpha^2]$$

  y en particular al hamiltoniano,

  $${\cal H}= \sum_{\alpha}[P_\alpha^2/2m_\alpha + m_\alpha \omega_\alpha^2 Q_\alpha^2]$$

  donde las coordenadas normales estn completamente desacopladas, de modo que

  $$Z=\prod_\alpha Z_\alpha$$

  donde $Z_\alpha$ es la funcin de particin del modo $\alpha$, y es fcilmente calculable ya que es la de *un solo oscilador armnico unidimensional* con frecuencia $\omega_\alpha$. Gracias a esto vamos a poder resolver fcilmente el problema de las vibraciones de los slidos (fonones en el tratamiento cuntico) y otras excitaciones de baja energa en una red llamadas cuasipartculas (magnones, excitones, polaritones, plasmones, polarones, anyones, etc). ## Generalidades

En el ejemplo anterior vimos que adems de la aparicin de correlaciones, el problema es que las sumas o el conteo se puede volver muy difcil en los casos en que hay interacciones y muchos microestados o grados de libertad.

* **Dificultad**: El nmero de microestados sobre el que hay que sumar para obtener las energas libres crece exponencialmemte rpido con $N$, ya que a energa fija $\Omega \sim e^{-S/k_B}$ con $S\propto N$.

* **Factorizacin** Qu son los problemas factorizables? (contexto cannico para fijar ideas)

  Por ejemplo, en el cannico, un sistema factorizable es uno con $N$ grados de libertad, donde $E(\omega)=E_1(_1)+E_2(_2)+\dots+E_M(_M)$, tal que un microestado $\omega$ de todo el sistema est dado por el producto externo de los microestados de los $M \propto N$ subsistemas, $\omega \equiv \{\omega_1,\omega_2,\dots,\omega_M\}$. Un ejemplo son un sistema de $N$ tomos que forman $M$ molculas no interactuantes, cada una de ellas con posiblemente varios tomos interactuantes.
  Entonces
  
  $$Z=\sum_\omega e^{-\beta E_\omega}
  =
  \sum_{\omega_1} e^{-\beta E_{1}(\omega_1)}  \sum_{\omega_2} e^{-\beta E_{2}(\omega_2)}\dots \sum_{\omega_M} e^{-\beta E_{M}(\omega_M)}
  \propto Z_1 Z_2 Z_3 \dots Z_M
  $$

  Notar que puede o no haber un $M!$ dividiendo dependiendo la indistinguibilidad de las partes, pero lo importante aqu es notar que podemos evaluar $Z$ facilmente para $N$ grande si cada $Z_i$ es computable. Notar que la factorizacin implica independencia de los subsistemas, $p(\omega_1, \dots, \omega_N)=p_1(\omega_1)\dots p_M(\omega_M)$.

  Ejemplos: Gas ideal, gases de molculas no interactuantes, espines o dmeros no interactuantes, pequeas vibraciones de un slido, el campo electromagntico en una cavidad. En todos $Z \propto Z_1^N$ o $Z \propto Z_1 Z_2\dots Z_N$, con cada $Z_i$ calculable analticamente (a veces es ms sencillo an porque todos los $Z_i$ iguales).
  
  Que tipo de factores $Z_i$ hay? representa cada uno necesariamente una particula o grado de libertad? En cualquier caso, lo importante es poder calcular los $Z_i$ analticamente. Los $Z_i$ puede ser funciones de particin de sistemas con un nmero finito de grados de libertad, y por lo tanto computables.

# Modelo de Ising

Los sistemas interactuantes descriptos por variables discretas definidas sobre una red son a menudo muy convenientes tanto para los clculos analticos como para las simulaciones numricas, y arrojan luz sobre muchas cosas generales, que van mucho ms alla de los detalles de los modelos.

Entre los modelo ms estudiados esta el [**modelo de Ising** ](https://en.wikipedia.org/wiki/Ising_model). Este modela la energa de un sistema de espines 1/2, o lo que es equivalente, de sistemas de dos niveles localizados en sitios de una red dada. En realidad es el modelo que invent Lenz y se lo di a Ising para que lo resuelva, quien concluy errneamente que no haba transicin de fase en ninguna circunstancia. El modelo de Ising ferromagntico es de hecho el primer modelo de mecnica estadstica que predijo una transicin de fase.  

Supongamos que tenemos variables $\sigma_i=\pm 1$ sobre una red, en presencia de un campo magntico $B$ y agregamos una interaccin o *acoplamiento* de magnitud $J_{ij}$ entre los sitios $i$ y $j$. La probabilidad de un microestado es $p_\omega = e^{-\beta H}/Z$ con $H$ dado por

$$H = -\sum_{i<j} J_{ij} \sigma_i \sigma_j - \sum_i B \sigma_i$$

donde $J_{ij}$ es la constante de interaccin entre el sitio $i$ y el $j$, y la tomamos positiva/negativa para favorecer/desfavorecer alineamiento del espn $i$ con el $j$. El campo $B$ es una notacin abreviada para $\mu B'$, donde $\mu$ es el momento magntico del espn y $B'$ el campo magntico. Este tipo de acoplamiento se suele llamar acoplamiento $Zeeman$. Fsicamente los $\sigma_i$ corresponden a los posibles valores que puede tomar la proyeccin sobre un eje el espn $1/2$ de una partcula. Cuando la energa depende solo de esta proyeccin es porque existe una fuerte anisotropa.

Formulado as, el problema es muy general, y las soluciones dependen crucialmente de la estructura de los acoplamientos $J_{ij}$. Es tan general que todava no dijimos en que red estn localizados estos espines, ni en que dimensin. De hecho, as escrito es un modelo de mecnica estadstica en un grafo arbitrario cuyas conexiones son de magnitud $J_{ij}$. Por lo tanto, hay una cuestin de **topologa** que hay que definir bien antes de resolver, as como la asignacin de los $J_{ij}$.
As como esta escrito, puede servir tanto para describir un ferromagneto o antiferromagneto, o un fludo sobre una red, o una aleacin binaria, o una memoria asociativa en una red neuronal, o la opinin binaria en una red social de agentes que se influencian, o el modelo de un vidrio de espines, etc, etc.

Si vamos a modelar por ejemplo un ferromagneto, es natural suponer que $J_{ij}=J>0$ para espines vecinos en la red asi favorecemos la alineacin de espines cercanos ($J<0$ favorece orden local antiferromagntico), y tomar $J_{ij}=0$ para todos los otros pares de espines. As llegamos al famoso [modelo de Ising ferromagntico](https://es.wikipedia.org/wiki/Modelo_de_Ising).

$$H = - J\sum_{\langle i,j \rangle} \sigma_i \sigma_j - \sum_i B \sigma_i$$

donde la notacin $\langle i,j \rangle$ significa que solo vamos a sumar sobre pares de espines vecinos en una red regular de algn tipo (hay que tener mucho cuidado de no sumar el mismo par de espines dos veces). Es decir, que vamos a considerar interacciones de corto alcance para empezar.

*Que podemos decir de este sistema sin resolver nada?*

Por un lado es claro que a $T\to \infty$, o ms precisamente $k_B T \gg J$, los espines se vuelven independientes y el sistema se comporta como un *paramagneto*. Es decir, tiene magnetizacin neta solamente como respuesta a un campo $B$ aplicado.

Por otro lado es de esperar que a $T\to 0$ el sistema busque minimizar $H$. Para eso, si $B=0$, es muy conveniente que $\sigma_i=\sigma_j$, ya que eso minimiza la energa de interaccin, que vale $-J$ por cada par. Por lo tanto, el estado fundamental a $B=0$ es degenerado:

$$E_{min}=-J N z/2,\;\; \sigma_i=1 \forall i,\; o \;\sigma_i=-1 \forall i,$$

donde $z$ es el nmero de vecinos de un espn (la red es regular, por lo tanto cada espn tiene el mismo nmero de vecinos), y el $1/2$ se usa para corregir el doble conteo de pares de espines interactuantes. Los estados que tienen esta energa mnima son dos, segn $\sigma_i=\pm 1$. Es decir el *estado fundamental* del sistema tiene degeneracin $g=2$. En otro lenguaje, podramos decir que el modelo tiene dos memorias que se recuperan cuando $T\to 0$.

El estado fundamental degenerado esta totalmente magnetizado, y sus magnetizaciones son $m=\pm N$ respectivamente.
Si prendemos el campo magntico $B>0$ la energa de uno de estos dos estados, el de $\sigma_i=1$ se hace menor a la del otro, y se rompe la degeneracin, quedando un estado fundamental nico,  

$$E_{min}=-J N z/2 -  B N,\;\; \sigma_i=1 \forall i$$

Este estado est totalmente magnetizado, y su magnetizacin es mxima $m=N$.




## Magnetizacin espontnea

Concentremosnos de nuevo en $B=0$. Supongamos que estamos en el estado totalmentre magnetizado con $\sigma_i=1 \forall i$. Que ocurre si aumentamos la temperatura? Una cosa que sabemos que ocurre seguro es que a $T\to \infty$, $m=0$, porque todos los estados se vuelven equiprobables. En particular, los estados de un sitio son equiprobables, y entonces $\langle \sigma_i \rangle=0, \forall i$. Entonces tenemos que $M=\mu N$ para $T\to 0$, y $M=0$ para $T\to \infty$.

Que ocurre a temperaturas intermedias?

* como decae $\langle M \rangle=N\langle \sigma_i \rangle$ aumentando $T$ desde cero?
* Cambia suavemente o bruscamente?
* Hay alguna temperatura especial?

Para saber la respuesta necesitaramos calcular,

$$\langle M \rangle = N \langle \sigma_i \rangle \equiv \partial_B \ln(Z)$$

vs $T$, por ejemplo, usando el ensemble cannico. El primer paso es calcular la funcin de particin cannica

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1} e^{-\beta H(\sigma_1,\dots,\sigma_N)}=
\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1} e^{\beta (J\sum_{\langle i,j \rangle} \sigma_i \sigma_j + \sum_i B \sigma_i)}
$$

Esta es una suma de $2^N$ configuraciones que no podemos factorizar de manera sencilla culpa de las interacciones. Si no fuera por ellas, ya vimos que $Z=Z_1^N$. Debemos adoptar entonces otra estrategia para poder hacer esta suma. Vamos a ver de hecho muchas estrategias, algunas exactas, aplicables para ciertas circunstancias muy especiales, y otras aproximadas, pero aplicables a muchas circunstancias.

Hay sin embargo, un problema muy importante que salta a la vista a $B=0$, que es que una configuracin dada tiene la misma probabilidad que la obtenida invirtiendo todos los espines $\sigma_i \to -\sigma_i$, y que tiene la misma magnetizacin pero opuesta. Esto significa que $m=0$. Entonces no habra nunca una magnetizacin espontanea, ni una transicin de fase segn la mecnica estadstica. Y no es una propiedad del modelo de Ising solamente. Matemticamente, aplicar la teora de ensembles da siempre $m=0$ si hay una simetra de inversin *temporal* (ojo, es una simetra temporal y no espacial; Se puede pensar como analoga para el espn, que invertir el tiempo es como invertir la direccin de la corriente en una bobina, su momento magntico cambia de direccin.). Pero la ausencia de magnetizacin espontnea, es decir sin campo $B$ aplicado, implica que no puede haber ferromagnetismo!

**Error: acabamos de demostrar que no existe el ferromagnetismo o magnetizacin espontnea para cualquier modelo con simetra de inversin temporal!**

Tenemos que volver a la fsica y tratar de interpretar correctamente las cosas.

Consideremos la probabilidad de observar una magnetizacin

\begin{align}
p(m_0)=\frac{\sum_{\sigma_1,\dots, \sigma_N=\pm 1} e^{-\beta H(\sigma_1,\dots, \sigma_N)} \delta
\left(\sum_{i=1}^N \sigma_i-m_0 \right)}{Z}
\end{align}

Como $m$ es una variable aleatoria, podemos pensar a $Q(m,B,T)=p(m)Z$ como una funcin de particin a magnetizacin fija,

$$Z(B,T)=\sum_{M=-N}^N Q(M,B,T)$$.

y a $A(m_0,T)=-kT \ln (Q(m_0))$ como una energa libre *con el vnculo de magnetizacion fija*, que representa el trabajo reversible requerido para cambiar la magnetizacin a temperatura fija, entre dos estados de equilibrio a distintas magnetizaciones.

Si dibujamos $A(m,T)$ para $B=0$ tiene tpicamente forma de doble pozo, con una barrera en $M=0$, reflejando el hecho de que hay dos estados igualmente estables termodinmicamente. Lo interesante es que en general va a tener una barrera separando los dos mnimos y **la magnitud de esta barrera es crucial**.

En otras palabras, $p(m_0)$ tiene a bajas temperaturas dos picos, uno cerca de $N\mu$ y el otro cerca de $-N\mu$. Si el sistema tiene magnetizacin positiva, para pasar a magnetizacin negativa debe necesariamente pasar espontneamente por una configuracin de magnetizacin cero, pero esta configuracin es poco probable cuando $T\to 0$ y $N\to \infty$. Esa es la barrera. En algunos casos la barrera es practicamente infranqueable a $T$ suficientemente baja, y tenemos magnetizacin espontnea y ferromagnetismo, y en otros no, y no tenemos ferromagnetismo a ninguna $T$.


Pensemos no en el ensemble, sino en un sistema en particular. Si en un momento dado el sistema esta cerca de un mnimo de la energa libre con una magnetizacin espontnea, para ir al otro mnimo es necesario que absorba energa suficiente como para pasar la barrera. Esto tiene que ocurrir como resultado de una fluctuacin trmica. El punto crucial es el siguiente: *Los sistemas con transicin de fase ferromagntica tienen una barrera que escalea como el nmero de espines*, por ejemplo como $N^{1/2}$ para $d=2$, o $N^{2/3}$ para $d=3$. Por lo tanto, en el lmite termodinmico, tal fluctuacin es improbable, porque la varianza de $m$ es pequea comparada con $m$. En particular, $p(m=0)\to 0$, y el sistema se pasa muchsimo tiempo cerca del mnimo inicial. Por lo tanto, a los fines prcticos, presenta magnetizacin espontnea. Sistemas con un nmero pequeo de espines tienen una barrera relativamente baja, y la magnetizacin flucta mucho ms fuertemente que en un sistema de $N \gg 1$. Esto implica, que existe una probabilidad no nula de invertirse en el tiempo. En $d=1$, sin embargo, *para cualquier nmero de espines $N$* veremos que la barrera tiende a cero por una cuestin entrpica, de modo que la magnetizacin se pasea ampliamente y se promedia a cero. La barrera tiene que ver con el fnomeno de ncleacin y crecimiento que necesitamos para que un sistema invierta su magnetizacin. Este es un fenmeno dinmico, pero podemos servirnos de la mecnica estadstica para entender las escalas de energa que entran en juego.

Pero como, no era que los sistemas que considerabamos para hacer mecnica estadstica eran ergdicos, que exploraban todo su espacio de fases o de configuraciones? Bueno, no, **en las transiciones de fase continuas hay usualmente una ruputura efectiva de la ergodicidad** y esta viene acompaada de una **ruptura espontnea de la simetra** .

Cuando hay campo $B$, en un sistema ferromagntico, uno de los dos mnimos de la energa libre a $m$ fijo esta ms bajo que el otro, y correponde a una mayora de los espines apuntando en la direccin del campo. Este es el mnimo estable, el estado termodinmico de equilibrio. Sin embargo, si preparamos el sistema en el otro mnimo, y $B$ no es muy grande, la barrera todava estar. Esta situacin es metastable y es la responsable de la histresis de os ferromagnetos cuando uno hace un ciclado en campo, desde valores negativos grandes a positivos grandes, y de vuelta. El campo donde la magnetizacin salta, es el campo coercitivo.

Veremos en lo que sigue que

* El modelo de Ising ferromagntico en $d=1$ tiene solucin exacta en el lmite termodinmico, pero **no** tiene transicin de fase.

* El modelo de Ising ferromagntico en $d=2$ tiene solucin exacta en el lmite termodinmico, y tiene transicin de fase.

* El modelo de Ising ferromagntico en $d>2$ no tiene solucin exacta en el lmite termodinmico pero tiene transicin de fase.

* El modelo de Ising ferromagntico en $d=\infty$ tiene solucin exacta en el lmite termodinmico y tiene transicin de fase.

## Solucin exacta del modelo de Ising en d=1 para $N \gg 1$

Las soluciones exactas de sistemas interactuantes no factorizables de muchos grados de libertad son escasas y usualmente representan casos demasiado idealizados. Sin embargo son muy valiosas ya que permiten poner a prueba esquemas de aproximacin analtica y numrica.

Empecemos con el caso de una red regular de sitios equiespaciados en una red.
En este caso

$$H= -J \sum_i \sigma_i \sigma_{i+1} - \mu B \sum_i\sigma_i$$

y notar que los primeros vecinos son los de la derecha y la izquierda, y que la suma sobre pares se puede hacer movindonos solamente en la direccin creciente del ndice del sitio.

La funcin de particin queda

$$Z=
\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1} e^{\beta (J \sum_i \sigma_i \sigma_{i+1} + \mu B \sum_i\sigma_i)}
$$

y todava parece difcil de sumar. Sin embargo, existe en este caso un truquito muy ingenioso. Rescribimos primero el hamiltoniano

$$H= -J \sum_{i=1}^N\sigma_i \sigma_{i+1} - \mu B \sum_{i=1}^N (\sigma_i+\sigma_{i+1})/2$$

y pensemos para empezar en condiciones de contorno peridicas, de tal forma que $\sigma_{N+1}\equiv \sigma_1$. Es decir, que la red tiene forma de anillo. Se pueden elegir otras condiciones de contorno sin problemas, pero esta nos va a resultar ms simple por ahora y por otro lado, esperamos que las condiciones de contorno no sean tan importantes para el comportamiento colectivo cuando $N\gg 1$. La simetrizacin del trmino de acoplamiento con $B$, o acoplamiento Zeeman, nos va a servir para escribir $Z$ en una forma muy compacta.

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1}
T_{\sigma_1,\sigma_2}T_{\sigma_2,\sigma_3}\dots T_{\sigma_N,\sigma_1}
$$

donde

$$T_{\sigma_i, \sigma_{i+1}}
= e^{\beta [J \sigma_i \sigma_{i+1}+B\mu(\sigma_i + \sigma_{i+1})/2]}
$$

Si pensamos a estas cantidades como elementos de una matriz, variando el valor de $\sigma_i=\pm 1$ y $\sigma_j=\pm 1$, tenemos una matriz de 2x2

$${\cal T}=
\begin{pmatrix}
T_{++} & T_{+-}\\
T_{-+} & T_{--}
\end{pmatrix}
=
\begin{pmatrix}
e^{\beta (J + B\mu)} & e^{-\beta J}\\
e^{-\beta J} & e^{\beta (J - B\mu)}
\end{pmatrix}
$$

que se denomina *matriz de transferencia*. Para ver la ventaja de representar en trminos de esta matriz, notemos que los dos ltimos factores en $Z$, los nicos que involucran la ltima sumatoria en $\sigma_N$, se pueden escribir como

$$
\sum_{\sigma_N=\pm 1} T_{\sigma_{N-1},\sigma_{N}}T_{\sigma_{N},\sigma_{1}}
\equiv [{\cal T}^2]_{\sigma_{N-1},\sigma_1}
$$

ahora, si hacemos la suma en $\sigma_{N-1}$ tenemos

$$
\sum_{\sigma_{N-1}=\pm 1} T_{\sigma_{N-2},\sigma_{N-1}}{\cal T}^2_{\sigma_{N-1},\sigma_{1}} \equiv {\cal T}^3_{\sigma_{N-2},\sigma_{1}}
$$

y asi uno podra seguir hasta la ltima suma y encontrar que


$$Z = \sum_{\sigma_1=\pm1} T^{N}_{\sigma_1,\sigma_1} \equiv \text{Tr}[{\cal T}^N]$$

<blockquote>
Una forma piola de ver lo mismo es pensar a la matriz de transferencia como un operador $\hat T$ sobre el espacio de estados de $\sigma$ y usar la notacin de bra-kets de mecnica cuntica

$$T_{\sigma_i,\sigma_j}=\langle \sigma_i | {\hat T} | \sigma_j  \rangle$$

entonces la funcin de particin adquiere una forma muy sugestiva

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1}
\langle \sigma_1| {\hat T} | \sigma_2 \rangle
\langle \sigma_2| {\hat T} | \sigma_3 \rangle

\langle \sigma_N| {\hat T} | \sigma_1 \rangle
$$

que se puede escribir as

$$Z=
\sum_{\sigma_1=\pm 1} \langle \sigma_1| {\hat T}
\left[\sum_{\sigma_2=\pm 1} | \sigma_2 \rangle
\langle \sigma_2| \right] {\hat T} \left[\sum_{\sigma_3=\pm 1}
| \sigma_3 \rangle \langle \sigma_3|\right]

\left[\sum_{\sigma_N=\pm 1}| \sigma_N \rangle \langle \sigma_N|\right] {\hat T} | \sigma_1 \rangle \\
=
\sum_{\sigma_1=\pm 1} \langle \sigma_1| {\hat T}
\left[\mathbb{1}\right] {\hat T} \left[\mathbb{1}\right]

\left[\mathbb{1}\right] {\hat T} | \sigma_1 \rangle
= \sum_{\sigma_1=\pm 1} \langle \sigma_1 | {\hat T}^N | \sigma_1 \rangle
$$
y obtenemos el mismo resultado. Sin embargo, la ltima expresin es muy sugerente porque es muy parecida a la que aparece en la formulacin de integrales de camino de la mecnica cuntica, donde se divide el operador de evolucin $e^{-iHt}$ en $N$ pedacitos...
</blockquote>


Para evaluar $Z$ recordemos que la traza de una matriz es igual a la suma de sus autovalores y que los autovalores de la matriz $\cal{T}$ son las potencias N-simas de los de la matriz $\cal{T}$. Por lo tanto, si $\lambda_1$ y
$\lambda_2$ son los autovalores de $\cal{T}$, llegamos a

$$Z= \lambda_1^N + \lambda_2^N$$

Notemos ahora que si $N\gg1$ solo importa el ms grande de los autovalores, supongamos que sea $\lambda_1>\lambda_2$, entonces

$$Z=\lambda_1^N+\lambda_2^N=\lambda_1^N(1+(\lambda_2/\lambda_1)^N)\to \lambda_1^N$$

Despus de algunas simplificaciones de $\lambda_1$, llegamos a

### Longitud de Correlacin

Una de las propiedades emergentes ms interesantes de las interacciones es la emergencia de correlaciones espaciales.
Calculemos la correlacin spin-spin en funcin de la distancia,

\begin{aligned}
\left\langle\sigma_i \sigma_{i+r}\right\rangle & =\frac{1}{Z} \sum_{\sigma_1= \pm 1} \sum_{\sigma_2= \pm 1} \cdots \sum_{\sigma_N= \pm 1} \sigma_i \sigma_{i+r} \prod_{j=1}^N \exp \left(K \sigma_j \sigma_{j+1}\right) \\
& =\frac{1}{Z} \operatorname{tr}\left[T^{i-1} \hat{\sigma}_z T^r \hat{\sigma}_z T^{N-i-r+1}\right]=\frac{1}{Z} \operatorname{tr}\left[\hat{\sigma}_z T^r \hat{\sigma}_z T^{N-r}\right]
\end{aligned}

En la segunda igualdad usamos el mismo truco de la matriz de Pauli que usamos cuando calculamos $\langle \sigma_i \rangle$.


\begin{aligned}
\left\langle\sigma_i \sigma_{i+r}\right\rangle & =\frac{1}{Z} \operatorname{tr}\left[\left(\begin{array}{cc}
0 & 1 \\
1 & 0
\end{array}\right)\left(\begin{array}{cc}
\lambda_{+}^r & 0 \\
0 & \lambda_{-}^r
\end{array}\right)\left(\begin{array}{cc}
0 & 1 \\
1 & 0
\end{array}\right)\left(\begin{array}{cc}
\lambda_{+}^{N-r} & 0 \\
0 & \lambda_{-}^{N-r}
\end{array}\right)\right] \\
& =\frac{1}{Z} \operatorname{tr}\left(\begin{array}{cc}
\lambda_{+}^{N-r} \lambda_{-}^r & 0 \\
0 & \lambda_{-}^{N-r} \lambda_{+}^r
\end{array}\right)=\frac{\lambda_{+}^{N-r} \lambda_{-}^r+\lambda_{-}^{N-r} \lambda_{+}^r}{\lambda_{+}^N+\lambda_{-}^N} .
\end{aligned}

\begin{equation}
\left\langle\sigma_i \sigma_{i+r}\right\rangle \approx\left(\frac{\lambda_{-}}{\lambda_{+}}\right)^r \equiv \mathrm{e}^{-r / \xi}
\end{equation}

Es decir que

\begin{equation}
\xi = [\ln(\lambda_+/\lambda_-)]^{-1} = -\frac{1}{\ln \tanh (\beta J)} = \frac{1}{\ln \coth (\beta J)}
\end{equation}

Este resultado es superinteresante. Por un lado, vemos que el mas grande de los autovalores de la matriz de transferencia nos da $Z$ y por el otro el cociente entre el ms chico y el ms grande nos da la longitud de correlacin, la cual decrece aumentando $T$ como es de esperarse.

Es notorio que $\xi$ diverge solo cuando $T\to 0$. Veremos luego que cuando $\xi$ diverge a una temperatura finita $T_c$, es la seal de una transicin de fase a esa temperatura.


Hay una conexin precisa entre la susceptibilidad magntica y la funcin de correlacin espacial. Recordemos que la suceptibilidad esta conectada a la varianza de la magnetizacin y que ese resultado es muy general e independiente de la forma de $H$ siempre y cuando aparezca un acoplamiento del tipo $M B$.
Es decir que
$$\chi = \frac{1}{k_B T}[\langle M^2 \rangle-\langle M \rangle^2]$$
En el caso de espines no interactuantes usamos el teorema del lmite central para escribir $\chi=N\chi_1$, con $\chi_1$ la susceptibilidad de un spin aislado. Con correlaciones, ya no podemos usar el mismo argumento y a cambio obtenemos,

\begin{align}
\chi = \frac{1}{k_B T}[\langle M^2 \rangle-\langle M \rangle^2]
=  
\frac{1}{k_B T}[
  \langle (\sum \sigma_i)^2 \rangle
  -
  \langle (\sum \sigma_i) \rangle^2
  ]
=
\frac{1}{k_B T}[
  \sum_{i,j} \langle \sigma_i \sigma_j \rangle
  -
  \sum_{i,j} \langle \sigma_i \rangle \langle \sigma_j \rangle
  ]
= \frac{1}{k_B T}[
  \sum_{i,j} [\langle \sigma_i \sigma_j \rangle - \langle \sigma_i \rangle \langle \sigma_j \rangle
  ]
  =
  \frac{1}{k_B T}[
  \sum_{i,j} g(|j-i|)
  ]
  =  
  \frac{N}{k_B T} \sum_r g(r) \sim \frac{N}{k_B T} \xi(T,J)
\end{align}


donde
$$g(r)=\langle \sigma_i \sigma_{i+r} \rangle- \langle \sigma_i \rangle \langle \sigma_j \rangle = \langle \sigma_i \sigma_{i+r} \rangle$$ usando ya que $\langle \sigma_i \rangle=0$ y $\xi=[\ln \coth(\beta J)]^{-1}$.

\begin{align}
\chi &= \frac{1}{k_B T}\left[\langle M^2 \rangle-\langle M \rangle^2\right] \\
&= \frac{1}{k_B T}
  \sum_{i,j} [\langle \sigma_i \sigma_j \rangle - \langle \sigma_i \rangle \langle \sigma_j \rangle
  ] \\
&=
  \frac{1}{k_B T}
  \sum_{i,j} g(|j-i|) \\
&=  
  \frac{N}{k_B T} \sum_r g(r) \sim \frac{N}{k_B T} \xi(T,J)
\end{align}



Notar que en un sistema infinito la funcin de correlacin depende slo de la distancia entre los dos espines en consideracin, es decir depende de $r$ y no de $i$, como resultado de la invariancia discreta de traslacin. Es decir, si nos trasladamos un parmetro de red, al siguiente sitio, el resultado no puede cambiar. Adems, podemos decir que $g(r) \equiv g(|r|)$ por simetra especular, es decir la correlacin no distingue el sentido del vector de desplazamiento.

### Orden espacial y magnetizacin espontnea

Vimos que a $B=0$, tenemos $\langle M \rangle =0$, para toda $T$. Esto **no** puede ser usado como argumento en contra de un orden ferromagtico, ya que ocurre en cualquier sistema con simetra de inversin temporal, en cualquier dimensin. Necesitamos entonces otros argumentos.

* Un argumento en contra de la transicin de fase, es que la solucin exacta no tiene signularidades a $T$ finita. En particular, la susceptibilidad, que debera divergir a $T_c$, diverge pero a $T=0$, como tambien lo hace el sistema no interactuante.

* Otro argumento que muestra que no hay orden ferromagntico es notar que la probabilidad de que un par de espines separados $r<\xi$ estn orientados en la misma direccin es alta por $g(r)$ es positiva y alta. Es decir, hay *dominios* con espines con el mismo signo. Sin embargo, para $r>\xi$, la correlacin se va a cero, significando hay una probabilidad apreciable de que a esa distancia el segundo espn este invertido la mitad de las veces. Pero si es as, es probable que exista un dominio de magnetizacin opuesta empezando ms all. Si tenemos un montn de dominios de tamao finito, con magnetizacin positiva o negativa, no queda otra que $\langle |M| \rangle =0$ si $B=0$.
Una forma de ver esto, es *renormalizar* el sistema, es decir, mirarlo desde mas lejos cada vez, es decir ir generando una serie de funciones de particin con menos grados de libertad (ver notas de E. Jagla).

* Supongamos por el contrario que hay una magnetizacin neta $\langle M \rangle>0$ a $T>0$. Entonces $\langle \sigma_i \sigma_{i+r} \rangle \sim 1$ para cualquier $r$, contradiciendo que la longitud de correlacin es finita.

* Otro argumento para la *ausencia de transicin de fase en $d=1$* es que
si partimos de todo magnetizado, el costo energtico de dos paredes que generen un dominio de espines up en un mar de espines down cuesta solamente $4J$. Por otro lado, como puedo poner esas paredes en cualquier posicin de las $N$ posibles, puedo ganar una entropa que crece con $N$ a cualquier $T=\epsilon>0$ para $N \gg 1$, con lo cual se hace altamente probable la configuracin excitada. Por lo tanto, tener paredes minimiza $F$. Pero tener muchas paredes no es consistente con un estado magnetizado. Es decir, si parto a $T=0$ totalmente magnetizado en un sentido, a $T>0$ ese orden se destruye.




**Conclusin: No hay transicin de fase ferromagntica en $d=1$**

### Mtodo de la Matriz de Transferencia

El mtodo que usamos para resolver el modelo de Ising 1d es muy poderoso y muy general, y adecuado para resolver una familia grande de problemas en $d=1$ con interacciones a primeros vecinos.

Para cualquier funcin de particin que podamos escribir como

$$Z=\sum_{s_1,\dots,s_N} e^{\sum_{i=1}^N B(s_i, s_{i+1})}
= \sum_{s_1,\dots,s_N} \prod_{i=1}^N e^{B(s_i, s_{i+1})}$$

podemos definir su matriz de transferencia

\begin{equation}
\left\langle s_i\right| T\left|s_j\right\rangle=\mathrm{e}^{B\left(s_i, s_j\right)}
\end{equation}

y calcular, en el caso de condiciones de contorno peridicas (si son abiertas hay una pequea correccin que no es importante si $N\gg 1$),

$$Z=tr[T^N] \approx \lambda_{max}^N$$

y la funcin de correlacin haciendo el cociente entre autovalores.

>El [teorema de Perron-Frobenius](https://en.wikipedia.org/wiki/Perron%E2%80%93Frobenius_theorem) nos garantiza que $\lambda_{max}$ no es degenerado porque $T$ es finita y con elementos positivos. Si por ejemplo tenemos esta matriz simtrica,
$$\left(
\begin{array}{cc}
 a & b \\
 b & c \\
\end{array}
\right)$$
sus autovalores son
$$\frac{1}{2} \left(-\sqrt{a^2-2 a c+4 b^2+c^2}+a+c\right),\frac{1}{2} \left(\sqrt{a^2-2 a c+4 b^2+c^2}+a+c\right)$$
El discriminante es $a^2-2 a c+4 b^2+c^2 = (a-c)^2 +4 b^2>0$, por lo tanto los autovalores son reales y diferentes. Como $a$ y $c$ son positivos, y la raz cuadrada es siempre menor a $a+c$, porque el determinante de la matriz es positivo, ambos autovalores son reales, positivos y diferentes.  


Por lo tanto, $Z$ ser una funcin analtica de los parmetros del hamiltoniano y de $T$, y solo podr exhibir una singularidad a $T=0$. Por lo tanto, no hay transicin de fase en $d=1$ en este tipo de modelos.

El formalismo no se limita a variables discretas. Puede ser aplicado sin problemas al hamiltoniano de Heisenberg por ejemplo

$$Z=\sum_{{\vec s}_1,\dots,{\vec s}_N} e^{-\beta[-J\sum_{i=1}^N {\vec s}_i.{\vec s}_{i+1}]} = tr[T^N]$$

donde ${\vec s}_i=(s_i^x,s_i^y,s_i^z)$ son vectores con componentes continuas. El gran desafo es diagonalizar $T$ que tiene dimensin $\infty \times \infty$, pero esto puede hacerse sin problemas encontrando los autovectores que en realidad son autofunciones en este caso...




### $Z$ como una suma sobre caminos...

Y que tiene que ver con la formulacin de la cuntica de Feynman?

El formalismo de la matriz de transferencia en mecnica estadstica tiene adems una conexin con la mecnica cuntica. Se puede mostrar por ejemplo, que el modelo de Ising 1d es isomorfo al problema de **un solo** espn cuntico en un campo transversal. La conexin se ve cuando uno usa la formulacin de la mecnica cuntica en trminos de integrales de camino de Feynman (ver por ejemplo cap 5.8 del libro de Chandler).  

Para ver que hay "caminos" escondidos recordemos esta forma

$$Z=\sum_{\sigma_1=\pm 1}\dots\sum_{\sigma_N=\pm 1}
\langle \sigma_1| {\hat T} | \sigma_2 \rangle
\langle \sigma_2| {\hat T} | \sigma_3 \rangle

\langle \sigma_N| {\hat T} | \sigma_1 \rangle
$$

Y piensen que $\sigma_i$ es el estado de un espn solito al "tiempo imaginario" $t=i$, es decir $\sigma(t=i)=\sigma_i$.
Viendolo as, $\hat T$ es un "operador de evolucin", que lleva al espn de un tiempo $i$ al siguente, y asi sucesivamente.

La suma entonces es una suma sobre todos los caminos posibles que salen y llegan al mismo estado $\sigma_1$ y al final sumo sobre estos estados. Es decir, la expresin de $Z$ en esta analoga es una suma sobre caminos seguidos por $\sigma(t)$m entre $t=1$ a $t=N$.
