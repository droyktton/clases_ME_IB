# Solución exacta para el modelo de Ising totalmente conectado

Este modelo considera que todos los espines están conectados con todos con un acoplamiento ferromagnético,

$$H = -\frac{J}{2N} \sum_{i=1}^N \sum_{j=1}^N \sigma_i \sigma_j - B \sum_{i=1}^N \sigma_i$$





image.png

* ¿Por qué se introduce $N$ en la intensidad de la interacción, mediante el coeficiente $J/N$?
¿Tiene importancia el término de autointeracción de un sitio consigo mismo?

Notar que la constante de acoplamiento esta dividida $N$, asi que se vuelve muy pequeña en el límite termodinámico. Aún así, la interacción nunca puede despreciarse. Dividir por $N$ es necesario sin embargo para mantener la energía extensiva. En el estado fundamental por ejemplo tenemos $E=(-J/2N)N^2-hN=-JN/2-hN$. Físicamente es plausible, ya que el número de pares aumenta como $N^2$, de modo que el acoplamiento tiene que ser más débil cuando más particulas o pares haya. Notar que estamos agregando autointeracción, lo que no estaría bien, pero todos los términos de autointeracción dan una constante, $\sum_{i=1}^{N}(J/2N)\sigma_i^2=J/2$, y no tiene por lo tanto relevancia en los promedios. Incluir la autointeracción tiene la ventaja de hacer la cuenta más sencilla.


Notemos que como la energía es una función de la magnetizacion total $M$, la funcion de partición se puede resolver como

$$Z \equiv e^{-\beta F(T,B)}= \sum_{M=-N}^{N} \text{[#configuraciones con M fija]} e^{-\beta((-J/2N)M^2 - MB)}
= \sum_{M}\frac{N!}{N_+! (N-N_+)!}e^{-\beta((-J/2N)M^2 - MB)}
= \sum_{M} e^{-\beta(-J M^2/2N - M B)+\ln \left[\frac{N!}{N_+! (N-N+)!}\right]}
= \sum_{M=-N}^{N} e^{-\beta {\tilde F}(\beta, M)}
$$

donde

$$
{\tilde F}(\beta, M)=
(-J/2N)M^2 - MB - \beta^{-1}\ln \left[\frac{N!}{N_+! (N-N+)!}\right] \\
\approx (-J/2N)M^2 - MB - \beta^{-1}\left[
N \ln N - N_+ \ln N_+ - N_- \ln N_-  
\right] \\
=
(-J/2N)M^2 - MB + \beta^{-1} N \left[
(N_+/N) \ln (N_+/N) + (N_-/N) \ln (N_-/N)  
\right]
$$


Usando que

$$M=(N_+ - N_-)$$
$$N = N_+ - N_-$$
$$m=M/N$$

tenemos
$$m+1=2(N_+/N)$$
$$1-m=2(N_-/N)$$

y definiendo ${\tilde f}(\beta, m)={\tilde F}(\beta, M)/N$

llegamos a

$$\boxed{{\tilde f}(m,\beta, B)=(-J/2)m^2 - m B + \beta^{-1}
\left[
\frac{1+m}{2} \ln \frac{1+m}{2} + \frac{1-m}{2} \ln \frac{1+m}{2}  
\right]}$$

Esta expresión luce como una energía libre, con una primera parte que tiene que ver con la energía a magnetización fija, y la otra que tiene que ver con la entropía a magnetización fija. Luce, pero no es, porque $m$ no es algo que controlemos. Solo controlamos $T$, y $B$. Pero, de alguna manera, describe el peso de los distintos valores de $m$ en la verdadera energía libre termodinámica. Veámoslo:

$$Z = \sum_{M=-N}^{N} e^{-\beta N \tilde{f}(\beta, m, B)}$$

Esta suma tiene muchos términos, pero hay uno que le gana por goleada al resto. Si ${\tilde f}(\beta,m,B)$ tiene un mínimo en $m^*(\beta,B)$, este dominará la suma. Para verlo notemos que

$$e^{-\beta N {\tilde f}(\beta, m^*)} \leq Z \leq 2N e^{-\beta N {\tilde f}(\beta, m^*)}$$

y como el logaritmo es una función creciente

$$-\beta N {\tilde f}(\beta, m^*) \leq \ln Z \leq -\beta N {\tilde f}(\beta, m^*)+\ln (2N)$$

$$-\beta {\tilde f}(\beta, m^*) \leq \ln Z/N \leq -\beta  {\tilde f}(\beta, m^*)+\ln (2N)/N$$

y en el límite $N\to \infty$,

$$\ln Z/N \to -\beta {\tilde f}(\beta, m^*)$$

Por lo tanto, la verdadera energía libre por partícula es

$$\boxed{f(T,B)={\tilde f}(T, m^*(T,B), B)}$$

donde $m^*=m^*(T,B)$ minimiza ${\tilde f}(\beta, m^*, B)$. Lo que acabamos de hacer se conoce como aproximación de punto silla, o *saddle node*.

Entonces $m^*$ satisface
$\partial_m {\tilde f}(\beta, m^*, B)=0$:

$$-J m - B + \beta^{-1}( (1/2)\ln ((1+m)/2) + 1 - (1/2)\ln ((1-m)/2) - 1)
=-J m - B + \beta^{-1} (1/2) \ln((1+m)/(1-m))
=-J m - B + \beta^{-1} \text{arctanh}(m)=0
$$

entonces

$$\text{arctanh}(m^*)= \beta[J m^*+B]
$$

$$\boxed{m^* = \tanh[\beta(J m^*+B)]}$$

resultando una ecuación trascendente o *autoconsistente* para $m^*$ vs $T$ y $B$.

con la condición de que sea estable, es decir realmente un mínimo (podría tener máximos o puntos silla). Pedimos entonces que $\partial^2_m{\tilde f}(\beta, m^*, B)>0$

$$\frac{T+m^{*2}-J}{1-m^{*2}}>0$$

de donde vemos que

* $m=0$ es estable si $T-J>0$, o sea si $T > T_c$. De lo contrario es inestable.

* si $T < T_c$, $m=\pm m^*$ y son estables pues $T+m^{*2}-J>0$ y $m^{*2}\leq 1$. De lo contrario son inestables. Esta conclusión es más facilmente apreciada dibujando $\tilde{f}$ vs $m$.


La magnetización por sitio $m$ tiene entonces que tener forma de tridente en función de $T$, con el "mango" estable bifurcando a $T_c$, donde aparecen dos "dientes" estables simétricos, y el diente central es inestable.

### Energía libre de Bragg-Williams

A la cuasi energía libre o "energía libre extendida" ${\tilde f}(m,\beta, B)$ se la conoce como *Energía libre de Bragg-Williams*

* El mínimo **global estable** de la energía de Bragg-Williams corresponde al estado de equilibrio termodinámico, es decir que en ese caso $m_{eq}\equiv m^*$.

* La energía libre de Bragg-Williams por partícula  es ${\tilde f}(m,T,B)$. Evaluada esta función en su mínimo **global estable** $m=m_{eq}(T,B)$ nos da la verdadera energía libre termodinámica por partícula $f(T,B)$.

El análisis de los extremos también es práctico hacerlo simplemente dibujando ${\tilde f}(m,T,B)$ vs $m$. Es equivalente a usar la ecuación trascendente, como hicimos anteriormente, pero tiene la ventaja que podemos ver la estabilidad de cada uno de los extremos, además de permitirnos visualizar mejor el significado de mestastabilidad.


## Soluciones gráficas

La ecuación autoconsistente para $m_{eq}$ puede ser resuelta numéricamente en función de $T$ y $B$. Es instructivo también dibujar la energía de Bragg-Williams para apreciar rápidamente la estabilidad de las soluciones encontradas y entender sus cambios de estabilidad o desaparición. Estos cambios o desapariciones son señal de transiciones de fase.

```python
#@title Solución interactiva para ${\tilde f}$ y sus extremos $m^*$.


import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


# Define the function f(x, y)
def f(M, T, h):
    return -M**2/2 - h*M + T*((1 + M) * np.log((1 + M)/2)/2 + (1 - M) * np.log((1 - M)/2)/2)

# Define the function f'(x, y)
def f1(M, T, h):
    return -h - M +T*(-np.log((1-M)/2)+np.log((1+M)/2))/2

# Define the function f''(x, y)
def f2(M, T, h):
    return - 1 + T*(1/(1-M)+1/(1+M))/2

def plot_free_energy(t,h):
  m = np.linspace(-0.999 , 0.999, 400)
  y = f(m,t,h)
  plt.plot(m, y, label=r"$F(m,t)$")
  plt.show()

def plot_contour(h):
  # Create a grid of (x, y) points
  t = np.linspace(0, 2, 400)
  m = np.linspace(-0.999 , 0.999, 400)
  M, T = np.meshgrid(m, t)

  # Evaluate the function on the grid
  Z = f1(M, T,h)

  # second derivative
  Z2 = f2(M, T,h)

  stable = Z2 > 0
  unstable = Z2 < 0


  filtered_Z = np.where(stable,Z,np.nan)
  plt.contour(M, T, filtered_Z, levels=[0], colors='blue')

  filtered_Z = np.where(unstable,Z,np.nan)
  plt.contour(M, T, filtered_Z, levels=[0], colors='red',linestyles='--')

  #print(stable.shape)

  # Customize plot
  plt.title(r"Contorno de $df/dm =  0$")
  plt.xlabel("M")
  plt.ylabel("T")
  plt.grid(True)
  plt.show()


interact(plot_contour, h=(-2.0, 2.0, 0.001))
interact(plot_free_energy,h=(-0.5, 0.5, 0.001),t=(0, 3.0, 0.1))

```

Es importante en estos gráficos diferenciar bien entre los extremos que no son mínimos, y cuando son mínimos si corresponden a equilibrios estables o metastables.

Recordemos que termodinámicamente la energía libre tiene que ser un mínimo global. Sin embargo, la experiencia muestra que los estados metaestables son muchas veces relevantes debido a los tiempos que tarda el sistema en darse cuenta de que hay otro mínimo al cual escaparse. Esto es debido a la presencia de barreras de energía libre.

## Transición de fase

Como podemos ver, el cambio de $m_{eq}$ en $T_c$ es no analítico para $B=0$. También vemos que $m_{eq}$ pega un salto variando $B$ alrededor de $B=0$ si $T< T_c$.

Por otro lado $m_{eq}$ es una derivada de una verdadera energía libre, $\partial_B f(T,B)$, donde $f(T,B)\equiv \tilde{f}(m_{eq}(T,B),T,B)$. Así que $f(T,B)$ es no analítica y finalmente hemos encontrado una **transición de fase** en el límite termodinámico.

### parámetro de orden
La magnetización es el denominado *parámetro de orden* de esta transición pues es el que me revela la naturaleza de las fases, en este casos la fase paramagnética, que no tiene magnetización espontánea a $B=0$, de la ferromagnética, que si la tiene.

En general el *parámetro de orden* de una transición es definido de forma tal que en una fase sea cero y en la otra finito. Por lo tanto su comportamiento es no analítico. Además cuantifica el grado de orden de la fase ordenada. Muchas veces el parámetro de orden finito está asociado a la ruptura de simetría pero no todas las veces.


En la figura vemos el diagrama $B$-$T$ del modelo de Ising (si bien resolvimos el totalment conectado es similar para todos los modelos de Ising en cualquier dimensión $d>1$). Al lado mostramos el diagrama de fases de un sistema totalmente diferente, simplemente para que aprecien la similitud de la linea azul.

* Cruzando la linea azul, hay un salto de magnetización en el ferromagneto mientras que hay un salto de densidad en el otro sistema asociado a la condensación. Estas son transiciones de primer orden.

* Ambas líneas azules terminan en un punto crítico. Como veremos, atravesarlo es una transición de segundo orden, y está caracterizada por exponentes críticos, como veremos luego.

* Se puede pasar continuamente de una fase a la otra rodeando el punto crítico.

## Comportamiento crítico cerca del punto $T=T_c$, $B=0$

Es interesante caracterizar bien la singularidad en $T_c$, $B=0$, porque contiene en general información universal, es decir que no depende de detalles del modelo.

Justo por debajo de una dada $T_c$, llamada temperatura crítica, que depende de parámetros microscópicos del hamiltoniano, la magnetización adquiere un valor finito que crece desde cero, hasta saturar a $1$ a bajas temperaturas. Se trata de una *transición de fase continua*. Para verlo mejor, es instructivo calcular la susceptibilidad

$$\chi = \lim_{B\to 0} \partial_B m_{eq}$$

para hacer el cálculo conviene pensar que $m_{eq} \ll 1$ por encima de $T_c$, donde solo hay una raíz. Entonces, asumiendo también que $B$ es pequeño (la susceptibilidad mide respuesta lineal, cosa que ocurre cuando $B$ es chico), podemos aproximar

$$
m_{eq} \approx
\beta J m_{eq} +\beta B \\
m_{eq} \approx \frac{\beta B}{1-\beta J}
$$

de donde vemos entonces que para $T>T_c$

$$
\boxed{
\chi \approx \frac{\beta}{1-\beta J} \approx (T-T_c)^{-1}}
$$

Es decir, que $\chi \to \infty$ cuando $T\to T_c$, y lo hace como una ley de potencia con exponente crítico $1$. Esta es una señal muy general de la existencia de una transición de segundo orden. Recordemos además que vale

$$\chi = \frac{1}{k_B T}\left[ \langle M^2 \rangle - \langle M \rangle^2  \right]$$

de modo que una divergencia en $T_c$ quiere decir fluctuaciones enormes. Usualmente esto tiene consecuencias muy observables. Por otro lado, como $M=\sum_i \sigma_i$,

$$\chi = \frac{1}{k_B T}\left[ \langle \sum_{ij} \sigma_i \sigma_j \rangle - \langle \sum_i \sigma_i \rangle^2  \right]
= \frac{1}{k_B T}\sum_{ij} \left[ \langle  \sigma_i \sigma_j \rangle - \langle \sigma_i \rangle \langle \sigma_j \rangle \right]
= \frac{1}{k_B T}\sum_{ij} \left[ \langle  \sigma_i \sigma_j \rangle - \langle \sigma \rangle^2 \right]\equiv
\frac{1}{k_B T}\sum_{ij} \left[ \langle  (\sigma_i-\langle \sigma \rangle)(\sigma_j-\langle \sigma \rangle) \right]
$$

asi que una divergencia de $\chi$ implica una covariancia grande o correlación muy fuerte entre los espines. Es decir, que sus fluctuaciones están "comunicadas". En el modelo presente no hay una noción de espacio, pero cuando la hay, las correlaciones tienden a decaer con la distancia, dentro de una distancia característica o longitud de correlación (recordar la longitud de correlación exacta del modelo de Ising en $d=1$, la cual solo tenía una divergencia en $T=0$ porque no hay transición en ese caso).



Otra cosa que vale la pena observar es que si $B=0$, para $T≲T_c$ podemos desarrollar así

$$
m_{eq} =
\tanh[\beta J m_{eq}]
\approx \beta J m_{eq} - (\beta J  m_{eq})^3/3
$$

como suponemos que $T < T_c$ y $ m_{eq} \neq 0$ tenemos

$$
1
\approx \beta J - (\beta J)^3  m_{eq}^2/3 \\
m_{eq}^2 = 3\frac{\beta J-1}{(\beta J)^3} \\
m_{eq}^2 = 3\frac{\beta J-1}{(\beta J)^2} \\
m_{eq}  = \pm \frac{3^{1/2}}{\beta J}\sqrt{\beta J-1}
$$

por lo tanto

$$\boxed{m_{eq}  \approx \pm (T_c-T)^{1/2}}$$

es decir, que la magnetización se anula como una ley de potencia.
Uno también podría determinar la susceptibilidad para $T < T_c$. Para eso hay que considerar $B$ o calcular $\langle \sigma^2 \rangle$.

Calculemos ahora el calor específico. Para eso primero hay que calcular el valor medio de la energía.

$$\langle E \rangle
= -\partial_\beta \ln Z
= \partial_\beta [\beta (-k_B T \ln Z)]
= N \partial_\beta [\beta f(T,B)]
= N f(T,B) + N \beta \partial_\beta f(T,B)
= N f(T,B)
+ N \beta \partial_\beta [{\tilde f}(m_{eq}(T,B),T,B)]
= N f(T,B)
+ N \beta \partial_m [{\tilde f}(m_{eq}(T,B),T,B)]_\beta \partial_\beta m_{eq} +
N \beta \partial_\beta [{\tilde f}(m_{eq}(T,B),T,B)]_m
$$

donde usamos ya que $m_{eq}$ es un extremo.
Recordemos que $f(T,B)={\tilde f}(m_{eq},T,B)$ con

$$\beta {\tilde f}(m,\beta, B)=\beta(-J/2)m^2 - \beta m B + \left[
\frac{1+m}{2} \ln \frac{1+m}{2} + \frac{1-m}{2} \ln \frac{1+m}{2}  
\right]$$

Asi que las cuentas se complican, pero, si uno esta interesado en el comportamiento cerca de $T_c$ para campos chicos lo más práctico sería tener un desarrollo a bajo orden en $m$ y $B$ de la energía libre.

### Energía libre de Landau

El comportamiento tipo leyes de potencia cerca del punto crítico en el diagrama $B-T$ se puede obtener a partir de una aproximación de la energía de Bragg-Williams para $m$ chico, ya que cerca del punto crítico $m_{eq}$ lo es necesariamente.

La aproximación es polinomial y la idea es truncar a un cierto orden de $m$. En este caso es suficiente truncar a orden $m^4$ para reproducir los mínimos de la de Bragg-Williams correctamente, y mantener la energía libre aproximada acotada para $|m|$ grande. Hagamos el desarrollo usando sympy,

```python
#@title Energía Libre aproximada de Landau a partir de la exacta de Bragg-Williams

from sympy import *

m = Symbol('m')
J = Symbol('J')
b = Symbol('\\beta')
B = Symbol('B')

def f(m,J,b,B):
    return -(J/2)*m**2 + (1/b)*( ((1+m)/2)*log(((1+m)/2)) + ((1-m)/2)*log(((1-m)/2)) ) -m*B

series(f(m,J,b, B),m)
```

Por lo tanto

$$
{\tilde f} \approx \frac{-\ln(2)}{\beta}+\frac{m^4}{12 \beta}+ \frac{m^2}{2} \left( -J + k_B T \right)-B m \\
=
\frac{-\ln(2)}{\beta}+\frac{m^4}{12 \beta}+ \frac{m^2}{2} k_B \left( T - T_c \right)-B m
$$

trabajar con este polinomio es mucho más práctico y es fácil convencerse de que *da los mismos resultados que antes si $m$ es chico*, en particular el comportamiento crítico.

Calculemos $m_{eq}$ haciendo $\partial_m {\tilde f}=0$.

$$\partial_m {\tilde f}= \frac{m^3}{3 \beta} + m k_B(T-T_c)-B=0$$

Esta ecuación cúbica se puede resolver y asi obtener $m_{eq}(T,B)$, que debería dar aproximadamente lo mismo que antes si $m_{eq} \ll 1$, cosa que en transiciones continuas esta garantizado cerca de $T_c$.

Si evaluamos la ecuación anterior en $T_c$ a $B>0$ tenemos $\frac{m^3}{3\beta}=B$. Entonces

$$\boxed{m_{eq} \sim B^{1/\delta}, \;\; \delta=3,\;\; T=T_c}$$


Si derivamos la ecuación anterior respecto de $B$, podemos obtener la susceptibilidad. Primero vemos que

$$\left[\frac{m_{eq}^2}{\beta} + k_B(T-T_c)\right]\partial_B m_{eq}=1$$

y cuando $B\to 0$ tenemos

$$\left[k_B(T-T_c)\right]\chi=1$$

de modo que

$$\boxed{\chi \sim |T_c-T|^{-\gamma}, \;\;\gamma=1}$$

Notar que usamos valor absoluto, sugiriendo que la divergencia es simétrica, pero no lo demostramos para $T < T_c$. Para demostrarlo basta notar que $\lim_{B\to 0} [\frac{m_{eq}^2}{\beta} + k_B(T-T_c)]\to 2 k_B(T_c-T)$, y entonces $\xi \sim (T_c-T)^{-1}$ desde $T < T_c$. Ojo que, en general, *no siempre son simétricos los exponentes alrededor de $T_c$*.

Mas sencillo es empezar con $B=0$ ya que inmediatamente vemos que $m=0$ es una solución. Las otras dos son soluciones de la cuadrática

$$\frac{m^2}{3 \beta} + k_B(T_c-T)=0$$

asi que

$$m_{eq} = (3\beta)^{1/2} k_B^{1/2} \sqrt{T_c-T}
\equiv  (3 T_c/T)^{1/2} \sqrt{1-T/T_c}
$$

igual a lo que obtuvimos antes. Cerca de $T_c$, el comportamiento está dominado por

$$\boxed{m_{eq} \sim (T_c-T)^{\tilde \beta},\;\;
{\tilde \beta=1/2},\;\; B=0}$$

Estas son las dos raíces estables cuando $T_c > T$. Caso contrario, estas raíces no son físicas, y la única estable es $m_{eq}=0$.

Calcular $⟨E⟩$ es ahora mucho más fácil.
Escribiendo

$$
\beta {f} =  \beta {\tilde f}(m_{eq},T,B) \approx
{-\ln(2)}+\frac{m_{eq}^4}{12}+ \frac{m_{eq}^2}{2} k_B \beta \left( T - T_c \right)-B m_{eq}
$$

Si $T ≳ T_c$, $m_{eq}=0$, entonces

$$
\beta {f} =  \beta {\tilde f}(m_{eq},T,B) \approx
{-\ln(2)} \\
\rightarrow \langle E \rangle = N\partial_\beta [\beta f] = 0
$$

Si $T ≲ T_c$ y $B=0$ en cambio

$$
\beta {f} =  \beta {\tilde f}(m_{eq},T,B) \approx
{-\ln(2)} + C_1(\beta)(T-T_c)^2 \\
\rightarrow \langle E \rangle
= N\partial_\beta [\beta f]
= N C_2(\beta) (T-T_c) + C'_1(\beta)(T-T_c)^2
$$

¡Asi que subiendo $T$ la energía media sube y pega un salto en $T_c$! Esto es permitido en una transición continua. De hecho se le llama de segundo orden porque recién las derivadas de potenciales termodinámicos, en este caso $⟨E⟩$, son discontinuas, pero los potenciales son continuos.

El calor específico $\partial_T ⟨E⟩$ es entonces
$C_V = C''' \Theta(T_c-T)$. Es decir, que pega un salto desde un valor constante a cero.
Si uno lo fuerza a ecribir como

$$\boxed{C_V \sim (T_c-T)^{-\alpha},\;\; \alpha=0,\;\; B=0}$$

para $T < T_c$, podemos concluir que $\alpha=0$ en este modelo, pero $\alpha$ es distinto de cero en general.

* Notar que en todos los cálculos de arriba hemos tomado como constantes todos los factores que no se anulan o no diverjen en $T_c$, porque nos interesa solo el escaleo con $|T-T_c|$, es decir la parte singular y dominante de la expresión.

* Notar también que el desarrollo de Landau no necesita ser exacto si lo que queremos es mirar el comportamiento crítico. Entonces nos basta con retener los términos que contengan $T-T_c$ y $B$ solamente. Eso quiere decir que podríamos haber trabajado sin problemas con este desarrollo
$$
{\tilde f} \approx a m^4 + b k_B \left( T - T_c \right)m^2 -B m
$$
con $a>0$ y $b>0$ constantes. Esto muestra que el comportamiento crítico es insensible a ciertos detalles del modelo.








## Exponentes críticos

Los exponentes $\alpha,\beta,\gamma,\delta$ que obtuvimos anteriormente para el modelo de Ising totalmente conectado son muy particulares y son los *exponentes de campo medio* de una gran familia de modelos, cuando se hacen aproximaciones. En general, los sistemas reales presentan exponentes distintos, salvo algunas excepciones, alrededor de un **punto crítico** cuya posición en el diagrama de fases hay que determinar en cada caso.

Hay más exponentes críticos importantes, pero requieren calcular funciones de correlación espaciales. En el modelo totalmente conectado no tienen mucho sentido.

## Diagramas de fase

El diagrama de fase del modelo de Ising es convenientemente el $B-T$, ya que estas son las variables o parámetros de control. Las figura muestra el mismo (donde $B=h$).

La línea azul indica que hay salto de la magnetización cuando una la cruza, mientras la línea punteada es un límite de metastabilidad. El punto azul se llama *punto crítico* y tiene propiedades universales como veremos luego.

## ¿Pseudo-energía libre?

A la energía de Bragg Williams $\tilde f$ le llamamos pseudo energía libre porque no es la energía libre termodinámica. Tiene una variable de más, $m$. Sin embargo, se convierte en la energía libre $G$ (de Gibbs si hay campo externo) cuando $m=m_{eq}$, que es el mínimo estable de $\tilde f$ respecto de $m$.

Se puede pensar que $e^{-\beta \tilde f}$ representa el peso en la función de partición de todas las configuraciones con $m$ fijo, y a $e^{-\beta G}$ el peso de la magnetización dominante.

## Memoria asociativa de Hopfield (comentario extra)

El modelo de Ising más general,
$$H = -\sum_{i<j} J_{ij} \sigma_i \sigma_j - \sum_i B_i \sigma_i$$
fue usado como modelo de memoria *neuronal* asociativa.
La idea es pensar que los $\sigma_i$ representan neuronas en distintos estados y a los $J_{ij}$ como la fuerza de sus conexiones sinápticas. Que es memoria asociativa quiere decir, que guarda patrones como si fueran recuerdos, pero además tiene la capacidad de asociarlos con patrones diferentes.

*En síntesis, es posible hacer que $H(\sigma_1,\dots,\sigma_N)$ tenga mínimos en estados que corresponden a "memorias", por ejemplo a los bits que definen una imagen blanco y negro.*

Si quieren ver una de las formas de hacerlo,
una de las reglas para asignar $J_{ij}$ es la de Hebb,
$$J_{ij}=\frac{1}{n}\sum_\mu \epsilon_i^\mu \epsilon_j^\mu$$
donde $\epsilon_i^\mu$ se puede pensar como el bit $i$ del patrón $\mu$. Si los bits $\sigma_i$ y $\sigma_j$ coinciden con $\epsilon_i^\mu$ y $\epsilon_j^\mu$ respectivamente, el par contribuye con la mínima energía posible.

Esto hace que la energía $H$ tenga mínimos locales cuando los $\sigma_i=\epsilon^\mu_i$, es decir cuando el microestado coincide con uno de los microestados recordados. Si empiezo cerca del patrón y bajo la temperatura, es muy probable que termine evolucionando hacia alguno de los $\mu=1,\dots,n$ patrones.

Una red neuronal está tipicamente muy conectada.
Una memoria asociativa binaria, o sea muy limitada, es la que vamos a estudiar a continuación. Solo puede recordar dos estados, todos "+" o todos "-".

El modelo de Hopfield es el "abuelo" de los modelos de AI generativa que se usan tanto hoy en día.
