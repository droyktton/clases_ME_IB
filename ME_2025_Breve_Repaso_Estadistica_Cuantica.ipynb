# Estadística Cuántica

* La estadística cuántica hace uso del **operador densidad** $\hat \rho$.
El promedio **cuántico-estadístico** de un observable $\hat O$ por ejemplo es
\begin{align}
\langle O \rangle &= \text{Tr}[\hat \rho \hat O] \\
\text{Tr}[\hat \rho]&=1
\end{align}
Donde $\langle \dots \rangle$ es un "doble promedio", en un **estado mixto**. Es decir, que no conozco bien el microestado cuántico por un lado, y a su vez en cada microestado no conozco bien el valor del observable (a menos claro que el microestado sea autoestado del observable).
Por ejemplo en el ensemble gran-canónico (pero como vimos se puede trabajar en cualquiera, con la $\hat \rho$ correspondiente)
\begin{align}
\hat \rho &=  \frac{e^{-\beta(\hat H-\mu {\hat N})}}{{\cal Z}}\\
{\cal Z} &= \text{Tr}[e^{-\beta (\hat H-\mu {\hat N})}]
\end{align}
Es conveniente claro está, el hacer la traza sobre autoestados de $\hat H$, pero no es obligatorio. Por ejemplo, si uno no los conoce, puede usar otros cualesquiera, por ejemplo autoestados del observable que uno quiera medir. Total, como es una traza, el resultado no depende de la base escogida. *Esto es análogo en el caso clásico a las transformaciones canónicas, que dejar invariantes las integrales en el espacio fásico*.

* Comparando el tratamiento cuántico y clásico, es interesante notar que la **introducción de $h$ en las integraciones en el espacio de fases clásico**, hace que la estadística cuántica converja exactamente al caso clásico a altas temperaturas, o a altos números cuánticos, y así justifica su introducción. Notar que en el caso clásico se integra sobre un continuo en el espacio fásico, mientras que en el cuántico se suma directamente sobre números cuánticos, mas o menos separados en energía. Por ejemplo, en el oscilador armónico unidimensional, uno usa $Z=\int (dq dp/h) \exp[-\beta H(q,p)]$, mientras que en el cuántico $Z=\sum_{n=0}^{\infty} \exp[-\beta E_n]$, con $E_n=(n+1/2)\hbar \omega$. Solo a $k_BT \gg \hbar \omega$ coinciden en sus predicciones, en particular el teorema de equipartición.

* Los **gases moleculares** de $N$ moléculas idénticas pero muy diluídos u a altas temperaturas pueden describirse en el canónico por ejemplo con $Z=Z_1^N/N!$ (luego vimos que esto estrictamente corresponde a la estadística de Maxwell-Boltzmann, que no discierne si las moléculas, como un todo, son fermiones o bosones). $Z_1$ contiene los grados traslacionales del centro de masa, proporcional a $V$, llevando a la ecuación de estado del gas ideal $PV=Nk_B T$. $Z_1$ sin embargo puede aún contener grados de libertad internos, que cuando son tratados cuánticamente, dan lugar a la **falla del teorema de equipartición** a temperaturas bajas (y esto ¡no es incompatible con la ecuación de estado del gas ideal!). Esto se ve muy bien en el **calor específico**, y sus contribuciones rotacionales y vibracionales, que se van "congelando" a distintas temperaturas características que pueden relacionarse con cantidades microscópicas de las moléculas, como interacciones y distancias entre átomos, sus masas, etc.

* Cuando tratamos con **partículas idénticas** tenemos que imponer la simetrización de las funciones de onda o estados porque *el estado producto tensorial no es físico*, aunque sea matemáticamente una solución de la ecuación de Schrodinger. El espacio de Hilbert simetrizado puede sin embargo ser descripto por combinaciones de todas las permutaciones de estos estados producto no físicos, logrando que estas sean simétricas o antisimétricas el intercambio de un par de partículas. Esto trae consecuencias a la estadística, en particular cuando, por ejemplo en $d=3$, $n \lambda_T^3$ se vuelve grande, donde $\lambda_T=h/\sqrt{2\pi m k_B T}$ es la **longitud de onda de De Broglie térmica** (importantísima escala de longitud ligada a la degeneración de intercambio), es decir altas densidades o temperaturas bajas.

* El estado de **Bosones y Fermiones idénticos y no interactuantes** puede describirse por los números de ocupación (números positivos) de estados de *una* partícula, $|n_1, n_2, \dots \rangle$, que en mecánica cuántica forman lo que se llama espacio de Fock, y generalizan, como un producto tensorial, el oscilador armónico (con sus operadore de creación y destrucción, y sus algebras definidas por conmutadores o anticonmutadores). Así, la **simetría de intercambio**, que requiere la mecánica cuántica de un sistema de partículas idénticas, está *automáticamente garantizada*, y cubre los dos únicos casos posibles de simetrización. Según el **teorema espín estadística**, cada caso depende de si el espín de la partícula es entero (bosón) o semientero (fermión).
No solamente eso, sino también que estos estados son autoestados de
$\hat H \equiv \sum_i \epsilon_i \hat n_i=\sum_i \epsilon_i {\hat c}_i^\dagger {\hat c}_i $,
es decir
$$\hat H |n_1, n_2, \dots \rangle = \left[\sum_{i=1}^{max} \epsilon_i n_i \right] |n_1, n_2, \dots n_{max}\rangle$$
lo cual es super conveniente para evaluar rápidamente la traza que da el promedio de cualquier observable.

* De los casos que vimos, solo los fonones (y otras excitaciones similares colectivas en redes como los magnones, etc) tienen un **número acotado de estados posibles** $max=M$, tal que un microestado autoestado de $\hat H$ sería $|n_1, n_2, \dots n_M \rangle$ donde $M$ es el número de modos. En general $max=\infty$.

* **Bosones** tienen permitido cualquier número $n_i=0,\dots, \infty$. La estadística que resulta es la de **Bose-Einstein**.

* **Fermiones** obedecen el principio de exlcusión de Pauli $n_i=0,1$.
La estadística que resulta es la de **Fermi-Dirac**.

* La **estadística de Maxwell-Boltzmann** es obtenida como caso límite de las de FD y BE cuando $n \lambda_T^3 \ll 1$ o $z\to 0$. En este límite, si son fermiones o bosones es irrelevante, y reobtenemos automáticamente la función de partíción con el factor $1/N!$ necesario para evitar la **paradoja de Gibbs**.


* Imponer los vínculos a los números de ocupación es engorroso en el microcanónico y canónico, pero al parecerse mucho a un problema de adsorción, es conveniente usar **Gran canónico** y fijar número medio de partículas con $\mu$.
\begin{align}
{\cal Z}_\eta &= \prod_i \sum^\eta_n e^{-\beta(\epsilon_i-\mu)n} = \prod_i {\cal Z}^\eta_i ⇒
\\
{\cal Z}_B &= \prod_i \frac{1}{1-e^{-\beta(\epsilon_i-\mu)}} \text{  (suma geométrica)}\\
{\cal Z}_F &= \prod_i [1+e^{-\beta(\epsilon_i-\mu)}]
\end{align}
donde la suma geométrica para bosones converge porque $e^{-\beta(\epsilon_{min}-\mu)}< 1$, o sea $\epsilon_{min} > \mu$, imponiendo condición a $\mu$.
La analogía con el problema de adsorción sería
  * "sitio con energía $\epsilon_i \equiv$ estado de una partícula con energía $\epsilon_i$".
  * "Fermi-Dirac $\equiv$ sitios con 0 o 1 partícula"
  * "Bose-Einsten $\equiv$ sitios con $0$ a $\infty$ partículas".


* Una función muy importante es el **número medio de ocupación** en cada estado posible $i$ de *una* partícula
\begin{align}
\langle n_i \rangle_\eta
&= \frac{\sum^\eta_n n \;e^{-\beta(\epsilon_i-\mu)n}}{{\cal Z}_\eta} = \partial_{\beta \mu} \ln {\cal Z}_\eta \\
\langle n_i \rangle_B  &= \frac{1}{e^{\beta(\epsilon_i-\mu)}-1} = \frac{1}{z^{-1}e^{\beta \epsilon_i}-1} \\
\langle n_i \rangle_F  &= \frac{1}{e^{\beta(\epsilon_i-\mu)}+1}= \frac{1}{z^{-1} e^{\beta \epsilon_i}+1}
\end{align}
donde $z=\exp(\beta \mu)$ es la **fugacidad** y $\eta$ indica si es boson o fermión, alterando el signo de algunas cosas ( $\eta=-1$ son fermiones y $\eta=1$ bosones), y los límites de las sumatorias sobre número de ocupación ($\sum_{n=0}^\eta$ quiere decir usar $\sum_{n=0}^1$ para fermiones y $\sum_{n=0}^\infty$ para bosones)

* Conociendo los $\langle n_i \rangle_\eta$ puedo calcular, por ejemplo
\begin{align}
\beta P_\eta &=  
\frac{1}{V}\sum_i \ln {\cal Z}_i^\eta
= \frac{1}{V}\sum_i [1-\eta z e^{-\beta \epsilon_i}]
\\
N_\eta &= \sum_i \langle n_i \rangle_\eta  \\
E_\eta &= \sum_i \epsilon_i \langle n_i \rangle_\eta
\end{align}

* Las anteriores sumas, si los estados se pueden describir con estados muy amontonados en energía cuando variamos algunos de los números cuánticos, por ejemplo $i \equiv ({\bf k},\sigma)$, son convenientemente escritas en términos de integrales vía la **densidad de estados** $g(\epsilon)=\sum_i \delta(ϵ-\epsilon_i)$.

* La densidad de estados de una partícula $g(\epsilon)$ es fácil de calcular para partículas en una caja microscópica en cualquier dimensión si la relación de dispersión tiene simetría esférica, por ejemplo $\epsilon_{{\bf k},\sigma}=a|{\bf k}|^\gamma$. En algunos problemas puede depender de $\sigma$, por ejemplo $\epsilon_{{\bf k},\sigma}=a|{\bf k}|^\gamma - \mu_B \sigma B$. La receta es
\begin{align}
\sum_i f(\epsilon_i) \equiv \sum_{\sigma}\sum_{\bf k} f(\epsilon_{{\bf k},\sigma}) \equiv \sum_\sigma \int \frac{d^3 k}{(2\pi/L)^d} f(\epsilon_{{\bf k},\sigma}) \equiv \int_{-\infty}^{\infty} g(\epsilon)d\epsilon f(\epsilon)
\end{align}
donde definimos $g(\epsilon)$ de tal forma que incluya la degeneración de espín (u otras variales) y también su soporte (es decir, que se vea donde es distinta de cero). Por ejemplo: sin campo, $\epsilon \geq 0$, con campo $\epsilon_{\sigma=1} \geq -\mu_B B$ y $\epsilon_{\sigma=1} \geq \mu_B B$.
Si por ejemplo $\epsilon_{{\bf k},\sigma}=f(|{\bf k}|)$ en $d=3$,
$$g(\epsilon)d\epsilon
= \frac{g_s 4\pi k^2 dk}{(2\pi/L)^3}
=\left[\frac{g_s 4\pi k(\epsilon)^2 (dk(\epsilon)/d\epsilon)}{(2\pi/L)^3}\right] d\epsilon
$$
y es claro que la cuenta se generaliza a dimensión $d$
$$
g(\epsilon)d\epsilon
=\left[\frac{g_s \Omega_d k(\epsilon)^{d-1} (dk(\epsilon)/d\epsilon)}{(2\pi/L)^d}\right] d\epsilon
$$
Y se puede terminar especificando una forma concreta, por ejemplo $\epsilon_{{\bf k},\sigma}=\hbar^2 k^2/2m$, de donde puedo obtener $k(\epsilon)=\sqrt{2m\epsilon}/\hbar$ para terminar el cálculo.

* *Los estados de una partícula no siempre están etiquetados por $i\equiv ({\bf k},\sigma)$. Por ejemplo, si en vez de caja tenemos trampa armónica, un estado sería $i \equiv (m_1,\dots,m_d)$, con $m_i=0,\dots$, y $\epsilon_i \equiv \hbar \omega_0(m_1+\dots+m_d+d/2)$. En este caso, pasar a una $g(ϵ)$ continua es una aproximación que vale si $\hbar \omega_0 \ll k_B T$, la aproximación de "niveles encimados"*.

* Cuando usamos $g(\epsilon)$ todo suele quedar hermosamente expresado en términos de las funciones especiales (machetearse sus propiedades siempre viene bien!)
\begin{align}
f^{\eta}_m(z) = \frac{1}{(m-1)!}\int_0^\infty \frac{dx\;x^{m-1}}{z^{-1}e^x-\eta}
\end{align}
y para el gas no relativista en tres dimensiones y con degeneración de espín en particular
\begin{align}
\beta P_\eta &= \frac{g}{\lambda_T^3} f_{5/2}^\eta(z) \\
n_\eta &= \frac{g}{\lambda_T^3} f_{3/2}^\eta(z) \\
\epsilon_\eta &= \frac{3}{2} P_\eta
\end{align}
donde $n_\eta=N_\eta/V$ y $\epsilon_\eta=E_\eta/V$. A esto se llega usando ${\cal Z}^\eta$ o $\Theta=-PV=-k_B T \ln {\cal Z}^\eta$ y derivando termodinámicamente. Notar que se puede generalizar para cualquier dimensión $d$ y a relaciones de dispersión del tipo $g(\epsilon)\sim \epsilon^\gamma$, pues entonces se general las $f_m^\eta(z)$.
  * *La entropía se puede calcular y tiene una expresión más complicada porque $S = -\partial_T \Theta|_{\mu,V}$, y al derivar tanto $\lambda_T$ como $z$ dependen de $T$.*

* En el contexto del gas no relativista en tres dimensiones y con degeneración de espín el *límite de Maxwell-Boltzmann* se alcanza en general cuando $z\to 0$, y es equivalente a pedir $n_\eta \lambda_T^3/g_s \ll 1$, o caso "no degenerado" (con respecto al intercambio de partículas). Notar nuevamente que se puede generalizar para cualquier dimensión $d$ y relaciones de dispersión del tipo $g(\epsilon)\sim \epsilon^\gamma$, pues entonces se general las $f_m^\eta(z)$.
Es útil entonces tener desarrollos del tipo
\begin{align}
f_m^\eta(z)=\sum_{\alpha=1}^{\infty} \eta^{\alpha+1}\frac{z^\alpha}{\alpha^m}
\end{align}

* Para estudiar correcciones cerca del **límite de Maxwell-Boltzmann** es útil poder fijar $n_\eta$ y $T$, y calcular $z$ en términos de estos.
El método es perturbativo:
  * Orden $\alpha=1$: $z = (n_\eta \lambda^3/g_s)$
  * Orden $\alpha=k$: Trunco la serie a orden $z^k$ y reemplazo todos los $z$ por el obtenido al orden anterior $\alpha=k-1$.
  * Es engorroso, asi que tipicamente no pasaremos de $\alpha=2$...

* **Efectos de degeneración** generales ocurren cuando $n_\eta \lambda_T^3/g_s \gtrsim 1$ y $z$ crece pero tenemos que tratar separadamente fermiones de bosones (recordar que $z\leq 1$ para bosones).

* Efectos de **degeneración cuántica de Fermiones a $T=0$** se traducen en la existencia de una **energía de Fermi** $\epsilon_F$
\begin{align}
\int_0^{\epsilon_F} g(\epsilon)d\epsilon = N
\end{align}
y como
\begin{align}
\int_0^{\infty} f(\epsilon-\mu) g(\epsilon)d\epsilon = N
\end{align}
donde $f(x)=(e^{\beta x}+1)^{-1}$ la **función de Fermi**, vemos que $\epsilon_F \equiv \mu(T=0)$. A $T=0$ todos los estados por debajo de $ϵ_F$ están ocupados formando el **mar de Fermi**, que cuando la relación de dispersión depende de $|{\bf k}|$ solamente tiene la forma de una esfera, la **esfera de Fermi** en el espacio recíproco. Recordar que para partículas no relativistas en $d=3$ con degeneración de espín, $g(\epsilon)\propto V g_s\sqrt{\epsilon}$.
De la energía de Fermi nace la **temperatura de Fermi** $T_F=\epsilon_F/k_B$,
que aparece naturalmente marcando el crossover de Fermi-Dirac hacia Maxwell-Boltzmann. La **presión de Fermi** $P_F = P(T=0)>0$ por el principio de exclusión de Pauli que fuerza a que hayan fermiones con momentos finitos golpeando las paredes (o la "superficie" de una estrella para estabilizarla de un colapso gravitatorio) aún a $T=0$.


* Efectos de degeneración cuántica de **Fermiones** a bajas $T$ y $\mu$ dado se trabajan convenientemente con el **desarrollo de Sommerfeld**.
\begin{align}
\int_{-\infty}^{\infty} f(\epsilon-\mu)\phi(\epsilon)d\epsilon \approx
\int_{-\infty}^{\mu} \phi(\epsilon)d\epsilon + \frac{\pi^2}{6} (k_B T)^2 \phi'(\mu) + \frac{7\pi^4}{360} (k_B T)^4 \phi'''(\mu) + \dots
\end{align}
  * Para calcular $\mu=\mu(T,n)$, usamos el desarrollo de Sommerfeld de $N$, que corresponde a $\phi(\epsilon)=g(\epsilon)$. En general es difícil despejar, pero si uno quiere la primer corrección, usamos $\mu = \epsilon_F + \delta \mu$, y resolvemos para $\delta \mu$ una ecuación lineal. Para llegar a la ecuación lineal uno cancela el $N$ del miembro izquierdo con el $\int_0^{\epsilon_F} g(\epsilon)d\epsilon$ del derecho.
  * La energía interna es un caso particular de Sommerfeld, con $\phi(\epsilon)=\epsilon g(\epsilon)$.
  * La presión puede ser obtenida de la densidad del desarrollo de Sommerfeld de la energía media.
  * *No siempre se puede usar el desarrollo de Sommerfeld* para ver las correcciones en fermiones a bajas $T$. Necesitamos que la función $\phi(\epsilon)$ sea finita en un entorno $k_B T$ alrededor de $\mu$, caso contrario no da ninguna corrección... El ejercicio 1 de la guía 10 es un ejemplo donde no se puede usar pero tampoco es necesario.


* **Truquito para fermiones** general para saber cualitativamente como se mueve $\mu$ cuando cambio $T$ (también funciona para $B$) a $N$ fijo. Consiste en
  * Dejarlo $\mu$ fijo y ensanchar la función de Fermi.
  * Mirar el cambio del número de partículas por debajo y por encima de $\mu$.
  * Si el cambio es mayor por encima de $\mu$, significa que $\mu$ baja aumentando $T$, para conservar el número de partículas.
  * Si el cambio es mayor por debajo de $\mu$, significa que $\mu$ aumenta aumentando $T$, para conservar el número de partículas.
  * Si los cambios son iguales, el $\mu$ no cambia con $T$, pues lo que saco por debajo de $\mu$ lo pongo por encima de $\mu$.

* La **susceptibilidad de Pauli** es la susceptibilidad independiente de $T$ que aparece debido al acoplamiento del momento magnético (proporcional a sus espines) de los electrones con un campo magnético. A temperatura cero la respuesta *paramagnética* se puede estimar usando la función de fermi y descomponiendo la densidad de estados por espín.
\begin{align}
M &= \mu_B (N_⇓-N_⇑) \\
&= \mu_B\kappa V \int_{\mu_B B}^{\epsilon_F} \sqrt{\epsilon-\mu_B B} \;d\epsilon -
\mu_B\kappa V  \int_{-\mu_B B}^{\epsilon_F} \sqrt{\epsilon+\mu_B B} \;d\epsilon \\
&= \mu_B\kappa V  \int_{0}^{\epsilon_F+\mu_B B} \sqrt{x} \;dx - \mu_B\kappa V \int_{0}^{\epsilon_F-\mu_B B} \sqrt{x} \;dx
 \\
&= \mu_B\kappa V \int_{0}^{\epsilon_F+\mu_B B} \sqrt{x} \;dx
-\mu_B\kappa V \int_{0}^{\epsilon_F-\mu_B B} \sqrt{x} \;dx\\
&= \mu_B\kappa V \int_{\epsilon_F-\mu_B B}^{\epsilon_F+\mu_B B} \sqrt{x} \;dx \approx 2 \kappa V \sqrt{\epsilon_F} \mu_B^2 B \\
&= g(\epsilon_F) \mu_B^2 B = \chi_{Pauli} B
\end{align}
donde hay que entender que $g(\epsilon)\sim \sqrt{\epsilon}$ es la densidad de estados *sin campo* y que $\epsilon_F \equiv \mu(T=0,B=0)$.

* Los **bosones** son partículas de espín entero que obedecen la estadística de Bose-Einstein.
  * ${\cal Z}=\prod_i {\cal Z}_i=\prod_i \sum_{n=0}^\infty e^{-\beta (ϵ_i-\mu)n}= \prod_i [e^{\beta(ϵ_i-\mu)}-1]^{-1}$, si $\mu>\min_i(\epsilon_i)$.
  * Del número de ocupación medio
$$\langle n_i \rangle = \frac{1}{e^{\beta (\epsilon_i-\mu)}-1}>0$$
concluimos también que $\mu < \min_i[\epsilon_i]$. Si son partículas en una caja sin campos exteriories $\mu \leq 0$, ya que $\epsilon_{i \equiv {{\bf k},\sigma}}=\hbar^2 k^2/2m$. Cuando $\mu \to  \min_i[\epsilon_i]$, $\langle n_0 \rangle \to \infty$, indicando un número macroscópico  de partículas en el fundamental. Notar que este límite particular es consistente con la no convergencia de la serie geométrica de ${\cal Z}_i = \sum_{n=0}^\infty e^{-\beta (ϵ_i-\mu)n}$, pero eso no quiere decir que no exista el límite termodinámico, ya que las cantidades intensivas son bien comportadas. Físicamente $\mu < 0$ pero se puede acercar mucho a $\mu=0$.
  * La mayoría de las propiedades termodinámicas quedan expresadas en términos de las funciones $f^+_m(z)$ de la fugacidad, llamadas *polilogaritmos*. Conviene saber sus propiedades (derivadas, límites, etc) y expansiones.
  * Para fijar ideas consideremos de aquí para abajo solo partículas no relativistas en una caja en $d=3$, pero todos los cálculos pueden ser particularizados a cualquier dimensión, para **cualquier relación de dispersión o bien cualquier estructura de estados de una partícula**.
  * Resulta que $n=\frac{g}{\lambda_T^3}f^{+}_{3/2}(z)\leq \frac{g_s}{\lambda_T^3}f^{+}_{3/2}(1) \approx \frac{g_s}{\lambda_T^3}2.612$, con $f^{+}_{3/2}(z)$ una función *monótonamente creciente*.
  * La **saturación** de $n\lambda_T^3/g_s$ indica que existe una $T_c(n)$, tal que $n \lambda_{T_c(n)}^3/g_s\approx 2.612$. Cuando $T \to T_c$, $\mu \to 0-$ y $z \to 1$.
  * Si $T \leq T_c(n)$, $n_{exc}=\frac{g_s}{\lambda_T^3}f^{+}_{3/2}(1)$, y $n_0=n-n_{exc}$.
  * Cuando $T < T_c(n)$, $z=1$, $n_0/n=1-(T_c/T)^{3/2}$, una **fracción macroscópica de partículas está en el estado fundamental**, mientras que una fracción $n_{exc}={g_s}{\lambda_T^{-3}}2.612 \sim T^{3/2}$ está en los excitados. La presión satura y se vuelve independiente de la densidad, porque proviene exclusivamente de las partículas en estados excitados.
  * Para $T > T_c(n)$, $n_0/n = 0$ (termodinámicamente) y $n_{exc}=n={g}\lambda_T^{-3}f^{+}_{3/2}(z)$. El gas se comporta como un gas ideal con correcciones que se asemejan a incluir interacciones atractivas.
  * Como fenómeno crítico, la **condensación de Bose-Einstein** es continua, con parámetro de orden $n_0/n$, y $n_0/n \sim (T_c(n)-T)$, o exponente crítico igual a $1$. La **compresibilidad isotérmica** diverge en $T_c$ y el **calor específico** es mayor que $(3/2)k_B$ para $T>T_c$, decae a cero como una ley de potencia $T^{3/2}$ para  $T < T_c$ y tiene un pico finito a $T_c$.
  * La fase condensada es en realidad mixta, y la transición tiene **calor latente**. No existe un estado de equilibrio para cualquier $P$ y $T$ porque si doy $P$ en la fase condensada de densidad fija $n$, ya estoy dando $T=T_c(n)$.
  * En el estado condensado hay una **función de onda macroscópica**, cuyo módulo cuadrado es aproximadamente la densidad de partículas. Esta densidad es diferente a la esperada por fluctuaciones térmicas solamente en la estadística de Maxwell-Boltzmann.
  * En la **trampa armónica isotrópica** de frecuencia $\omega$ (o constante de resorte $\kappa=m\omega^2$), tenemos
    * $E = \hbar \omega (n_x+n_y+n_z+3/2)$.
    * Degeneración de cada nivel de energía es $g_n=(n+1)(n+1)/2$, con $n=n_x+n_y+n_z$, lo que implica $g(ϵ)\sim ϵ^2$.
    * $n_0/n = 1-(T/T_c)^3$.
    * $T_c(n)\approx 0.94 (\hbar \omega/k_B) N^{1/3}$.

  Los experimentos se hacen típicamente en una *trampa armónica anisotrópica* porque eso brinda mas formas de discernir las partículas en el estado condensado y en los excitados.    

* **Los fotones y los fonones son bosones** con la restricción $\mu=0$. Su número no se conserva. La energía media de estos gases es
\begin{align}
E = \sum_{{\bf k},\sigma} \hbar \omega_{{\bf k},\sigma} (\langle n_{{\bf k},\sigma} \rangle+1/2)
= E_0 + \sum_{{\bf k},\sigma} \hbar \omega_{{\bf k},\sigma} \langle n_{{\bf k},\sigma} \rangle
\end{align}
donde $E_0$ es la energía de punto cero total, que es suma de las energías de punto cero de cada modo $\hbar \omega_{{\bf k},\sigma}/2$. En el caso de fonones $E_0$ es una constante finita, mientras que en el de fotones es una constante *infinita*. El número medio de fotones/fonones es
\begin{align}
N = \sum_{{\bf k},\sigma} \langle n_{{\bf k},\sigma} \rangle
\end{align}
y comparando con lo de arriba vemos que cada uno aporta una energía $\hbar \omega_{{\bf k},\sigma}$ al modo ${{\bf k},\sigma}$. *La enegía de punto cero es una propiedad del vacío de fotones/fonones*.

* Los **fonones** tienen un número acotado de estados de una partícula (de un fonón), que coincide con el número de modos del sólido, el cual es a su vez proporcional al número de átomos del sólido que es una *cantidad fija* $N$. No olvidar que modos con el mismo ${\bf k}$ pero distinta polarización son modos distintos o estados distintos de un fonón. En dimensión $d=3$ pueden haber hasta $3$ polarizaciones (dos transversales y una longitudinal).
  * Los **modos normales** se obtienen haciendo una transformación canónica a coordenadas normales, que es lo mismo que resolver un problema de autovalores y autovectores generalizados. Para un sólido cristalino tal problema de diagonalización o desacoplamiento del hamiltoniano, se hace automáticamente usando **transformada discreta de Fourier** gracias a la simetría de traslación discreta. Este es en realidad un problema de mecánica que sirve de input para la mecánica estadistica clásica y cuántica de las vibraciones.
  * Un fonón es un cuanto de energía de un oscilador armónico con frecuencia $\omega_{{\bf k},\alpha}$.
  * En general, la suma o integral sobre ${\bf k}$ se termina en una superficie llamada **zona de Brillouin**, que depende de la red subyacente.
  * El **modelo de Debye** más simple consiste en aproximar la relación de dispersión de la forma simple $\omega_{{\bf k},\alpha}=v_\alpha |{\bf k}|$, y más simple aún si $v_\alpha=v$ y las ondas tienen la misma velocidad independiente de la polarización. *Así, la zona de Brillouin es simplemente una esfera*.
  * Si los fonones están en un sólido macroscópico tiene sentido trabajar con la densidad de estados por unidad de frecuencia $g(\omega)$ que se calcula en forma análoga a $g(\epsilon)$.
  * En el **modelo de Debye** hay que forzar que la densidad de estados de un número finito de ellos $\int_0^{\omega_D} g(\omega)d\omega=g_s N$, con $g_s$ el número de polarizaciones posibles y $N$ el número de átomos.
  * No lo vimos, pero para redes con un solo tipo de átomo hay una sola banda, la acústica, pero si hay más de un tipo, pueden aparecer otras... los modos ópticos.

* Los **fotones** tienen un número infinito de estados de una partícula (de un fotón), pues las ondas electromagnéticas no tienen una longitud de onda mínima permitida.
  * La **relación de dispersión** en el vacío es siempre $\omega_{{\bf k},\sigma}=c |{\bf k}|$, para todo $k$.
  * $\alpha=1,2$ en $d=3$ pues las ecuaciones de Maxwell prohíben las ondas longitudinales electromagnéticas.
  * El **espectro de cuerpo negro** me dice cual es la densidad de energía media por frecuencia. Es decir, el número medio de fotones alrededor de una dada frecuencia, $g(\omega)[e^{\beta \hbar \omega}-1]^{-1}d\omega$, multiplicado por $\hbar \omega/V$. De ahí sale el espectro de Planck.
  * La *catástrofe ultravioleta* aparece al despreciar la discretización de la energía, que es justamente más importante a $\omega$ altas.
  * El espectro de Planck es esperable observarlo en la emisión de ondas de cualquier cuerpo en equilibrio a temperatura $T$ que no refleje. Salvo esa aclaración *Es universal*. Una realización es el agujerito de una cavidad, pues este dificilmente refleje.
  * La **ley de Stefan-Boltzmann** sale de integrar el espectro de energía de Planck. La ley de Wien sale de mirar el máximo.  
  * Los fotones ejercen una **presión de radiación**, que como ya sabiamos en general, es proporcional a la densidad de energía electromagnética.
  * En la presión y energía del gas de fotones hay constantes infinitas debidos a la energía de punto cero o "fluctuaciones cuánticas del vacío de fotones". En general, se pueden dejar fuera de la descripción (salvo en algunos casos que no vimos, de cavidades muy pequeñas...Efecto Casimir).
