### Entrop√≠a de Shannon

La expresi√≥n $S = I/N$ se llama *entrop√≠a de Shannon* de la distribuci√≥n $p_i$, y cumple la propiedad

$$
S = -\sum_i p_i \log_2(p_i) \geq 0.
$$

Como veremos m√°s adelante, tiene una conexi√≥n con la *entrop√≠a termodin√°mica* en el contexto de la mec√°nica estad√≠stica, cuando las "letras" son microestados f√≠sicos completos de un sistema.

Adem√°s de ser positiva, la entrop√≠a es aditiva para conjuntos de variables independientes, e invariante ante transformaciones biyectivas (si transformamos varias variables en una, $S$ disminuye).

Otra forma √∫til de escribirla es:

$$
S = -\sum_i p_i \log_2(p_i) = \langle \log_2(1/p_i) \rangle
$$

Es decir, que $S$ es el promedio de la *informaci√≥n del evento $i$*, $\log_2(1/p_i)$. Cuanto m√°s improbable es el evento, mayor sorpresa produce que ocurra, y m√°s informaci√≥n aporta. Por otro lado, si el evento es muy probable, la sorpresa es baja, y tambi√©n lo es la informaci√≥n que brinda.

### Derivaci√≥n usando combinatoria

El juego de letras ya nos ense√±a c√≥mo hacer una demostraci√≥n de la f√≥rmula de la entrop√≠a de Shannon. Supongamos que tenemos $M$ posibles valores para una variable aleatoria que ocurren con probabilidades $p_i$, no necesariamente uniformes. Queremos construir un mensaje de $N \gg 1$ s√≠mbolos.

El n√∫mero *t√≠pico* de mensajes que puedo construir **no es** $M^N$, ya que la mayor√≠a de esos mensajes son altamente improbables. En el l√≠mite de $N \gg 1$, se cumple $N_i \approx p_i N$ (ley de los grandes n√∫meros / l√≠mite multinomial), por lo tanto el n√∫mero de mensajes t√≠picos es:

$$
g = \frac{N!}{\prod_{i=1}^{M} N_i!}
$$

Cada uno de estos mensajes t√≠picos es equiprobable, y su informaci√≥n total es

\begin{align}
\log_2 g &= \log_2 N! - \sum_i \log_2 N_i! \\
&\approx N \log_2 N - N - \sum_i (N_i \log_2 N_i - N_i) \\
&= N \log_2 N - \sum_i N_i \log_2 N_i \\
&= N \log_2 N - \sum_i \left( N_i \log_2(N_i/N) + N_i \log_2 N \right) \\
&= -N \sum_i (N_i/N) \log_2(N_i/N) \\
&= -N \sum_i p_i \log_2 p_i
\end{align}

As√≠, obtenemos el resultado buscado. Como $S = \log_2 g / N$, entonces

$$
S = -\sum_i p_i \log_2 p_i
$$

### Validaci√≥n estad√≠stica

La demostraci√≥n anterior se basa en que la distribuci√≥n multinomial

$$
p_N(\{N_1, \dots, N_M\}) = \frac{N!}{N_1! \dots N_M!} p_1^{N_1} \dots p_M^{N_M}
$$

tiende, para $N_i \gg 1$, a concentrarse muy fuertemente alrededor de $N_i = p_i N$, ya que:

\begin{align}
\langle N_i \rangle &= p_i N \\
\text{Var}(N_i) &= N p_i (1 - p_i) \\
\Rightarrow \frac{\sqrt{\text{Var}(N_i)}}{\langle N_i \rangle} &\to 0 \quad \text{cuando } N \to \infty
\end{align}

Entonces, los mensajes t√≠picos cumplen $N_i = p_i N$ con gran precisi√≥n. Podemos verlo expl√≠citamente usando la aproximaci√≥n de Stirling $\ln N! \approx N \ln N - N$ para $N \gg 1$:

\begin{align}
\ln [p_N(\{N_1, \dots, N_M\})]
&= \ln \frac{N!}{\prod_i N_i!} + \sum_i N_i \ln p_i \\
&\approx N \ln N - N - \sum_i (N_i \ln N_i - N_i) + \sum_i N_i \ln p_i \\
&= - \sum_i N_i \ln(N_i / (p_i N)) = 0 \quad \text{cuando } N_i = p_i N
\end{align}

As√≠, $p_N(\{N_i = p_i N\}) \to 1$ en el l√≠mite. Todas estas configuraciones t√≠picas tienen la misma probabilidad, y hay $g$ de ellas, lo que justifica usar la entrop√≠a de distribuciones uniformes sobre ese subconjunto t√≠pico.

---

### Propiedades de la entrop√≠a de Shannon

Se puede demostrar que $S$ es la √∫nica funci√≥n $H(p_1,\dots,p_\Omega)$ que cumple ciertas [propiedades matem√°ticas muy generales](https://en.wikipedia.org/wiki/Entropy_(information_theory)#Characterization), y puede derivarse axiom√°ticamente. Algunas propiedades importantes:

- $0 \leq S(X) \leq \log_2(\Omega)$, donde $\Omega$ es el n√∫mero de valores posibles de $X$.
  - $S = 0$ si $p_i = \delta_{ij}$ (distribuci√≥n determinista).
  - $S = \log_2(\Omega)$ si $p_i = 1/\Omega$ (uniforme).
- Invarianza bajo transformaciones biyectivas $X \to F(X)$: $S$ no cambia.
- Si $F(X)$ no es biyectiva (reduce granularidad), $S$ disminuye.
- Subaditividad: $S(X,Y) \leq S(X) + S(Y)$.
- Aditividad: $S(X,Y) = S(X) + S(Y)$ si $X$ y $Y$ son independientes.
- Continuidad en las probabilidades $p_i$.

## Principio de m√°xima entrop√≠a

Vimos que dada una distribuci√≥n de probabilidad $p_i$ de una variable aleatoria que puede tomar $\Omega$ valores posibles, podemos calcular su contenido de informaci√≥n, que es a su vez el m√≠nimo indispensable de memoria que necesito para guardarlos, y la m√°xima compresi√≥n posible.

Ahora supongamos que **NO** conocemos las $p_i$ y que queremos inferirlas o estimarlas, en base a la informaci√≥n a priori que tenemos sobre ella. Por ejemplo, queremos las $p_i$ para un dado, justo o cargado.

Como principio, es razonable que la elecci√≥n de las $p_i$ sea lo m√°s desprejuiciada posible con la informaci√≥n real que tenemos de la variable, es decir la que maximice la ignorancia sobre la misma, o la que maximice la informaci√≥n que nos entregue conocer su valor, o que maximice el n√∫mero de preguntas que tenemos que hacer para adivinarla. En otras palabras, necesitamos que maximice la entrop√≠a de Shannon.  


Este es el [principio de m√°xima entrop√≠a](https://en.wikipedia.org/wiki/Principle_of_maximum_entropy) y fue propuesto en un [art√≠culo](https://drive.google.com/file/d/1bTOUCoY4WJRu76XMUpR4fBNTu-9OwRIx/view?usp=sharing) por Edwin T. Jaynes en 1957 como m√©todo de inferencia Bayesiana para la mec√°nica estad√≠stica.

Para eso tenemos que pensar primero $S$ como funci√≥n de las $p_i$,

$$
S[p_1,p_2,\dots,p_\Omega]
$$

y derivar respecto de las $p_i$ con la restricci√≥n de que $\sum p_i=1$. Eso se puede hacer usando multiplicadores de Lagrange:

$$
\partial_{p_j}\left[-\sum p_i \ln p_i + \lambda \sum_i p_i\right]=0
$$

con

$$
\sum p_i=1
$$

Entonces:

$$
\partial_{p_j}\left[-\sum p_i \ln p_i + \lambda \sum_i p_i\right]=
\left[
  -\sum_i (1+\ln p_i)\delta_{ij} + \lambda \sum_i \delta_{ij}
\right] = 1 + \ln p_j + \lambda = 0
$$

o sea que $p_j=\text{const}$, y como tiene que estar normalizada, $p_j=1/\Omega$.

Es decir, si no tenemos ninguna informaci√≥n a priori sobre una variable aleatoria, lo m√°s conservador o m√°s desprejuiciado ser√≠a tomar todas las $p_i$ iguales, lo que corresponde a maximizar la entrop√≠a sin ninguna otra restricci√≥n que no sea $\sum_i p_i=1$.

Se puede demostrar adem√°s que **el m√°ximo valor de $S$ posible en funci√≥n de las $p_i$ es la que corresponde a la distribuci√≥n uniforme**, es decir

$$
S=\ln(\Omega)
$$

donde $\Omega$ es la cantidad de s√≠mbolos o letras (comparar la expresi√≥n con el epitafio de Boltzmann).

Cuando poseemos m√°s informaci√≥n sobre la variable aleatoria (sea alg√∫n momento o una probabilidad parcial) podemos entonces enunciar un principio m√°s general para estimar las $p_i$ sin un sesgo:

*Las $p_i$ m√°s "desprejuiciadas" o menos sesgadas son aquellas que maximizan la entrop√≠a usando lo que sabemos de ellas en forma de v√≠nculos. Es decir, es la distribuci√≥n con desinformaci√≥n m√°xima.*

Si tenemos por ejemplo $m$ restricciones lineales:

$$
\sum_i p_i =1\\
\sum_i p_i f_k(i) = F_k,\; k=1,\dots,M
$$

y maximizamos con restricciones:

$$
\partial_{p_j}
\left[
  \sum_i p_i \log_2(1/p_i)
  + \lambda_0 (\sum_{i} p_i -1)
  + \sum_{k=1}^M \lambda_k (\sum_{i} p_i f_k(i) - F_k)
\right]=0
$$

obtenemos

$$
p_i = \frac{\exp[\lambda_1 f_1(i)+\dots+\lambda_M f_M(i)]}{Z(\lambda_1,\dots,\lambda_M)}
$$

donde $\lambda_0$ desapareci√≥ porque ya forzamos la normalizaci√≥n:

$$
Z=\sum_i \exp[\lambda_1 f_1(i)+\dots+\lambda_M f_M(i)]
$$

Para obtener los multiplicadores de Lagrange usamos las ecuaciones de v√≠nculo:

$$
\sum_i p_i f_k(i) \equiv  
\frac{\partial_{\lambda_k} \sum_i \exp[\lambda_1 f_1(i)+\dots+\lambda_M f_M(i)]}{Z(\lambda_1,\dots,\lambda_M)}\\
= \partial_{\lambda_k} \ln Z(\lambda_1,\dots, \lambda_M) = F_k \equiv
\langle f_k \rangle
$$

La entrop√≠a para *distribuciones continuas* se puede generalizar a:

$$
S=-\int dx\; p(x)\ln p(x)
$$

donde se entiende que en el l√≠mite $p(x)=0$, entonces $p(x)\ln p(x)$ se anula.

Esta definici√≥n tiene el inconveniente (que no tiene la distribuci√≥n discreta) de que no es invariante ante un cambio de variable (en una o varias dimensiones). Sin embargo, si las variables consisten en pares de variables din√°micas can√≥nicamente conjugadas el jacobiano es la identidad, y por tanto se garantiza la invariancia.  
No va a ser un problema si el sistema es cu√°ntico ya que los microestados est√°n determinados por n√∫meros cu√°nticos. Sin embargo, una alternativa cuando el jacobiano no es la identidad, es usar la entrop√≠a relativa:

$$
S=-\int dx \;p(x)\ln \left(\frac{p(x)}{q(x)}\right)
$$

donde $q(x)$ es otra distribuci√≥n de referencia, que puede ser simplemente un modelo. Tambi√©n se llama a esta entrop√≠a *distancia de Kullback-Leibler*. Se puede demostrar que es positiva y que es invariante ante cambios generales de variables.

En los casos que nos interesan, que son bien el caso discreto o el caso del espacio de fases, podemos tomar $q(x)$ constante sin embargo. Para maximizar $S$ de una variable continua debemos recurrir al c√°lculo funcional, o a las [derivadas funcionales](https://en.wikipedia.org/wiki/Functional_derivative), en forma similar a lo que se hace en mec√°nica anal√≠tica para derivar las ecuaciones de Euler-Lagrange a partir de la acci√≥n.

Finalmente, vale la pena notar que  
[*Muchas distribuciones continuas famosas pueden obtenerse maximizando la entrop√≠a de Shannon con simples restricciones.*](https://en.wikipedia.org/wiki/Maximum_entropy_probability_distribution)

Como ejemplo de aplicaci√≥n del principio de m√°xima entrop√≠a, consideremos una variable aleatoria continua $-\infty < x < \infty$, de la que s√≥lo conocemos su media $m=\langle x \rangle$ y su varianza $\sigma^2=\langle [x-m]^2 \rangle$. La *funcional* que hay que maximizar es:

$$
{\tilde S}[p(x);\lambda_1,\lambda_m, \lambda_\sigma]
=-\int_{-\infty}^{\infty} dx' \;p(x')\ln (p(x'))
+ \lambda_1 \left[\int_{-\infty}^{\infty} dx' \;p(x') -1 \right]
+ \lambda_m \left[\int_{-\infty}^{\infty} dx' \;p(x') x'-m \right]
+ \lambda_\sigma \left[\int_{-\infty}^{\infty} dx' \;p(x') [x'-m]^2 - \sigma^2 \right]
$$

donde $\lambda_1$, $\lambda_m$ y $\lambda_\sigma$ son multiplicadores de Lagrange.

La *derivada funcional* respecto de $p(y)$ es:

$$
\frac{\delta {\tilde S}}{\delta p(y)} =
-\ln (p(y)) - 1
+ \lambda_1
+ \lambda_m y
+ \lambda_\sigma [y-m]^2 = 0
$$

Por lo tanto:

$$
p(y)=
\exp[1
- \lambda_1
- \lambda_m y
- \lambda_\sigma [y-m]^2]
= \frac{\exp[-(y-m)^2/2\sigma^2]}{\sqrt{2\pi \sigma^2}}
$$

donde en el √∫ltimo paso fuimos r√°pido porque si elegimos $\lambda_m=0$, $\lambda_\sigma=1/2\sigma^2$, y $\exp[1-\lambda_1]= 1/\sqrt{2\pi\sigma^2}$ tenemos asegurados los tres v√≠nculos simult√°neamente:

- $\int_{-\infty}^\infty p(x)dx=\langle 1 \rangle=1$
- $\langle x \rangle=m$
- $\langle [x-m]^2 \rangle=\sigma^2$

Entonces:

- *La distribuci√≥n con media y varianza fijas m√°s "desprejuiciada" de todas, ¬°es la gaussiana!*
- *Cualquier otra distribuci√≥n con igual media e igual varianza ¬°tiene informaci√≥n extra!*


# Mensajes

* La **entrop√≠a de Shannon** de una variable aleatoria discreta o car√°cter con una distribuci√≥n de probabilidad $p_i$ para el valor i-√©simo que puede tomar est√° dada por (si la medimos en bits):

  $$
  S = -\sum_i p_i \log_2(p_i) \geq 0
  $$

* $S$ es tambi√©n la informaci√≥n que brinda, en promedio, conocer el valor que toma $X$, o el grado de sorpresa promedio que existe cuando uno conoce el valor.

* $S$ es tambi√©n el l√≠mite de compresi√≥n de un mensaje de $N$ caracteres, por unidad de car√°cter, en el caso de que no haya correlaciones entre caracteres.

* $S \geq 0$ es una funci√≥n continua de las $p_i$. Cuando la variable aleatoria es continua, $S$ es una funcional de la densidad de probabilidad $p(x)$.

* La distribuci√≥n uniforme, $p_i = \text{cte}$, da el m√°ximo de $S$. Cuanto m√°s concentrada la distribuci√≥n, menor la "sorpresa", y menor $S$.

* El **principio de maximizaci√≥n de entrop√≠a** nos permite estimar $p_i$ de la forma m√°s desprejuiciada posible teniendo solamente en cuenta v√≠nculos conocidos (normalizaci√≥n + otras condiciones).

* En mec√°nica estad√≠stica t√≠picamente, la variable aleatoria es multidimensional, discreta o continua, con una distribuci√≥n de probabilidad conjunta con muchas variables.

* Para mec√°nica estad√≠stica es importante saber que si tengo dos variables aleatorias independientes $P(X,Y) = P(X)P(Y)$, la entrop√≠a de la variable conjunta es la suma de las entrop√≠as de cada variable:  
  $$
  S_{X,Y} = S_X + S_Y
  $$
  Si no son independientes, en general se cumple:
  $$
  S_{X,Y} \leq S_X + S_Y
  $$
  Lo cual se generaliza f√°cilmente a muchas variables.
