Esto es consistente con una longitud de correlación finita a toda temperatura finita.

$$F = 2M J - k_B T \ln \binom{N}{M}

$$  
No conocemos $x$ en equilibrio, pero podemos encontrarla minimizando esta energía libre con respecto a $x$. Queda como ejercicio ver que es $x_{eq}=1/(e^{\beta 2 J}+1)$. Lo importante es notar que es finita si $T>0$, asi que *siempre* habrá un número macroscópico de paredes $M_{eq}=xN$, incompatible con orden magnético de largo alcance.



* El modelo de Ising totalmente conectado si tiene transición ferromagnética. Generar dominios cuesta mucha energía comparado con el aumento de entropía, y la energía libre es no analítica a una $T_c>0$ bien definida.

## Modelo de Ising en $d=2$ en una red cuadrada


image.png

El modelo de Ising en dos dimensiones en una red cuadrada

$$H = - J\sum_{\langle i,j \rangle} \sigma_i \sigma_j - \sum_i \mu B \sigma_i$$

fue resuelto exactamente por Onsager en 1944, quien recibió más tarde el premio Nobel por esta y otras aportaciones a la física estadística. El método es muy avanzado y trabajoso, así que nos limitaremos a describir los resultados. Fue muy importante este trabajo, ya que fue la *primera predicción de una transición de fase en dimensión finita*.

> Notemos que en una red cuadrada en dimensión $d=2$, $z=4$, pero este resultado cambiaría si tuvieramos una red triangular, donde $z=6$ en vez de el $z=4$ de la red cuadrada. El resultado exacto que presentamos aquí es para la red cuadrada. Sin embargo, cualitativamente los resultados son similares para otras redes, y hay resultados que son exactamente los mismos, estos son resultados universales.


Anteriormente vimos que en $d=1$ no había transición de fase, mientras que en $d=\infty$ o espines con infinitos vecinos, hay una transición de fase. Así que la gran pregunta es ahora, ¿para que dimensiones hay transición de fase?
Recordemos que Ising resolvio el caso $d=1$ y con un argumento erróneo concluyo que en $d=2$ no había transición de fase.

Un argumento fenomenológico nos permite aseverar que si hay transición en dimensión $d$, debería haber en dimensión $d+1$. La razón es que el término de interacción se vuelve cada vez más importante, ya que las vecindades de los espines son cada vez más grandes, y cuando se ponen de acuerdo en apuntar para una sola dirección es difícial que las fluctuaciones térmicas los hagan cambiar de opinión a temperaturas bajas. Ya habíamos discutido anteriormente de esto, pensando en excitaciones del estado ferromagnético en forma de paredes. En dos dimensiones las paredes ya no son puntos, sino líneas con una energía elástica, por lo tanto deja de ser obvio si la contribución entrópica de las mismas es tan importante comparado con su costo energético.

Entonces, podemos reformular la pregunta ¿Cuál es la menor dimensión para la que hay transición de fase?

Onsager mostró analíticamente que en $d=2$ hay transición, aunque su solución vale solo a campo cero $B=0$ (no se conoce solución analítica con campo). Por lo tanto hay transición para $d \geq 2$. Por supuesto, la $T_c$ puede depender de $d$, como otras cosas que ya veremos...
La energía libre que encontró Onsager, para una red cuadrada pero con una interacción vertical $J_1$ y una horizontal $J_2$, queda expresada analíticamente en términos de dos integrales

### Solución de Onsager a campo nulo $B=0$ en la red cuadrada

Es posible resolver el modelo en el caso anisotrópico a $B=0$, con una constante de interacción distinta en cada dirección, horizontal o vertical. Si definimos el parámetro $k$

$$k =\frac{1}{\sinh\left(2 K\right)\sinh\left(2 L\right)}$$

donde $K=\beta J_1$ y $L=\beta J_2$, La *energía libre de Helmholtz* por sitio $F$ puede ser expresada como

$$-\beta F = \frac{\log(2)}{2} + \frac{1}{2\pi}\int_{0}^{\pi}\log\left[\cosh\left(2 K\right)\cosh\left(2 L\right)+\frac{1}{k}\sqrt{1+k^{2}-2k\cos(2\theta)}\right]d\theta$$

Para el caso isotrópico $J_1 = J_2=J$, la energía por sitio queda

$$U = - J \coth(2 \beta J) \left[ 1 + \frac{2}{\pi} (2 \tanh^2(2 \beta J) -1) \int_0^{\pi/2} \frac{1}{\sqrt{1 - 4 k (1+k)^{-2} \sin^2(\theta)}} d\theta \right]$$

y la magnetización

\begin{align}
M =
\begin{cases}
\left[ 1 - \sinh^{-4}(2 \beta J) \right]^{1/8} & T<T_c\\
0 & T>T_c
\end{cases}
\end{align}

*¡Una curiosidad sobre ésta expresión de $M$ es que Onsager lo publicó en 1948 sin una prueba! Recién en 1952 Yang publicó la prueba.*

la singularidad está en

$$\sinh\left(\frac{2J_1}{kT_c}\right)\sinh\left(\frac{2J_2}{kT_c}\right) = 1/k=1$$

o sea donde se va a cero $M$, y para el caso isotropico $k =\frac{1}{\sinh^2\left(2 \beta J \right)}$ da la temperatura crítica
$$T_c = \frac{2J}{k\ln(1+\sqrt{2})} = (2.269185\cdots)\frac{J}{k}$$


>Una observación importante para hacer es que el resultado vale para cualquier signo de $J$, aunque hayamos estado suponiendo $J>0$ más arriba. Si $J < 0$ también hay una transición pero *antiferromagnética*, donde no hay magnetización neta pero si una "staggered", donde los espines están para arriba o para abajo como en un tablero de ajedrez. Esta última solución es sin embargo válida para la red cuadrada y otras redes, pero no para la red triangular por ejemplo, porque en ese caso hay frustración y el sistema queda en un estado vidrioso...  



```python
# prompt: plot the function [ 1 -  1/(sinh(2 /x))^4 ]^{1/8} as a function of x

import numpy as np
import matplotlib.pyplot as plt

# Define the function
def magnetization(x):
  """Calculates the magnetization as a function of x."""
  if x <= 0:
    return 0
  if x < 2.27:
    return (1 - 1 / (np.sinh(2 / x))**4)**(1/8)
  else:
    return 0

# Generate x values
x_values = np.linspace(0.1, 2.5, 100)

# Calculate y values
y_values = [magnetization(x) for x in x_values]

# Create the plot
plt.plot(x_values, y_values,'-r')
plt.xlabel("T/J")
plt.ylabel("Magnetization")
plt.title("Ising 2d square lattice, Tc/J=2.269...")
plt.grid(True)
plt.show()
```

```python
# prompt: plot the Onsager free energy of the 2d Ising model at zero field as a function of  T/J

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def free_energy(T_J):
  """Calculates the Onsager free energy for the 2d Ising model at zero field."""
  beta_J = 1 / T_J
  K = beta_J
  L = beta_J

  def integrand(theta):
    k = 1 / (np.sinh(2 * K) * np.sinh(2 * L))
    return np.log(np.cosh(2 * K) * np.cosh(2 * L) + (1 / k) * np.sqrt(1 + k**2 - 2 * k * np.cos(2 * theta)))

  integral_result, _ = quad(integrand, 0, np.pi)
  beta_F = np.log(2) / 2 + integral_result / (2 * np.pi)
  return -beta_F


# Generate T/J values
T_J_values = np.linspace(0.1, 6, 100)

# Calculate free energy values
F_values = [free_energy(T_J) for T_J in T_J_values]

# Create the plot
plt.plot(T_J_values, F_values*T_J_values, '-b')
plt.xlabel("T/J")
plt.ylabel("F/NJ")
plt.title("Onsager Free Energy of the 2d Ising Model at Zero Field")
plt.grid(True)
plt.show()
```

```python
# prompt: plot the Onsager internal energy of the 2d Ising model at zero field as a function of  T/J

import matplotlib.pyplot as plt
import numpy as np
def internal_energy(T_J):
  """Calculates the Onsager internal energy for the 2d Ising model at zero field."""
  beta_J = 1 / T_J
  K = beta_J

  def integrand(theta):
    k = 1 / (np.sinh(2 * K)**2)
    return 1 / np.sqrt(1 - 4 * k * (1 + k)**(-2) * np.sin(theta)**2)

  integral_result, _ = quad(integrand, 0, np.pi / 2)
  U = - np.tanh(2 * K)**(-1) * (1 + (2 / np.pi) * (2 * np.tanh(2 * K)**2 - 1) * integral_result)
  return U

# Generate T/J values
T_J_values = np.linspace(0.1, 6, 100)

# Calculate internal energy values
U_values = [internal_energy(T_J) for T_J in T_J_values]

# Create the plot
plt.plot(T_J_values, U_values, '-g')
plt.xlabel("kT/J")
plt.ylabel("U/N J")
plt.title("Onsager Internal Energy of the 2d Ising Model at Zero Field")
plt.grid(True)
plt.show()
```

```python
# prompt: plot the specific heat of the 2d Ising model at zero field as a function of  T/J

import matplotlib.pyplot as plt
import numpy as np
def specific_heat(T_J):
  """Calculates the specific heat of the 2d Ising model at zero field."""
  beta_J = 1 / T_J
  K = beta_J

  def integrand(theta):
    k = 1 / (np.sinh(2 * K)**2)
    return 1 / np.sqrt(1 - 4 * k * (1 + k)**(-2) * np.sin(theta)**2)

  integral_result, _ = quad(integrand, 0, np.pi / 2)
  U = - np.tanh(2 * K)**(-1) * (1 + (2 / np.pi) * (2 * np.tanh(2 * K)**2 - 1) * integral_result)

  # Calculate the derivative numerically
  dT = 0.00001
  U_plus = internal_energy(T_J + dT)
  C = (U_plus - U) / dT

  return C

# Generate T/J values
T_J_values = np.linspace(0.1, 4, 100)

# Calculate specific heat values
C_values = [specific_heat(T_J) for T_J in T_J_values]

# Create the plot
plt.plot(T_J_values, C_values, '-m')
plt.xlabel("T/J")
plt.ylabel("C/k")
plt.title("Specific Heat of the 2d Ising Model at Zero Field")
plt.grid(True)
plt.show()
```

### Exponentes críticos

Partiendo de la expresión

\begin{align}
M =
\begin{cases}
\left[ 1 - \sinh^{-4}(2 \beta J) \right]^{1/8} & T<T_c\\
0 & T>T_c
\end{cases}
\end{align}

podemos determinar el exponente $\beta$. Para eso notamos que
como $k =\frac{1}{\sinh^2\left(2 \beta J \right)}$, y $k(T_c)=1$, tenemos

\begin{align}
M &=
\left[ 1 - \sinh^{-4}(2 \beta J) \right]^{1/8} =
\left[ k(T_c)^2 - k(T)^2 \right]^{1/8} \\
&=
\left[ k(T_c)+ k(T) \right]^{1/8}
\left[ k(T_c)- k(T) \right]^{1/8} \\
&\approx  
\left[ 2k(T_c) \right]^{1/8}
\left[ k'(T_c)(T_c-T) \right]^{1/8}
\sim (T_c-T)^{1/8}
\end{align}

donde en el último paso usamos que $k'(T_c)=\sqrt{2}\ln(1+\sqrt{2})^2>0$.

Por lo tanto

$$
\boxed{
\beta=1/8
}
$$,

diferente del $\beta=1/2$ del modelo totalmente conectado. Sin embargo, $\beta=1/8$ para cualquier modelo en una red en $d=2$, por más que $T_c$ no coincida. Esto muestra que los exponentes críticos suelen ser *universales*. De hecho, son tan universales que pueden coincidir con los exponentes de sistemas totalmente diferentes, no basados en espines.

## Simulaciones Numéricas

En mecánica estadística se emplean muchos métodos de simulación, pero los más conocidos son el de dinámica molecular, y el de MonteCarlo. El primero se trata simplemente de resolver las ecuaciones de Newton de las partículas, o las ecuaciones de movimiento del sistema. El segundo es más artificial, porque no necesariamente imita la dinámica real de un sistema, pero es muy general y aplica a una gran variedad de casos, con variables continuas o discretas en espacios continuos o discretos. La idea de cualquier simulación en equilibrio térmico es generar configuraciones compatibles con la disrtibución de Boltzmann para poder calcular promedios estadísticos aproximadamente.

Por su generalidad, a continuación describiremos el método de MonteCarlo brevemente.

### Enumeración Exacta

Si el sistema es pequeño, uno podría en principio hacer cálculos exactos por enumeración, con ayuda de la computadora.

¿Es posible hacer las sumas para calcular la función o promedios de distintas cantidades en el modelo de Ising de $N \times N$ sitios?
Para hacerlo deberíamos ser capaces de

* Enumerar *todas* las configuraciones
* Calcular la energía de cada una para calcular el *peso de Boltzmann*.
* Sumar sobre todas las configuraciones, la propiedad por el peso de Boltzmann, y dividir por la función de partición.

Si bien la función de partición se puede calcular, y la energía libre de Helmoltz, si bien exacta, no tenemos una expresión analítica de la misma, así que los promedios termodinámicos los tenemos que hacer sumando también.

### Monte Carlo

El método de Monte Carlo es básicamente una forma de generar secuencialmente muchas configuraciones de un sistema tal que la frecuencia de aparición de las configuraciones sea compatible con el peso de Boltzmann $\sim \exp[-\beta E]$. Es similar a una evolución temporal estocástica pero no necesariamente refleja la verdadera dinámica del sistema, pues lo que nos interesa es muestrear. Para generar esta secuencia se usan números aleatorios para ir generando cambios desde una configuración a la siguiente. Esto quiere decir que cualquier valor medio lo podemos calcular simplemente como

$$\langle O \rangle =
\sum_{\omega} \frac{e^{-\beta E_{\omega}}}{Z} O_{\omega}
\approx \frac{1}{M}\sum_{n_{MC}=n_{eq}}^{M+n_{eq}} O_{n_{MC}}
$$
si $M$ es lo suficientemente grande, de forma que el promedio sea realizado sobre un conjunto suficientemente grande de configuraciones. Notar que un sistema con muchas variables tiene un número enorme de configuraciones permitidas, así que normalmente visitaremos solo un subconjunto, y esperamos que sean las más importantes. Notar además típicamente es necesario descartar $n_{eq}$ pasos de equilibración para descorrelacionarnos de la condición inicial, ya que el muestreo de Boltzmann se espera en el estado estacionario, salvo que tengamos un método de muestreo directo. Para tener buena estadística tenemos que hacer en general muchos pasos de MC. Para un número no muy grande $M$, el estado no se descorrelaciona de su condición inicial, y de alguna manera su evolución con $n$ imita la dinámica de no equilibrio.


El observable $O$ puede ser cualquier función de las coordenadas microscópicas del sistema. Por ejemplo, para el modelo de Ising

$$O = O(\sigma_1,\dots, \sigma_N)$$

Un caso particular es $O=\sigma_1$, en cuyo caso $\langle O \rangle$ es simplemente la magnetización por sitio. En este caso, en la simulación calcularemos

$$\langle O \rangle =
\sum_{n=n_{eq}}^{M+n_{eq}}
O(\sigma^n_1,\dots, \sigma^n_N)
$$

donde $\nu^n \equiv \{\sigma^n_1,\dots, \sigma^n_N\}$ representa la configuración al "tiempo" $n$.

*¿Como se generan las configuraciones con su peso de Boltzmann?*

Sería ideal tener un método de muestreo directo de configuraciones pesadas con el factor de Boltzmann. Algo así como cuando uno directamente muestrea una variable aleatoria con la distribución deseada.


>Supongamos una partícula en un potencial 1d $U(x)=k x^2/2$. Usando la [*Ley fuerte de los grandes números*](https://en.wikipedia.org/wiki/Law_of_large_numbers) tenemos $\langle O(x) \rangle = \int dx\; p(x)O(x) \approx \sum_{i}^M O(x_i)/M$ con $x_i$ muestreado de $p(x)\sim e^{-\beta k x^2/2}$, para lo cual existen métodos rápidos basados en cambios de variables. Si $M\gg 1$ la suma coneverge al valor de expectación. Este es un ejemplo de *muestreo directo*.     


Lamentablemente, para muchas variables, el método de muestreo directo se hace muy difícil sino imposible de encontrar. Por suerte tenemos un truco de muestreo indirecto pero muy general, las *cadenas de Markov*.


Una *cadena de Markov* es un proceso que conecta, en cada paso, una configuración $\nu$ con otra $\nu'$ según una regla estocástica, definida por una probabilidad de transición. La probabilidad de estar en una configuración está dada por una ecuación de movimiento de primer orden, que se conoce como *ecuación maestra*

\begin{align}
\dot{p}_\nu = \sum_{\nu'} [-w_{\nu \nu'} p_\nu
+ w_{\nu'\nu} p_{\nu'}]
\end{align}

donde $w_{\nu \nu'}$ es la probabilidad de transición por unidad de tiempo, de ir de $\nu \to \nu'$. Notar entonces que lo que alimenta la probabilidad de estar en $\nu$ a tiempo $t$ es el balance entre la contribución positiva de la probabilidad de estar en $\nu'$ y caer en $\nu$ a tiempo $t+dt$, y la contribución negativa de estar en $\nu$ e irse a cualquier otra configuración $\nu'$. Notar también que se tiene que cumplir
\begin{align}
0 = \sum_\nu \dot{p}_\nu = \sum_{\nu \nu'} [-w_{\nu \nu'} p_\nu
+ w_{\nu'\nu} p_{\nu'}]
\end{align}
pues $\sum_\nu {p}_\nu =1$ de modo que la matriz de transición tiene que tener ciertas propiedaes.


Toda la astucia del MonteCarlo, consiste en elegir las probabilidades de transición para que cuando lleguemos al estado estacionario y $\dot p_\nu=0$ para todo $\nu$, sea $p_\nu \sim e^{-\beta E_\nu}$ (o lo que corresponda al ensemble, que aquí suponemos canónico). En el estado estacionario se tiene que cumplir entonces

\begin{align}
0 = \sum_{\nu'} [-w_{\nu \nu'} e^{-\beta E_\nu}
+ w_{\nu'\nu} e^{-\beta E_{\nu'}}]
\end{align}

Hay muchas soluciones pero eligiremos la de "balance detallado", es decir la que anula cada sumando. Es decir

\begin{align}
\frac{p_\nu}{p_\nu'}
=e^{-\beta (E_\nu'-E_\nu)}
\equiv e^{-\beta \Delta E_{\nu \nu'}}
= \frac{w_{\nu \nu'}}{w_{\nu' \nu}}
\end{align}

Todavía hay muchas formas de elegir las $w_{\nu \nu'}$ pero elijiremos la de Metropolis por su sencillez.

\begin{align}
w_{\nu \nu'} \propto
\begin{cases}
1 & \Delta E_{\nu \nu'} \leq 0 \\
\exp[-\beta \Delta E_{\nu \nu'}] & \Delta E_{\nu \nu'} > 0
\end{cases}
\end{align}

o si prefieren
\begin{align}
w_{\nu \nu'} \propto
\text{min}[1, \exp[-\beta \Delta E_{\nu \nu'}]
\end{align}


O sea que, si proponemos un cambio de configuración $\nu \to \nu'$ y este baja la energía, se acepta con probabilidad 1, pero si sube la energía se acepta con una probabilidad tanto más chica cuanto mayor sea el incremento de energía. Si $T=0$ tiene sentido porque solo se acepta bajar la energía. Si $T>0$ uno en cambio minimiza la energía libre.

En la práctica, una simulación del modelo de Ising por ejemplo, consiste en

* Partir de una configuración $\nu \equiv \{\sigma_1,\dots, \sigma_N\}$.

* Proponer un cambio en la misma, por ejemplo, invertir uno de los espines $\sigma_i \to -\sigma_i$ y calcular $\Delta E$.

* Si $\Delta E< 0$ aceptar el cambio, es decir
$\nu' \equiv \{\sigma_1,\dots, -\sigma_i, \dots\sigma_N\}$

* Reempezar desde arriba con la nueva configuración.

*Se puede demostrar que la cadena de Markov contruída como lo hicimos siempre converge a un estacionario único, gracias a las propiedades de la matriz de transición.*

Más información en la [wikipedia](https://en.wikipedia.org/wiki/Metropolis%E2%80%93Hastings_algorithm), y de ahí a los libros de física computacional, por ejemplo el de W. Krauth, "Statistical Mechanics, Algorithms and Computations".
