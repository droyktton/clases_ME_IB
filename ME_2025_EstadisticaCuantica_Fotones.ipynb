#Fotones y cuerpos negros

Entender la interacción de la radiación con la materia es fundamental. El campo electromagnético está en todos lados, existe en el vacío, e interactúa con la materia cargada, la cual puede estar distintas fases y reaccionar de manera diferente en cada una. La interacción entre partículas cargadas está mediada por el campo electromagnético. Es de esperar entonces que exista también un **estado de equilibrio termodinámico del campo electromagnético**, y que contribuya a las propiedades termodinámicas.


Todos los cuerpos absorben, reflejan y emiten radiación. Cuando reflejan luz visible podemos verlos con nuestros ojos. Cuando reflejan mejor parte del espectro visible decimos que el objeto tiene un color, y tenemos sensores (conos) en la retina que pueden distinguirlos y pasar esa información al cerebro. Sin embargo, aún en completa oscuridad para nosotros, los objetos siguen reflejando y emitiendo radiación invisible, tipicamente infraroja, lo cual podremos explicar luego.

En general, el espectro de la radiación reflejada, absorbida o emitida depende de propiedades intrínsecas de los materiales, y de sus fases. Por ejemplo los gases diluídos tienen líneas discretas en los espectros de emisión o absorción, debidas a los gaps entre los niveles de energía permitidos. En cualquier caso, la tasa de absorción debe ser igual a la de emisión pues debe existir un equilibrio. Hay sin embargo un límite donde el espectro se vuelve *continuo y universal*, y es el de los "[cuerpos negros](https://en.wikipedia.org/wiki/Black-body_radiation)".

Un cuerpo negro es una aproximación a un cuerpo que absorbe toda la radiación que incide sobre él. Impilcito está que el tamaño del objeto es *mucho mayor que todas las longitudes de onda a las que prestaremos atención*. Es un absorbedor perfecto, pero también un emisor perfecto, pues el equilibrio exige que ambas la tasa de absorción y emisión sean idénticas sino las paredes no estarían en equilibrio con el campo electromagnético. En este caso, la radiación absorbida o existente dentro del material logra equilibrar térmicamente y lo que vemos emitido refleja esa equilibración térmica. Decimos que un cuerpo tiene una emisividad igual a la unidad. En otras palabras, nos interesa esa parte de la radiación emitida que equilibró termicamente, porque como veremos, es *universal*.

Una realización práctica de un cuerpo negro es un agujerito en una cavidad, pues toda la radiación que entra en él queda prácticamente atrapada adentro, y toda la radiación que emite es el resultado de muchas absorciones, emisiones y reflexiones dentro de la cavidad. Notablemente, las paredes interiores de la cavidad no necesitan ser negras!, podrían ser espejos de hecho, o tener cualquier color. Lo importante es que la radiación interactúe lo suficiente con la materia en equilibrio para desarrollar un espectro de equilibrio adentro de la cavidad.

El espectro de emisión de un cuerpo negro depende solo de la temperatura y es universal. Por esto es usado para medir la temperatura de cuerpos cercanos o muy lejanos como estrellas o planetas, con instrumentos tales como pirómetros ópticos, camaras infrarojas, o telescopios espaciales (estos especializados en distintas partes del espectro, ver por ejemplo el reciente [James Webb](https://en.wikipedia.org/wiki/James_Webb_Space_Telescope)). De hecho, podemos hoy en día medir la fiebre de alguien sin ponerle un termómetro o saber cual fue la temperatura del universo primigenio, de cuando el plasma de electrones y protones empezaban a formar moléculas de hidrógeno, y de paso tener evidencia del Big Bang y de la expansión del universo.  

La radiación del cuerpo negro fue el fenómeno que dió origen a las primeras ideas sobre mecánica cuántica allá por el 1900 y que terminó explicando un montón de otros fenómenos que no encajaban con la teoría clásica, ni con la mecánica newtoniana o relativista, ni con el electromagnetismo. La teoría clásica, como estaba, predecía una *catastrofe ultravioleta*, diciendo que un cuerpo caliente debería emitir enormes cantidades de radiación de alta frecuencia. Es decir, predecía que un pedacito de carbón debería emitir rayos X, rayos gamma, etc.

Recién 25 años del problema de la radiación de cuerpo negro, tuvimos una formulación general y universal de la mecánica cuántica.




## Fluctuaciones térmicas del campo electromagnético

Un sólido puede a menudo aproximarse como un medio continuo, y presenta fluctuaciones térmicas. A los modos de vibración los podemos pensar como estados de cuasipartículas cuánticas, los fonones, cada una contribuyendo con un cuanto a la energía de cada uno de esos estados. Los fonones (y otras tantas cuasipartículas) necesitan un medio meterial para existir. En cambio, aún en el vacío aboluto pueden haber fluctuaciones térmicas del campo electromagnético.

A diferencia de los sistemas de partículas debil o fuertemente interactuantes es sin embargo raro pensar en el equilibrio térmico del campo electromagnético.

* ¿Que energía contiene? Sabemos que clásicamente contiene una densidad de energía que es suma de términos cuadráticos, uno asociado al campo eléctrico y otro al magnético.

* ¿Como se excita térmicamente el campo electromagnético? ¿Como se excitan todos sus modos?  ¿Como se comparte esa energía entre todos los modos si cada uno es independiente del otro? Lo hace principalmente interactuando con la materia cargada.

* Dentro de una cavidad, si meto luz monocromática de un laser por ejemplo, ¿no se deberia absorber toda completamente en un ratito? ¡Aunque pusiera espejos desaparecería luego de un rato! Lo que sucede es que luego de un rato hay un espectro continuo, con todos los colores y frecuencias con una intensidad que depende de la temperatura.

La radiación interactúa con la materia a temperatura $T$, que la absorbe y la reemite, pero el asunto va más allá: aún a $T=0$ hay fluctuaciones cuánticas del campo electromagético, la famosa energía del punto cero. Así que el campo electromagnético siempre fluctúa. Y llevándolo más lejos aún, por la relatividad especial, si hay energía pueden aparecer partículas masivas como ser pares electrón-positrón, de modo que ni siquiera podemos asegurarnos el vacío de materia.

Un duda importante que aparece es:

* ¿Como vamos a tratar el campo electromagnético cuánticamente si no es un sistema mecánico? ¿Tiene un hamiltoniano acaso? Y si así fuera, ¿vale aplicar cuantización canónica? Por otro lado, acaso tienen todos los hamiltonianos cuánticos venir de una regla de cuantización de un sistema clásico?

* Tiene coordenadas conjugadas? Tiene momentum el campo electromagnético? Pero no era que el fotón tiene masa cero?

El campo electromagnético es el ejemplo más conocido de campo clásico que no está construído sobre un sistema mecánico. Sin embargo se puede construir una lagrangiana del campo y luego un hamiltoniano, al cual podemos cuantizar canónicamente... Pero en el fondo lo único que necesitamos es que el *modelo* cuántico tenga el límite clásico conocido, el de Maxwell.



## Cuantización del campo electromagnético [lectura opcional]

Dentro de la cavidad valen las ecuaciones de Maxwell
\begin{align}
&\nabla \cdot \mathbf{E} = 0, &
\nabla \cdot \mathbf{B} = 0, \notag\\
&\nabla \times \mathbf{E} = - \frac{\partial \mathbf{B}}{\partial t}, &
\nabla \times \mathbf{B} = \mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t} \notag
\end{align}

entonces el campo adentro, como no hay cargas ni corrientes, satisface las ecuaciones de onda

\begin{align}
&\nabla^2 \mathbf{E} - \frac{1}{c^2} \frac{\partial^2 \mathbf{E}}{\partial t^2} = 0, &
\nabla^2 \mathbf{B} - \frac{1}{c^2} \frac{\partial^2 \mathbf{B}}{\partial t^2} = 0, &
c = \frac{1}{\sqrt{\mu_0 \varepsilon_0}} \notag
\end{align}


>En el espacio libre, tenemos ondas, y los dos términos dan igual contribución, ya que $c=\mu_0\epsilon_0$ y $E=cB$.

---
En notacion covariante, definimos el cuadrivector potencial y el tensor de campo y las ecuaciones de Maxwell se condensan todas en
\begin{align}
A^\mu &= \left( \frac{\Phi}{c}, \mathbf{A} \right), \\
F^{\mu\nu} &= \partial^\mu A^\nu - \partial^\nu A^\mu, \\
\partial_\mu F^{\mu\nu} &= \mu_0 J^\nu, \\
\end{align}

El tensor del campo electromagnético recordemos que es

$$F^{\mu \nu}=\partial^\mu A^\nu - \partial^\nu A^\mu$$, en términos del potencial cuadrivector $A^\mu=(\phi, {\vec B})$ y $\partial_\mu = \partial/\partial x_\mu$. En términos de los campos eléctrico y magnético tenemos

$$
F^{\mu\nu} =
\begin{bmatrix}
0 & -\frac{E_x}{c} & -\frac{E_y}{c} & -\frac{E_z}{c} \\
\frac{E_x}{c} & 0 & -B_z & B_y \\
\frac{E_y}{c} & B_z & 0 & -B_x \\
\frac{E_z}{c} & -B_y & B_x & 0
\end{bmatrix}.
$$

Vale la pena recordar que las ecuaciones de Maxwell se pueden derivar de la lagrangiana del campo electromagnético,

$$\cal L= -\frac{1}{4\mu_0}F^{\mu \nu}F_{\mu \nu}-J_\mu A^\mu$$

Si elegimos el gauge de Lorentz
\begin{align}
\partial_\mu A^\mu = 0,
\end{align}
reobtenemos la ecuacion de ondas en notacion covariante
\begin{align}
\Box A^\nu &= \partial_\mu \partial^\mu A^\nu = \frac{1}{c^2} \frac{\partial^2 A^\nu}{\partial t^2} - \nabla^2 A^\nu = \mu_0 J^\nu,
\end{align}
Por lo tanto dentro de la cavidad
\begin{align}
\Box A^\nu &= 0 \notag
\end{align}

Recordemos también que la densidad de energía del campo es la componente $T^{00}$ del cuadri-tensor energía-momentum

$$\mu_0 T^{\mu \nu}=F^{\mu \alpha}F^{\nu}_{\alpha} - \frac{\eta^{\mu \nu}}{4}F^{\alpha \beta}F_{\alpha \beta}$$

donde $\eta^{\mu \nu}$ es el tensor métrico, que en relatividad especial es muy sencillo.

---

La parte espacial del tensor es
\begin{align}
&T^{ij} = \varepsilon_0 \left( E^i E^j - \frac{1}{2} \delta^{ij} \mathbf{E}^2 \right) + \frac{1}{\mu_0} \left( B^i B^j - \frac{1}{2} \delta^{ij} \mathbf{B}^2 \right), \notag\\
\end{align}
y la presión sobre una pared con normal ${\bf n}$ se obtiene asi
\begin{align}
&P = \hat{\mathbf n} \cdot \mathbf{T} \cdot \hat{\mathbf n} = \hat n_i \, T^{ij} \, \hat n_j, \notag\\
\end{align}

Por ejemplo para paredes conductoras perfectas tenemos
\begin{align}
&\mathbf{E}_\parallel = 0, \quad \mathbf{B}_\perp = 0, \quad
P = \frac{\varepsilon_0}{2} E_\perp^2 + \frac{1}{2\mu_0} \mathbf{B}_\parallel^2 \notag
\end{align}

que también podríamos haber obtenido usando el vector de Pointing.

> *La radiación electromagnética transmite momentum porque tiene un campo magnético*
Si ponemos una carga de pueba $q$, el campo eléctrico $E$ la mueve oscilatoriamente de arriba para abajo, con fuerza $F_E=q E$, pero el campo magnético tuerce su trayectoria con la fuerza de Lorentz en la misma dirección de la radiación, en la dirección del famoso vector de Pointing.

La densidad de energía del campo electromagnético en una cavidad de volumen $V$ es

\begin{align}
T^{00} = \frac{E}{V} = \frac{\epsilon_0}{2}E^2 + \frac{1}{2\mu_0} B^2
\end{align}


## Modos

En una cavidad, podemos describir al campo continuo con modos de Fourier del potencial vector. Asi pasamos del continuo al discreto, pues ${\bf k}$ toma, si bien infinitos valores posibles, valores discretos.

Por ejemplo si hay condiciones de contorno periódicas

$${\bf k}=(2\pi/L_1,\dots,2\pi/L_3)$$

Deberíamos tener primero el hamiltoniano del campo electromagnético. Tal cosa existe. Uno puede empezar por la formulación lagrangiana y luego encontrar el hamiltoniano, cuyas coordenadas y momentos tienen correlación con los campos electromagnéticos y con el momento que transporta la radiación.

La derivación requiere de teoría de campos clásica, usando ${\bf A}$ como grados de libertad del campo, en ausencia de cargas...

En el gauge de Coulomb tenemos $\nabla.{\bf A}=0$ y en Fourier nos queda
$$\mathbf{A}(\mathbf{r}, t) = \sum_\mathbf{k}\sum_{\mu=\pm 1} \left(\mathbf{e}^{(\mu)}(\mathbf{k}) a^{(\mu)}_\mathbf{k}(t) e^{i\mathbf{k}\cdot\mathbf{r}} + \bar{\mathbf{e}}^{(\mu)}(\mathbf{k})   \bar{a}^{(\mu)}_\mathbf{k}(t) e^{-i\mathbf{k}\cdot\mathbf{r}} \right)$$
donde
$$\mathbf{e}^{(\pm 1)}  \equiv \frac{\mp 1}{\sqrt{2}}(\mathbf{e}_x \pm i \mathbf{e}_y) \qquad \hbox{with}\quad \mathbf{e}_x\cdot\mathbf{k} = \mathbf{e}_y\cdot\mathbf{k} = 0.$$
y asi me aseguro que se cumpla la condición de gauge.
Son ondas electromagnéticas básicamente, así que las amplitudes son $a_{\bf k}(t)=e^{i\omega_{\bf k} t}$,
con $\omega_{\bf k}=c |{\bf k}|$. Llamemos $\omega \equiv \omega_{\bf k}$ para condensar un poco la notación...

$$
\begin{align}
\mathbf{E}(\mathbf{r},t) &= i \sum_{\mathbf{k}}{\sum_{\mu =\pm 1} \omega {\left( {\mathbf{e}^{(\mu )}}(\mathbf{k})a_{\mathbf{k}}^{(\mu )}(t){e^{i\mathbf{k}\cdot \mathbf{r}}}-{{\overline{\mathbf{e}}}^{(\mu )}}(\mathbf{k})\bar{a}_{\mathbf{k}}^{(\mu )}(t){{e}^{-i\mathbf{k}\cdot \mathbf{r}}} \right)}} \\
\mathbf{B}(\mathbf{r},t) &= i \sum_{\mathbf{k}} \sum_{\mu =\pm 1} \left \{ \left (\mathbf{k}\times {{\mathbf{e}}^{(\mu )}}(\mathbf{k}) \right ) a_{\mathbf{k}}^{(\mu )}(t) e^{i\mathbf{k}\cdot \mathbf{r}} - \left (\mathbf{k}\times {{\overline{\mathbf{e}}}^{(\mu )}}(\mathbf{k}) \right )\bar{a}_{\mathbf{k}}^{(\mu )}(t){{e}^{-i\mathbf{k}\cdot \mathbf{r}}} \right \}
\end{align}
$$

La cuantización se hace reemplazando las amplitudes de Fourier por operadores de creación y destrucción
\begin{align}
a^{(\mu)}_\mathbf{k}(t)  &\to  \sqrt{\frac{\hbar}{2 \omega V\epsilon_0}}  a^{(\mu)}(\mathbf{k}) \\
\bar{a}^{(\mu)}_\mathbf{k}(t)  &\to  \sqrt{\frac{\hbar}{2 \omega V\epsilon_0}}  {a^\dagger}^{(\mu)}(\mathbf{k}) \\
\end{align}
con
\begin{align}
\left [ a^{(\mu)}(\mathbf{k}), a^{(\mu')}(\mathbf{k}') \right] & = 0 \\
\left [ {a^\dagger}^{(\mu)}(\mathbf{k}),  {a^\dagger}^{(\mu')}(\mathbf{k}')\right ] &=0 \\
\left [ a^{(\mu)}(\mathbf{k}), {a^\dagger}^{(\mu')}(\mathbf{k}')\right ] &= \delta_{\mathbf{k},\mathbf{k}'} \delta_{\mu,\mu'}
\end{align}
Entonces obtenemos campos cuantizados...
\begin{align}
\mathbf{A}(\mathbf{r}) &= \sum_{\mathbf{k},\mu} \sqrt{\frac{\hbar}{2 \omega V\epsilon_0}} \left \{ \mathbf{e}^{(\mu)}  a^{(\mu)}(\mathbf{k}) e^{i\mathbf{k}\cdot\mathbf{r}} +\bar{\mathbf{e}}^{(\mu)}  {a^\dagger}^{(\mu)}(\mathbf{k}) e^{-i\mathbf{k}\cdot\mathbf{r}} \right \} \\
\mathbf{E}(\mathbf{r}) &= i\sum_{\mathbf{k},\mu} \sqrt{\frac{\hbar\omega}{2  V\epsilon_0}} \left \{ \mathbf{e}^{(\mu)} a^{(\mu)}(\mathbf{k}) e^{i\mathbf{k}\cdot\mathbf{r}} - \bar{\mathbf{e}}^{(\mu)}  {a^\dagger}^{(\mu)}(\mathbf{k}) e^{-i\mathbf{k}\cdot\mathbf{r}} \right \} \\
\mathbf{B}(\mathbf{r}) &= i\sum_{\mathbf{k},\mu} \sqrt{\frac{\hbar}{2 \omega V\epsilon_0}} \left \{ \left (\mathbf{k}\times\mathbf{e}^{(\mu)} \right )  a^{(\mu)}(\mathbf{k}) e^{i\mathbf{k}\cdot\mathbf{r}} - \left (\mathbf{k}\times\bar{\mathbf{e}}^{(\mu)} \right ) {a^\dagger}^{(\mu)}(\mathbf{k} ) e^{-i\mathbf{k}\cdot\mathbf{r}} \right \}
\end{align}


El hamiltoniano del campo es
\begin{align}
H&=\frac{1}{2} \epsilon_0 \iiint_{V}{\left( {{\left| E(\mathbf{r},t) \right|}^{2}}+{{c}^{2}}{{\left| B(\mathbf{r},t) \right|}^{2}} \right)}{{\text{d}}^{3}}\mathbf{r} \\
&= V \epsilon_0 \sum_{\mathbf{k}}\sum_{\mu =\pm 1} \omega^2 \left (\bar{a}_{\mathbf{k}}^{(\mu )}(t)a_{\mathbf{k}}^{(\mu )}(t)+a_{\mathbf{k}}^{(\mu )}(t)\bar{a}_{\mathbf{k}}^{(\mu )}(t) \right ).
\end{align}

y usando conmutadores finalmente nos queda
\begin{align}
H &= \frac{1}{2}\sum_{\mathbf{k},\mu=\pm 1} \hbar \omega \left ({a^\dagger}^{(\mu)}(\mathbf{k}) a^{(\mu)}(\mathbf{k}) + a^{(\mu)}(\mathbf{k}) {a^\dagger}^{(\mu)}(\mathbf{k})\right) \\
&= \sum_{\mathbf{k},\mu} \hbar \omega \left ({a^\dagger}^{(\mu)}(\mathbf{k})a^{(\mu)}(\mathbf{k}) + \frac{1}{2}\right ) \\
&= \sum_{\mathbf{k},\mu} \hbar \omega_{\bf k} \left ({a^\dagger}^{(\mu)}(\mathbf{k})a^{(\mu)}(\mathbf{k}) + \frac{1}{2}\right )
\end{align}


## Gas de fotones

Ahora que el campo electromagnético es una colección de osciladores cuánticos, ¡podemos hacer toda la mecánica estadística!

A primera vista parece igual al caso de los fonones, y de hecho ahora tendremos fotones a cambio, pero hay algunas diferencias

* Son discretos los ${\bf k}$ pero no hay $k_{max}$ porque es un continuo... y esto lleva a la *catastrofe ultravioleta* en el caso clásico. O sea... no hay mecánica estadística clásica para los fotones, proque el teorema de equipartición nos daría $\infty k_B T$.

* Las polarizaciones permitidas son solo dos, no hay modos longitudinales del campo electromagnético porque $\nabla. {\bf E}=0$.

* La relación de dispersión para fotones (es el espacio vacío, homogéneo e isótropo) es siempre $\omega_{\bf k}=c |{\bf k}|$, donde $c$ es la velocidad de la luz. Para todo $k$ y no solamente para $k$ pequeño como los fonomes acústicos.

El hamiltoniano del campo electromagnético es entonces de nuevo una colección de osciladores armónicos cuánticos
\begin{align}
H
&= \sum_{\mathbf{k},\alpha} \hbar \omega_{{\bf k},\alpha} \left (a^\dagger_\alpha(\mathbf{k})a_\alpha(\mathbf{k}) + \frac{1}{2}\right ) \\
&= \sum_{\mathbf{k},\alpha} \hbar c |{\bf k}| \left (a^\dagger_\alpha(\mathbf{k})a_\alpha(\mathbf{k}) + \frac{1}{2}\right )
\end{align}
donde ${\bf k}$ barre sobre todos los modos posibles, que son infinitos pero discretos, y $\alpha=1,2$ son las polarizaciones posibles. En la última igualdad ya usamos la relación de dispersión, que dice que la velocidad de fase es $c$, que la propagación es isotrópica (depende del módulo $|{\bf k}|$) e independiente de la polarización (no depende de $\alpha$). Notar que la relación de dispersión es lineal siempre.




## Enegía interna, Capacidad calorífica y presión

Nuevamente, al igual que los fonones, podemos ahora pensar en fotones, cada uno contribuyendo con un cuanto de energía $\hbar \omega_{{\bf k},\alpha}$ a cada uno de los modos ${\bf k},\alpha$. Como no hay restricción para el número de cuantos en cada modo, o para el número cuántico de cada oscilador armónico, el número de ocupación medio de fotnes es idéntico que para los fonones,

$$\langle n_{i} \rangle = \frac{1}{e^{\beta \epsilon_i}-1}$$

es decir, que el fotón es un bosón con $\mu=0$ pues el número de fotones no se conserva. Hasta aquí es parecido a los fonones, pero veremos que las relaciones de dispersión para cada modo son diferentes, además de que el número total de modos es diferente.

* Para fotones tenemos, en el espacio isótropo vacío
$$\omega_{{\bf k},\alpha}=c |{\bf k}|$$
a diferencia de los fonones que tienen esa forma solo para $k$ pequeño comparado con la inversa del parámetro de red $1/a$, o aproximación acústica (recordar la relación de dispersión de la cadena elástica, del tipo $\sim |\sin(ka)|$).   
* Para fonones, hay un $k$ máximo en cada dirección, delimitado por la primera zona de Brillouin, que tiene su origen en el número grande pero finito de modos de vibración. El campo electromagnético en cambio, puede vibrar con longitudes de onda arbitrariamente cortas, pues es un campo continuo, no un campo sobre una red. ¡Esto tiene consequencias!

La energía media se puede calcular sumando sobre los ${\bf k}$ y las polarizaciones $\alpha$, o sumando sobre frecuencias, a partir de la densidad de estados $g(\omega)$. Como los $k\sim 1/L$ en realidad podemos integrar en frecuencias como antes,
$$
\begin{align}
\langle H \rangle &=
\int d\omega g(\omega) \hbar/2
+ \int d\omega\; g(\omega) \frac{\hbar \omega}{e^{\beta \hbar \omega}-1} \\
&=  
E_0
+ \int d\omega\; g(\omega) \frac{\hbar \omega}{e^{\beta \hbar \omega}-1} \\
\end{align}
$$
Notar que la expresión es idéntica que la de los fonones, pero $E_0$ (independiente de $T$) y $g(\omega)$ son diferentes para fotones.

La densidad de estados en tres dimensiones es el número de estados $g(\omega)d\omega$ es dos veces (una por cada polarización) el número de valores de ${\bf k}$ en un casquete de area $4 \pi k^2 dk$, con $\omega=ck$. Es decir

\begin{align}
g(\omega)d\omega = 2 (4 \pi k^2 dk) \frac{V}{(2\pi)^3}
= (k^2 dk)\frac{V}{\pi^2} = \frac{V }{\pi^2 c^{3}} \omega^2 d\omega
\end{align}

Y como no hay $k_{max}$, no hay $\omega$ límite!

Entonces
$$
\begin{align}
\langle H \rangle/V &=
E_0/V
+ \frac{1}{\pi^2 c^{3}} \int_0^\infty d\omega\; \omega^2 \frac{\hbar \omega}{e^{\beta \hbar \omega}-1} \\
&= E_0/V
+ \frac{1}{\pi^2 c^{3}} \beta^{-4}\hbar^{-3} \int_0^\infty dx\; \frac{x^3}{e^{x}-1} \\
&= E_0/V
+ \frac{V \hbar c}{\pi^2 c^{4}} \beta^{-4}\hbar^{-4} \int_0^\infty dx\; \frac{x^3}{e^{x}-1} \\
&=
E_0/V
+ \frac{\hbar c}{\pi^2 } \left( \frac{k_B T}{\hbar c} \right)^4 \int_0^\infty dx\; \frac{x^3}{e^{x}-1}
\end{align}
$$

En mathematica
```Integrate[x^3/(Exp[x]-1),{x,0,Infinity}]```
esa integral da $\pi^4/15$
entonces
\begin{align}
\frac{\langle H \rangle}{V}
&= \frac{E_0}{V} + \frac{\hbar c \pi^2}{15} \left( \frac{k_B T}{\hbar c} \right)^4 \\
&= \frac{E_0}{V} + \frac{\pi^2}{15} \left( \frac{k_B T}{\hbar c} \right)^3 k_B T
\end{align}

Hay un temita con la energía de punto cero
\begin{align}
\frac{E_0}{V}
&= \frac{1}{\pi^2 c^{3}}  \int_0^\infty d\omega\; \omega^2 \frac{\hbar \omega}{2}
= \frac{\hbar}{\pi^2 c^{3}}  \int_0^\infty d\omega\; \frac{\omega^3}{2} = \infty
\end{align}
pero por suerte, como solo importan las diferencias de energías, este infinito es inofensivo...

También podemos calcular la energía libre. Para eso primero notamos que la $Z$ canónica se factoriza en los modos ${\bf k},\alpha$,
\begin{align}
Z = \prod_{{\bf k},\alpha} \frac{e^{-\beta \hbar c k/2}}{1-e^{-\beta \hbar c k}}
\end{align}
donde hemos usado que la de un modo es
\begin{align}
Z_{{\bf k},\alpha}=\sum_{n=0}^\infty
e^{-\beta(\hbar c k(n+1/2))} =
e^{-\beta(\hbar c k/2)} \frac{1}{1-e^{-\beta(\hbar c k)}}
\end{align}
de modo que la energía libre es suma sobre modos
\begin{align}
F &= -k_B T \sum_{{\bf k},\alpha} \ln \left[\frac{e^{-\beta \hbar c k/2}}{1-e^{-\beta \hbar c k}} \right]
= \sum_{{\bf k},\alpha} \left[ \frac{\hbar c k}{2}
+ k_B T \ln [1-e^{-\beta \hbar c k}] \right] \\
&\approx
\sum_\alpha \int \frac{dk_x dk_y dk_z}{(2\pi)^3/V}
\left[ \frac{\hbar c k}{2}
+ k_B T \ln [1-e^{-\beta \hbar c k}] \right]
\\
&=
\int_0^\infty d\omega \; g(\omega)
\left[ \frac{\hbar \omega}{2}
+ k_B T \ln [1-e^{-\beta \hbar \omega}] \right] \\
&= \frac{V }{\pi^2 c^{3}} \int_0^\infty d\omega \; \omega^2
\left[ \frac{\hbar \omega}{2}
+ k_B T \ln [1-e^{-\beta \hbar \omega}] \right] \\
\end{align}
derivando respecto al volumen obtengo la presión de la radiación sobre las paredes de la cavidad,
\begin{align}
P
&= -\partial_V F =
-\frac{1}{\pi^2 c^{3}} \int d\omega \; \omega^2
\left[ \frac{\hbar \omega}{2}
+ k_B T \ln (1-e^{-\beta \hbar \omega}) \right] \\
&= P_0  + \frac{1}{3}\frac{(\langle H \rangle-E_0)}{V}
\end{align}
La presión de punto cero $P_0$, que sobrevive aún a $T\to 0$, es infinita. Su origen son las fluctuaciones de punto cero, o sea, *¡las fluctuationes del vacío!*.  No es problema que sea infinita porque lo que importan son las diferencias de presión. Si el volumen de la cavidad es muy grande estas presiones son practicamente iguales de los dos lados de la pared... Sin embargo, cuando la cavidad es pequeña, por ejemplo dos placas metálicas (es más sencilla la cuenta si hay nodos en las placas) separadas por una pequeña distancia (micro o nanometrica), se puede medir una fuerza, llamada *fuerza de Casimir*. El calculo requiere de algunos trucos de regularización porque hay que restar infinitos... [pero se mide!!](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7838308/).
Este es sin embargo el único caso donde la energía o presión de punto cero es relevante. Para la termodinámica no lo es, porque tiene que ver com cambios.

La relación entre presión y energía es la misma que uno puede obtener clásicamente para un gas de partículas relativistas. En general, si $\epsilon \sim p^s$, la presión es $P=(s/d)(E/V)$ como vimos en los problemas. Para el caso de fotones en tres dimensiones, $s=1$, $d=3$. Esto es consistente con el concepto del campo atrapafo como un **gas de fotones** como partículas ultrarelativistas, sin masa en reposo.


## Ley de Stefan
Si hacemos un agujerito a la cavidad, tenemos lo análogo a la efusión, pero esta vez dejamos escapar fotones de distintas energías, que nos dicen como está repartida la energía del campo en la cavidad entre las distintas frecuencias.
Mirar la radiación que sale del agujerito es ideal, porque es casi exclusivamente una muestra de lo de equilibrado térmicamente adentro, sin contribuciones de reflexiones, etc.

La energía que se escapa por unidad de área y tiempo es simplemente la densidad de energía que hay adentro o densidad de fotones, multiplicado por la componente media de la velocidad perpendicular al agujero, $\langle c_\perp \rangle$. Los fotones que salen por el mismo lo hacen con la misma probabilidad en todas las direcciones, pero a nosotros nos interesan aquellos que salen por el agujero. Si ponemos el eje Z perpendicular al agujero, eso quiere decir direcciones que están en un angulo sólido que va entre $\theta \in [0,\pi/2]$ y $\phi \in [0,2\pi]$ en esféricas.

\begin{align}
\langle c_\perp \rangle = \frac{c}{4\pi} \int_0^{\pi/2} [2\pi \sin(\theta)d\theta]  \cos(\theta) = \frac{c}{4\pi} \frac{2\pi}{2}=\frac{c}{4}
\end{align}

entonces la energía radiada por unidad de tiempo y área (¡del agujero!) es

$$\phi = \frac{c}{4} \frac{\langle H\rangle - E_0}{V} = \frac{\pi^2}{60} \frac{k_B^4 T^4}{\hbar^3 c^2} = \sigma T^4$$

donde $\sigma$ es la **constante de Stefan**

$$\sigma=5.67 \times 10^{-8} W m^{-2} K^{-4}$$

y podemos usar esto para estimar la emisión total de un objeto opaco de una cierta área.
**¡¡¡Notar que es una constante universal!!!!**



> Para entender porque el flujo no es simplemente $\phi=c E/V$, pero $\phi=(c/4)E/V$ uno puede primero ver que la probabilidad de que un fotón apunte en una dirección $[\theta,\theta+d\theta]$ dada es $\sin(\theta)d\theta/2$ pues es el cociente entre el área de una cintita $2\pi \sin\theta d\theta$ en la esfera unidad y el área total $4\pi$ de esa esfera. Pero luego, hay  que notar que la cantidad de energía que sale en cada dirección en un $dt$ es $(E/V) A c dt \cos\theta$, pues fotones muy rasantes pasando por el agujero hay menos que incidiendo normalmente. Luego integramos sobre los fotones con direcciones hacia el agujero  $\int_0^{\pi/2} (\sin \theta/2) ((E/V) A c dt \cos\theta)=(E/V)A c dt/4$. Por lo tanto, por unidad de área y de tiempo, tenemos $(E/V)c/4$.

## Distribución de Planck
Se puede medir también el espectro de energía radiada en función de frecuencia $\omega=2\pi \nu$.  Si usamos que
$${\cal E}(\omega,T)=
\frac{\hbar}{\pi^2 c^2} \frac{\omega^3}{e^{\beta \hbar \omega}-1}
$$
es la densidad de energía por unidad de frecuencia, la potencia espectral emitida por unidad de área del agujerito está dada por
$$I(\omega,T)=
\frac{c}{4}{\cal E}(\omega,T)
= \frac{\hbar}{4\pi^2 c} \frac{\omega^3}{e^{\beta \hbar \omega}-1}
$$
Esta es la famosisima distribución de Planck que desato la revolución cuántica, Einstein mediante.

* A frecuencias relativamente bajas, temperaturas relativamente altas, $\hbar \omega \ll k_B T$,
$$I(\omega,T)\approx
\frac{\hbar}{4\pi^2 c} \frac{\omega^2}{{\beta \hbar}}= \frac{\hbar}{4\pi^2 c} k_B T \omega^2
$$
es la Ley de Raleigh Jeans, que puede ser estimada con argumentos clásicos:

  Por equipartición, si hay $g(\omega)d\omega$ estados de frecuencia $\omega$ y por cada uno hay un $k_B T$, tenemos una intensidad por unidad de frecuencia del orden de $\sim w^2 k_B T$. Si esto valiera siempre, el espectro debería divergir a altas frecuencias: la **catastrofe ultravioleta**. Y nos estariamos quemando vivos. El problema es que no hay corte para las frecuencias de los fonones y por eso diverge.

* La ley de Planck, que básicamente está basada en la cuantización de la energía de los osciladores salva la catástrofe, ya que los efectos cuánticos son relevantes a bajas $T$ o altas $\omega$, es deicir cuando $k_B T < \hbar \omega$. En ese límite, la intensidad cae con frecuencia, pues los osciladores armónicos tienen gaps de energía muy grandes comparados con $k_B T$, similar a lo que pasaba con los grados de libertad más duros de las moléculas que se iban durmiendo antes que los otros...
$$I(\omega,T)=
\frac{\hbar}{4\pi^2 c} \omega^3 e^{-\beta \hbar \omega}
$$

Solo los cuerpos negros emiten este espectro, pero más allá de eso es universal, una propiedad termodinámica del campo de radiación en equilibrio térmico. Cuerpos no negros también emiten radiación térmica, pero el espectro debe ser corregido según las características de la superficie que lo emite.

### Estimación de la Temperatura de la Tierra

image.png

* El sol, a una distancia $D$ de la Tierra, emite aproximadamente una potencia total
$$P_{s}=4\pi R_{s}^2 \sigma T_{s}^4$$

* De esa potencia radiada en todas las direcciones a nosotros nos llega
$$
P_{s} \frac{\pi R_t^2}{4\pi D^2}
$$

* No absorbemos todo porque algo se refleja
$$
P_{abs} = (1-\alpha)P_{s} \frac{\pi R_t^2}{4\pi D^2}
$$
donde $\alpha$ es el *albedo* de la Tierra que es mas o menos $\alpha \approx 0.3$.

* Para que haya un equilibrio y la tierra no se caliente ni enfrie en promedio, necesitamos que emita lo que absorbe. Si suponemos para la Tierra una temperatura efectiva $T$, y una emisividad $\epsilon$. La emisividad la necesitamos porque no es un cuerpo negro perfecto. Notar que como $T \ll T_{s}$ la Tierra emite casi todo en el infrarojo. Entonces
$$
P_{emi} = 4\pi R_t^2 \epsilon \sigma T^4 =
(1-\alpha)4\pi R_{s}^2 \sigma T_{s}^4 \frac{\pi R_t^2}{4\pi D^2} =
P_{abs}
$$
y nos queda (se va $\sigma$ y $R_t$)
\begin{align}
T =
T_{s}  \left(\frac{(1-\alpha)}{\epsilon} \frac{R_s^2}{4 D^2}\right)^{1/4}
\end{align}

* Si usamos que
\begin{align}
T_{\rm s} &= 5772 \ \mathrm{K}\\
R_{\rm s} &= 696.340 \times 10^6 \ \mathrm{m}\\
%R_{\rm t} &= 6.371 \times 10^6 \ \mathrm{m}\\
D &= 1.496 \times 10^{11} \ \mathrm{m} \\
\alpha &= 0.309 \\
\epsilon &=1
\end{align}

````python
import numpy as np

Ts=5772
Rs=6.957E8
D=1.496E11
alpha=0.309
epsilon=1
Rt=6.371E6

Tt = Ts*pow((1.0-alpha)*Rs**2/(4*epsilon*D**2),0.25)

print("Tt=",Tt,"K =",(Tt-273.15),"C")
````
