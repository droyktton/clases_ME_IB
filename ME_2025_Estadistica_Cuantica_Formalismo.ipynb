En estas condiciones —bajas temperaturas, altas densidades o incluso en propiedades macroscópicas a temperatura ambiente— se requieren sumas sobre autoestados cuánticos en lugar de integrales clásicas, y el teorema de equipartición deja de ser válido.

A la aparición de los *números cuánticos* para etiquetar estados, y a la cuantización de la energía, se suma el problema de la *indistinguibilidad de partículas idénticas*. Estas propiedades generan fenómenos colectivos sin análogo clásico, incluso en ausencia de interacciones y aún cuando la discretización de los estados cuánticos se aproxime en forma diferencial.

La estadística cuántica no se limita a escalas microscópicas. El calor específico de los sólidos con sus contribuciones electrónicas y/o vibracionales, la radiación de cuerpo negro o la estabilidad de ciertas estrellas sólo pueden comprenderse en este marco y ambos efectos, la cuantización y la simetría de intercambio tienen efectos notables. En lo que sigue estudiaremos gases cuánticos, fonones y fotones, así como fenómenos macroscópicos originados en la naturaleza cuántica de los constituyentes microscópicos.

En resumen, los efectos cuánticos que estudiaremos son consecuencia de:

* La discretización de la energía de los estados estacionarios, presente incluso para particulas distinguibles.
  * Invalida el teorema de equipartición, el calor especifico clásico.
  * Hace que valga la tercera ley (si el estado fundamental no es degenerado).
  * Etc, etc, etc.

* La simetría de permutación o intercambio en sistemas de partículas idénticas.
  * Los gases no interactuantes por ejemplo ya no cumplen $PV=N k_B T$, ¡y pueden tener transiciones de fase!.
  * El $N!$ aparece sin forzarlo, por la naturaleza de los estados.
  * Etc, etc, etc.

Antes de avanzar con estos efectos, necesitamos un formalismo más general para hacer estadística cuántica.

## Microestados cuánticos: *puros*

En mecánica cuántica, los microestados relevantes ya no se especifican por las variables clásicas $\{q_i,p_i\}$, sino por vectores normalizados del espacio de Hilbert, $|\Psi\rangle$, que pueden expandirse en una base ortonormal $\{|n\rangle\}$:

$$
|\Psi\rangle = \sum_n \langle n|\Psi\rangle |n\rangle,
\quad \langle \Psi|\Psi\rangle=1 .
$$

En la representación de coordenadas, la normalización adopta la forma

$$
\int dr_1\dots dr_N \, |\Psi(r_1,\dots,r_N)|^2 = 1 .
$$

Los observables clásicos $O(q,p)$ se reemplazan por operadores hermíticos $\hat O$, definidos en términos de $\hat q,\hat p$ con los conmutadores $[\hat q_k,\hat p_l]=i\hbar\delta_{kl}$. El resultado de una medición de $\hat O$ es uno de sus autovalores, con probabilidades determinadas por el estado $|\Psi\rangle$. El valor esperado es

$$
\langle \hat O\rangle = \langle \Psi|\hat O|\Psi\rangle,
$$

lo cual, en una base, se interpreta como una forma matricial que refleja la naturaleza de superposición de los microestados.

Al microestado $| \Psi \rangle$ se lo denomina *estado puro*. Lo que tenemos que encontrar es un formalismo que nos permita hacer preducciones estadísticas cuando no podemos saber exactamente en que microestado puro esta el sistema, sino solamente una probabilidad de estar en el.

## Ejemplo: oscilador armónico en el ensemble canónico

Consideremos como ejemplo el **oscilador armónico cuántico**

$$
\hat H = \frac{\hat p^2}{2m} + \frac{1}{2} m \omega^2 \hat x^2 .
$$

Los autoestados del oscilador unidimensional están etiquetados por un número cuántico $|n\rangle$, con autoenergías

$$
\hat H |n\rangle = E_n |n\rangle,
\qquad E_n = \hbar \omega \left(n+\tfrac{1}{2}\right).
$$

La función de partición canónica resulta

$$
Z = \sum_{n=0}^\infty e^{-\beta E_n}
= \frac{e^{-\beta \hbar \omega/2}}{1-e^{-\beta \hbar \omega}} .
$$

De aquí, la energía promedio es

$$
\langle E \rangle
= Z^{-1}\sum_{n=0}^\infty E_n \, e^{-\beta E_n}
= -\frac{d \ln Z}{d\beta}.
$$

En el límite $T \to 0$, se obtiene $\langle E \rangle \to \hbar \omega/2$, la energía del estado fundamental.  
En contraste, en el caso clásico $\langle E \rangle \to 0$ cuando $T \to 0$.

---

### Más allá de la energía: el problema de otros observables

¿Qué ocurre si queremos calcular, por ejemplo, el valor de expectación de $\hat x^2$?

* A temperatura cero ($T\to 0$), el sistema ocupa con probabilidad $1$ el estado fundamental $|0\rangle$.  
  Allí $\langle 0|\hat x|0\rangle=0$, pero $\langle 0|\hat x^2|0\rangle \neq 0$: la posición sigue indeterminada.  

* Para $T>0$, la situación no mejora: $\hat x$ o $\hat x^2$ **no tienen valores definidos en los autoestados** $|n\rangle$, por lo que necesitamos calcular un promedio cuántico incluso dentro de cada microestado.

Esto muestra que en la mecánica estadística cuántica hay en realidad un **doble promedio**:

1. **Cuántico**, debido a que en un estado puro los observables no siempre son números definidos.  
2. **Estadístico**, por nuestra ignorancia respecto de en qué estado puro está el sistema.  

En conjunto, el sistema está en un **estado mixto**.

---

### Notación para los promedios

Para distinguir ambos promedios (al menos por ahora):

- $\langle \hat O \rangle$: valor esperado cuántico en un estado puro.  
- $\overline{\langle \hat O \rangle}$: promedio estadístico sobre la mezcla.

Por ejemplo, para $\hat x^2$ en el ensemble canónico,

$$
\overline{\langle \hat x^2 \rangle}
= \sum_n p_n \langle n|\hat x^2|n\rangle,
\qquad p_n = \frac{e^{-\beta E_n}}{Z}.
$$

---

### Hacia una formulación más general

A primera vista, esta fórmula parece requerir conocer todos los autoestados de $\hat H$.  
Esto no siempre es práctico. Sin embargo, podemos reescribirla en una forma más general:

$$
\overline{\langle \hat x^2 \rangle}
= \text{Tr}\!\left[\frac{e^{-\beta \hat H}}{Z}\,\hat x^2\right].
$$

Definimos entonces el **operador densidad canónico**:

$$
\hat \rho = \frac{e^{-\beta \hat H}}{Z},
\qquad Z = \text{Tr}\!\left[e^{-\beta \hat H}\right].
$$

y de manera compacta:

$$
\overline{\langle \hat O \rangle} = \text{Tr}[\hat \rho \,\hat O].
$$

---

### Propiedades clave del operador densidad

- $\text{Tr}(\hat \rho) = 1$ (normalización).  
- La traza es **invariante bajo cambio de base**.  
- Por lo tanto, la expresión para $\overline{\langle \hat O \rangle}$ no depende de la representación elegida.  

En su forma más general, usando cualquier base

$$
\hat \rho = \sum_\alpha p_\alpha |\Psi_\alpha\rangle \langle \Psi_\alpha |,
$$

donde $p_\alpha$ es la probabilidad de encontrar al sistema en el microestado $|\Psi_\alpha\rangle$.  

---

### Pregunta natural

**¿Cómo determinar las $p_\alpha$ en un ensemble dado?**  

En el caso canónico, si $|\Psi_\alpha\rangle$ son autoestados de $\hat H$, se demuestra que

$$
p_\alpha = \frac{e^{-\beta E_\alpha}}{Z},
$$

en completa analogía con la estadística clásica.  
Esto asegura que la **mecánica estadística cuántica recupera la clásica en el límite correspondiente**, tema que exploraremos más adelante.


## Macroestados cuánticos: *mixtos*

En mecánica estadística no alcanza con describir al sistema mediante un microestado puro \$|\Psi\rangle\$. En general debemos considerar **estados mixtos**, es decir, combinaciones probabilísticas de microestados con pesos \$p\_\alpha\$.

El valor esperado de un observable \$\hat O\$ involucra entonces un **doble promedio**:

* uno **cuántico**, propio de cada estado \$|\Psi\_\alpha\rangle\$,
* y otro **estadístico**, que refleja nuestra falta de conocimiento sobre qué microestado ocupa el sistema.

Esta descripción se formaliza mediante el **operador densidad**:

$$
\hat\rho=\sum_\alpha p_\alpha|\Psi_\alpha\rangle\langle\Psi_\alpha|,
\qquad \langle \hat O\rangle=\mathrm{Tr}(\hat\rho\hat O),
\quad \mathrm{Tr}\,\hat\rho=1 .
$$

En equilibrio, \$\hat\rho\$ adopta formas específicas según el ensemble considerado:

* **Microcanónico:**
  \$\hat\rho=\Omega(E)^{-1}\sum\_n \delta(E-E\_n)|n\rangle\langle n|\$
  (todos los estados de energía \$E\$ son igualmente probables).

* **Canónico:**
  \$\hat\rho=Z^{-1}e^{-\beta\hat H}, \qquad Z=\mathrm{Tr}(e^{-\beta\hat H})\$.

* **Gran canónico:**
  \$\hat\rho=\mathcal{Z}^{-1}e^{-\beta(\hat H-\mu \hat N)}\$.

La entropía asociada es la **entropía de von Neumann**,

$$
S=-k_B\,\mathrm{Tr}(\hat\rho\ln\hat\rho),
$$

análoga a la entropía de Shannon y fundamento tanto de la termodinámica cuántica como de la teoría de la información.

Si queremos saber, cual es la **probabilidad de que el sistema este en un microestado puro** $| \Psi \rangle$, hacemos simplemente

$$P_{\Psi} = \langle \Psi |\hat \rho| \Psi \rangle
$$

En el canónico, eso significa

$$P_{\Psi} =Z^{-1} \langle \Psi |e^{-\beta \hat H}| \Psi \rangle
= \text{Tr}\left[ \hat \rho  |\Psi \rangle \langle \Psi|  \right]
$$

Si lo expresamos en la base de autoestados de $\hat H$ tenemos,

$$P_{\Psi} = \langle \Psi |e^{-\beta \hat H}| \Psi \rangle
= \sum_{n,m} \langle \Psi | n \rangle \langle n |e^{-\beta \hat H}| m \rangle \langle m | \Psi \rangle
= \sum_{n} |\langle n| \Psi \rangle|^2 e^{-\beta E_n}
$$


---

Algunas observaciones clave:

* El formalismo es **independiente de la base** del espacio de Hilbert: \$\hat\rho\$ se puede expresar en cualquier representación, aunque en la práctica suele usarse la base de autoestados de \$\hat H\$ (si está disponible).

* El operador (o matriz) densidad es útil **más allá del equilibrio térmico**. Permite describir experimentos en los que no se logra preparar el sistema en un estado puro, sin necesidad de suponer acoplamiento a un baño ni equilibrio termal.

* La incerteza en los resultados de las mediciones tiene un **doble origen**: la naturaleza probabilística de la mecánica cuántica y nuestro desconocimiento estadístico sobre el estado exacto del sistema.




## Ejemplo: Sistema cuántico de dos estados

Consideremos el hamiltoniano

$$
\hat H = -W \left(|0\rangle \langle 1 |+|1\rangle \langle 0|\right)
$$

que describe un sistema con solo dos estados posibles, $|0\rangle$ y $|1\rangle$.  
Este tipo de hamiltoniano puede modelar, por ejemplo:  

* un **bit cuántico** (qubit),  
* o una partícula que "salta" entre dos sitios localizados.  


Notemos que en la base natural $\{|0\rangle, |1\rangle\}$, el hamiltoniano **no está diagonalizado**.

---

### Acción del hamiltoniano sobre la base

Aplicando $\hat H$ a cada estado de la base:

$$
\hat H |0\rangle = -W|1\rangle, \\
\hat H |1\rangle = -W|0\rangle.
$$

Esto significa que, si el sistema comienza en un estado puro (digamos $|0\rangle$), inmediatamente después de un tiempo infinitesimal $dt$ puede encontrarse en una **superposición coherente** de su estado inicial y del otro estado.  

> Para verlo más claro, recordemos la ecuación de Schrodinger
$$\hat H | \psi(t) \rangle dt = i \hbar (|\psi(t+\Delta t)\rangle-|\psi(t)\rangle)$$
lo que es equivalente a
$$(1 - i \hbar^{-1} \hat H dt) | \psi(t) \rangle  = |\psi(t+d t)\rangle$$
y asi podemos identificar a $1 + i \hbar \hat H dt$ como el generador infinitesimal de traslaciones temporales.
Si aplicamos el generador infinitesimal a los estados de la base, podemos ver su acción, sobre $|\psi(t)\rangle \equiv |0 \rangle$,
$$(1 - i \hbar^{-1} \hat H dt) | 0 \rangle  = | 0 \rangle + i \hbar^{-1} W dt | 1 \rangle$$
implicando que $|\psi(t+dt)\rangle \equiv | 0 \rangle + i \hbar^{-1} W dt | 1 \rangle$, es decir, una superposición de amplitudes de probabilidad, de seguir en el mismo estado con la de cambiar de estado. Vamos perdiendo así información de donde esta la partícula.

>Es más,  cuando $\hat H$ no depende de $t$, podemos obtener fácilmente el operador de evolución entre $t=0$ hasta cualquier otro tiempo $t$, $\hat U \equiv e^{i\hbar {\hat H} t}$, tal que
$\hat U | \psi(0) \rangle  = |\psi(t)\rangle$.



---

### En contacto con un baño térmico

El interés ahora es estudiar qué ocurre cuando este sistema se encuentra en contacto con un baño térmico.  
Para ello conviene **diagonalizar el hamiltoniano**. En la base $\{|0\rangle, |1\rangle\}$, la representación matricial es

$$
\hat H \equiv
\begin{pmatrix}
0 & -W \\
- W & 0
\end{pmatrix}.
$$

La función de partición canónica se define como

$$
Z = \text{Tr}\!\left[e^{-\beta \hat H}\right]
= \langle 0 | e^{-\beta \hat H} | 0 \rangle
+ \langle 1 | e^{-\beta \hat H} | 1 \rangle.
$$

A primera vista, este cálculo parece complicado, ya que involucra una función exponencial de un operador que **no está diagonal en la base elegida**. Sin embargo, diagonalizar $\hat H$ nos dará directamente los autoestados y autovalores, y con ellos podremos obtener $Z$ de manera sencilla.


Es fácil ver que los autoestados normalizados son

$$
|a \rangle = [| 0\rangle +| 1\rangle]/\sqrt{2}\\
|b \rangle = [| 1\rangle -| 0\rangle]/\sqrt{2}\\
$$

Y podemos chequear que son efectivamente los autoestados

$$ \hat H |a\rangle = [- W | 1\rangle  - W | 0\rangle]/\sqrt{2}
= -W |a\rangle
$$

$$ \hat H |b\rangle = [- W | 0\rangle  + W | 1\rangle]/\sqrt{2}
= W |b\rangle
$$

con autoenergías,

$$
E_{a}= - W,\;\;\;
E_{b}= + W
$$

Entonces, la función de partición es (usamos solo las autoenergías, pero los autoestados pueden ser necesarios para otros valores medios)

$$
Z=
e^{-\beta W}+
e^{\beta W}
= 2 \cosh(\beta W)
$$

y vale la pena insistir que este resultado no depende de la base de microestados que usamos, ya que $Z$ es una traza. Sin embarfo es mucho más difícil calcular $Z$ en la base original,

$$Z = \langle 0 |e^{-\beta \hat H}| 0 \rangle + \langle 1 |e^{-\beta \hat H}| 1 \rangle$$

La probabilidad de cada autoestado es entonces

$$p_{a}=e^{-\beta W}/Z, \;\;\; p_{b}=e^{\beta W}/Z$$

y el sistema esta en un **macroestado** mixto.
En un instante dado, el sistema está en un estado cuántico puro, no necesariamente un autoestado, pero como lo desconocemos, usamos un ensemble. Por ejemplo, la probabilidad de estar en $| \Psi \rangle$ es


$$p_\Psi = |\langle a | \Psi \rangle|^2 e^{-\beta E_a}+|\langle b | \Psi \rangle|^2 e^{-\beta E_b} \equiv \langle \Psi | \hat \rho | \Psi \rangle = \text{Tr}\left[\hat \rho | \Psi \rangle \langle \Psi |\right] $$

Lo que significa que el sistema puede estar en cualquier combinación lineal de los autoestados de $\hat H$ que forman base.

Ojo, entonces, que sumemos sobre base de autoestados para calcular $Z$ no significa que el sistema este instantáneamente solo en autoestados. No es como el caso clásico, donde si sumamos sobre todos los microestados posibles. En el caso cuántico sumamos solo sobre los estados de una base, la cual es, solo por conveniencia, la de autoestados del hamiltoniano.


> **Mapeo interesante** Notar que en la base que diagonaliza $\hat H$, el sistema es equivalente a un sistema clásico de un espin $\sigma=\pm 1$ con campo magnético efectivo $\propto W$. Es decir, un sistema con dos niveles separados por un gap, por lo tanto no debería ser raro que la funcion de partición, energía media y calor específico vs temperatura, tengan la pinta siguiente. Este *mapeo* entre sistemas permite conectar lo que uno sabe de uno al otro, y es muy común en mecánica estadística, y se puede llevar muy lejos...





```python
#@title Funcion de partición, energía media, y calor específico
import numpy as np
import matplotlib.pyplot as plt

kB = 1.0    # elegir unidades tales que kB = 1 si se desea
eps = 1.0   # epsilon = alpha = beta
gamma = 0.5 # coupling (>=0)

def Z(beta, eps, gamma):
    return 2.0*np.exp(-beta*eps)*np.cosh(beta*gamma)

def E_mean(beta, eps, gamma):
    return eps - gamma*np.tanh(beta*gamma)

def C_v(T, eps, gamma):
    beta = 1.0/(kB*T)
    return kB*(beta*gamma)**2 * (1.0/np.cosh(beta*gamma))**2

# rango de temperaturas
T = np.linspace(0.05, 5.0, 300)
beta = 1.0/(kB*T)

Z_vals = Z(beta, eps, gamma)
E_vals = E_mean(beta, eps, gamma)
C_vals = C_v(T, eps, gamma)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(T, Z_vals)
plt.xlabel("T")
plt.title("Z(T)")

plt.subplot(1,3,2)
plt.plot(T, E_vals)
plt.xlabel("T")
plt.title(r"$\langle E\rangle$")

plt.subplot(1,3,3)
plt.plot(T, C_vals)
plt.xlabel("T")
plt.title(r"$C_V$")

plt.tight_layout()
plt.show()
```

## Ejemplo: 📦 Partícula en una caja 3D

## 1. Planteo del problema
Potencial:
$$
V(x,y,z) =
\begin{cases}
0 & \text{si } 0 < x < L_x,\; 0 < y < L_y,\; 0 < z < L_z, \\
\infty & \text{fuera de la caja}.
\end{cases}
$$

Ecuación de Schrödinger:
$$
-\frac{\hbar^2}{2m} \nabla^2 \psi(x,y,z) = E \psi(x,y,z),
\qquad \psi = 0 \text{ en las paredes}.
$$

---

## 2. Separación de variables
Suponemos
$$
\psi(x,y,z) = X(x)Y(y)Z(z).
$$

Cada función satisface el problema 1D:
$$
-\frac{\hbar^2}{2m} \frac{d^2 X}{dx^2} = E_x X,
\quad X(0)=X(L_x)=0,
$$
y análogamente para $Y(y),Z(z)$.

---

## 3. Soluciones

Funciones de onda normalizadas:
$$
\psi_{n_x,n_y,n_z}(x,y,z) =
\sqrt{\frac{8}{L_x L_y L_z}}
\sin\!\left(\frac{n_x \pi x}{L_x}\right)
\sin\!\left(\frac{n_y \pi y}{L_y}\right)
\sin\!\left(\frac{n_z \pi z}{L_z}\right),
$$

con $n_x,n_y,n_z = 1,2,3,\dots$.

Energías:
$$
E_{n_x,n_y,n_z} =
\frac{\hbar^2 \pi^2}{2m}
\left(
\frac{n_x^2}{L_x^2} +
\frac{n_y^2}{L_y^2} +
\frac{n_z^2}{L_z^2}
\right).
$$

---

## 4. Propiedades
1. **Cuantización**: solo valores discretos permitidos para $(n_x,n_y,n_z)$.  
2. **Degeneración**: en caja cúbica $L_x=L_y=L_z=L$, distintos tripletes dan la misma energía.  
3. **Límite clásico**: para cajas grandes, los niveles se hacen casi continuos.

## Ejemplo: Particula en una caja 2D

```python
#@title Instalar para visualización interactiva
!pip install ipywidgets
from google.colab import output
output.enable_custom_widget_manager()
```

```python
#@title 1 Partícula en una caja 2d (mas facil de visualizar que 3d)

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# Función de onda en 2D (partícula en una caja cuadrada de lado L)
def psi_2d(nx, ny, L=1.0, res=200):
    x = np.linspace(0, L, res)
    y = np.linspace(0, L, res)
    X, Y = np.meshgrid(x, y, indexing="ij")

    psi_val = (
        np.sqrt(4/L**2)
        * np.sin(nx*np.pi*X/L)
        * np.sin(ny*np.pi*Y/L)
    )
    return X, Y, psi_val

# Plot interactivo
def plot_wavefunction(nx=1, ny=1):
    X, Y, psi_val = psi_2d(nx, ny)

    plt.figure(figsize=(6,5))
    plt.contourf(X, Y, psi_val, 100, cmap="RdBu")
    plt.colorbar(label=r"$\psi(x,y)$")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Wavefunction for n=({nx},{ny})")
    plt.show()

# Widgets para elegir los números cuánticos
interact(
    plot_wavefunction,
    nx=IntSlider(min=1, max=6, step=1, value=1),
    ny=IntSlider(min=1, max=6, step=1, value=1)
)
```

## Estado cuántico vs Estado clásico: para **pensar**...

Para fijar ideas, pensemos en un gas con $N$ partículas en un espacio de dimensión $d$.

---

* En el caso **clásico**, un microestado corresponde a un punto en el espacio de fases, de dimensión $2dN$ (posiciones y momentos de todas las partículas).

* En el caso **cuántico**, un microestado es un vector en un espacio de Hilbert.  
  En la representación de coordenadas, es una función de onda que depende de $dN$ variables, lo que implica un espacio de dimensión infinita.

---

* En un sistema **clásico**, cualquier observable tiene un valor bien definido en un microestado dado.  

* En un sistema **cuántico**, un observable no tiene en general un valor definido en un microestado, salvo que dicho microestado sea un autoestado del observable.

---

* Un microestado cuántico puro puede siempre expresarse como una combinación lineal (superposición) de estados de una base.  

* En el caso clásico, las **transformaciones canónicas** dejan invariante la función de partición.  
* En el caso cuántico, los **cambios de base** en el espacio de Hilbert dejan invariante la función de partición.

---

* La **incerteza clásica** proviene de no conocer exactamente el microestado (todas las posiciones y momentos).  
  Si lo conociéramos con precisión, podríamos predecir la evolución temporal de forma determinista.  

* La **incerteza cuántica** proviene de no conocer exactamente el estado cuántico.  
  Si conociéramos la función de onda completa, la ecuación de Schrödinger nos permitiría predecir la evolución temporal de manera determinista.

---

* **Atención:** tanto la mecánica clásica como la cuántica son teorías *deterministas*.  
  La diferencia está en la interpretación de los observables: en la clásica siempre tienen valores definidos, en la cuántica solo si el estado es un autoestado del observable.  
  El determinismo se pierde cuando no conocemos el microestado inicial con precisión, o cuando el sistema interactúa con un entorno desconocido.

---

* Un **macroestado clásico** es un conjunto de microestados posibles, cada uno con una probabilidad determinada por el ensemble estadístico correspondiente.  

* Un **macroestado cuántico** no es una superposición de microestados, sino un **estado mixto**: asignamos probabilidades a diferentes microestados cuánticos.  
  En equilibrio térmico, esas probabilidades están dadas por los factores de Boltzmann, si los microestados considerados son autoestados del hamiltoniano.

---

* El formalismo de la **matriz densidad** no solo es útil en mecánica estadística de equilibrio, sino también en cualquier situación de la mecánica cuántica en la que aparezcan estados mixtos, ya sea por ignorancia sobre el estado inicial o por interacción con un entorno.
